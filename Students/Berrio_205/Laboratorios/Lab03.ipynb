```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def mapa_logistico(r, x0, n):
    x = np.zeros(n)  
    for i in range(0, n):
        xn = r * x0 * (1-x0)
        x[i] = x0
        x0 = xn
    return x
```

```python
mapa = mapa_logistico(1, 0.5, 100)
```

```python
mapa
```

```python
n = np.arange(0, 100, 1)
n
```

```python
plt.plot(n, mapa, 'r*')
plt.xlabel('Iteraciones')
plt.ylabel('$f(x_n)$')
plt.show()
```

```python
r = np.array([1.9, 2.9, 3.1, 3.5, 3.9,4.0])
numpoints = 100 # Numero sugerido de iteraciones
x0 = 0.5
n = np.arange(0, numpoints, 1)

for i in r:
    mapas_r = mapa_logistico(i, x0, numpoints)
    plt.plot(n, mapas_r, '--', label = 'r = %0.1f' %i)
    
plt.legend()
plt.show()
    
```

```python
def gradiente_descendente(alfa, x0, Np):

    # Funciones
    x = lambda x: x ** 2
    dx = lambda x: 2 * x
    
    for i in range(Np):
        gradient = dx(x0)
        x0 = x0 - alfa * gradient
    return x0
        
    
 # Parámetros
x0 = np.random.randint(1,10)
alfa = 0.1
Np = 100

x_estimado = gradiente_descendente(alfa, x0, Np)

print('Estimación de x (preimagen) en el mínimo: ', x_estimado)
print('Estimación de f(x) (imagen) en el mínimo: ', x_estimado**2)

```

```python
X = np.linspace(-30,30,10)
FX = X**2
DX = 2*X

plt.figure(figsize=(6,4))
plt.plot(X,FX,"--*", label= '$f(x) = x^2$')
plt.plot(X,DX,"--*", label= '$f´(x) = 2x$')
plt.xlabel("preimagen", fontsize=14)
plt.xticks(fontsize=14)
plt.ylabel("imgagen", fontsize=14)
plt.yticks(fontsize=14)
plt.title('Intersección entre curvas $f(x) y f´(x)$')
plt.grid()
plt.legend()
plt.show()
```

```python
X_ = gradiente_descendente(alfa, x0, Np)
DX_ = 2*X_

plt.figure(figsize=(6,4))
plt.plot(X_,DX_,"--*", label= 'Algoritmo gradiente descendente')
plt.xlabel("valores preimagen", fontsize=14)
plt.xticks(fontsize=14)
plt.ylabel("Valores imgagen", fontsize=14)
plt.yticks(fontsize=14)
plt.title('Algoritmo gradiente descendente')
plt.grid()
plt.legend()
plt.show()
```
