```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f= lambda x: (x-1)**2
g= lambda x: x-(x-1)**2
I= lambda x: x # función identidad
```

```python
x=np.linspace(0,2)
```

```python
plt.plot(x,f(x),label="f(x)")
plt.plot(x,g(x),label="g(x)")
plt.plot(x,I(x),label="I")
plt.title("Gráfica de las funciones")
plt.legend()
plt.grid()
```

```python
p0= 1.6 # ejemplo de como itera

for i in range(0,10):
  pi=g(p0)
  p0=pi
  print(p0)
```

```python
p0=1.6
epsilon= 1E-10
p_inicial= g(p0)
Niter=0

while (abs(p_inicial-g(p_inicial)))>=epsilon and Niter<=100:
  p_inicial= g(p0)
  p0=p_inicial
  Niter=Niter+1

print(g(p0))
```

```python
from scipy import optimize

raiz= optimize.fixed_point(g,p0) # verificar con libreria
raiz
```

```python
p0=1.6
epsilon= 1E-10


def punto_fijo(p0,f,g):
  p_inicial=g(p0)
  Niter=0

  while (abs(p0-g(p_inicial)))>=epsilon and Niter<=100:

    p_inicial= g(p0)
    p0= p_inicial
    Niter=Niter+1

  return g(p0)

print(punto_fijo(p0,f,g))
```
