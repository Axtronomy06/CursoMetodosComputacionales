# [Atractor de lorentz](https://en.wikipedia.org/wiki/Lorenz_system)



Dada el siguiente sistema de ecuaciones, asociadas a un modelo simplificado de la atomosfera, encontrar los valores de x, y, z.

\begin{equation}
\frac{\mathrm{d}x}{\mathrm{d}t} = \sigma (y-x)
\end{equation}


\begin{equation}
\frac{\mathrm{d}y}{\mathrm{d}t} = x(\rho-z)-y
\end{equation}

\begin{equation}
\frac{\mathrm{d}z}{\mathrm{d}t} = xy-\beta z
\end{equation}

El modelo anteror permite describir de un a manera simplificada el movimiento atmosférico de masas de aire, los valores x, y, z se corresponden a posiciones de un volumen que contiene  una masa de aire, y los párametros $\sigma$, $\beta$ y  $\rho$ hacen referencia a, las siguientes cantidades fisicas:

\begin{equation}
\sigma = \frac{\mathrm{viscosidad}}{\mathrm{conductividad\ térmica }},
\end{equation}

$\rho$ hace refencia al término de convección entre el suelo y el techo del volumen discretizado.

\begin{equation}
\rho =\frac{g\beta}{v \alpha} (T_s-T_{\infty})L^3
\end{equation}

y $\beta=\frac{\mathrm{Ancho\ del\ cubo}}{\mathrm{Altura\ de\ la\ caja}}$ es un termino geométrico.

````python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
````

````python
# Condición inicial 1

sigma = 10
beta = 8/3
rho = 28

y = [1.5,1.5,1.5]
````

````python
def funcion(y,t, sigma, beta, rho):
    dUdt = [sigma*(y[1]-y[0]),y[0]*(rho-y[2])-y[1] , y[0]*y[1] - beta*y[2]]
    return dUdt
````

````python
t = np.linspace(0,30,10000)
````

````python
# Las demás condiciones

y_1 = [1,1,1]
y_2 = [1.001,1.001,1.001]
````

````python
sol = odeint(funcion, y, t,args=(sigma,beta,rho))
````

````python
sol_1 = odeint(funcion, y_1, t,args=(sigma,beta,25))
sol_2 = odeint(funcion, y_2, t,args=(sigma,beta,28))
````

````python
fig = plt.figure()
ax = fig.add_subplot(projection="3d")
ax.plot(sol[::,0], sol[::,1], sol[::,2],color="c",label="Sol1")
ax.plot(sol_1[::,0], sol_1[::,1], sol_1[::,2],color="violet",label="Sol2")
ax.plot(sol_2[::,0], sol_2[::,1], sol_2[::,2],color="mediumpurple",label="Sol3")
ax.legend()
ax.set_title("Atractor de Lorentz")
plt.draw()
plt.show()
````

La solución parece ser un atractor de Lorentz ya que presenta un comportamiento caótico.
