```python
import numpy as np
import matplotlib.pyplot as plt
```

Prueba de la interpolaci√≥n

```python
x=np.array([1,4,5])

y=np.array([3,2,1])
```

```python
x_valor= np.linspace(x[0],x[1],100)
```

```python
m1 = (y[1]-y[0])/(x[1]-x[0])

recta1 = y[0]+ m1*(x_valor-x[0])
recta1
```

```python
m2= y[1]-y[2]/x[1]-x[2]

recta2= y[1]+m2*(x_valor-x[1])
```

```python
plt.figure()
plt.scatter(x,y,color="red")
plt.plot(x,y,color="red")
plt.plot(x_valor,recta1, color="blue")
plt.title("Iinterpolaci√≥n lineal")
plt.grid()
```

Generalizaci√≥n

```python
x_prueba= np.array([0,2, 4, 5])
y_prueba= np.array([3,5, 4, 5])
```

```python
x_=np.linspace(x_prueba[0],x_prueba[1],100)
```

```python
for i in range(0,len(x_prueba)-1):
    m_= (y_prueba[i+1]-y_prueba[i])/(x_prueba[i+1]-x_prueba[i])
    recta_= y_prueba[i]+m_*(x_-x_prueba[i])
    print(recta_)
```

```python
def ecuacion_recta(x_prueba,y_prueba):

  for i in range(0,len(x_prueba)-1):
    m_= (y_prueba[i+1]-y_prueba[i])/(x_prueba[i+1]-x_prueba[i])
    recta_= y_prueba[i]+m_*(x_-x_prueba[i])
    return recta_


general=ecuacion_recta(x_prueba,y_prueba)
general
```

```python
plt.scatter(x_prueba,y_prueba,color="red")
plt.plot(x_prueba,y_prueba,color="red")
plt.plot(x_,general,color="blue")
plt.title("Iinterpolaci√≥n lineal")
plt.grid()
```
