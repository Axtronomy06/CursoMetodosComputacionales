```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url="https://raw.githubusercontent.com/hernansalinas/autogradesMetodosComputacionales/main/Dataset/Temperature_Bogota_Barcelona.csv"
```

```python
# Punto 1
df=pd.read_csv(url)
```

```python
df
```

```python
# Punto 2
df["AvgTempCelcius"]=(df.AvgTemperature-32)/1.8
```

```python
df
```

```python
# Filtro para Bogotá

filtro_b= ((df.City=="Bogota"))
df1= df[filtro_b].copy() # Filtro avanzado
df1.reset_index()
```

```python
# Punto 3

plt.figure(figsize=(10, 6))
plt.plot(df1["Year"], df1["AvgTempCelcius"], marker="o",color="darkviolet")
plt.title("Temperatura de Bogotá durante los años 1995-2020")
plt.xlabel('Año')
plt.ylabel("Temperatura (°C)")
plt.grid(0.2)
plt.show()
```

```python
# Filtro para Barcelona

filtro_bar= ((df.City=="Barcelona"))
df2= df[filtro_bar].copy()
df2.reset_index()
```

```python
# Punto 4

plt.figure(figsize=(10, 6))
plt.plot(df2["Year"], df2["AvgTempCelcius"], marker="o",color="mediumturquoise")
plt.title("Temperatura de Barcelona durante los años 1995-2020")
plt.xlabel('Año')
plt.ylabel("Temperatura (°C)")
plt.grid(0.2)
plt.show()
```

```python
# Mayor temperatura par Bogotá y Barcelona

mBar = df2["AvgTempCelcius"].max()
mBo = df1 ["AvgTempCelcius"]. max()
```
