```python
from  scipy.misc import derivative
import numpy as np
import matplotlib.pyplot as plt
```

# **Ejemplo 1**

```python
def f(x,a):
    return a*np.sin(x)+(np.tan(x))**2 # diverge en pi/2
```

```python
epsilon=0.1
x= np.linspace(0,np.pi/2-epsilon,100)
```

```python
plt.plot(x,f(x,1),c="r")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
```

```python
x0= np.linspace(0,1,100)
a=1
```

```python
g=derivative(f, x0, dx= 1E-9, args=(a,)) # se le puede pasar un parámetro a
g
```

```python
# Calcularlo a lo largo del dominio  x=[min(x),max(x)]

plt.plot(g,c="r")
plt.title("Derivada")
plt.grid()
```

# **Ejemplo 2**

```python
g= lambda x: x**2
```

```python
N= 100
x0_prueba= np.linspace(0,3,N)

h= (max(x0_prueba)-min(x0_prueba))/N

derivada_1= np.zeros(len(x0_prueba))

def derivada (x0_prueba,h):

  for i in range (0,len(x0_prueba)):
    derivada= 1/(2*h)*(g(x0_prueba[i]+h)-g(x0_prueba[i]-h))
    derivada_1[i]=derivada
  return derivada_1
```

```python
 derivada(x0_prueba,h)
```

```python
plt.plot(x0_prueba, derivada(x0_prueba,h),c="darkviolet")
plt.title("Gráfica de la derivada")
plt.grid()
```

```python
plt.plot(x0_prueba,g(x0_prueba),c="darkviolet")
plt.plot(x0_prueba, derivada(x0_prueba,h),c="blue")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gráfica de la función y la derivada")
plt.grid()
```
