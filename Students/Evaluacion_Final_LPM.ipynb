```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
#Condiciones iniciales:

Tinf = 300
A = 5e-5
p=0.102
K = 200
Tb = 313
h=20
a = (h*p)/K*A

#Theta inicial:
thetao= Tb- Tinf

#Longitud:
xo = 0
xf = 1e-6

#Definiciòn de puntos:
n = 1000
dx = (xf - xf)/n
```

```python

x = np.linspace(xo,xf,n)

w = np.zeros(len(x))
theta = np.zeros(len(x))

w[0] = -(Tb - Tinf) * a**(1/2)
theta[0] = thetao

for i in range(n-1):
  theta[i+1] = theta[i] + dx *w[i+1]
  w[i+1] = w[i] + dx*((a**2)*theta[i])


# Solución númerica
Thetanum = theta + Tinf
# Solución analítica
Thetana = Tinf + (Tb - Tinf)*np.exp(-x*np.sqrt(a))


```

```python
plt.plot(x,Thetanum, label='Sol numérica')
plt.plot(x,Thetana, label='Sol analítica', linestyle='dashed')
plt.title("Temperatura en ala de avión")
plt.xlabel("x")
plt.ylabel("T")
plt.grid()
plt.legend()
plt.show()
```
