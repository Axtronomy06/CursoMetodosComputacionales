# Ejercicio: Simulación de un Péndulo Simple con el Método de Euler y RungeKutta

## Objetivo
Simular el movimiento de un péndulo simple  con fricción utilizando el método de Euler y Runge Kutta para resolver las ecuaciones diferenciales que describen su movimiento.

## Descripción del Péndulo Simple
Un péndulo simple consiste en una masa puntual suspendida de un hilo inextensible y sin masa en el cual  la masa oscila bajo la influencia de la gravedad.

## Ecuaciones del Movimiento
Las ecuaciones de movimiento de un péndulo simple de longitud L, en presencia de  fricción, vienen dadas por :

1. Ecuación diferencial de la posición angular:

\begin{equation}
\frac{d^2\theta}{dt^2} + \frac{g}{L}\sin(\theta) + \gamma L \frac{d\theta}{dt}= 0
\end{equation}

donde $\theta$ es el ángulo de desplazamiento, $g=9.8$m/s la aceleración de la gravedad  y  \gamma la constante de amortigüamiento.

2. La ecuación diferencial se puede separar en dos ecuaciones de primer orden:

\begin{equation}
\frac{d\theta}{dt} = \omega
\end{equation}
   
\begin{equation}
\frac{d\omega}{dt} = -\frac{g}{L}\theta - \gamma L \omega
\end{equation}


## Parámetros
- $g$: Aceleración debido a la gravedad (9.81 $m/s^2$)
- $L$: Longitud del hilo del péndulo (en metros)
- $\theta_0$: Ángulo inicial de desplazamiento (en radianes)
- $\omega_0$: Velocidad angular inicial (en radianes/segundo)
- $\Delta t$: Paso de tiempo para la simulación (en segundos)
- $t_{max}$: Tiempo máximo de simulación (en segundos)

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
````

**1)** condiciones iniciales

````python
#condiciones iniciales
g= 9.8 #aceleracion gravedad [m/s**2]
L= 1 #longitud del pendulo [m]
λ = 0.2 #constante amortiguamiento
theta0= np.pi/12 #angulo inicial de desplazamiento [rad]
omega0= 0
t0=1 #[s]
tmax= 40 #tiempo maximo de simulacion [s]
dt= 0.01 #paso de tiempo para la simulacion [s]
````

**2)** funcion con funciones a integrar

````python

def funcion(w,t, g, L, λ):
  theta, omega = w

  dw0dt= omega
  dw1dt= (-g/L)*np.sin(theta) - λ*L*omega
  dwdt= np.array([dw0dt, dw1dt])
  return dwdt
````

````python
#metodo euler
def euler(theta0, omega0, t0, tmax, dt, g, L,λ):
  N= int((tmax-t0)/dt)
  time= np.linspace(t0 ,tmax,N)
  w = np.zeros((N,2))
  w[0]= np.array([theta0, omega0])

  for i in range(0, N-1):
    w[i+1]= w[i] + dt* funcion(w[i] ,i, g,L,λ)

  return time, w

time, w= euler(theta0, omega0, t0, tmax, dt, g, L, λ)
````

````python
plt.plot(time, w[:, 0], label='angulo', color='r')
plt.plot(time, w[:, 1], label='velocidad angular', color='b')
plt.xlabel('Tiempo [s]')
plt.ylabel('Theta [rad] / Omega [rad/s]')
plt.legend()
plt.title('Movimiento del pendulo con el Metodo Euler')
plt.grid()
plt.show()
````

````python
def funcion_rk(U,t, g, L, λ):
  theta, omega = U

  dU0dt= omega
  dU1dt= (-g/L)*np.sin(theta) - λ*L*omega
  dUdt= np.array([dU0dt, dU1dt])
  return dUdt
````

````python
def rugenkutta(theta0, omega0, t0, tmax, dt, g, L, λ ):
    N= int((tmax-t0)/dt)
    tiempo= np.linspace(t0 ,tmax,N)
    U = np.zeros((N,2))
    U[0]= np.array([theta0, omega0])

    for i in range(0, N-1):
      k1 = dt * funcion_rk(U[i], tiempo[i], g, L, λ)
      k2 = dt * funcion_rk(U[i] + 0.5 * k1, tiempo[i] + 0.5 * dt, g, L, λ)
      k3 = dt * funcion_rk(U[i] + 0.5 * k2, tiempo[i] + 0.5 * dt, g, L, λ)
      k4 = dt * funcion_rk(U[i] + k3, time[i] + dt, g, L, λ)
      U[i+1]= U[i] + (k1 + 2*k2 + 2*k3)/6

    return time, U

tiempo, U= rugenkutta(theta0,omega0, t0, tmax, dt, g, L, λ)
````

````python
plt.plot(tiempo, U[:, 0], label='angulo', color='m')
plt.plot(tiempo, U[:, 1], label='velocidad angular', color='g')
plt.xlabel('Time [s]')
plt.ylabel('Theta [rad] / Omega [rad/s]')
plt.legend()
plt.title('Movimiento del pendulo con Rugenkutta')
plt.grid()
plt.show()
````

````python
#comparar solucion con odeint

def funcion_ode(m,t, g, L, λ):
  theta, omega = m

  dm0dt= omega
  dm1dt= (-g/L)*np.sin(theta) - λ*L*omega
  dmdt= np.array([dm0dt, dm1dt])
  return dmdt


N= int((tmax-t0)/dt)
t= np.linspace(t0,tmax,N)
m0= [theta0, omega0]

solucion= odeint(funcion_ode, m0,  t, args=(g, L, λ))
````

````python
plt.plot(t, solucion[:, 0], label='angulo', color='m')
plt.plot(t, solucion[:, 1], label='velocidad angular', color='g')
plt.xlabel('Time [s]')
plt.ylabel('Theta [rad] / Omega [rad/s]')
plt.legend()
plt.title('Movimiento del pendulo con Odeint')
plt.grid()
plt.show()
````

````python
#superposicion de las graficas para notar diferencias
plt.plot(time, w[:, 0], '--',  label='angulo metodo euler', color='r')
plt.plot(tiempo, U[:, 0], '--',label='angulo metodo rugenkutta', color='m')
plt.plot(t, solucion[:, 0], '--',label='angulo con odeint', color='g')
plt.title('Movimiento de pendulo simple con friccion')
plt.xlabel('Tiempo [s]')
plt.ylabel('Angulo theta [rad]')
plt.grid()
plt.legend()
plt.show()
````

````python
#solucion teorica
def solucion_teorica(theta0, omega0, tmax, dt, g, L):

  fase= np.sqrt(g/L)
  tm= np.arange(t0,tmax, dt)

  theta1 = theta0 * np.cos(fase * tm) + (omega0 / fase) * np.sin(fase * tm)
  omega1 = -theta0 * fase * np.sin(fase * tm) + omega0 * np.cos(fase * tm)

  return tm, theta1, omega1
tm, theta1, omega1 = solucion_teorica(theta0, omega0, tmax, dt, g, L)
````

````python
plt.plot(tm, theta1, label='angulo')
plt.plot(tm, omega1, label='velocidad angular')
plt.title('Solucion Teorica')
plt.grid()
plt.legend()
````
