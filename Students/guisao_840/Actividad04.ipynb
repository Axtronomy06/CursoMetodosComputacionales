```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url='https://raw.githubusercontent.com/hernansalinas/autogradesMetodosComputacionales/main/Dataset/Temperature_Bogota_Barcelona.csv'
```

```python
df=pd.read_csv(url)
df
```

```python
df['celcius']= (df.AvgTemperature-32/1.8) #nueva columna
```

```python
tf=df['City']=='Bogota'
df1=df[tf].copy()
df1
```

```python
#gráfico de AvgTempCelcius como función del año para la ciudad de Bogotá
grados=df1.celcius.unique()
años=df1.Year.unique()
plt.figure()
plt.plot(años,grados, '.g-')
plt.title('Temperatura Bogotá', fontsize=30)
plt.xlabel('Years', fontsize=30)
plt.ylabel('Celcius', fontsize=30)
plt.grid()
```

```python
tf1=df['City']=='Barcelona'
df2=df[tf1].copy()
df2
```

```python
#gráfico de AvgTempCelcius como función del año para la ciudad de Barcelona
grados=df2.celcius.unique()
años=df2.Year.unique()
plt.figure()
plt.plot(años,grados, '.g-')
plt.title('Temperatura Barcelona', fontsize=30)
plt.xlabel('Años', fontsize=25)
plt.ylabel('Celsius', fontsize=25)
plt.grid()
```
