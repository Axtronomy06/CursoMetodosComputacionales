````python
%run -i {pathL}/test02.py

def numeros_defectivos(n):
  divisores=[]

  for i in range(1,n):
    if n%i==0:
      divisores.append(i)

  suma = sum(divisores)


  if suma<n:
    return True

  return False

numeros_defectivos(18)
````

````python
%run -i {pathL}/test04.py

def divisores_propios(n):
  divisores=[]
  for i in range(1,n):
    if n%i==0:
      divisores.append(i)
  return divisores

def get_combinations(arr, k):
    result = []

    def combine(current, start):
        if len(current) == k:
            result.append(current.copy())
            return

        for i in range(start, len(arr)):
            current.append(arr[i])
            combine(current, i + 1)
            current.pop()

    combine([], 0)
    return result


def suma_divisores(n):
  suma=0
  for element in n:
    suma+=element
  return suma

def numeros_semiperfectos(n):
  numbers=divisores_propios(n)

  for i in range(1,len(numbers)):
    combinations=get_combinations(numbers,i)

    for element in combinations:
      suma= suma_divisores(element)
      if suma==n:
        return True
  return False



numeros_semiperfectos(15)
````

Sólo para comprobar tu código,  puedes hacer uso del comando sort de python.
```python
<<< b = [22, 32, 42, 12, 22, 31, 41, 11, 12, 232, 24, 12, 22]
```

````python
def quicksort(arr):
  if len(arr)<=1:
    return arr

  pivot=arr[len(arr)//2 ]
  arr_left=[]
  arr_middle=[]
  arr_right=[]

  for n in arr:
    if n<pivot:

      arr_left.append(n)
    if n>pivot:

      arr_right.append(n)
    if n==pivot:
      arr_middle.append(n)
  return quicksort(arr_left) + arr_middle + quicksort(arr_right)

quicksort([22, 32, 42, 12, 22, 31, 41, 11, 12, 232, 24, 12, 22])
````
