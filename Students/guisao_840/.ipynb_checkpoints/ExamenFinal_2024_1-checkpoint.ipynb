## Ecuaciones de movimiento

### Ecuaciones de movimiento del centro de masa

De acuerdo con la segunda ley de Newton, la ecuación de movimiento a lo largo de la vertical ($y$) puede ser escrita como:

\begin{equation}
m\frac{d^2y}{dt^2} = N - mg
\end{equation}

donde $m$ es la masa de la lápiz, $N$ es la fuerza normal y $g$ es la aceleración de la gravedad. Escribiendo las ecuaciones en términos del ángulo $\theta$, tenemos que:

\begin{equation}
-m\frac{l}{2}\cos \theta \left(\frac{d\theta}{dt} \right)^2 - m\frac{l}{2}\sin \theta \frac{d^2\theta}{dt^2} = N - mg
\end{equation}

### Ecuaciones de rotación del centro de masa

La sumatoria de momentos de fuerza se puede escribir como sigue:

\begin{equation}
I_c\frac{d^2\theta}{dt^2} = N\frac{l}{2} \sin \theta
\end{equation}

donde $I_c = \frac{1}{12}ml^2$ es el momento de inercia de una lápiz de masa $m$ y longitud $l$ respecto a un eje perpendicular a la lápiz en su centro de masa.

Reemplazando el valor de la normal en la ecuación anterior, obtenemos que la ecuación de movimiento para la posición angular está dada por:

\begin{equation}
\frac{d^2 \theta}{dt^2} = \left( \frac{\frac{2g}{l} - \left(\frac{d\theta}{dt}\right)^2\cos \theta }{\frac{1}{3} + \sin^2 \theta} \right)\sin \theta
\end{equation}

Podemos resolver la ecuación anterior si conocemos las condiciones iniciales: $\theta = \theta_0$ en $t = 0$ y $\frac{d\theta}{dt} = \omega_0 = 0$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

**1)** vamos a reescribir la ecuacion para uqe quede una de primer orden y utilizar odeint

```python
#condiciones iniciales
g= 9.8 #[m/s**2]
L=0.1 #[m]
theta0= np.pi/12
omega0= 0
y= (L/2)*np.cos(theta0)
y0= [theta0, omega0]
time_ini=0
time_max=8
dt= 0.01
N= int((time_max-time_ini)/dt)
t= np.linspace(time_ini,time_max, N)

def caida_lapiz(Y,t,g, L): #Y es un vector que contiene angulo inicial y velocidad angular inicial
  theta= Y[0]
  omega= Y[1]
  dU0dt= omega
  dU1dt= ( ( (2*g/L) - omega**2 * np.cos(theta) ) / (1/3 + np.sin(theta)**2) ) * np.sin(theta)
  dUdt= np.array([dU0dt, dU1dt])
  return dUdt


solucion_caida_lapiz= odeint(caida_lapiz, y0, t, args= (g, L))

solucion_theta= solucion_caida_lapiz[:,0]
solucion_omega= solucion_caida_lapiz[:,1]

#para graficar hasta cuando llega el lapiz a pi/2 vamos a crear una funcion que garantice que se grafique hasta este angulo

def theta_max(solucion_caida_lapiz,t):
  theta= solucion_theta
  indice= np.argmax(theta>=np.pi/2)
  if theta[indice]>=np.pi/2:
    return solucion_caida_lapiz[:indice +1 ], t[:indice+1]
  else:
    return solucion_caida_lapiz,t

#finalmente la solucion hasta pi/2 es

solucion_caida, time_caida = theta_max(solucion_caida_lapiz,t)
thetaf= solucion_caida[:,0] #almacena las posiciones angulares
omegaf= solucion_caida[:,1] #almacena la velocidad angular
```

**2)** graficar

```python
plt.figure(figsize=(16, 6))

plt.subplot(2, 2, 1)
plt.plot(time_caida, thetaf, label=r'$\theta(t)$')
plt.xlabel('Tiempo [s]')
plt.ylabel('Ángulo [rad]')
plt.title(r'Ángulo $\theta$ vs. Tiempo')
plt.grid()
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(time_caida, omegaf, label=r'$\omega(t)$', color='orange')
plt.xlabel('Tiempo [s]')
plt.ylabel('Velocidad Angular [rad/s]')
plt.title(r'Velocidad Angular $\omega$ vs. Tiempo')
plt.grid()
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(thetaf, omegaf, label=r'$\omega(\theta)$', color='green')
plt.xlabel(r'Ángulo $\theta$ [rad]')
plt.ylabel(r'Velocidad Angular $\omega$ [rad/s]')
plt.title(r' velocidad angular $\omega$ vs. angulo $\theta$')
plt.grid()
plt.legend()
```

**3)** la solucion es valida desde un tiempo 0 hasta el tiempo en que el lapiz toca el piso, es decir cuando el angulo es pi/2 y esto pasa

```python
indice_valido = np.argmax(solucion_theta>= np.pi/2)
time_valido = time_caida[indice_valido]

print(f'el tiempo final cuando el angulo es pi/2 es: {time_valido:.2f} [s]')
```

**4)** grafica desde diferentes angulos de caida

```python
#tomamos el codigo del primer punto, para hacer un bucle y variar el angulo
g= 9.8 #[m/s**2]
L=0.1 #[m]
omega0= 0
y= (L/2)*np.cos(theta0)
y0= [theta0, omega0]
time_ini=0
time_max=10
dt= 0.01
N= int((time_max-time_ini)/dt)
t= np.linspace(time_ini,time_max, N)

def caida_lapiz(Y,t,g, L): #Y es un vector que contiene angulo inicial y velocidad angular inicial
  theta= Y[0]
  omega= Y[1]
  dU0dt= omega
  dU1dt= ( ( (2*g/L) - omega**2 * np.cos(theta) ) / (1/3 + np.sin(theta)**2) ) * np.sin(theta)
  dUdt= np.array([dU0dt, dU1dt])
  return dUdt

#angulos de caida
angle_ini= [np.pi/12, np.pi/6, np.pi/4, np.pi/3]

plt.figure(figsize=(8,4))

for i,theta0 in enumerate(angle_ini):
  y1=[theta0,omega0]
  caida=odeint(caida_lapiz,y1, t, args=(g,L)) #resuelve la ecuacion para cada angulo

  sol_theta= caida[:,0]
  indice_val= np.argmax(sol_theta>= np.pi/2) #para asegurar que se grafique hasta pi/2
  time_val= t[indice_val]

  plt.plot(t[:indice_valido+1], sol_theta[:indice_valido+1], label=r'$\theta_0 = {}$'.format(theta0))

plt.xlabel('Tiempo [s]')
plt.ylabel(r'Ángulo $\theta$ [rad]')
plt.title('Posición Angular como Función del Tiempo para Diferentes Ángulos Iniciales')
plt.grid()
plt.legend()
```

**5)** interpolacion para hallar el tiempo de caida de cada angulo

```python
from scipy.optimize import newton

angle_ini= [np.pi/12, np.pi/6, np.pi/4, np.pi/3]
time_caidas= []

for theta0 in angle_ini:
  y0=[theta0,omega0]
  sol_fall_lapiz= odeint(caida_lapiz,y0,t, args=(g, L))
  sol_theta= sol_fall_lapiz[:,0]

  def raiz(time):
    return np.interp(time,t,sol_theta) - np.pi/2

  def raiz_prime(time):
    h=1e-5
    return (raiz(time + h) - raiz(time)) / h

  ini,fin= np.min(time_caida), np.max(time_caida)
  try:
    time_fall= newton(raiz, x0= ini+fin/2, fprime= raiz_prime)
    time_caidas.append(time_fall)
    print(f'el tiempo de caida para el angulo {theta0:.2f}rad es: {time_fall:.5f}seg')
  except RuntimeError as e:
    print(f'no se pudo encontrar el tiempo de caida para el angulo {theta0:.2f}rad = {e}')
```

**6)** validar las soluciones de tiempo de caida

```python
from scipy.integrate import quad
angulos=[np.pi/12, np.pi/6, np.pi/4, np.pi/3]

def time_fall_teorico(theta0, l= L, g= g):

  integral,_ = quad(lambda theta: np.sqrt((1/3 + (np.sin(theta)**2))/ (np.cos(theta0) - np.cos(theta))), theta0, np.pi/2)
  return (1/2)* np.sqrt(l/g)*integral

raiz=[]
for angulo in angulos:
  time_teorico= time_fall_teorico(angulo)
  raiz.append(time_teorico)
  print(f"el tiempo teorico de caida para cada theta inicial {angulo:.2f} rad es: {time_teorico:.5f}")
```

**7)** graficar T en funcion en funcion del angulo inicial

```python
plt.figure(figsize=(8,4))
plt.plot(angulos, raiz, 'o-', color='m', label=r'$T(\theta_0)$')
plt.title('Tiempo Teorico vs angulos iniciales')
plt.xlabel('angulo inicial [rad]')
plt.ylabel('tiempos [seg]')
plt.grid()
plt.legend()
```
