*CAMPO ELECTRICO*





```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#campo electrico para carga positiva
def campo_q1(xp,yp,x1,y1,q1):
  r1=np.array([x1,y1]) #vector posicion a la carga positiva
  rp=np.array([xp,yp])
  delta_q1= rp-r1

  magnitud_deltaq1= np.linalg.norm(delta_q1) #norma
  unitario1=(1/magnitud_deltaq1)*(delta_q1)

  magnitud_campoq1=(9*10**(-9))*(q1/(magnitud_deltaq1)**2)
  campo_q1=(magnitud_campoq1)*(unitario1)

  return campo_q1

#ahora campo para carga negativa
def campo_q2(xp,yp,x2,y2,q2):
  rp=np.array([xp,yp])
  r2=np.array([x2,y2]) #vector posicion a la carga negativa
  delta_q2=rp-r2

  magnitud_deltaq2= np.linalg.norm(delta_q2)
  unitario2=(1/magnitud_deltaq2)*(delta_q2)
  magnitud_campoq2=(9*10**(-9))*(q2/(magnitud_deltaq2)**2)
  campo_q2=(magnitud_campoq2)*(unitario2)

  return campo_q2
```

```python
#campo electrico generalizado
def campo_electrico(carga1,xp,yp,x,y):
  r1=np.array([x,y])
  rp=np.array([xp,yp])
  delta_carga1= rp-r1
  magnitude_deltacarga=np.linalg.norm(delta_carga1)
  unitario1=(1/magnitude_deltacarga)*(delta_carga1)

  magnitude_campocarga=(9*10**(-9))*(carga1/(magnitude_deltacarga)**2)
  campo_carga=(magnitude_campocarga)*(unitario1)

  return campo_carga
#campo_electrico(4,3,7,-1,0)
```

*Gráfica del Campo Eléctrico*

```python
q1=2
q2=-2
x1=-1
x2=1
y1=0
y2=0
xp=0
yp=0
```

```python
E1=campo_q1(xp,yp,x1,y1,q1)
E2=campo_q2(xp,yp,x2,y2,q2)
E = E1 + E2
xp1=0.5
yp1=0.5

plt.figure()
plt.title('Campo Eléctrico', fontsize=24)
plt.grid()
plt.plot(x1,y1,"ro")
plt.plot(x2,y2,"ro")
plt.plot(xp1,yp1,"bo")
plt.quiver(xp1,yp1,E[0],E[1])
```

```python
#grafica generalizada
def campo_electrico(carga1,xp,yp,x,y):
  r1=np.array([x,y])
  rp=np.array([xp,yp])
  delta_carga1= rp-r1
  magnitude_deltacarga=np.linalg.norm(delta_carga1)
  unitario1=(1/magnitude_deltacarga)*(delta_carga1)

  magnitude_campocarga=(9*10**(-9))*(carga1/(magnitude_deltacarga)**2)
  campo_carga=(magnitude_campocarga)*(unitario1)

  return campo_carga


#xp1=0.5
#yp1=0.5
x= np.arange(-6,6,1)
y=np.arange(-6,6,1)

plt.figure(figsize=(5,5))
plt.plot(x1,y1,"ro")
plt.plot(x2,y2,"ro")

for xp1 in x:
  for yp1 in y:
    E1=campo_q1(xp1,yp1,x1,y1,q1)
    E2=campo_q2(xp1,yp1,x2,y2,q2)

    E=E1 + E2
    plt.quiver(xp1,yp1,E[0], E[1])

plt.show()
```
