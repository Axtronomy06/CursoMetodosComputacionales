```python
import numpy as np
import matplotlib.pylab as plt
import json
```

```python
gravity=9.8

def component_x(x0, V_i, t, theta):
  position_x= x0 + (V_i*np.cos(theta))*t
  return position_x

def component_y(y0, V_i, t, theta):
  position_y= y0 + (V_i*np.sin(theta)*t)- (0.5*gravity*t**2)
  return position_y

def Flight_time(V_i, theta):
  Tv= (2*V_i*np.sin(theta))*(1/gravity)
  return Tv
```

```python
v0=2 #m/s
theta = 30
theta_rad=np.deg2rad(theta)
Tv= Flight_time(v0, theta_rad)

t=np.linspace(0,Tv,100)
x=component_x(0, v0, t, theta_rad)
y=component_y(0, v0, t, theta_rad)
```

**GRÁFICA TIRO PARABÓLICO**



```python
fig,ax=plt.subplots()
ax.plot(x,y, 'b')
ax.set_xlabel('x')
ax.set_ylabel('y')

```

```python
#generalizacion del codigo para la lectura de archivo
def read_file(nombre_archivo):
  with open(nombre_archivo, 'r') as archivo:
    d=json.load(archivo)
  return d

  gravity=9.8

def component_x1(x0, velocity, t, angle):
  angle_rad=np.deg2rad(angle)
  position_x1= x0 + (velocity*np.cos(angle_rad)*t)
  return position_x1

def component_y1(y0, velocity, t, angle):
  angle_rad=np.deg2rad(angle)
  position_y1= y0 + (velocity*np.sin(angle_rad))*t- (0.5*gravity*(t**2))
  return position_y1

def Flight_time1(velocity, angle):
  angle_rad=np.deg2rad(angle)
  Tv1= (2*velocity*np.sin(angle_rad))/gravity
  return Tv1

print(read_file('condicionesinicales.json'))
```

```python
d1= read_file('condicionesinicales.json')
v0_1= d1['velocity']
theta1=d1['angle']
Tv_1= Flight_time1(v0_1,theta)

t1=np.linspace(0,Tv_1,100)
x1=component_x1(0,v0_1,t1,theta1)
y1=component_y1(0,v0_1,t1,theta1)
```
