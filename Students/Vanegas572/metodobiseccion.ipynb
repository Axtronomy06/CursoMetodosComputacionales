Método de busqueda de raíces

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = lambda x : x*(np.exp(x))-1
```

```python
f(2)
```

```python
x = np.linspace(-1,1,100)
```

```python
#@title Código que utiliza el método de bisección para encontrar raíces


#vamos a darle un intervalo a la funcion para encontrar nuestra raiz de forma aproximada
a = 0.4    #defino un punto en x de tal forma que la funcion evaluada en ese punto sea negativa
b = 0.6    #defino otro punto donde la funcion sea positiva
c = a
e = 1e-5  #defino un epsilon para definir un intervalo de error ya que nunca será exactamente 0 la función
iter = 0
maxiter=1000

while ((abs(f(c))>=e) and ((f(a)*f(b)) < 0) and (iter<maxiter)):    #itera hasta que el valor de la función en un punto medio sea mayor a mi epsilon
    c = (a+b)*0.5                                                   #encontramos el punto medio
    punto_medio = f(c)                                              #calculamos el valor de la función en ese punto
    if (punto_medio>0):                                             #si el valor es positivo entonces se redefine mi valor inicial b
      b=c
    else:
      a=c                                                           #si es negativo se redefine mi valor inicial a
    iter=iter + 1
    if (iter==maxiter):
      print("No se alcanza la posicion deseada")
print(c, f(c))
```

```python
#@title Graficamos la función para visualizar la ubicación de las raíces
plt.plot(c,f(c), "o", color = "orange", label = "Punto donde se encuentra la raíz")
plt.plot(x,f(x), color = "mediumblue")
plt.axhline(0, color='dimgray', linestyle='--')
plt.axvline(0, color='dimgray', linestyle='--')
plt.title("Función f(x) = x$e^x - 1$")
plt.ylabel("f(x)")
plt.xlabel("x")
plt.grid()
plt.legend()
plt.show()
```

```python
#@title Utilizando la libreria scipy verificamos el resultado
from scipy import optimize
```

```python
root = optimize.bisect(f,a,b)
root
```
