```python
import pandas as pd
import matplotlib.pyplot as plt
```

1. Leer el url

```python
url = "https://raw.githubusercontent.com/hernansalinas/autogradesMetodosComputacionales/main/Dataset/Temperature_Bogota_Barcelona.csv"
```

```python
df = pd.read_csv(url)
df
```

2. Agregar una nueva columna con la temperatura en grados centrígados

```python
df["AvgTempCelcius"] = (df.AvgTemperature-32)/1.8
df
```

3. Vamos a realizar una gráfica donde se muestre la temperatura en función de los años para la ciudad de Bogotá y para la de Barcelona

```python
#comienzo sacando un dataframe con la información de bogotá
a = df[df["City"]=="Bogota"]
a
```

```python
#otro con la de barcelona
b = df[df["City"]=="Barcelona"]
b
```

4. Sacar la temperatura máxima de ambas ciudades

```python
#comienzo sacando la temperatura máxima para bogotá
mbogota = a["AvgTempCelcius"].max()
mbogota
```

```python
#y saco la fila donde se ubica ese máximo para poner extraer su información
año_maximob = a.loc[a["AvgTempCelcius"] == mbogota]
año_maximob
```

```python
#igual para barcelona
mbarcelona = b["AvgTempCelcius"].max()
mbarcelona
```

```python
año_maximobar = b.loc[b["AvgTempCelcius"] == mbarcelona]
año_maximobar
```

5. Gráfica donde se ilustran también la temperatura máxima de cada ciudad

```python
plt.plot(a["Year"],a["AvgTempCelcius"], color = "rebeccapurple", label="Bogota")
plt.plot(año_maximob["Year"],a["AvgTempCelcius"].max(), marker = "o", color="darkorange")
plt.plot(b["Year"],b["AvgTempCelcius"], color = "darkgreen", label="Barcelona")
plt.plot(año_maximobar["Year"],b["AvgTempCelcius"].max(), marker = "o", color="orangered")
plt.title("Temperatura en función del año")
plt.ylabel("Temperatura °C",labelpad=10)
plt.xlabel("Años",labelpad=10)
plt.legend()
plt.grid()
```
