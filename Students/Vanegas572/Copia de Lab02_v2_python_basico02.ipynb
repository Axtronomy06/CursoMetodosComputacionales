````python
#funcion que me dice si un numero es defectivo

a = int(input("Ingrese un número entero mayor o igual a 2 "))
def numeros_defectivos(a):
  x = []
  y = []
  for divisor in range(1,a):                            #por cada numero en el rango (excluimos el 2)
    if (a % divisor) == 0:                              #si el residuo entre el numero del rango y n es 0 entonces
      x.append(divisor)                                 #agregar ese numero a la lista vacia
  #print(x)
  contador = 0                                          #iniciamos un contador
  for i in x:                                           #por i en la lista de divisores
    contador = contador + i                             #contador se redefine como ese numero
  if contador < a:                                      #despues de sumar todos los numeros si es menor a a, entonces es defectivo
    return True
  else:
    return False

numeros_defectivos(a)

````

````python
#numeros semiperfectos

a = int(input("Ingrese un número entero: "))
def numeros_semiperfectos(a):
  x = []
  for divisor in range(1,a):
    if (a % divisor) == 0:
      x.append(divisor)                                     #se sacan los divisores como anteriormente
  y = list(reversed(x))                                     #crearemos la variable y que sera la lista de los numeros de x pero invertida, pues asi se llega mucho mas rápido a la suma de a
  contador4 = 0                                             #iniciamos un contador para que cuente el numero de suma que hara el siguiente ciclo
  l = []
  for i in y:                                               #por cada i en la lista revertida
    contador4 = contador4 + i                               #contador será contador más ese i pero teniendo en cuenta dos condiciones
    if contador4 < a:                                       #si al sumar, contador es menor a a
      l.append(i)                                           #se agrega a la lista ese numero
    if contador4 > a:                                       #si la suma se pasa de a
      contador4 = contador4 - i                             #a el contador se le resta ese numero que nos pasó de a
    if contador4 == a:                                      #si la suma es exactamente a entonces es semiperfecto
      l.append(i)
      #print(l)
      #print(contador4,"soy contador")
      return True
  else:
    return False

numeros_semiperfectos(a)

````


7. Diseñar un programa en el que entrado dos números `a`  y `b` retorne una variable booleana **True** o **false**.

  [Números amigos](https://es.wikipedia.org/wiki/N%C3%BAmeros_amigos) `a` y `b` tales que a es la suma de los divisores propios de `b` y viceversa.
    La rutina se debera llamar numeros_amigos








````python
#funcion que me dice si dos numeros son amigos o no

a = int(input("Ingrese un número a: "))
b = int(input("Ingrese un número b: "))
def numeros_amigos(a,b):
  x = []
  y = []
  for divisor in range(1,a):
    if (a % divisor) == 0:
      x.append(divisor)                                       #se le sacan los divisores a a
  #print(x)
  contadorx = 0
  for i in x:
    contadorx = contadorx + i                                 #se suman los divisores
  #print(contadorx)
  for divisor2 in range(1,b):                                 #igualmente para b
    if (b % divisor2) == 0:
      y.append(divisor2)
  #print(y)
  contadory = 0
  for k in y:
    contadory = contadory + k
  #print(contadory)
  if contadorx == b and contadory == a:                       #si los contadores son iguales al numero opuesto entonces son amigos
    return True
  else:
    return False

numeros_amigos(a,b)
````
