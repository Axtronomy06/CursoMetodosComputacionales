# Este método nos ayuda a encontrar las raíces de una función utilizando aproximaciones por medio de las rectas  tangentes y secantes.

## Primer método: Rectas Tangentes

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#definamos una función lambda

f = lambda x : (x**2) - 3
```

```python
#definimos un arreglo de números para visualizar la función

x = np.linspace(-5,5,100)
```

```python
#definimos la derivada de f(x)

df = lambda x : 2*x
```

```python
#@title Programamos la recurrencia

x_o = 4
epsilon = 1e-5
iter = 0
maxiter = 1000

while (abs(f(x_o))>=(epsilon)) and (iter<maxiter):
  a = f(x_o)
  b = df(x_o)
  x_o = (x_o) - (a/b)
print(x_o, f(x_o))
```

```python
#@title Graficamos la función y su raíz
plt.plot(x_o, f(x_o), "o", color = "deeppink", label = "Ubicación de la raíz")
plt.plot(x,f(x), color = "limegreen", label = "Función")
plt.plot(x, df(x), color = "skyblue", label = "Derivada de la función")
plt.title("Función f(x) = $x^2 -3$")
plt.axhline(0, color='dimgray', linestyle='--')
plt.axvline(0, color='dimgray', linestyle='--')
plt.ylabel("f(x)")
plt.xlabel("x")
plt.grid()
plt.legend()
plt.show()
```

```python
#@title Verificamos el resultado
from scipy import optimize
```

```python
root = optimize.newton(f,4,df)
root
```

# Como vemos las dos raíces son muy similares

## Segundo método: Rectas Secantes

```python
#@title Programamos la recurrencia

x_o = 4
x_1 = 0.1
epsilon = 1e-3
iter = 0
maxiter = 1000

while (abs(f(x_o))>=(epsilon)) and (iter<maxiter):
  a = f(x_o)
  b = f(x_1)
  x_2 = (x_o) - ((a*(x_o-x_1))/(a-b))
  x_o, x_1 = x_2, x_o

print(x_2, f(x_2))
```

```python
#@title Gráfica con el método de la secante
plt.plot(x_2, f(x_2), "o", color = "deeppink", label = "Ubicación de la raíz")
plt.plot(x, f(x), color = "limegreen", label = "Función")
plt.plot([4, 0.1],[f(4), f(0.1)] , "-", color = "coral", label = "Recta secante inicial")
plt.title("Función f(x) = $x^2 -3$")
plt.axhline(0, color='dimgray', linestyle='--')
plt.axvline(0, color='dimgray', linestyle='--')
plt.ylabel("f(x)")
plt.xlabel("x")
plt.grid()
plt.legend()
plt.show()
```
