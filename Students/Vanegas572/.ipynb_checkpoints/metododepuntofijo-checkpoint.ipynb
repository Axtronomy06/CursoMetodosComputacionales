```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#definimos nuestra función a la que le vamos a sacar la raíz
f = lambda x: (x-1)**2
```

```python
#definimos la función x para utilizarla
h = lambda x: x
```

```python
#definimos la resta de las funciones
g = lambda x: h(x) - f(x)
```

```python
x = np.linspace(-25, 25, 1000)
```

```python
#@title Recurrencia para el método del punto fijo
p_0 = 1.5
epsilon = 1e-8

while ((abs(h(p_0) - g(p_0))) >= epsilon):
  a = g(p_0)
  p_0 = a

print(p_0)
```

```python
#@title Grafica para visualizar las funciones
plt.plot(p_0, f(p_0), "o", color = "darkorange")
plt.plot(x, f(x), color = "deeppink" , label="$f(x) = (x-1)²$")
plt.plot(x, h(x), color = "black", label = "f(x) = x")
plt.plot(x, g(x), color = "forestgreen", label= "f(x) = x - (x-1)²")
plt.axhline(0, color = "dimgray", linestyle = "--")
plt.axvline(0, color = "dimgray", linestyle = "--")
plt.title("Método del punto fijo")
plt.ylabel("f(x)")
plt.xlabel("x")
plt.grid()
plt.legend()
plt.show()
```

```python
from scipy import optimize
```

```python
#@title Verificamos que nuestro código funcione, para esto se utilizó la misma tolerancia que maneja la función de optimize $(1e^-8)$
root = optimize.fixed_point(g, p_0)
root
```
