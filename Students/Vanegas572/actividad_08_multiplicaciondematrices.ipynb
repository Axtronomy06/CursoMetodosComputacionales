### Vamos a diseñar un programa que multiplique matrices

```python
#@title Realizaremos la multiplicación manual de 2 matrices de dimensión 2 × 2

matriz_1 = np.array([[1, 2], [3, 4]])
matriz_2 = np.array([[1, 2], [3, 4]])

#verificar que si se pueden multiplicar

dim_1 = np.shape(matriz_1)
dim_2 = np.shape(matriz_2)
print(dim_1, dim_2)

if dim_1[1] == dim_2[0]:
  print(f"la matriz será de {dim_1[0], dim_2[1]}")
else:
  print("las matrices no pueden ser multiplicadas")
```

```python
#@title Vamos a construir el código vectorizado, utilizando arreglos numéricos

#creamos una matriz de zeros para modificarla
matriz_multiplicada = np.zeros((dim_1[0], dim_2[1]))
matriz_multiplicada
```

```python
#@title Realizamos las operaciones manualmente para generalizarlo después

matriz_multiplicada[0][0] = matriz_1[0][0] * matriz_2[0][0] + matriz_1[0][1] * matriz_2[1][0]
matriz_multiplicada[0][1] = matriz_1[0][0] * matriz_2[0][1] + matriz_1[0][1] * matriz_2[1][1]
matriz_multiplicada[1][0] = matriz_1[1][0] * matriz_2[0][0] + matriz_1[1][1] * matriz_2[1][0]
matriz_multiplicada[1][1] = matriz_1[1][0] * matriz_2[0][1] + matriz_1[1][1] * matriz_2[1][1]
```

```python
np.matmul(matriz_1, matriz_2)                                                   #verificamos que la multiplicación es correcta
```

```python
from numpy.matrixlib.defmatrix import matrix
#@title Ahora vamos a generalizar el código

def multiplicacion_matrices(A, B):
  elemento = 0
  for i in range(0, dim_1[0]):
    for k in range(0, dim_2[1]):
      a  = A[i]
      print(a)
      b = B[:,k]
      print(b)
      matriz_multiplicada = a * b
      print(matriz_multiplicada)
      elemento = elemento + matriz_multiplicada
      print(elemento)

  return elemento

multiplicacion_matrices(matriz_1, matriz_2)
```
