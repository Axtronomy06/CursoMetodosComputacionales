```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
#@title Primero realizaremos la integral para 4 puntos, es decir, tendremos 3 trapecios
#supongamos cuatro puntos y la funcion x
x1, x2, x3, x4 = 2, 5, 7, 20
y1, y2, y3, y4 = 2, 5, 7, 20
g = lambda x: x

#vamos a calcular el area de cada trapecio y las sumaremos
tra1 = (g(x2) + g(x1)) * ((x2 - x1)/2)
tra2 = (g(x3) + g(x2)) * ((x3 - x2)/2)
tra3 = (g(x4) + g(x3)) * ((x4 - x3)/2)
trat = tra1 + tra2 + tra3
trat
```

```python
x = np.array([x1, x2, x3, x4])
y = np.array([y1, y2, y3, y4])
```

```python
plt.plot(x, y, color = "seagreen", label = "√Årea bajo la curva: " + str(trat))
plt.fill(x, y)
plt.title("$f(x) = x$")
plt.ylabel("f(x)")
plt.xlabel("x")
plt.axhline(0, color = "dimgray", linestyle = "--")
plt.axvline(0, color = "dimgray", linestyle = "--")
plt.legend(loc = "upper center")
plt.grid()
plt.show()
```

```python
#@title Vamos a crear una rutina que nos permita calcular una integral utilizando la regla del trapecio

def integral(xi, xf, f, n):                     #la funcion recibe un intervalo, una funcion y un n√∫mero de trapecios
  dominio = np.linspace(xi, xf, n + 1)
  area_total = 0
  for i in range(n):
    trap = (f(dominio[i + 1]) + f(dominio[i])) * ((dominio[i + 1] - dominio[i])/2)
    area_total = area_total + trap
    #plt.plot(dominio, f(dominio), "-.")
    #plt.fill(dominio, f(dominio))
  return area_total

integral(x1, x4, g, 3)
```

### Como se puede observar, el valor que retorna la funci√≥n concuerda con el valor obtenido calculando el √°rea para los 4 trapecios

```python
#@title Por √∫ltimo verificamos nuestra funcion "integral" para 4 funciones mas, utilizando un intervalo en com√∫n
f = lambda x: np.sin(x)
h = lambda x: np.log(x**2)
t = lambda x: np.exp(-x)
l = lambda x: 3*(x**2) + x - 1
intervalo = np.linspace(-2, 2, 100)
intervalolog = np.linspace(1e-4, 2, 100)
fig, ax = plt.subplots(figsize=(10, 5), nrows=2, ncols=2)

ax[0, 0].plot(intervalo, f(intervalo), label = str(integral(-2, 2, f, 20)))
ax[0, 1].plot(intervalolog, h(intervalolog), label = str(integral(1, 2, h, 20)))
ax[1, 0].plot(intervalo, t(intervalo), label = str(integral(-2, 2, t, 20)))
ax[1, 1].plot(intervalo, l(intervalo), label = str(integral(-2, 2, l, 20)))

ax[0, 0].set_title("sen(x)")
ax[0, 1].set_title("$ln(x^{2})$")
ax[1, 0].set_title("$e^{-x}$")
ax[1, 1].set_title("$3x^2 + x -1$")

ax[0, 0].grid()
ax[0, 1].grid()
ax[1, 0].grid()
ax[1, 1].grid()

ax[0, 0].legend()
ax[0, 1].legend()
ax[1, 0].legend()
ax[1, 1].legend()

ax[0, 0].axhline(0, color = "dimgray", linestyle = "--")
ax[0, 0].axvline(0, color = "dimgray", linestyle = "--")
ax[0, 1].axhline(0, color = "dimgray", linestyle = "--")
ax[0, 1].axvline(0, color = "dimgray", linestyle = "--")
ax[1, 0].axhline(0, color = "dimgray", linestyle = "--")
ax[1, 0].axvline(0, color = "dimgray", linestyle = "--")
ax[1, 1].axhline(0, color = "dimgray", linestyle = "--")
ax[1, 1].axvline(0, color = "dimgray", linestyle = "--")

plt.suptitle("Integrales num√©ricas en el intervalo mostrado", fontsize=13)
plt.tight_layout()
plt.show()
```
