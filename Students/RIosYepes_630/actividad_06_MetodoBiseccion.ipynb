```python
# Estudiante: Alejandro Rios Yepes (CC1036966630)

# Definición de la función para calcular la raíz utilizando el método de la bisección.
def bisection(a, b, f, tolerancia):
    # Comprueba si el signo de la función en los extremos del intervalo es el mismo.
    if f(a) * f(b) > 0:
        raise ValueError("La función no cumple con los requisitos del método de la bisección en el intervalo dado.")

    # Realiza la iteración del método de la bisección hasta que la diferencia entre a y b sea menor que la tolerancia.
    while (b - a) / 2 > tolerancia:
        # Calcula el punto medio del intervalo.
        c = (a + b) / 2

        # Si la función en el punto medio es exactamente igual a cero, se ha encontrado una raíz exacta y se retorna.
        if f(c) == 0:
            return c  # Encontramos una raíz exacta
        # Si el producto de la función en a y c es negativo, significa que hay una raíz en el intervalo [a, c].
        elif f(a) * f(c) < 0:
            b = c
        # De lo contrario, la raíz está en el intervalo [c, b].
        else:
            a = c

    # Cuando la diferencia entre a y b sea menor que la tolerancia, se calcula el punto medio final y se retorna como la raíz aproximada.
    return (a + b) / 2

# Ejemplo de uso con una función:
def f(x):
    return x**3 - 5*x**2 + 10

a = 0
b = 2
tolerancia = 0.0001

# Llama a la función de bisección para encontrar la raíz de la nueva función en el intervalo [a, b] con la tolerancia especificada.
raiz = bisection(a, b, f, tolerancia)

# Imprime la raíz aproximada.
print(f"La raíz aproximada es: {raiz}")
```

```python
# Comparar su resultado empleando la libreria optmize de scipy.
import scipy.optimize as opt

def f(x):
    return x**3 - 5*x**2 + 10

a = 0
b = 2
tolerancia = 0.0001

# Llama a la función de bisección para encontrar la raíz de la nueva función en el intervalo [a, b] con la tolerancia especificada.
raiz_biseccion = bisection(a, b, f, tolerancia)

# Utiliza la función root_scalar de scipy.optimize para encontrar la raíz de la función.
raiz_scipy = opt.root_scalar(f, bracket=[a, b], method='bisect', xtol=tolerancia)

# Imprime ambos resultados.
print(f"Raíz aproximada con bisección: {raiz_biseccion}")
print(f"Raíz aproximada con scipy.optimize: {raiz_scipy.root}")
```
