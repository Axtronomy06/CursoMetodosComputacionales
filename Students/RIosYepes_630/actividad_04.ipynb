*Para los siguientes datos de temperatura media como función del tiempo de los últimos años de Bogotá y Barcelona (Temperature_Bogota_Barcelona) realice lo siguiente:*

**1) Leer el dataset desde la url, con el comando pd.read_csv.**



```python
import pandas as pd
import matplotlib.pylab as plt

url="https://raw.githubusercontent.com/hernansalinas/autogradesMetodosComputacionales/main/Dataset/Temperature_Bogota_Barcelona.csv"
```

```python
df=pd.read_csv(url)
```

```python
df
```

**2) Adicionar una nueva columna AvgTempCelcius con la temperatura en grados celcius. $ C=(F-32)/1.8$**

```python
df["AvgTemperatureCelsius"]=(df.AvgTemperature-32)/1.8
df
```

**3) Realizar un gráfico de AvgTempCelcius como función del año para la ciudad de Bogotá.**

```python
bogota_data = df[df['City'] == 'Bogota']
plt.figure(figsize=(10, 5))
plt.plot(bogota_data['Year'], bogota_data['AvgTemperatureCelsius'], label='Bogotá', marker='o')
plt.title('Temperatura Promedio en Bogotá')
plt.xlabel('Año')
plt.ylabel('Temperatura (°C)')
plt.legend()
plt.grid(True)
plt.show()
```

**4) Realizar un gráfico de AvgTempCelcius como función del año para la ciudad de Bogotá.**

```python
barcelona_data = df[df['City'] == 'Barcelona']
plt.figure(figsize=(10, 5))
plt.plot(barcelona_data['Year'], barcelona_data['AvgTemperatureCelsius'], label='Barcelona', marker='o')
plt.title('Temperatura Promedio en Barcelona')
plt.xlabel('Año')
plt.ylabel('Temperatura (°C)')
plt.legend()
plt.grid(True)
plt.show()
```

**5) Determine la mayor temperatura para Bogotá y Barcelona.**

```python
# Determinar la mayor temperatura para Bogotá y Barcelona
max_temp_bogota = bogota_data['AvgTemperatureCelsius'].max()
max_temp_barcelona = barcelona_data['AvgTemperatureCelsius'].max()

print(f"La mayor temperatura en Bogotá fue {max_temp_bogota:.2f}°C.")
print(f"La mayor temperatura en Barcelona fue {max_temp_barcelona:.2f}°C.")
```
