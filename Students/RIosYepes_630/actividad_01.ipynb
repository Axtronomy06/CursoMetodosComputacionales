```python
def doble_factorial(n):
    # Definición de la función llamada doble_factorial que recibe un parámetro n.
    if n < 0:
        raise ValueError("Error. El doble factorial no está definido para números negativos.")
        # Verifica si n es negativo.
    if n == 0 or n == 1:
        return 1
        # Si n es 0 o 1, devuelve 1. (Casos base del doble factorial).

    resultado = 1
    # Inicializa una variable llamada resultado en 1 para almacenar el producto del doble factorial.
    paso = 2 if n % 2 == 0 else 1
    # Calcula el paso (incremento/decremento) basado en la paridad de n. Si n es par, el paso es 2, si es impar, el paso es 1.
    while n >= paso:
        # Inicia un bucle while que se ejecutará mientras n sea mayor o igual al paso.
        resultado *= n
        # Multiplica el resultado actual por n.
        n -= 2
        # Reduce el valor de n en 2 unidades (para avanzar al siguiente número con la misma paridad).

    return resultado
    # Devuelve el resultado final del doble factorial.

# Ejemplos de llamadas a la función doble_factorial
print(doble_factorial(6))  # Resultado = 15 (6 x 4 x 2)
print(doble_factorial(7))  # Resultado = 105 (7 x 5 x 3 x 1)
print(doble_factorial(8))  # Resultado = 384 (8 x 6 x 4 x 2)
```

```python
def lucas(n):
    # Manejamos los casos base directamente
    if n == 0:
        return 2
        # Si n es 0, devolvemos el primer término de la sucesión: 2.
    elif n == 1:
        return 1
        # Si n es 1, devolvemos el segundo término de la sucesión: 1.

    # Si n es mayor que 1, usamos recursión para calcular los términos de la sucesión
    return lucas(n - 1) + lucas(n - 2)
    # El n-ésimo término de Lucas es la suma del (n-1)-ésimo y (n-2)-ésimo términos.

# Ejemplos de llamadas a la función lucas
print(lucas(9))   # Debe devolver 76
print(lucas(10))  # Debe devolver 123
print(lucas(11))  # Debe devolver 199
```

```python
def binomial(n, k):
    if k < 0 or k > n:
        return 0
    # Si k es negativo o mayor que n, el coeficiente binomial es 0.

    result = 1
    # Inicializamos el resultado en 1.

    k = min(k, n - k)
    # Elegimos el valor más pequeño entre k y n - k, ya que C(n, k) = C(n, n - k).

    for i in range(k):
        result *= (n - i)
        result //= (i + 1)
    # Calculamos el coeficiente binomial utilizando la fórmula C(n, k) = n! / (k! * (n - k)!).

    return result

# Ejemplos de llamadas a la función binomial
print(binomial(6, 4))  # Debe devolver 15
print(binomial(7, 2))  # Debe devolver 21
```
