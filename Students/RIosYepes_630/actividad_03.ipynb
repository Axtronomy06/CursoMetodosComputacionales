```python
# Crear archivo json
import json

datos = {
    "angulo": 45,
    "velocidad_inicial": 20
}

with open('datos.json', 'w') as json_file:
    json.dump(datos, json_file)
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt

# Cargar datos desde el archivo JSON
with open('datos.json') as json_file:
    data = json.load(json_file)

# Extraer datos del archivo JSON
angulo_deg = data['angulo']  # Ángulo en grados
velocidad_inicial = data['velocidad_inicial']  # Rapidez inicial en m/s

# Constantes
g = 9.81  # Aceleración debido a la gravedad (m/s^2)
angulo_rad = np.radians(angulo_deg)  # Convertir el ángulo a radianes

# Calcular componentes de velocidad inicial
v0x = velocidad_inicial * np.cos(angulo_rad)
v0y = velocidad_inicial * np.sin(angulo_rad)

# Calcular tiempo de vuelo
T = 2 * v0y / g

# Crear un array de tiempos
t_values = np.linspace(0, T, num=100)

# Calcular trayectoria
x_values = v0x * t_values
y_values = v0y * t_values - 0.5 * g * t_values**2

# Graficar trayectoria
plt.plot(x_values, y_values)
plt.xlabel('Distancia horizontal (m)')
plt.ylabel('Altura (m)')
plt.title('Trayectoria de un objeto lanzado')
plt.grid(True)
plt.show()
```
