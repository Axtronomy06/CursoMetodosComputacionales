````python
# Problema 3

import seaborn as sns
import matplotlib.pyplot as plt

# Paso 1: Carga el archivo CSV en un DataFrame
df = pd.read_csv("Crimes_-_2019.csv")

# Paso 2: Muestra estadísticas generales del DataFrame
print(df.describe())

# Paso 3: Elimina las columnas "PrimaryType" y "Date"
#df.drop(["Primary Type", "Date"], axis=1, inplace=True)

# Paso 4: Transforma la columna "Date" en tipo Fecha
df["Date"] = pd.to_datetime(df["Date"], format="%m/%d/%Y %I:%M:%S %p", errors="coerce")

# Paso 5: Elimina las filas con valores NaT en la columna "Date"
df = df.dropna(subset=["Date"])

# Verifica si hay al menos una fila en el DataFrame antes de continuar
if not df.empty:
    # Paso 6: Calcula los casos totales para cada tipo de crimen y ordénalos ascendente por valores
    crime_counts = df.groupby("Primary Type").size().reset_index(name="Casos Totales")
    crime_counts = crime_counts.sort_values(by="Casos Totales", ascending=True)

    # Paso 7: Muestra las filas 100 a 120
    print(crime_counts.iloc[100:121])

    # Paso 8: Realiza una visualización utilizando seaborn con un gráfico de tipo barplot
    plt.figure(figsize=(12, 8))
    sns.set_style("whitegrid")
    ax = sns.barplot(x="Casos Totales", y="Primary Type", data=crime_counts.iloc[100:121])
    plt.xlabel("Casos Totales")
    plt.ylabel("Tipo de Crimen")
    plt.title("Casos Totales por Tipo de Crimen (Filas 100-120)")
    plt.show()
else:
    print("El DataFrame está vacío después de eliminar filas con valores NaT.")
````

````python
# Problema 4

# 1) Leer el conjunto de datos y filtrar los datos para Colombia
datos = pd.read_csv('country_vaccinations.csv')  # Leer el archivo CSV en un DataFrame llamado "datos"
datos_colombia = datos[datos['country'] == 'Colombia']  # Filtrar las filas donde el país es "Colombia"

# 2) Crear un archivo .csv con los datos para Colombia
datos_colombia.to_csv('colombia_vacunas.csv', index=False)  # Guardar los datos de Colombia en un archivo CSV

# 3) Realizar una comparación con los países latinoamericanos de la cantidad de vacunados
paises_latinoamericanos = ['Argentina', 'Brasil', 'Chile', 'México', 'Perú']  # Lista de países latinoamericanos
datos_latinoamerica = datos[datos['country'].isin(paises_latinoamericanos)]  # Filtrar los datos para los países de la lista

# 4) Determinar el número de vacunados en el intervalo [1.5E6, 2.0E6] usando máscaras
mascara = (datos['total_vaccinations'] >= 1.5E6) & (datos['total_vaccinations'] <= 2.0E6)  # Crear una máscara de datos en el intervalo especificado
vacunados_en_intervalo = datos[mascara]  # Filtrar los datos usando la máscara
````
