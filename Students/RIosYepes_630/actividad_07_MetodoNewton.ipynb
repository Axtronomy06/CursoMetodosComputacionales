```python
# Definición de la función para calcular una raíz utilizando el método de Newton-Raphson.
def newton(x0, f, df, epsilon, max_iter=100):
    """
    Calcula una raíz de la función f utilizando el método de Newton-Raphson.

    Args:
    x0: Punto inicial de evaluación.
    f: Función para la cual se busca una raíz.
    df: Derivada de la función f.
    epsilon: Tolerancia para la convergencia.
    max_iter: Número máximo de iteraciones (opcional, valor predeterminado es 100).

    Returns:
    x: Valor de la raíz aproximada.
    """
    x = x0
    for i in range(max_iter):
        x_nuevo = x - f(x) / df(x)
        if abs(x_nuevo - x) < epsilon:
            return x_nuevo
        x = x_nuevo
    raise Exception("El método de Newton-Raphson no convergió después de {} iteraciones.".format(max_iter))

# Función f(x) = x^3 - 5x^2 + 10
def f(x):
    return x**3 - 5*x**2 + 10

# Derivada de f(x) = x^3 - 5x^2 + 10
def df(x):
    return 3*x**2 - 10*x

x0 = 2.0  # Punto inicial de evaluación.
epsilon = 0.0001  # Tolerancia para la convergencia.

# Llama a la función newton para encontrar la raíz de la función de ejemplo.
raiz = newton(x0, f, df, epsilon)

# Imprime la raíz aproximada.
print(f"La raíz aproximada es: {raiz}")
```

```python
import scipy.optimize as opt

# Función f(x) = x^3 - 5x^2 + 10
def f(x):
    return x**3 - 5*x**2 + 10

# Derivada de f(x) = x^3 - 5x^2 + 10
def df(x):
    return 3*x**2 - 10*x

x0 = 2.0  # Punto inicial de evaluación.
epsilon = 0.0001  # Tolerancia para la convergencia.

# Llama a la función newton para encontrar la raíz de la función de ejemplo.
raiz_newton = newton(x0, f, df, epsilon)

# Utiliza la función root_scalar de scipy.optimize para encontrar la raíz de la función.
raiz_scipy = opt.root_scalar(f, fprime=df, bracket=[x0 - 1, x0 + 1], xtol=epsilon)

# Imprime ambos resultados.
print(f"Raíz aproximada con Newton-Raphson: {raiz_newton}")
print(f"Raíz aproximada con scipy.optimize: {raiz_scipy.root}")
```
