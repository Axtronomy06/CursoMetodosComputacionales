### Teoria de números


Diseñar un programa en el que entrado  un numero `a`  retorne una variable booleana True or false si cumple que es:

2. [Números defectivo](https://es.wikipedia.org/wiki/Número_defectivo) : la suma de los divisores propios es menor que el número.

  La rutina se deberá llamar números_defectivos



3. [Números abundantes](https://es.wikipedia.org/wiki/Número_abundante): la suma de los divisores es mayor que el número.

  La rutina se deberá llamar números_abundantes



4. [Números semiperfectos](https://es.wikipedia.org/wiki/N%C3%BAmero_semiperfecto) la suma de todos o algunos de los divisores propios es igual al número.

  La rutina se debera llamar numeros_semiperfectos

5. [Números perfectos](https://es.wikipedia.org/wiki/N%C3%BAmero_perfecto) la suma de todos sus divisores propios, excepto el mismo numero, es igual al número.

   La rutina se deberá llamar numeros_perfectos


6. [Números primos](https://es.wikipedia.org/wiki/N%C3%BAmero_primo) el número es divisible unicamente por sí mismo y por 1.
   La rutina se deberá llamar numeros_primos

````python
# Solución
# Estudiante: Alejandro Rios Yepes

def calcular_divisores(n):
    """
    Función auxiliar para calcular la lista de divisores propios de un número.
    """
    divisores = []
    for i in range(1, n):
        if n % i == 0:
            divisores.append(i)
    return divisores

def numeros_defectivos(a):
    """
    Verifica si un número es defectivo.
    """
    suma_divisores = sum(calcular_divisores(a))
    return suma_divisores < a

def numeros_abundantes(a):
    """
    Verifica si un número es abundante.
    """
    suma_divisores = sum(calcular_divisores(a))
    return suma_divisores > a

def numeros_semiperfectos(a):
    """
    Verifica si un número es semiperfecto.
    """
    divisores = calcular_divisores(a)
    n = len(divisores)
    for i in range(1, 2**n):
        suma_subset = sum(divisores[j] for j in range(n) if (i >> j) & 1)
        if suma_subset == a:
            return True
    return False

def numeros_perfectos(a):
    """
    Verifica si un número es perfecto.
    """
    suma_divisores = sum(calcular_divisores(a))
    return suma_divisores == a

def numeros_primos(a):
    """
    Verifica si un número es primo.
    """
    if a <= 1:
        return False
    for i in range(2, int(a**0.5) + 1):
        if a % i == 0:
            return False
    return True

# Pedir al usuario que ingrese un número
numero = int(input("Ingrese un número: "))

# Comprobar y mostrar los resultados
print(f"{numero} es un número defectivo: {numeros_defectivos(numero)}")
print(f"{numero} es un número abundante: {numeros_abundantes(numero)}")
print(f"{numero} es un número semiperfecto: {numeros_semiperfectos(numero)}")
print(f"{numero} es un número perfecto: {numeros_perfectos(numero)}")
print(f"{numero} es un número primo: {numeros_primos(numero)}")
````


7. Diseñar un programa en el que entrado dos números `a`  y `b` retorne una variable booleana **True** o **false**.

  [Números amigos](https://es.wikipedia.org/wiki/N%C3%BAmeros_amigos) `a` y `b` tales que a es la suma de los divisores propios de `b` y viceversa.
    La rutina se debera llamar numeros_amigos




P2. Retornar los 10 primeros pares de numeros perfectos, semiperfectos, amigos



````python
def calcular_divisores(n):
    """
    Función auxiliar para calcular la lista de divisores propios de un número.
    """
    divisores = []
    for i in range(1, n):
        if n % i == 0:
            divisores.append(i)
    return divisores

def numeros_amigos(a, b):
    """
    Verifica si dos números son amigos.
    """
    suma_divisores_a = sum(calcular_divisores(a))
    suma_divisores_b = sum(calcular_divisores(b))
    return suma_divisores_a == b and suma_divisores_b == a

# Pedir al usuario que ingrese dos números
numero_a = int(input("Ingrese el primer número: "))
numero_b = int(input("Ingrese el segundo número: "))

# Comprobar y mostrar el resultado
resultado = numeros_amigos(numero_a, numero_b)
print(f"Los números {numero_a} y {numero_b} son amigos: {resultado}")
````

````python
def quicksort(v):
    """
    Implementación del algoritmo Quicksort para ordenar un arreglo.
    """
    if len(v) <= 1:
        return v

    pivote = v[len(v) // 2]  # Elegir un pivote (en este caso, el elemento en medio)
    izquierda = [x for x in v if x < pivote]  # Elementos menores que el pivote
    medio = [x for x in v if x == pivote]  # Elementos iguales al pivote
    derecha = [x for x in v if x > pivote]  # Elementos mayores que el pivote

    return quicksort(izquierda) + medio + quicksort(derecha)  # Combinar los subarreglos ordenados

# Pedir al usuario que ingrese un arreglo de números separados por espacios
numeros_entrada = input("Ingrese un arreglo de números separados por espacios: ")
numeros = [int(x) for x in numeros_entrada.split()]

# Ordenar el arreglo usando Quicksort
numeros_ordenados = quicksort(numeros)

# Mostrar el arreglo ordenado
print("Arreglo ordenado:", numeros_ordenados)
````
