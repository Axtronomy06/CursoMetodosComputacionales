```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sc
import pandas as pd
```

```python
f=lambda x,a,b:a*np.sin(x)+b*np.tan(x)**2
a=1
b=2
x=np.linspace(-1,1,200000)
y=sc.misc.derivative(f,x,dx=1E-7,args=(a,b))
plt.grid()
plt.axhline(y=0,color="black")
plt.ylim(-1,1)
plt.plot(x,f(x,a,b))
plt.plot(x,y)
```

```python
@np.vectorize
def prima(f,x0):
  delta=0.00001
  der=1/(2*delta)*(f(x0+delta)-f(x0-delta))
  return der
```

```python
g=lambda x:np.sin(x)**2
x=np.linspace(-2,2,100)
plt.plot(x,g(x),label="sin(x)**2")
plt.plot(x,prima(g,x),label="derivada")
plt.ylim(-2,2)
plt.xlim(-2,2)
plt.grid()
plt.axhline(y=0,color="black")
plt.axvline(x=0,color="black")
plt.legend()
```
