# 1) Mapa logistico

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
def mapalogistico(x, r):
    return r * x * (1 - x)
# Funcion del mapa logistico seguido por su formula
```

```python
# esta celda pide los inputs para que la formula funcione perfectamente
x_0 = float(input("Ingrese un numero entre 0 y 1: "))
while x_0<=0 or x_0>=1:
  x_0 = float(input("Ingrese un numero entre 0 y 1: "))
r = float(input("Ingrese un nùmero entre 0 y 4: "))
while r<=0 or r>=4:
  r = float(input("Ingrese un nùmero entre 0 y 4: "))
x_prev = x_0
```

```python
# aca se grafica punto por punto el mapa logistico
plt.figure()
fig, ax = plt.subplots()
for n in range(0, 100):
    x_1 = mapalogistico(x_prev, r)
    plt.plot([n, n + 1], [x_prev, x_1], "g.-")
    x_prev = x_1
ax.grid()
plt.show()
```

# 2) Mapa de henon

```python
# Formula para el mapa de henon, roma como parametros las posicion x y y iniciales en el plano y los parametros a,b
def mapadehenon(x_0,y_0,a,b):
  x_n = 1-a*x_0**2 + y_0
  y_n = b*x_0
  return x_n,y_n
```

```python
# En esta celda se generan todas las listas necesarias para la iteracion
N=100000
N_0=np.arange(0,N+1)
X=np.zeros(N+1)
Y=np.zeros(N+1)
X[0]=0
Y[0]=0
a=1.4
b=0.3
```

```python
# Aca se generan todas las iteraciones, cabe aclarar que dependen de N
for h in range(N):
  x_1 ,y_1 = mapadehenon(X[h],Y[h],a,b)
  X[h+1] = x_1
  Y[h+1] = y_1
```

## Gráfica de x_n en función de N

```python
plt.style.use('dark_background')
plt.plot(N_0,X,'.', color='white', alpha = 0.8, markersize=0.3)
plt.show()
```

## Gráfica de y_n en Función de N

```python
plt.style.use('dark_background')
plt.plot(N_0,Y,'.', color='white', alpha = 0.8, markersize=0.3)
plt.show()
```

## Gráfica de x_n y y_n

```python
plt.style.use('dark_background')
plt.plot(X,Y,'.', color='white', alpha = 0.8, markersize=0.3)
plt.show()
```

## 3) Gradiente Descendente

```python
# En esta celda se define la funcion y su derivada
funcion = lambda x:x**2
derivada = lambda x:2*x
```

```python
# aca las iteraciones y el alpha utilizado
w=random.randint(0,10)
alpha=0.01
iteraciones=15
```

```python
# Se halla el valor del minimo utilizando un ciclo for
f=[]
for i in range(iteraciones):
  winicial=w
  w=winicial-alpha*derivada(winicial)
  f.append(funcion(w))
print("El minimo de la funcion ingresada es: ",f[iteraciones-1])
```

```python
# Se definen varios arreglos para poder graficar las funciones
x=np.arange(-50,50)
yf=np.zeros(100)
yd=np.zeros(100)
```

```python
for h in range(100):
  yf[h]=funcion(x[h])
  yd[h]=derivada(x[h])
```

```python
plt.style.use('classic')
plt.plot(x,yf,"r-")
plt.plot(x,yd,"b-")
plt.legend(["x^2","2x"])
plt.axhline(y=0,color="black")
plt.axvline(x=0,color="black")
plt.show
```

```python
##Para graficar las distintas funciones se tiene que cambiar el parametro alpha
##esta celda grafica la relacion de las aproximaciones que se generan con respecto a la cantidad de iteraciones
iteraciones1=np.arange(0,iteraciones)
plt.plot(iteraciones1,f)
```
