```python
import numpy as np
import math as mt
import matplotlib.pyplot as plt
```

```python
f=lambda x:np.tan(x)-x+1
d=lambda x:(1/(np.cos(x)**2))-1
```

```python
def newton(f,d,x_0,eps):
  while abs(f(x_0))>eps:
    x_1=x_0-(f(x_0)/d(x_0))
    x_0=x_1
  return x_1,f(x_1)
```

```python
print(newton(f,d,-1,1E-4))
```

```python
x=np.linspace(-1,1,100)
```

```python
plt.plot(x,f(x),"r")
plt.axhline(y=0)
plt.grid()
```
