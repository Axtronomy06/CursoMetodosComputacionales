## CÃ¡lculo del campo electrico de un dipolo

```python
import matplotlib.pyplot as plt
import numpy as np

#Esta celda me genera todas las coordenadas x,y

x = np.arange(-10, 10, 1)
y = np.arange(-10, 10, 1)
X,Y = np.meshgrid(x, y)
```

```python
#Esta celca calcula el campo electrico para cada par de coordenadas x,y

k=9*(10**(-9)) #Constante k con unidades (N* (m**2))/C**2
q1=10   #Carga q1 del dipolo
q2=-10  #Carga q2 del dipolo
EX1=k*q1*(1/((X-2)**2+Y**2+10**(-15))**(3/2))*(X-2) #Coordenada x de la carga 1
EX2=k*q2*(1/((X+2)**2+Y**2+10**(-15))**(3/2))*(X+2) #Coordenada x de la carga 2
EXT=EX1+EX2 #Calculo del campo electrico Total en el eje x
Ey1=k*q1*(1/((X-2)**2+Y**2+10**(-15))**(3/2))*Y #Coordenada y de la carga 1
Ey2=k*q2*(1/((X+2)**2+Y**2+10**(-15))**(3/2))*Y #Coordenada y de la carga 1
EYT=Ey1+Ey2 #Calculo del campo electrico Total en el eje y
```

```python
fig, ax = plt.subplots()


plt.plot(2,0,"ro")
plt.plot(-2,0,"ro")
q = ax.quiver(X, Y, EXT,EYT,scale=0.0000009,color="green")
ax.quiverkey(q, X=0.3, Y=1.1, U=10,
             label='Quiver key, length = 10', labelpos='E')

plt.show()
```
