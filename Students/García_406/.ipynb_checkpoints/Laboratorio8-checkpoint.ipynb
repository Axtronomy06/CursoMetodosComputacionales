```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as sc
import pandas as pd
```

```python
##Graficacion de datos
t=[1,1.01,1.02,1.03,1.04]
i=[3.1,3.12,3.14,3.18,3.24]
plt.plot(t,i,"r-")
```

```python
derv=np.diff(i)/np.diff(t)
derv
```

```python
##Funcion que saca el promedio entre dos puntos
def promedio(x):
  prom=np.zeros(len(x)-1)
  for i in range(len(x)-1):
    a=x[i]
    b=x[i+1]
    prom[i]=(a+b)/2
  return prom
```

```python
##Calculos
def fem(t,C,deriv):
  L=0.98 ##herios
  R=0.142 ##ohms
  C=promedio(C)
  e=np.zeros(len(t)-1)
  for i in range(len(t)-1):
    e[i]=L*deriv[i]+R*C[i]
  return e
v=promedio(t)
```

```python
##Grafica t vs fem
plt.plot(v,fem(t,i,derv),"r-")
plt.grid()
```

##Punto 5: Pendulo

```python
m = 20 #kg
d = 1 #m
I=(1/3)*(d**2)*m
theta = 30 # Grados
g = 9.8 #m/s**2
```

```python
j=lambda x,k:1/(1-(k*np.sin(x))**2)
```

```python
def p0(I,m,g,d):
  return 2*np.pi*(I/(m*g*d))**(1/2)
```

```python
@np.vectorize
def pendulo(I,m,g,d,theta):
  k=np.sin(np.deg2rad(theta/2))
  x=np.linspace(0,np.pi/2,1000)
  y=j(x,k)
  integral=sc.trapezoid(y,x)
  P=(2*p0(I,m,g,d)/np.pi)*integral
  return P
```

```python
theta=np.linspace(0.01,90,1000)
```

```python
error=abs(pendulo(I,m,g,d,theta)-p0(I,m,g,d))/pendulo(I,m,g,d,theta)
```

```python
plt.plot(theta,pendulo(I,m,g,d,theta),color="black",label="Periodo de un pendulo")
plt.axhline(y=p0(I,m,g,d),color="red",label="P constante")
plt.title("Periodo vs angulo")
plt.xlabel("Angulo (θ)")
plt.ylabel("Periodo (s)")
plt.grid()
plt.legend()
```

```python
plt.plot(theta,error*100,color="orange",label="Error")
plt.title("Error vs angulo")
plt.ylabel("Error (%)")
plt.xlabel("Angulo (θ)")
plt.legend()
plt.grid()
```

```python
m1=10
d1=10
I1=(1/3)*(d1**2)*m1
m2=20
d2=20
I2=(1/3)*(d2**2)*m2
```

```python
plt.plot(theta,pendulo(I1,m1,g,d,theta),color="black",label="10 m")
plt.plot(theta,pendulo(I2,m2,g,d,theta),color="red",label="20 m")
plt.title("Comparación distintas longitudes")
plt.ylabel("Periodo (s)")
plt.xlabel("Angulo (θ)")
plt.legend()
plt.grid()
```
