```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate as sc
```

```python
def interpolacion(x,y):
  f=lambda x,m,b:m*x+b
  for i in range(len(x)-1):
    m=(y[i+1]-y[i])/(x[i+1]-x[i])
    b=y[i]-(m*x[i])
    x_=np.linspace(x[i],x[i+1],100)
    y_=f(x_,m,b)
    plt.plot(x_,y_)
  return plt.plot(x,y,".",color="red")
```

```python
x=np.arange(1,50)
y=100*np.random.rand(49,1)
```

```python
interpolacion(x,y)
```

```python
sc.interp1d(x,y)
```
