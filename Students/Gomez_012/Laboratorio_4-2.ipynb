## Cancelación sustractiva



**serieSn**. Un ejemplo de la cancelacion substractiva es en la suma de series alternantes. Considere la serie $\mathrm{S}_N$,  escrita  matematicamente de tres formas diferentes:

-
\begin{equation}
\mathrm{S}_N^{(1)} = \sum_{n = 1}^{2N} (-1)^n \frac{n}{n+1}
\end{equation}

- Sumando números pares e impares.

\begin{equation}
\mathrm{S}_N^{(2)} = -\sum_{n = 1}^{N}\frac{2n-1}{2n} + \sum_{n=1}^{N} \frac{2n}{2n+1}
\end{equation}

- Combinando las series de una forma analítica

\begin{equation}
\mathrm{S}_N^{(3)} = \sum_{n = 1}^{N}\frac{1}{2n(2n+1)}
\end{equation}



Las tres sumas $\mathrm{S}_N^{(1)}$, $\mathrm{S}_N^{(2)}$ y $\mathrm{S}_N^{(3)}$ son matemáticamente iguales, pero los resultados  pueden tener resultados numéricos diferentes.


- Escribir un programa que realice el cálculo de $\mathrm{S}_N^{(1)}$, $\mathrm{S}_N^{(2)}$ y $\mathrm{S}_N^{(3)}$

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
## esta función me calcula la función $f(x) = e ^{-x}$ mediante una seríe de taylor para N iteraciones.
def exp(x,N):
  e = 0
  for n in range(N):
    e = e + (-1) ** n * ( ( x ** n ) / math.factorial(n) )
  return e

prueba1 = exp(x=5,N = 1000) # exp(-5)
print(prueba1)
```

```python
## esta función iteraciones me permite calcular la exponencial con una precisión dada por la expresión matematica de arriba.
## la ventaja es que requiere unicamente el valor a evaluar.
def iteraciones(x):
  exp_acu = list()
  eo = 1
  exp_acu.append(eo)
  j = 1
  contador = 0
  while abs( ( x ) ** (j+1) / math.factorial(j+1)) >= abs( (1E-7) * sum(exp_acu)):
    contador+=1
    e1 = (-1) ** j * ( ( x ** j) / math.factorial(j))
    j+=1
    exp_acu.append(e1)
  return contador,sum(exp_acu)

## devuelve la cantidad de iteraciones, y el valor respectivo de la función.
```

```python
## vamos a comparar la función iteraciones con np.exp(x) , para  0<x<1

## esta función me devuelve un array con los errores absolutos, para un array de entrada
## que contiene los valores de x para los cuales vamos a verificar el error.
def errores(valores):
  errores = np.zeros(len(valores))
  for i in range(len(valores)):
    error = abs(iteraciones(valores[i])[1] - np.exp(-1 * valores[i]))
    errores[i] = error
  return errores

valores = np.arange(0,1,0.02)

errores(valores)





#print(errores)

## como se aprecia la diferencia es bastante pequeña, tendiendo a cero.

## grafica de aquellos valores.
plt.figure(figsize = (10,8))
plt.title("Errores absolutos")
plt.xlabel("Valores de x")
plt.ylabel("Error absoluto")
plt.plot(valores,errores(valores),"ro",color = "red",label = r"$|exp_{esperado}-exp_{real}|$")
plt.legend()
```

Cabe señalar que en la anterior grafica se aprecía el error asociado a la precisión definida por la ecuación $\left| \frac{(-x)^{N+1}}{(N+1)!} \right | \le \left| 10^{-7} \sum_{N=0}^{N} \frac{(-x)^{n}}{n!} \right|$. Entonces, para valores pequeños de x, es decir, $x\rightarrow0$ el error es mayor, a medida que estos valores crecen a 1, esté error disminuye progresivamente.

```python
## función iteraciones dada para un valor de x = 10

valores2 = np.arange(10,10.8,0.02)
valores_e = np.zeros(len(valores2))
for i in range(len(valores2)):
  valores_e[i]=iteraciones(valores2[i])[1]

errores(valores2)


#fig, ax = plt.subplots(1,3,figsize = (10,8),sharex = True, sharey = True)

plt.figure(figsize = (6,4))
plt.plot(valores2,valores_e,color = "blue",label = r"$\sum_{i=1}^{N}(-1)^{n}(\frac{x^{n}}{n!})$")
plt.grid(True,linestyle = "dashed")
plt.xlabel("X")
plt.title("Seríe de Taylor")
plt.xlim(10,10.8)
plt.ylim(np.min(valores_e),np.max(valores_e))
plt.legend(fontsize="medium")
plt.show()
plt.close()

plt.figure(figsize = (6,4))
plt.plot(valores2,np.exp(-1*valores2),color= "orange",label="f(x)=exp(-x)")
plt.legend()
plt.title("Valor real de exp(-x)")
plt.grid(True,linestyle = "dashed")
plt.xlim(10,10.8)
plt.ylim(np.min(np.exp(-1*valores2)),np.max(np.exp(-1*valores2)))
plt.xlabel("X")
plt.show()
plt.close()

plt.figure(figsize = (6,4))
plt.plot(valores2,errores(valores2),color="red",label = r"$|\sum_{i=1}^{N}(-1)^{n}(\frac{x^{n}}{n!}) - exp(-x)|$")
plt.legend(fontsize = "small")
plt.title("Error entre aproximación y real")
plt.xlabel("X")
plt.xlim(10,10.8)
plt.ylim(np.min(errores(valores2)),np.max(errores(valores2)))
plt.grid(True,linestyle = "dashed")
plt.show()
plt.close()






```

```python
## función iteraciones pero empleando 1 / exp(x)

def iteraciones2(x):
  exp_acu = list()
  eo = 1
  exp_acu.append(eo)
  j = 1
  contador = 0
  while abs( ( x ) ** (j+1) / math.factorial(j+1)) >= abs( (1E-7) * sum(exp_acu)):
    contador+=1
    e1 =   ( x ** j) / math.factorial(j)
    j+=1
    exp_acu.append(e1)
  return contador,1 / sum(exp_acu)


# comparando valores para x grande y la cancelación substractiva.

def errores2(valores):
  errores = np.zeros(len(valores))
  for i in range(len(valores)):
    error = abs(iteraciones2(valores[i])[1] - np.exp(-1 * valores[i]))
    errores[i] = error
  return errores


def errores3(valores):
  errores = np.zeros(len(valores))
  for i in range(len(valores)):
    error = abs(iteraciones(valores[i])[1] - iteraciones2(valores[i])[1])
    errores[i] = error
  return errores

```

```python
### para realizar aquella comparación vamos a emplear la función iteraciones1, iteraciones2, np.exp(-x) y el error entre ellas.

### rango de valores
x1 = np.arange(10,20,0.4)
valores_e1 = np.zeros(len(x1)) ## esto es el array que contendrá los valores de la función interaciones1
valores_e2 = np.zeros(len(x1)) ## esto es el array que contendrá los valores de la función iteraciones2.
for i in range(len(x1)):
  valores_e1[i] = iteraciones(x1[i])[1]
  valores_e2[i] = iteraciones2(x1[i])[1]


plt.figure(figsize = (6,4))
plt.plot(x1,valores_e1,label = r"$e^{-x} = \sum_{n=1}^{N}(-1)^{n}(\frac{x^{n}}{n!})$")
plt.legend(fontsize= "small")
plt.xlabel("X")
plt.title("Serie de Taylor")
plt.grid(True,linestyle ="dashed")
plt.show()
plt.close()




plt.figure(figsize = (6,4))
plt.plot(x1,valores_e2,label = r"$\frac{1}{e^{x}}=e^{-x} = \frac{1}{\sum_{n=1}^{N}(\frac{x^{n}}{n!})}$",color = "red")
plt.legend(fontsize= "large")
plt.xlabel("X")
plt.xlim(np.min(x1),np.max(x1))
plt.ylim(np.min(valores_e2),np.max(valores_e2))
plt.title("Serie de Taylor")
plt.grid(True,linestyle ="dashed")
plt.show()
plt.close()



plt.figure(figsize = (6,4))
plt.plot(x1,np.exp(-1*x1),label = r"np.exp(-x)",color = "orange")
plt.legend(fontsize= "large")
plt.xlabel("X")
plt.title(r"$f(x) = e^{-x}$")
plt.grid(True,linestyle ="dashed")
plt.xlim(np.min(x1),np.max(x1))
plt.ylim(np.min(np.exp(-1*x1)),np.max(np.exp(-1*x1)))
plt.show()
plt.close()


plt.figure(figsize = (6,4))
plt.plot(x1,errores(x1),label = r"$|\sum_{n=1}^{N}(-1)^{n}(\frac{x^{n}}{n!}) - np.exp(-x)|$",color = "gold")
plt.legend()
plt.xlabel("X")
plt.title("Errores entre seríe de Taylor exp(-x) y real.")
plt.grid(True,linestyle ="dashed")
plt.xlim(np.min(x1),np.max(x1))
plt.ylim(np.min(errores(x1)),np.max(errores(x1)))
plt.show()
plt.close()

plt.figure(figsize= (6,4))
plt.plot(x1,errores2(x1),label = r"$|\frac{1}{\sum_{n=1}^{N}(\frac{x^{n}}{n!})} - np.exp(-x)|$",color = "green")
plt.legend()
plt.xlabel("X")
plt.title("Errores entre seríe de Taylor 1/exp(x) y real.")
plt.ylim(np.min(errores2(x1)),np.max(errores2(x1)))
plt.grid(True,linestyle ="dashed")
plt.xlim(np.min(x1),np.max(x1))
plt.show()
plt.close()

plt.figure(figsize = (6,4))
plt.plot(x1,errores3(x1),label = r"$|\frac{1}{\sum_{n=1}^{N}(\frac{x^{n}}{n!})} - \frac{1}{\sum_{n=1}^{N}(\frac{x^{n}}{n!})}|$",color = "black")
plt.legend()
plt.xlabel("X")
plt.xlim(np.min(x1),np.max(x1))
plt.ylim(np.min(errores3(x1)),np.max(errores3(x1)))
plt.title("Errores entre seríe de Taylor 1/exp(x) y exp(-x).")
plt.grid(True,linestyle ="dashed")
plt.show()
plt.close()

```

En este caso se observa un comportamiento bastante interesante entre las funciónes $f(x) = \frac{1}{e^{x}}$ y $f(x) = e^{-x}$. Si bien matematicamente pueden ser iguales, computacionalmente vemos que el error a atraves de $f(x) = e^{-x}$ incrementa a medida que los valores de la función crece. ## Cancelación substractiva.

```python
## programación de las seríes. S1.

def sn1(N):
  sn1 = 0
  for n in range(1,2*N+1):
    sn1 = sn1 + (-1) ** n * ( n / (n + 1))
  return sn1

## programación de la seríe s2.

def sn2(N):
  sn2 = 0
  for n in range(1,N+1):
    sn2 = sn2 + ( - ( (2*n-1) / (2*n) ) + ((2*n)/(2*n + 1 )) )
  return sn2

## programación de la serie de s3.

def sn3(N):
  sn3 = 0
  for n in range(1,N+1):
    sn3 = sn3 + (1 / (( 2*n ) * (2*n+1)))
  return sn3
```

```python
## en este codigo vamos a analizar la convergencia de las 3 seríes, tomando valores y granficando para los 3 casos.
N = 2000 ## esto sería para realizar 200 iteraciones. Es decir, vamos a graficar las sumas parciales para cada caso.
xn = np.arange(1,N+1,1)
valores_sn1 = np.zeros(N)
valores_sn2 = np.zeros(N)
valores_sn3 = np.zeros(N)

for i in range(N):
  valores_sn1[i] = sn1(i)
  valores_sn2[i] = sn2(i)
  valores_sn3[i] = sn3(i)

plt.figure(figsize = (12,10))
plt.plot(xn,valores_sn1,"ro",color = "red",label=f"Iteraciones: {N}")
plt.xlabel("Iteraciones")
plt.legend()
plt.ylabel(r"Serie $S_{N}^{1}$")
plt.title(r"$S_{N}^{1} = \sum_{n=1}^{2N} (-1)^{n} ( \frac{n }{n+1 })$")
plt.grid(True)

plt.figure(figsize = (12,10))
plt.plot(xn,valores_sn2,"ro",color = "orange",label=f"Iteraciones: {N}")
plt.xlabel("Iteraciones")
plt.ylabel(r"Serie $S_{N}^{2}$")
plt.title(r"$S_{N}^{2} = -\sum_{n=1}^{N} \frac{2n-1}{2n} + \sum_{n=1}^{N}\frac{2n}{2n+1}$")
plt.legend()
plt.grid(True)


plt.figure(figsize = (12,10))
plt.plot(xn,valores_sn2,"ro",color = "black",label=f"Iteraciones: {N}")
plt.xlabel("Iteraciones")
plt.ylabel(r"Serie $S_{N}^{3}$")
plt.title(r"$S_{N}^{1} = \sum_{n=1}^{N} \frac{1}{2n(2n+1)}$")
plt.legend()
plt.grid(True)
```

 En este caso para valores de 200 iteraciones, que comienzan a ser considerables iteraciones se observa como la sucesión converge a un valor de 0.30 aproximadamente. Inclusive, las tres se observan iguales. Para 2000 iteraciones se observa la misma convergencia. Concluyendo así que la serie converge experimentalmente.

```python
## grafica del error.

def error1(N):
  errores = list()
  for n in range(1,int(N)+1,1):
    error = abs(sn1(n) - sn3(n) ) / abs(sn3(n))
    errores.append(error)
  return errores

def error2(N):
  errores = list()
  for n in range(1,int(N)+1,1):
    error = abs(sn2(n) - sn3(n) ) / abs(sn3(n))
    errores.append(error)
  return errores

```
