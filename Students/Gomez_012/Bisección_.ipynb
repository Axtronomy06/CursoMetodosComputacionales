```python
## intervalo [0.5,0.6]
## epsilon = 10^-5

## esta función, calcula la cantidad de interaciones necesarias para una aproximación dada "x"
def cantidad(a,b,x):
  return 1/ np.log(2) * np.log( (b-a)/x )

## esta función permite aproximar la raíz mediante el método de bisección, considerando
## el intervalo, a,b y su error e.
def biseccion(f,a,b,x):
  pilist = list()
  if f(a)*f(b) > 0:
      msg = "En el intervalo se debe encontrar una raíz"
      return msg
  else:

    for i in range(int(cantidad(a,b,x))):
      if f(a) * f(b) < 0:
        pi = (a+b)/2
        #plt.plot(pi,f(pi),"^",color = "orange")
        pilist.append(pi)
        if f(a)*f(pi) <0:
          a = a
          b = pi
        else:
          if f(pi)*f(b) <0:
            a = pi
            b = b
    return pilist[-1]

biseccion(f,a = 0.5,b = 0.6,x = 1E-9)


#scp.bisect(f=f2,a=0,b=2,xtol= 1E-9)

```
