# Deducción de la interpolación.
Sea $H ⊂ ℝ \times ℝ $, donde $H := \left\{ (x1,y1),(x2,y2),....,(xn,yn)   \right\}$ De los cuales deseamos construir un interpolación lineal para aproximar los posibles valores entre dos par de puntos mediante una linea recta. De ese modo, empleando la ecuación punto pendiente tenemos que:

\begin{equation}
  y - y_{0} = m(x - x_{0})
\end{equation}
Donde $(x_{0},y_{0})\in H$, luego $m = \frac{y_{1}-y_{0}}{x_{1}-x_{0}}$, para el siguiente punto $(x1,y1)\in H$. Obteniendo así la ecuación:
\begin{equation}
 y = y_{0} + (\frac{y_{1}-y_{0}}{x_{1}-x_{0}})(x - x_{0})
\end{equation}
Si se repite este procedimiento para cada par de dos puntos consecutivos que pertenencen a $H$, tendriamos que la ecuación de la recta sería:

\begin{equation}
  y = y_{n} + (\frac{y_{n+1}-y_{n}}{x_{n+1}-x_{n}})(x - x_{n})
\end{equation}

```python
## esta es la función generalizada
## x,y son numpy array con el conjunto de datos.
def interpolacion(x,y):
  plt.plot(x,y,"^", color = "black",label= "interpolacion")
  plt.legend()
  plt.xlabel("x")
  plt.ylabel("y")
  for i in range(len(x)):
    if i!=len(x)-1:
      m_i = (y[i+1]-y[i])/(x[i+1]-x[i])
      if x[i+1]>x[i]:
        x_i = np.arange(x[i],x[i+1],0.0001)
        a = x[i]
        b = y[i]
        f = lambda x,m_i,a,b : m_i*(x-a) + b
        plt.plot(x_i,f(x_i,m_i,a,b),color ="blue")
        plt.grid(True)
      elif x[i]>x[i+1]:
        x_i = np.arange(x[i+1],x[i],0.0001)
        a = x[i]
        b = y[i]
        f = lambda x,m_i,a,b : m_i*(x-a) + b
        plt.plot(x_i,f(x_i,m_i,a,b),color ="blue")
        plt.grid(True)
      elif (x[i] == x[i+1]) and (y[i] == y[i+1]):
        plt.plot(x[i],y[i],"^",color = "black")
      elif (x[i] == x[i+1])  and (y[i]<y[i+1]):
        y_i = np.arange(y[i],y[i+1],0.0001)
        x_i_1 = np.full(len(y_i),x[i])
        plt.plot(x_i_1,y_i,color = "blue")
      elif (x[i] == x[i+1]) and (y[i]>y[i+1]):
        y_i_1 = np.arange(y[i+1],y[i],0.0001)
        x_i_2 = np.full(len(y_i_1),x[i])
        plt.plot(x_i_2,y_i_1,color = "blue")


x = np.array([4,6,6,7,8,9,0,1,2,3,4])
y = np.array([4,6,9,1,2,3,4,5,6,7,8])
interpolacion(x,y)
```
