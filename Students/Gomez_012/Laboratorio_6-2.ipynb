# 4. Fixed point methods.

When a new planet is discovered, there are different methods to estimate its physical properties. Many times is only possible to estimate either the planet mass or the planet radius and the other property has to be predicted through computer modelling.

If one has the planet mass, a very rough way to estimate its radius is to assume certain composition (mean density) and a homogeneous distribution (a very bad assumption!). For example, for the planet [Gliese 832c](http://es.wikipedia.org/wiki/Gliese_832_c) with a mass $M= 5.40 M_{\oplus}$, if we assume an earth-like composition, i.e. $\bar \rho_{\oplus} = 5520\ kg/m^3$, we obtain:

$$R_{g832c} = \left( \frac{3 M_{g832c}}{ 4 \pi \bar\rho_{\oplus} } \right)^{1/3} \approx 1.75 R_{\oplus}$$

That would be the planet radius if the composition where exactly equal to earth's.

A more realistic approach is assuming an internal one-layer density profile like:

$$\rho(r) = \rho_0 \exp\left( -\frac{r}{L} \right)$$

where $\rho_0$ is the density at planet centre and $L$ is a characteristic lenght depending on the composition. From numerical models of planet interiors, the estimated parameters for a planet of are $M= 5.40 M_{\oplus}$ are approximately $\rho_0 = 18000\ kg/m^3$ and $L = 6500\ km$.

Integrating over the planet volume, we obtain the total mass as

$$M = 4\pi \int_0^R \rho(r)r^2dr$$

This is a function of the mass in terms of the planet radius.

Solving the equation $M(R) = M_{g832c}$ it would be possible to find a more realistic planet radius. However when using numerical models, it is not possible to approach the solution from the left side as a negative mass makes no sense.

## Metodo de Bisección.
Sea $f(x)$ una función. Entonces, nos interesa resolver la ecuación asociada a
 \begin{equation}
    f(x) = 0
  \end{equation}
Lo cual equivale a la raíz de aquella función. Para ello existe un método conocido como el metodo de Bisección. Para aplicarlo debemos considerar las dos siguientes hipotesis:
*  $f(x)\in C[a,b]$
* ∃ $p \in [a,b]$ tal que $f(p) = 0$

Las cuales indican que la función debe ser continua para el intervalo $[a,b]$, en el cual tenemos la certeza de que existe una única raíz de la función $f$. Estas condiciones pueden ser satisfechas gracias a el T.V.I. El cual, indica que si $f(x) \in C[a,b] $ y $ k \in f(x), x \in [a,b]$, entonces existe un $p$ tal que $f(p) = k$. Por último, la unicidad de aquella raiz viene definida por la monotonia de la función $f$, es decir si $f'(x)>0 $ $∀ x\in [a,b]$ ó $f'(x)<0$ $ ∀ x\in [a,b]$.
Una vez confirmamos que se satisface aquellas dos anteriores condiciones se garantiza la convergencia de la sucesión generada por puntos medios $P_{n+1} = \frac{a_{n}+b_{n}}{2}$ al punto p. Donde $\left\{ a_{n} \right\}_{n=0}^{\infty}$ y $\left\{ b_{n} \right\}_{n=0}^{\infty}$ son sucesiones generadas a partir de los extremos del intervalo en cada caso que se va delimitando a medida que se busque el cero de la función.

# Deducción del método de Newton.
El metodo de Newton es ampliamente usado, y uno de los mas poderosos por su rapida convergencia a la raiz de la función. Este método cuenta con una interpretación geometrica. Y, su deducción parte de las series de Taylor. El teorema para emplear y garantizar la convergencia de esta sucesión es bastante complejo de demostrar, así que solo se tratará la deducción de su formula.

Partiendo del grafico tenemos que $f\in C[a,b]$ donde $∃ p \in [a,b]$ tal que $f(p)=0$, así para un $p_{0}$ cerca de p, existe un $p_{1}$ que será $y = f'(p_{0})(x-p_{0})+f(p_{0}) \cap y = 0$(intercepto entre la recta tangente en el punto $(p_{0},f(p_{0})$ y el eje x). De ese modo $x = p_{1} = p_{0}-\frac{f(p_{0})}{f'(p_{0})}$. Si $f(p_{1})\neq 0$, entonces repetimos el proceso, consiguiendo hasta que $f(p_{n+1})\approx 0$. Generando asi la sucesión de puntos: $P_{n+1} =P_{n} - \frac{f(P_{n})}{f'(P_{n})}$, donde $\lim_{n \rightarrow \infty } p_{n+1} = \lim_{n \rightarrow \infty } p_{n} - \frac{f(p_{n})}{f'(p_{n})} = p $

  



## Teorema
Sea $f$ una función tal que $f(x) \in C^{2}[a,b] $ (es decir, $f$ pertenece al espacio vectorial de las funciones, donde cada función y sus primeras dos derivadas son continuas sobre el intervalo $[a,b]$). Además, $∃p\in[a,b]$ tal que $f(p)=0$ y $f'(p)\neq 0$, entonces existe un $δ>0$ tal que la sucesión $P_{n+1} = P_{n}-\frac{f(P_{n})}{f'(P_{n})}$, $n\geqslant 0$ converge a $p$ para cualquier aproximación inicial $p_{0} \in [p-\delta,p+\delta]$.

## Bosquejo de la demostración.
Sea $f$ una función tal que $f(x) \in C^{2}[a,b] $ y $∃p\in[a,b]$ tal que $f(p)=0$. Ahora, ¿Quien debe ser δ?. Como $\frac{df}{dx}$ es continua para $[a,b]$, en particular $\frac{df}{dx}$ continua para $p$(esto garantiza cercanía para la derivada, al rededor de p) además $f'(p)\neq 0$, entonces existe un intervalo I, tal que $f'(x)\neq 0 ∀ x\in I$. Así, aproximando la función $f$ mediante una serie de Taylor centrada en un $p_{0}\in I$, existe un $β(x)$ entre $x$ y $p_{0}$  tal que:
\begin{equation}
  f(x) \approx f(p_{0}) + f'(p_{0})(x-p_{0}) + \frac{f''(\beta(x))(x-p_{0})^{2}}{2!}
\end{equation}
Evaluando la función en $p$, se tiene que:

\begin{equation}
 f(p)=0 \approx f(p_{0}) + f'(p_{0})(p-p_{0}) + \frac{f''(\beta(x))(p-p_{0})^{2}}{2!}
\end{equation}
Como $p_{0}\in I$ entonces, $p_{0}$ esta lo suficientemente cerca de $p$, consecuentemente $(p-p_{0})^{2}\approx 0$, y como $f''(\beta(x))$ se encuentra acotada por el intervalo $I$, se obtiene:
\begin{equation}
   0 \approx f(p_{0}) + f'(p_{0})(p-p_{0})
\end{equation}
\begin{equation}
   p \approx p_{0}-\frac{f(p_{0})}{f'(p_{0})}  
\end{equation}
Deduciendo de este modo la formula anteriomente presentada.

# Diferencias del método de Newton y la Secante.
Si bien el método de la secante se puede considerar un caso especial del método de Newton, cuentan con una interpretación geometrica completamente distinta. Comencemos por la diferencia dada de la ecuación:
\begin{equation}
p_{n+1} = p_{n} - \frac{f(p_{n})}{f'(p_{n})}
\end{equation}
Por definición podemos reescribir $f'(p_{n})$ del siguiente modo:
\begin{equation}
  f'(p_{n}) = \lim_{x \rightarrow \ p_{n} } \frac{f(x) - f(p_{n})}{x-p_{n}}
\end{equation}
Ahora, para un punto $p_{n-1}$ lo suficientemente cerca de $p_{n}$, podemos realizar la siguiente aproximación:
\begin{equation}
 f'(p_{n}) \approx  \frac{f(p_{n}) - f(p_{n-1})}{p_{n}-p_{n-1}}
\end{equation}
Y, así reescribir el metodo de Newton como:
\begin{equation}
 p_{n+1} = p_{n} - \frac{f(p_{n})(p_{n}-p_{n-1})}{f(p_{n}) - f(p_{n-1})}
\end{equation}
Ahora, geometricamente esto se puede interpretar bajo ciertas condiciones similares al método de Newton, pero hay una diferencia sustancial.
Captura de pantalla 2023-09-17 a la(s) 5.50.14 p.m..png
El metodo de la secante viene dado para funciones $f\in C^{2}[a,b]$, nuevamente.Donde $∃ p\in [a,b]$ tal que $f(p)=0$ y $f'(p) \neq 0$. Sea $p_{0} \in I$ y $p_{1} \in I$ (recordemos que $I$, es el conjunto al rededor de p, tal que $f'(x)\neq 0 ∀ x\in I$). Así, empleando la formula de la recta punto pendiente, podemos definir la recta que pasa por los puntos $(p_{o},f(p_{0})$ y $(p_{1},f(p_{1}))$ del siguiente modo:
\begin{equation}
 y -f(p_{0}) = \frac{ f(p_{1})-f(p_{0}) }{p_{1} - p_{0}}(x-p_{0})
\end{equation}
La interceptión de está recta y $y = 0$, nos otorga la siguiente iteración del metodo de la secante, la cual se define por la ecuación:
\begin{equation}
 p_{2} = p_{0} - \frac{f(p_{0})(p_{1}-p_{0})}{f(p_{1})-f(p_{0})}
\end{equation}
Sí $f(p_{2}) \neq 0 $, tenemos que debemos repetir el procedimiento. Aquella sucesión de puntos de aproximación viene dada por la recursión:
\begin{equation}
 p_{n+1} = p_{n} - \frac{f(p_{n})(p_{n}-p_{n-1})}{f(p_{n}) - f(p_{n-1})}
\end{equation}
Ahora bien, si se compará esta deducción geometrica con la deducción geometrica realizada para el método de Newton, son completamente distintas, solo comparten el hecho de que se interceptan con el eje x. Una ultima diferencia de ambos metodos viene definida por el orden de convergencia. Para el caso de los ceros simples(ceros de multiplicidad 1) el método de Newton tiene un orden de convergencia cuadratico, por el contrario del método de la secante que cuenta con uno de aproximadamente 1.62. Ya en el caso de ceros  con multiplicidad $m\geqslant 2$ para ambos métodos el orden de convergencia cambia.

# Metodo de punto fijo.
Un punto fijo de una función $f$ es un punto donde $f(x) = x$. Es decir, evaluar la función para el punto $x$ produce el mismo punto.
Para relatar un pseudocodigo que genere la sucesión de puntos $\left\{ p_{n} \right\}_{n=0}^{\infty}$ que convergen al punto fijo debemos considerar el teorema de existencia y unidad de punto fijo.  
## T.E.U.P.F.
Sea $f \in C^{1}[a,b]$, sí $f(x)⊂[a,b] ∀ x\in[a,b]$, podemos garantizar la existencia de un punto fijo en el intervalo $[a,b]$. Además, sí $|f'(x)|<k$ para $0<k<1$, podemos garantizar su unicidad.

Este teorema parte del hecho la función $f$ se debe interceptar con la recta $y = x$, entonces si la función $f$ se encuentra delimitada por la región $A : \left\{ (x,y)\in \mathbb{R}^{2}:\left[ a, b \right] \times \left[ a, b \right] \right\}⟺ f(x)⊂[a,b] ∀x\in[a,b]$ y esta es continua sobre dicho intervalo, debe encontrarse en algún momento con la recta $y = x$. Y, el hecho del valor absoluto de la derivada menor que un valor $k$ se requiere para garantizar que no se da el teorema de valor medio en aquel intervalo.

Ahora bien, si se satisfacen las condiciones de esté teorema podemos garantizar la convergencia al punto fijo de la sucesión de puntos definida por la recurrencia:
\begin{equation}
p_{n+1} = f(p_{n})
\end{equation}
Para cualquier $p_{0}\in [a,b]$

# 3. Movimiento de proyectiles en presencia de viento.

Supongamos que un proyectil es lanzado desde el origen con una rapidez $v=vo( \cos \theta , \sin \theta $. Supongamos además que el efecto del viento sobre el proyectil es el de una fuerza proporcional a la velocidad relativa entre el viento y el proyectil. Teniendo presente que si u es la velocidad del viento y $\alpha$ el ángulo que hace dirección de esta velocidad con el eje x, se puede entrontrar las siguientes ecuaciones diferenciales, asociada al estado de movimiento del [proyectil](https://iopscience.iop.org/article/10.1088/0143-0807/36/2/025016/meta):


\begin{equation}
m\frac{dv_x}{dt} = -mb(v_x-u\cos \alpha )
\end{equation}

\begin{equation}
m\frac{dv_y}{dt} = -mg-mb(v_x-u\sin \alpha ),
\end{equation}

donde b es el párametro de rozamiento. La soluciones a las ecuaciones anteriores vienen dadas por:

Posición en x
\begin{equation}
x=(u\cos \alpha) t + \frac{1}{b} (v_0\cos \theta_0 -u \cos\alpha)(1-\mathrm{exp}(-bt))
\end{equation}

Posicion en y
\begin{equation}
y= \frac{1}{b}\left(\frac{g}{b}+v_0 \sin \theta_0 -u\sin\alpha \right)(1-\mathrm{exp}(-bt))-\left( \frac{g}{b}-u\sin\alpha\right)t
\end{equation}

# Radio del planeta $G_{832c}$
Para calcular esté radio empleamos la ecuación $M(R)= M$, donde
$M(R) =  4\pi \int_0^R \rho(r)r^2dr$. Para resolver aquella ecuación reemplazamos la función $\rho(r) = \rho_0 \exp\left( -\frac{r}{L} \right)$ en la integral y la resolvemos para cualquier valor del radio,y resolvemos la ecuación con aquella masa total ya brindada por el problema. La solución de la integral sería :
\begin{equation}
 M(R) =  4\pi \int_0^R \rho(r)r^2dr ⟺ 4\pi \int_0^R \rho_0 \exp\left( -\frac{r}{L} \right)r^2dr
\end{equation}
\begin{equation}
 M(R)= 4πρ_{0}(-2L^{3}e^{(\frac{-R}{L})} - 2L^{2}Re^{(\frac{-R}{L})} - R^{2}Le^{\frac{-R}{L}} + 2L^{3})
\end{equation}
obteniendo así la ecuación
\begin{equation}
4πρ_{0}(-2L^{3}e^{(\frac{-R}{L})} - 2L^{2}Re^{(\frac{-R}{L})} - R^{2}Le^{\frac{-R}{L}} + 2L^{3}) = M
\end{equation}
\begin{equation}
 -2L^{3}e^{(\frac{-R}{L})} - 2L^{2}Re^{(\frac{-R}{L})} - R^{2}Le^{\frac{-R}{L}} + 2L^{3} - \frac{M}{4πρ_{0}} = 0
\end{equation}
Si reemplazamos las valores de las constantes, podemos hallar el radio.

````python
# parametros.
l = 6500 * (1E3) ## (m)
radio_tierra = 6371*(1E3) ## (m)
masa_tierra = 5.972 * (1E24) ## kg
po = 18000 ## kg/m^{3}

## parametros expresados en terminos de tierra.

M_tierra = 5.40
l_radio_tierra = l / radio_tierra
po_ra_ma_tierra = 18000 * (radio_tierra)**3 * ( 1 / masa_tierra)

## función asociada a la ecuación que queremos resolver
## expresando todo en terminos del radio de la tierra y su masa.


def f(R,L,po,M):
  r1 = 2 * L**3 * np.exp(-R/L)
  r2 = 2 * L ** 2 * R * np.exp(-R/L)
  r3 = R**2 * L * np.exp(-R/L)
  r4 = 2*L**3
  r5 = M / (4 * np.pi * po)
  r = -r1-r2-r3+r4-r5
  return r

f_1 = lambda R : f(R,L = l_radio_tierra,po = po_ra_ma_tierra,M = M_tierra)

## graficando la función para buscar el cero.
R = np.arange(0,3,1)
plt.figure(figsize = (8,6))
plt.title("Función de f(x) en terminos de radios de la tierra")
plt.xlabel("Radio de la tierra")
plt.plot(R,f_1(R),label = r"$ -2L^{3}e^{(\frac{-R}{L})} - 2L^{2}Re^{(\frac{-R}{L})} - R^{2}Le^{\frac{-R}{L}} + 2L^{3} - \frac{M}{4πρ_{0}} = 0$")
plt.legend(fontsize = "small")
plt.grid(True)

````
