## Solución analitica de la segunda EDO

\begin{equation}   
m \frac{d^2x}{dt^2}+kx + bv=0
\end{equation}

Es una EDO de segundo orden de coeficientes constantes lineal homogenea, por lo tanto la solución es de la forma
$x(t) = e^{rt}$, así que si se deriva y reemplazamos en la EDO se hallaran las constantes $r_1$ y $r_2$ tal que $C.F.S := \left\{e^{r_1t},e^{r_2t}\right\}$, si y solo si $r_1,r_2 \in ℝ $ y $m_1 \neq m_2$.

Por el contrario, si $r_1,r_2 \in ℝ $ y $r_1=r_2$, el conjunto fundamental de solución sería: $C.F.S = \left\{e^{r_1t},te^{r_1t}\right\}$

Por último, si tenemos que $r_1,r_2 \in ℂ$, y $r_1 = a+\beta i,r_2 = a-\beta i$ entonces el conjunto fundamental de solución sería $C.F.S = \left\{e^{at}cos(\beta t),e^{at}sin(\beta t)\right\}$


Luego $m>0,k>0,b>0$, $v = \frac{dx}{dt}$, y así se obtiene la siguiente EDO:

\begin{equation}
m \frac{d^2x}{dt^2}+b\frac{dx}{dt}+kx =0
\end{equation}

Derivadas:
\begin{equation}
\frac{dx}{dt} = re^{rt}
\end{equation}

\begin{equation}
\frac{dx}{dt} = r^{2}e^{rt}
\end{equation}
Reemplazando en la EDO

\begin{equation}
m(r^{2}e^{rt})+b(re^{rt})+ke^{rt} = 0 ⟺ e^{rt}(mr^{2}+br+k)=0
\end{equation}
Como $e^{rt} \neq 0$ $\forall r \in ℝ$, entonces obtenemos la ecuación caracteristica

\begin{equation}
mr^{2}+br+k=0
\end{equation}
Y, las soluciones vienen dadas por la formula
\begin{equation}
\frac{-b\pm\sqrt{b^{2}-4mk}}{2m}
\end{equation}
Y, dependiendo del valor que tome el discriminante ($b^{2}-4mk$) tendriamos los tres casos mencionados anteriormente del $C.F.S$, que constituyen el sub espacio vectorial de soluciones de la EDO homogenea.

# Metodo de Rugen Kutta 4

Este metodo busca resolver ecuaciones diferenciales de la siguiente forma:

\begin{equation}
\frac{dy}{dt} = f(t,y)
\end{equation}

Lo diferente radica en la evolución en el tiempo de la ecuación diferencial, teniendo cuidado el mapa de puntos de la ecuación diferencial.

\begin{equation}
y_{n+1} = y_{n}+\frac{h}{6}(k_{1}+2k_{2}+2k_{3}+k_{4})
\end{equation}

Donde:

\begin{equation}
k_{1} = f(t_{n},y_{n})
\end{equation}

\begin{equation}
k_{2} = f(t_{n}+\frac{h}{2},y_{n}+\frac{h}{2}k_{1})
\end{equation}

\begin{equation}
k_{3} = f(t_{n}+\frac{h}{2},y_{n}+\frac{h}{2}k_{2})
\end{equation}

\begin{equation}
k_{4} = f(t_{n}+h,y_{n}+hk_{3})
\end{equation}

Significando que el proximo punto en aquel campo de soluciones tenemos que es un promedio de las posibles soluciones

## Cambio de variable

Para resolver está ecuación diferencial introducimos las variables $u_{1}(t)=x(t)$ y $u_{2}(t) = x'(t)$. Si derivamos cada variable tendriamos que:

$u_{1}'(t) = x'(t)$ y
$u_{2}'(t) = x''(t)⟺ u_{2}'(t) =-\lambda x'(t)-w^{2}x(t) $ $⟺ u_{2}'(t)=-\lambda u_{2}(t)-w^{2}u_{1}(t)$. Ahora bien, si se define las funciones $\vec{u(t)} : ℝ ⟹ ℝ^{2}$ y $\vec{F(t)}: ℝ\times ℝ^{2} ⟹ ℝ^{2}$, como:


\begin{equation}
\vec{u(t)} = (u_{1}(t),u_{2}(t))
\end{equation}


\begin{equation}
\vec{F(t,\vec{u(t)})} = (f_{1}(t,\vec{u(t)}),f_{2}(t,\vec{u(t)}))
\end{equation}

donde $f_{1}(t,\vec{u(t)})=u_{2}(t)$ y $f_{2}(t,\vec{u(t)})=-\lambda u_{2}(t)-w^{2}u_{1}(t)$. Podemos reescribir aquella ecuación diferencial de orden superior como $\vec{u'(t)}=\vec{F(t,\vec{u(t)})}$, es decir:



\begin{equation}
\begin{bmatrix}
\frac{\mathrm{d}u_1}{\mathrm{d}t}\\
\frac{\mathrm{d}u_2}{\mathrm{d}t}\\
\end{bmatrix}=
\begin{bmatrix}
\ u_{2}(t)\\
-\lambda u_{2}(t)-w^{2}u_{1}(t)\\
\end{bmatrix}
\end{equation}

Luego, si introducimos el vector $\vec{W_{i}}=\vec{u(t_{i})}$ el método de Ruggen-Kutta 4 para el sistema de ecuaciones diferenciales se puede escribir como:

\begin{equation}
\vec{W_{i+1}} = \vec{W{i}}+\frac{h}{6}(\vec{K_{1}}+2\vec{K_{2}}+2\vec{K_{3}}+\vec{K_{4}})
\end{equation}
donde:
\begin{equation}
\vec{K_{1}} = \vec{F(t_{i},\vec{W_{i}})}

````python
### Generalización del método de RK4

def RK4_(funct,y0,h_,*params):
  x = np.zeros(len(t_))
  x[0] = y0[0]
  vx = np.zeros(len(t_))
  vx[0] = y0[1]
  for i in range(len(t_)-1):
    k1 = funct1(y0,t_[i],*params)
    k2 = funct1(y0+(h_/2)*k1,t_[i]+h_/2,*params)
    k3 = funct1(y0+(h_/2)*k2,t_[i]+h_/2,*params)
    k4 = funct1(y0+h*k3,t_[i]+h_,*params)
    wi = y0+(h_/6)*(k1+2*k2+2*k3+k4)
    y0[0],y0[1] = wi[0],wi[1]
    x[i+1], vx[i+1] = y0[0], y0[1]
  return x,vx
````
