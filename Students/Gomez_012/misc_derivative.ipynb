```python
import sympy as sp
import scipy as sc
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x : np.sin(x) + (np.tan(x)*np.tan(x))
g = lambda x : x
x = np.linspace(0,1,100)
a = 1
b = 1
plt.figure(figsize = (10,8))
plt.plot(x,f(x), linewidth =2, color = "black")
plt.grid(True,linestyle = "dashed")
plt.plot(x,sc.misc.derivative(f,x))
sc.misc.derivative(g,x)
```

```python
a = 1
f2 = lambda x,a : a*x**2 + 5
x2 = np.linspace(1,10,100)
plt.plot(x2,f2(x2,a), label = r"$f(x) = x^{2}$")
plt.plot(x2,sc.misc.derivative(f2,x2,args = (a,)))
```

```python
f3 = lambda x,a,b : a* np.sin(x) + b * (np.tan(x) * np.tan(x))
epsilon = 0.3
a,b = 2,3
dominio = np.linspace(0, ( np.pi / 2 - epsilon),100000)
plt.figure(figsize = (10,8))
plt.plot(dominio,f3(dominio,a,b), label = fr"$f(x)= {a}sin(x)+{b}tan^{{2}}(x)$", linewidth= 2)
plt.grid(True, linestyle = "dashed")
plt.plot(dominio, sc.misc.derivative(f3,dominio,dx = 1E-5,args=(a,b, )), label =fr"f'(x) = -{a}sin(x) + 6tan(x)sec^{{2}}(x)" , color ="red")
plt.legend()
plt.show()
```

```python
def derivada(f,x0,h):
  T1 = (1) / (2 * h)
  diff_sup = x0+h
  diff_inf = x0-h
  derivative = T1 * (f(diff_sup) - f(diff_inf))
  return derivative

y = lambda x : (np.sin(x))**2
yp = lambda x : 2*np.sin(x) * np.cos(x)
x0 = 2
h = 1E-5
print(yp(x0))
print(derivada(y,x0,h))

def derivada2(f,x):
  derivatives = np.zeros(len(x) - 1)
  k = 0
  for i in range(1,len(x)):
      h = x[i] - x[i-1]
      diff_sup = x[i] + h
      diff_inf = x[i] - h
      T1 = (1) / ( 2 * h)
      derivative = T1 * (f(diff_sup) - f(diff_inf) )
      derivatives[k] = derivative
      k+=1
  return derivatives

valores = np.linspace(1,2,100)
valores2 = np.linspace(1,2,99)
plt.figure(figsize = (6,4))
plt.plot(valores2,derivada2(y,valores), label = "Derivada calculada", color = "blue", linewidth = 2.5)
plt.grid(True)
plt.plot(valores,yp(valores),color = "red", linewidth = 2, label = r"f'(x) = 2x")
plt.legend(fontsize = "small")
plt.show()

```
