```python
## Funci칩n para campo electrico.
import numpy as np
import matplotlib.pyplot as plt
# condiciones iniciales
xp = 0
yp = 1
k = 1 # constante electrica

  # xp ser치 la componente en x donde deseo averiguar el campo electrico
  # yp ser치 la componente en y donde deseo averiguar el campo electrico

# carga 1
x1 = -1
y1 = 0
q1 = 2

# carga 2
x2 = 1
y2 = 0
q2 = -2



def campo_electrico(xp,yp,x1,y1,q1):
  # Esta funci칩n sirve para determinar un campo electrico de una carga puntual.
  deltax = xp -x1
  deltay = yp - y1
  deltar = np.sqrt(deltax**2 + deltay**2 )
  ux = deltax / deltar
  uy = deltay / deltar

  Ex = (k*q1/deltar**2) * ux
  Ey = (k*q1/deltar**2) * uy
  E = np.array([Ex,Ey])
  return E

```

```python
from IPython.core.shellapp import CaselessStrEnum
eje_x = np.linspace(-10,10,10)
eje_y = np.linspace(-10,10,10)
X,Y = np.meshgrid(eje_x,eje_y)
for i in range(len(eje_x)):
  for j in range(len(eje_y)):
    campo_cada_punto_q1 = campo_electrico(X[i,j],Y[i,j],x1,y1,q1)
    campo_cada_punto_q2 = campo_electrico(X[i,j],Y[i,j],x2,y2,q2)
    campo_total_cada_punto = campo_cada_punto_q1 + campo_cada_punto_q2
    plt.quiver(X[i,j],Y[i,j],campo_total_cada_punto[0],campo_total_cada_punto[1],angles = "xy",scale_units = "xy")
plt.plot(x1,y1,"ro")
plt.plot(x2,y2,"ro")
plt.title("Campo electrico de un dipolo")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True, linestyle = "dashed")
```
