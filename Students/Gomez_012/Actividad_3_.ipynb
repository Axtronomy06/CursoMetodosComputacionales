```python
# condiciones iniciales
theta = 0.5 # pi*rad
vo = 2 # [m/s]
g = 9.8 # [ m / s**2 ]
xo = 1
yo = 0
import numpy as np
import matplotlib.pyplot as plt

# funciones de t
x = lambda xo,vo,t,theta : xo + vo*np.cos(theta)*t
y = lambda yo,vo,t,theta : yo + vo*np.sin(theta)*t - ( 1 / 2 )* g * ( t ** 2)

#tiempo final y limite para graficar
def tv(vo,theta):
  global tv
  tv = ( 2 * vo * np.sin(theta) / g )
  return tv
tf = tv(vo,theta)
print(tf)
t = np.linspace(0,tf,100)

# valor de x en maximo

x_max = x(xo,vo,tv,theta)
print(x_max)
# valor de y maximo

y_max = y(yo,vo,tv/2,theta)
print(y_max)
#dominio (valores de x)n1")

X = x(xo,vo,t,theta)

# valores de y
Y = y(yo,vo,t,theta)

## grafica de x vs y
plt.figure(figsize = (6,4))
plt.plot(X,Y, color = "red")
plt.title("Movimiento parabolico")
plt.xlabel("Distancia en el eje x")
plt.ylabel("Distancia en el eje y")
plt.show()
```

## Condiciones iniciales guardadas en un documento tipo json
Construir el diccionario con el nombre que necesitamos, y a partir de ese documento dar las condiciones iniciales.

```python

def read_file(namefile):
  with open(namefile,"r") as archivo:
    d = json.load(archivo)
  return d ## funciones que me permite leer documentos tipo json.


w = read_file("condiciones_iniciales.json")
```

```python
w
```

```python

vo1 = w["Velocidad"]
xo1 = w["xo"]
yo1 = w["yo"]
def radianes(x):
  return np.deg2rad(x)
theta1 = radianes(w["theta"])




```

```python
def tv(vo,theta):
  global tv
  tv = ( 2 * vo * np.sin(theta) / g )
  return tv
tf1 = tv(vo1,theta1)
t2 = np.linspace(0,tf1,100)
```

```python
#dominio (valores de x)n1")

X1 = x(xo1,vo1,t2,theta1)

# valores de y
Y1 = y(yo1,vo1,t2,theta1)

## grafica de x vs y
plt.figure(figsize = (6,4))
plt.plot(X1,Y1, color = "red")
plt.title("Movimiento parabolico")
plt.xlabel("Distancia en el eje x")
plt.ylabel("Distancia en el eje y")
plt.show()
```
