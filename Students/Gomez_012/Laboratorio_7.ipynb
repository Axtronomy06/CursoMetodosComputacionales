```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate as sci
import pandas as pd
```

```python
## Datos de la paloma
datosx = np.array([0.9,1.3,1.9,2.1,2.6,3.0,3.9,4.4,4.7,5,6,7,8,9.2,10.5,11.3,11.6,12,12.6,13,13.3])
datosy = np.array([1.3,1.5,1.85,2.1,2.6,2.7,2.4,2.15,2.05,2.1,2.25,2.3,2.25,1.95,1.4,0.9,0.7,0.6,0.5,0.4,0.25])
diccionario = {"x":datosx,"y":datosy}
pd.DataFrame(diccionario)
```

```python
x = np.linspace(min(datosx),max(datosx),200)
poly =  sci.lagrange(datosx,datosy)
plt.figure(figsize = (14,4))
plt.plot(datosx,datosy,"ro")
plt.grid(True)
```

```python

plt.figure(figsize = (14,4))
plt.plot(datosx,datosy,"ro")
plt.plot(x,poly(x),label= "paloma" )
plt.grid(True)
plt.legend()
```

```python
poly2 = sci.CubicSpline(datosx,datosy)
plt.figure(figsize = (14,4))
plt.grid(True)
plt.plot(x,poly2(x))
```

```python
np.diff(datosy)
np.diff(datosx)
dydx = np.diff(datosy)/np.diff(datosx)

datosx2 = np.array([1.3,1.9,2.1,2.6,3.0,3.9,4.4,4.7,5,6,7,8,9.2,10.5,11.3,11.6,12,12.6,13,13.3])
datosy2 = np.array([1.5,1.85,2.1,2.6,2.7,2.4,2.15,2.05,2.1,2.25,2.3,2.25,1.95,1.4,0.9,0.7,0.6,0.5,0.4,0.25])

poly3 = sci.CubicHermiteSpline(datosx2,datosy2,dydx)
```

```python
x2 = np.linspace(min(datosx2),max(datosx2),200)
plt.figure(figsize = (14,4))
plt.plot(datosx,datosy,"ro")
plt.grid(True,linestyle = "dashed")
plt.plot(x2,poly3(x2), color = "black",label = "Interpolate Hermite",linewidth = 5)
plt.plot(x,poly2(x), color = "red", label = "Interpolate Cubic.Splite",linewidth = 3)
plt.plot(x,poly(x),label= "Interpolate Lagrange",color = "blue", linewidth = 2 )
plt.title("Interpolaciones paloma")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
```
