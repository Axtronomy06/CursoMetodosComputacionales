```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as scp

def puntofijo(f,po,epsilon):
  g = lambda x : x - f(x) ## f tiene que ser una función
  puntos_fijos = np.linspace(po-5,po+5,100)
  j = 0
  g1 = list()
  g1.append(po)
  g1.append(g(po))
  contador = 0
  while abs(g1[-1] - g1[-2]) >= epsilon:
    pn1 = g(g1[-1])
    g1.append(pn1)

  return g1[-1]

f = lambda x : (x-1)**2
g = lambda x : x - f(x)
#puntofijo(f,po=1.5,epsilon = 1E-9)

x = np.arange(-2,2,0.02)

plt.plot(x,g(x))
plt.plot(x,x)
print(puntofijo(f,po = 1.5,epsilon = 1E-12))
plt.axvline(puntofijo(f,po = 1.5, epsilon = 1E-12))
scp.fixed_point(g,x0=1.5)

```
