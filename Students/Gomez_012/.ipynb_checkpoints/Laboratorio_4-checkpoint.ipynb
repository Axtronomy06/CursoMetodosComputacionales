## Cancelación sustractiva



**serieSn**. Un ejemplo de la cancelacion substractiva es en la suma de series alternantes. Considere la serie $\mathrm{S}_N$,  escrita  matematicamente de tres formas diferentes:

-
\begin{equation}
\mathrm{S}_N^{(1)} = \sum_{n = 1}^{2N} (-1)^n \frac{n}{n+1}
\end{equation}

- Sumando números pares e impares.

\begin{equation}
\mathrm{S}_N^{(2)} = -\sum_{n = 1}^{N}\frac{2n-1}{2n} + \sum_{n=1}^{N} \frac{2n}{2n+1}
\end{equation}

- Combinando las series de una forma analítica

\begin{equation}
\mathrm{S}_N^{(3)} = \sum_{n = 1}^{N}\frac{1}{2n(2n+1)}
\end{equation}



Las tres sumas $\mathrm{S}_N^{(1)}$, $\mathrm{S}_N^{(2)}$ y $\mathrm{S}_N^{(3)}$ son matemáticamente iguales, pero los resultados  pueden tener resultados numéricos diferentes.


- Escribir un programa que realice el cálculo de $\mathrm{S}_N^{(1)}$, $\mathrm{S}_N^{(2)}$ y $\mathrm{S}_N^{(3)}$
- Estudiar la convergencia de la serie para los tres casos.
- Considere  que $S_N^{(3)}$ es la solución exacta, realizar dos graficos log-log, del error relativo en función
del número de términos ($\mathrm{Log_{10}} E$ VS $\mathrm{Log_{10}} N$), para $S_N^{(1)}$ y $S_N^{(2)}$. Donde,  $E = |(S_N^{(i)} - S_N^{(3)})/S_N^{(3)}|$, i = 1, 2. Considere valores de $N = 1\times 10^{6}$.
- Analizar y discutir los resultados. Tenga en cuenta la discusión realizada sobre teoría de errores en la página

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
## esta función me calcula la función $f(x) = e ^{-x}$ mediante una seríe de taylor para N iteraciones.
def exp(x,N):
  e = 0
  for n in range(N):
    e = e + (-1) ** n * ( ( x ** n ) / math.factorial(n) )
  return e

prueba1 = exp(x=5,N = 1000) # exp(-5)
print(prueba1)
```

```python
## esta función iteraciones me permite calcular la exponencial con una precisión dada por la expresión matematica de arriba.
## la ventaja es que requiere unicamente el valor a evaluar.
def iteraciones(x):
  exp_acu = list()
  eo = 1
  exp_acu.append(eo)
  j = 1
  contador = 0
  while abs( ( x ) ** (j+1) / math.factorial(j+1)) >= abs( (1E-7) * sum(exp_acu)):
    contador+=1
    e1 = (-1) ** j * ( ( x ** j) / math.factorial(j))
    j+=1
    exp_acu.append(e1)
  return contador,sum(exp_acu)

## devuelve la cantidad de iteraciones, y el valor respectivo de la función.
```

```python
## vamos a comparar la función iteraciones con np.exp(x) , para  0<x<1

valores = np.arange(0,1,0.02)
errores = np.zeros(len(valores))

for i in range(len(valores)):
  error = abs(iteraciones(valores[i])[1] - np.exp(-1 * valores[i]) )
  errores[i] = error


#print(errores)

## como se aprecia la diferencia es bastante pequeña, tendiendo a cero.

## grafica de aquellos valores.
plt.figure(figsize = (10,8))
plt.title("Errores absolutos")
plt.xlabel("Valores de x")
plt.ylabel("Error absoluto")
plt.plot(valores,errores,"ro",color = "red",label = r"$|exp_{esperado}-exp_{real}|$")
plt.legend()
```

```python
## función iteraciones dada para un valor de x = 10

valor2 = iteraciones(10)
real = np.exp(-10)
# error absoluto
print(abs( valor2[1] - real))

## una diferencia pequeña.
```

```python
## función iteraciones pero empleando 1 / exp(x)

def iteraciones2(x):
  exp_acu = list()
  eo = 1
  exp_acu.append(eo)
  j = 1
  contador = 0
  while abs( ( x ) ** (j+1) / math.factorial(j+1)) >= abs( (1E-7) * sum(exp_acu)):
    contador+=1
    e1 =   ( x ** j) / math.factorial(j)
    j+=1
    exp_acu.append(e1)
  return contador,1 / sum(exp_acu)

## comparando errores para valores grandes, x = 10.

valor3 = iteraciones2(10)
real2 = np.exp(-10)

## error absoluto
print(abs(valor3[1] - real2))
## comparación de la substranción cancelativa para ambas funciones.
print(abs(valor3[1] - real2) <  abs( valor2[1] - real) )

x2 = np.arange(10,15,1)
errores2 = list()
apreciacion_ambos = list()
for i in range(len(x2)):
  error2 = abs( iteraciones2(x2[i])[1] - np.exp(-1*x2[i])  )
  errores2.append(error2)
  veracidad = abs(iteraciones2(x2[i])[1] - np.exp(-1*x2[i])) <  abs( iteraciones(x2[i])[1] - np.exp(-1*x2[i]))
  apreciacion_ambos.append(veracidad)

print(errores2)
print(apreciacion_ambos)

## los resultados indica que para valores mayores a 10, es mucho mas preciso 1 / exp(x) que exp(-x)

```

```python
## errores para valores de x donde 1<x<10.

x3 = np.arange(1,10,0.5)
errores3 = list()
for i in range(len(x3)):
  error3 = abs( iteraciones2(x3[i])[1] - np.exp(-1*x3[i])  )
  errores3.append(error3)

## errores para valores de x donde 10<x<100

x4 = np.arange(10,100,0.5)
errores4 = list()
for i in range(len(x4)):
  error4 = abs( iteraciones2(x4[i])[1] - np.exp(-1*x4[i])  )
  errores4.append(error4)

# esta función me permite indicar cuando la serie pierde precisión y cuando la serie no converge mas.
# lo cual depende de la apreciación definida para cuando deja de converger.

def convergencia2(x,apreciacion):
  x5 = np.arange(1,x,0.5)
  errorn = list()
  for i in range(len(x5)):
    erroresn = abs ( iteraciones2(x5[i])[1] - np.exp(-1 * x5[i]) )
    errorn.append(erroresn)
    if erroresn > 0.5:
      msg = "La serie no converge."
      return msg
  msg2 = "La serie converge"
  precision = max(errorn)
  if precision > apreciacion:
    msg1 = f"La serie empieza a perder precisión para valores como {x}"
    return msg2,msg1
  else:
    msg3 = "La serie tiene buena precisión"
    return msg2,msg3

# convergencia2(101) para este valor ya hay divergencia.


convergencia2(10,apreciacion = 1E-8)
```

```python
## graficas de los errores vs cantidad de iteraciones.  Para distintos valores de x.

## esta función me produce los errores dado cierta iteración N.
def errores_grafica(x,N):
  errores = list()
  for i in range(N):
    error = abs( exp(x,i) - np.exp(-1*x) )
    errores.append(error)
  return errores

ro = [2,4,6,8,10,12]
color = ["red","blue","orange","green","black","gold"]

def grafica_subplots(ro,N):
  variable = 0
  fig,axes = plt.subplots(2,3, figsize = (10,8),sharex=True,sharey=True) ## esto me crea una figura para un numpy array de la mitad de la cantidad de valores de r
  plt.suptitle("Error vs Iteraciones",fontsize = 16) # el titulo para todos los subplots
  for i in range(len(ro)):## este ciclo for se usa para acceder a cada valor de r y usarlo para cada grafica
    valor = errores_grafica(ro[i],N) # este sería el valor del eje y para cada uno de los distintos r
    x = [x for x in range(0,len(valor),1)]
    if i == 3: ## esto es para que pasemos a la siguiente fila
      variable=+1
    if i >= 3:  ## este condicional es para que podamos organizar el tamaño del ciclo con la posición de cada una de las figuras
      axes[variable,i-3].plot(x,valor,color = color[i],label = fr"$ e^{-ro[i]} = \sum_{{n=0}} (-1)^{{n}}\cdot \frac{{ {ro[i]}^{{n}} }}{{n!}}$") ## la leyenda y que queremos graficar en cada figura
      axes[variable,i-3].set_xlabel("Iteraciones") ## nombre en el eje x
      axes[variable,i-3].set_ylabel(r"Errores") ## nombre en el eje y
      axes[variable,i-3].legend(fontsize = "small")
    else: ## este es para la primera fila.
      axes[variable,i].plot(x,valor,color = color[i],label = fr"$ e^{-ro[i]} = \sum_{{n=0}} (-1)^{{n}}\cdot \frac{{ {ro[i]}^{{n}} }}{{n!}}$")
      axes[variable,i].set_xlabel("Iteraciones")
      axes[variable,i].set_ylabel(r"Errores")
      axes[variable,i].legend(fontsize = "small")

  for ax in axes.flat:
    ax.label_outer()
  #fig.supxlabel("Cantidad de datos")
  #fig.supylabel(r"$X_{n+1}$")

errores_grafica(ro[-1],N = 100)
```

## Cancelación substractiva.

```python
## programación de las seríes. S1.

def sn1(N):
  sn1 = 0
  for n in range(1,2*N+1):
    sn1 = sn1 + (-1) ** n * ( n / (n + 1))
  return sn1

## programación de la seríe s2.

def sn2(N):
  sn2 = 0
  for n in range(1,N+1):
    sn2 = sn2 + ( - ( (2*n-1) / (2*n) ) + ((2*n)/(2*n + 1 )) )
  return sn2

## programación de la serie de s3.

def sn3(N):
  sn3 = 0
  for n in range(1,N+1):
    sn3 = sn3 + (1 / (( 2*n ) * (2*n+1)))
  return sn3
```

```python
## grafica del error.

def error1(N):
  errores = list()
  for n in range(1,int(N)+1,1):
    error = abs(sn1(n) - sn3(n) ) / abs(sn3(n))
    errores.append(error)
  return errores

def error2(N):
  errores = list()
  for n in range(1,int(N)+1,1):
    error = abs(sn2(n) - sn3(n) ) / abs(sn3(n))
    errores.append(error)
  return errores

```

```python

y1 = error1(1000)
x = np.arange(1,len(y1)+1,1)
plt.figure(figsize = (10,8))
plt.loglog(x,y1, label = r"$ \frac{|S_{N}^{1} - S_{N}^{3}|}{|S_{N}^{3}|}$", color = "blue")
plt.loglog(x,y1)
plt.legend()
plt.title("LogE vs LogN")
plt.grid(True, linestyle = "dashed")
plt.xlabel("Iteraciones")
plt.ylabel("Error relativo")

```

```python
y2 = error2(1000)
x = np.arange(1,len(y2)+1,1)
plt.figure(figsize = (10,8))
plt.loglog(x,y2, label = r"$ \frac{|S_{N}^{2} - S_{N}^{3}|}{|S_{N}^{3}|}$", color = "blue")
plt.loglog(x,y2)
plt.legend()
plt.title("LogE vs LogN")
plt.grid(True, linestyle = "dashed")
plt.xlabel("Iteraciones")
plt.ylabel("Error relativo")
```
