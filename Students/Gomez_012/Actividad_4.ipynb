```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Punto 1. Leer la tabla de datos
url = "https://raw.githubusercontent.com/hernansalinas/autogradesMetodosComputacionales/main/Dataset/Temperature_Bogota_Barcelona.csv"
df = pd.read_csv(url)
datos = pd.DataFrame(df)

## Punto 2. Agregar una nueva columna al DataFrame.
datos["AvgTempCelcius"] = ( datos.AvgTemperature -32 ) / 1.8
filtro_Barcelona = datos.City == "Barcelona"
filtro_Bogota =  datos.City == "Bogota"
barcelona = datos[filtro_Barcelona]
bogota = datos[filtro_Bogota] ## esto me crea una nueva tabla para la condición definida.


```

```python
## Punto 3. Crear graficas para "AvgTempCelcius" vs "Bogota"

## Creamos los datos del eje x
eje_xbo = bogota.Year
## datos para el eje y
eje_ybo = bogota.AvgTempCelcius
plt.figure(figsize = (6,4))
plt.plot(eje_xbo,eje_ybo, label = "$C°$")
plt.title("Temperatura promedio (c°) vs año")
plt.xlabel("Años")
plt.ylabel("Temperatura (c°)")
plt.legend()
plt.show()
```

```python
## Punto 3. Crear graficas para "AvgTempCelcius" vs "Bogota"

## Creamos los datos del eje x
eje_xba = barcelona.Year
## datos para el eje y
eje_yba = barcelona.AvgTempCelcius
plt.figure(figsize = (6,4))
plt.plot(eje_xba,eje_yba,label = "$C°$")
plt.title("Temperatura promedio (c°) vs año")
plt.xlabel("Años")
plt.ylabel("Temperatura (c°)")
plt.legend()
plt.show()
```
