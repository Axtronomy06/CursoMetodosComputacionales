```python
l = 'https://raw.githubusercontent.com/hernansalinas/Curso_Metodos_Computacionales/main/Entregas/Parciales/Parcial1/dataset/Temperature_Bogota_Barcelona.csv'
df = pd.read_csv(l)

df = df.drop(['Month', 'Day', 'Unnamed: 0'], axis = 1)

df['AvgTempCelcius'] = (df['AvgTemperature'] - 32) / 1.8

df
```
