```python
# Problema 1:modelar sistemas físicos
# Construir un gráfico que muestre la variación de presión con la profundidad
# Función P=Pa+pgy donde Pa=presión atmosférica, p=densidad del agua, y=profundidad

Pa = 101325 #atm
ρ = 997 #Kg/m**3
g = 9.776 #m/s**2


def presion(y):
  P = (Pa + (ρ*g*y)) / 101325 # Dividimos entre 101325 para que la presion nos quede en unidades de atmosferas
  return P
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,30,100) #(inicio,fin,número de pasos)___----->dominio de la función y step
h = presion(x)

plt.figure(figsize=(6,4))
plt.plot(x,h,".-",label=f"La presion a 30 metros es {presion(30):.2} atm")
plt.xlabel("Profundidad debajo del nivel del mar (m)",fontsize=15)
plt.ylabel("Presion (atm)",fontsize=15)
plt.xticks(fontsize=14)  #tamaño de los valores de los ejes
plt.title("Comportamiento de la presion en funcion de la produndidad")
plt.legend()
plt.grid()
```

```python
def tiro(d, theta, v0):
  tfx = (d) / (v0*np.cos(theta))
  tfy = (v0 * (np.sin(theta)) + np.sqrt (((v0)**2) * ( (np.sin(theta))**2) - 2*g*h)) / g

  if f'{tfx:.3}' == f'{tfy:.3}':
    return 'El jugador encesto'
  else:
    return 'El jugador no encesto'

tiro(10,(3*np.pi)/9,11.68466)
```

```python
v0 = 11.68466
theta = np.pi/3
d = 10
h = 3
tfx = (d) / (v0*np.cos(theta))
tfy = (v0*np.sin(theta) + np.sqrt((v0**2)*(np.sin(theta)**2) + 2*g*h))/g
t = np.linspace(0,tfx,10000)
x = v0*np.cos(theta)*t
y = v0*np.sin(theta)*t-(g * (t**2))/2

plt.plot(x,y, label = "Trayectoria del balon")
plt.ylabel("Altura (m)",fontsize=15)
plt.xlabel("Posicion horizontal (m)",fontsize=15)
plt.xticks(fontsize=14)  #tamaño de los valores de los ejes
plt.title("Trayectoria del balon lanzado")
plt.legend()
plt.grid()
```
