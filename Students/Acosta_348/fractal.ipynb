```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
x,y = 0,0

x_vals = []
y_vals = []

num_puntos = 150000
for _ in range(num_puntos):
  r = random.random()
  if r <= 0.01:
   x_nuevo = 0
   y_nuevo = 0.16*y
  elif 0.01 < r <= 0.86:
   x_nuevo = 0.85*x + 0.04*y
   y_nuevo = -0.04*x + 0.85*y + 1.6
  elif 0.86 < r <= 0.93:
   x_nuevo = 0.2*x - 0.26*y
   y_nuevo = 0.23*x + 0.24*y + 0.44
  elif 0.93 < r <= 1:
   x_nuevo = -0.15*x + 0.28*y
   y_nuevo = 0.26*x + 0.24*y +0.44

  x_vals.append(x_nuevo)
  y_vals.append(y_nuevo)

  x = x_nuevo
  y = y_nuevo

plt.figure(figsize = (6,10))
plt.scatter(x_vals,y_vals, s = 0.1, color = "green")
plt.title("helecho de barnsley")
plt.axis("off")
plt.show()
```
