```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange
```

```python
f = lambda x: 1/(1+25*x**2)
x_ = np.linspace(-1,1,1000)
N = 4
n = N+1
x = np.linspace(-1,1,n) #puntos a interpolar
POL = lagrange(x,f(x)) # funcion de interpolaci√≥n
xl = np.linspace(np.min(x),np.max(x))
```

subplotear para cada uno de los n = 4,8,12,16,20

```python
plt.plot(x_,f(x_),'b-',label = "Funcion original" )
plt.plot(x,f(x),'ro',label =  "Puntos")
plt.plot(xl,POL(xl),'k-',label = "Polinomio de lagrange")
plt.legend()
```

```python
g = lambda z: np.tan(z)
N = 10
n = N+1
z = np.linspace(-np.pi/4,np.pi/4,n)
from scipy.interpolate import CubicSpline
poltan = lagrange(z,g(z)) # funcion de interpolaci√≥n
zl = np.linspace(np.min(z),np.max(z))
z_ = np.linspace(-np.pi/4,np.pi/4,1000)
plt.plot(z_,g(z_),'b-',label = "Funcion original" )
plt.plot(z,g(z),'ro',label =  "Puntos")
plt.plot(zl,poltan(zl),'k-',label = "Polinomio de lagrange")
plt.legend()
```

```python
g = lambda z: np.tan(z)
N = 3
n = N+1
z = np.linspace(-np.pi/4,np.pi/4,n)
from scipy.interpolate import CubicSpline
poltan = lagrange(z,g(z)) # funcion de interpolaci√≥n
zl = np.linspace(np.min(z),np.max(z))
z_ = np.linspace(-np.pi/4,np.pi/4,1000)
plt.plot(z_,g(z_),'b-',label = "Funcion original" )
plt.plot(z,g(z),'ro',label =  "Puntos")
plt.plot(zl,poltan(zl),'k-',label = "Polinomio de lagrange")
plt.legend()
```

```python
zintlineal = np.interp(z_,z,g(z))
plt.plot(z_,g(z_),'b-',label = "Funcion original" )
plt.plot(z,g(z),'ro',label =  "Puntos")
plt.plot(z_,zintlineal,'k-',label = "Int Lineal")
```

```python
import matplotlib.pyplot as plt
import urllib.request
pato='https://3.bp.blogspot.com/-8_LTrCIxVSQ/VxU-5Gq2ffI/AAAAAAAAIto/LhCDEwHKTiAbLr7IhizQwGV13YTwgpbDwCLcB/s400/pato1.png' #URL
urllib.request.urlretrieve(pato, 'pato.jpg')#Decargar la imagen
imagen_pato=plt.imread('pato.jpg') #Variable con la imagen del pato
#fig, ax = plt.subplotsb ()
#plt.figsize()
a = np.array([0.9,1.3,1.9,2.1,2.6,3.0,3.9,4.4,4.7,5.0,6.0,7.0,8.0,9.2,10.5,11.3,11.6,12.0,12.6,13.0,13.3])
Pa = np.array([1.3,1.5,1.85,2.1,2.6,2.7,2.4,2.15,2.05,2.1,2.25,2.3,2.25,1.95,1.4,0.9,0.7,0.6,0.5,0.4,0.25])
plt.plot(a,Pa,'ro', label = "Pato Puntos")
plt.imshow(imagen_pato, extent = (0.4,14,-6.8,3.4),alpha = 0.4)

polpato = lagrange(a,Pa) # funcion de interpolaci√≥n
al = np.linspace(np.min(a),np.max(a),10000)

plt.plot(al,polpato(al),'k-',label = "Polinomio de lagrange")

from scipy.interpolate import CubicSpline

PatoCubico = CubicSpline(a,Pa)

plt.plot(al,PatoCubico(al),'m-',label="Splines cubicos")
plt.legend()
```
