```python
#Condiciones Iniciales
theta_0 = np.pi/12
theta_f = np.pi/2
ta = 0
tb = 17
w_0 = 0
L = 0.1
g = 9.8
d_t = 0.0001

X = np.array([theta_0,w_0])
Num = int((tb-ta)/d_t)

t = np.linspace(ta,tb,Num)


def model(X,t,L,g):
    O,W = X
    dodt = W
    dwdt = ((2*g/L) - (dodt**2)*np.cos(O))/(1/3 + np.sin(O)**2)*np.sin(O)
    return np.array([dodt,dwdt])
```

```python
def model(X,t,L,g):
    O,W = X
    dodt = W
    dwdt = ((2*g/L) - (dodt**2)*np.cos(O))/(1/3 + np.sin(O)**2)*np.sin(O)
    return np.array([dodt,dwdt])
```

```python
ode = odeint(model,X,t,args=(L,g))
theta,w =ode[:,0],ode[:,1]
theta,w
```

```python
thetay = np.array([])
for i in range(len(theta)):
  if theta[i] < theta_f:
    thetay = np.append(thetay, theta[i])
  else:
    thetay = np.append(thetay, theta[i])
    break
```
