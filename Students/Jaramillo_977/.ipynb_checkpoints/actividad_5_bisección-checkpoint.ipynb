```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
```

```python
x=np.linspace(0,1,100)

fx=lambda x: x*np.exp(x)-1

plt.figure()
plt.plot(x,fx(x))

plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.plot(x,0*x)
```

```python
def bisection(a, b, f, e=0):
  if f(a)>f(b):
    a1=a
    a2=b
    a=a2
    b=a1
  err=abs(a-b)
  c=(a+b)/2
  iter=0

  itermax=np.log2(err/1e-16)
  while abs(f(c))>e and itermax:
    c=(a+b)/2
    if f(c)<0:
      a=c
    else:
      b=c
    iter+=1
  return c

f=fx
a=0
b=1
e=1e-10

print("solución aproximada con el método de la bisección: a x*e^x-1", bisection(a, b, f, e), '\n'+
      "solución utilizando optimize.", optimize.bisect(f, a, b, xtol=e))

# la solución de la función bisection y la fución optimize.bisect de scipy con el la bisect dan resultados ligeramente
# distintos debido al parámeto rtol que es una tolerancia relativa a b.
```
