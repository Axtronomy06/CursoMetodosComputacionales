```python
from sympy import *
import matplotlib.pyplot as plt
import numpy as np
```

```python
def pos_x(theta,v_0,t):
  return v_0*np.cos(theta)*t

def pos_y(theta,v_0,t):
  g=9.8
  return v_0*np.sin(theta)*t-1/2*g*t**2

def tiempo_de_vuelo(theta,v_0):
  g=9.8
  return 2*v_0*np.sin(theta)/g
```

```python
theta=3/4*np.pi
v_0=2
t=tiempo_de_vuelo(theta,v_0)
n=300

plt.figure()
for i in range(1,n):
  plt.plot(pos_x(theta,v_0,t/n*i),pos_y(theta,v_0,t/n*i),"r.")
plt.show
```

```python
t=np.linspace(0, t, 100)
x = pos_x(theta, v_0, t)
y = pos_y(theta,  v_0, t)

fig, ax=plt.subplots()
ax.plot(x,y)
```
