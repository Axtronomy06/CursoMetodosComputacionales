```python
import numpy as np
import scipy as sc
import pylab as plt
```

```python
f = lambda x: x**2 + 115
```

```python
def simpson(f,a,b,n):

  h = (b - a) / n
  X = np.linspace(a,b,n)
  Y = f(X)

  S1, S2 = Y[2:n:2], Y[1:n-1:2]

  I = (h/3) * ( Y[0] + (2*S1.sum()) + (4*S2.sum()) + Y[-1] )

  return I
```

```python
simpson(f,0,1,1000000)
```

```python
X = np.linspace(0,1,10)
Y = f(X)

sc.integrate.simpson(Y,X)
```
