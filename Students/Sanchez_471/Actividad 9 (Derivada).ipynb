```python
import numpy as np
import scipy as sc
import pylab as plt
```

```python
a, b = 1, 1
```

```python
f = lambda x: a*np.sin(x) + b*(np.tan(x))**2
g = lambda x: x**2 / 2
```

**Forma vectorial a un conjunto de datos**

```python
def deri(f,X):

  h = (np.max(X) - np.min(X)) / len(X)
  df = (f(X[0:len(X)+1] + h) - f(X[0:len(X)+1] - h)) / (h*2)

  return df
```

```python
X = np.linspace(-10,10,10000)

dgg = deri(g,X)

plt.plot(X,g(X), label = 'g(x)')
plt.plot(X,dgg, label = 'dg_dx')
plt.legend()
plt.xlabel('eje x')
plt.ylabel('eje y')
plt.grid()
plt.show()
```

```python
X = np.linspace(-1,1,1000000)

dgg = deri(f,X)

plt.plot(X,f(X), label = 'f(x)')
plt.plot(X,dgg, label = 'df_dx' )
plt.legend()
plt.xlabel('eje x')
plt.ylabel('eje y')
plt.grid()
plt.show()
```

**Derivada punto a punto con un h independiente al conjunto de datos**

```python
def deriva(f,x,h):

  df = (f(x + h) - f(x - h)) / (h*2)

  return df
```

```python
X = np.linspace(-10,10,10000)

dgg = deriva(g,X,2e-06)

plt.plot(X,g(X), label = 'g(x)')
plt.plot(X,dgg, label = 'dg_dx')
plt.legend()
plt.xlabel('eje x')
plt.ylabel('eje y')
plt.grid()
plt.show()
```

```python
X = np.linspace(-1,1,1000000)

dgg = deriva(f,X,2e-06)

plt.plot(X,f(X), label = 'f(x)')
plt.plot(X,dgg, label = 'df_dx' )
plt.legend()
plt.xlabel('eje x')
plt.ylabel('eje y')
plt.grid()
plt.show()
```

**Usando librerias**

```python
X = np.linspace(-10,10,1000)

dgg = sc.misc.derivative(g,X)

plt.plot(X,g(X), label = 'g(x)')
plt.plot(X,dgg, label = 'dg_dx')
plt.legend()
plt.xlabel('eje x')
plt.ylabel('eje y')
plt.grid()
plt.show()
```
