```python
import numpy as np
import pylab as plt
import scipy as sc
import pandas as pd
```

**PUNTO #1**

**A**

```python
def caida(t,g,k,s0,m):

    T1 = (m * g * t) / k
    T21 = ((m**2) * g) / (k**2)
    T22 = 1 - np.exp( ((-k) * t) / m )

    S = s0 - T1 + (T21 * T22)

    return S

def caida_prima(t,g,k,s0,m):

    T1_prima = (m * g) / k
    T21_prima = ((m**2) * g) / (k**2)
    T22_prima = (np.exp( ((-k) * t) / m ) * k) / m

    dS = ((-1) * T1_prima) + (T21_prima * T22_prima)

    return dS
```

```python
g = 9.8
k = 0.148
s0 = 900
m = 0.25
t = np.linspace(0,60,1000)

plt.plot(t,caida(t,g,k,s0,m), label = "S(t)", color = "skyblue")
plt.xlabel("Segundos")
plt.ylabel("Metros")
plt.title("Altura vs Tiempo")
plt.grid()
plt.legend()
plt.show()
```

**B**

```python
g = 9.8
k = 0.148
s0 = 900
m = 0.25

print("El tiempo de vuelo usando Secante es: "+str(sc.optimize.newton(caida,4, args = (g,k,s0,m)))+"s")
print("El tiempo de vuelo usando Newton es: "+str(sc.optimize.newton(caida,55, fprime = caida_prima, args = (g,k,s0,m)))+"s")
print("El tiempo de vuelo usando Punto Fijo es: "+str(sc.optimize.fixed_point(caida,55, args = (g,k,s0,m)))+"s")
```

**C**

```python
g = 9.8
k = 0.148
s0 = 900
M = np.linspace(0.25,1,10)

for m in M:
  print("El tiempo de vuelo usando Secante para m = "+str(round(m,2))+ "kg es: "+str(sc.optimize.newton(caida,4, args = (g,k,s0,m)))+"s")

```
