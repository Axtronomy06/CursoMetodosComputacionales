```python
import numpy as np
import matplotlib.pylab as plt
import json
```

```python
def x(x0,v0,t,theta):

  xf = x0 + (v0*np.cos(theta))*t

  return xf
```

```python
def y(y0,v0,t,theta):

  yf = y0 + (v0*np.sin(theta)*t) - ((9.8*t**2) / 2)

  return yf
```

```python
def tiempo_vuelo(v0,theta):

  tf = (2*v0*np.sin(theta))*(1/9.8)

  return tf
```

```python
v0 = 6
theta = np.pi/4
tf = tiempo_vuelo(v0,theta)

T = np.linspace(0,tf,500)

xx = np.zeros(len(T))
yy = np.zeros(len(T))

for t in np.arange(0,len(T),1):

    xx[t] = x(0,v0,T[t],theta)
    yy[t] = y(0,v0,T[t],theta)
```

**GRÁFICA TIRO PARABÓLICO**



```python
fig,ax = plt.subplots()
ax.plot(xx,yy, 'b')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()

```

```python
def read_file(nombre_archivo):
  with open(nombre_archivo, 'r') as archivo:
    d=json.load(archivo)
  return d
```

```python
d = read_file("E:\Codes\condicionesinicales.json")
v0 = d ['velocity']
theta = (d['angle'] * np.pi) / 180
tf = tiempo_vuelo(v0,theta)

t = np.linspace(0,tf,500)
xx = x(0,v0,t,theta)
yy = y(0,v0,t,theta)
```
