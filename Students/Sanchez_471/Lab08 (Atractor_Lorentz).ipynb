```python
from scipy.integrate import odeint as od
import numpy as np
import pylab as plt
```

*Condiciones iniciales*

- $\sigma = 10.0$, $\beta = 8./3$ ,$\rho = 28$  con condiciones iniciales,  $x_0,y_0,z_0 = [1.5,1.5,1.5]$

- $\sigma = 10.0$, $\beta = 8./3$ ,$\rho = 25$  con condiciones iniciales,  $x_0,y_0,z_0 = [1.5,1,5,1.5]$


- $\sigma = 10.0$, $\beta = 8./3$ ,$\rho = 25$  con condiciones iniciales,  $x_0,y_0,z_0 = [1,1,1]$

- $\sigma = 10.0$, $\beta = 8./3$ ,$\rho = 25$  con condiciones iniciales,  $x_0,y_0,z_0 = [1001,1.001,1.001]$

```python
r1 = [1.5,1.5,1.5]
sigma1, beta1, ro1 = 10, 8/3, 28

r2 = [1.5,1.5,1.5]
sigma2, beta2, ro2 = 10, 8/3, 25

r3 = [1,1,1]
sigma3, beta3, ro3 = 10, 8/3, 25

r4 = [1.001,1.001,1.001]
sigma4, beta4, ro4 = 10, 8/3, 25

T = np.linspace(0,30,10000)
```

*Sistema de ecuaciones*


$$\frac{\mathrm{d}x}{\mathrm{d}t} = \sigma (y-x)$$

$$\frac{\mathrm{d}y}{\mathrm{d}t} = x(\rho-z)-y$$

$$\frac{\mathrm{d}z}{\mathrm{d}t} = xy-\beta z$$


```python
def dr_dt(r0, t, sigma,beta,ro):

    x, y, z = r0

    dx_dt = sigma * ( y - x )
    dy_dt = ( x * ( ro - z ) ) - y
    dz_dt = ( x * y ) - ( beta *z )

    return np.array((dx_dt,dy_dt,dz_dt))
```

```python
X1 = od(dr_dt, r1, T, args = (sigma1, beta1, ro1))[:,0]
Y1 = od(dr_dt, r1, T, args = (sigma1, beta1, ro1))[:,1]
Z1 = od(dr_dt, r1, T, args = (sigma1, beta1, ro1))[:,2]

X2 = od(dr_dt, r2, T, args = (sigma2, beta2, ro2))[:,0]
Y2 = od(dr_dt, r2, T, args = (sigma2, beta2, ro2))[:,1]
Z2 = od(dr_dt, r2, T, args = (sigma2, beta2, ro2))[:,2]

X3 = od(dr_dt, r3, T, args = (sigma3, beta3, ro3))[:,0]
Y3 = od(dr_dt, r3, T, args = (sigma3, beta3, ro3))[:,1]
Z3 = od(dr_dt, r3, T, args = (sigma3, beta3, ro3))[:,2]

X4 = od(dr_dt, r4, T, args = (sigma4, beta4, ro4))[:,0]
Y4 = od(dr_dt, r4, T, args = (sigma4, beta4, ro4))[:,1]
Z4 = od(dr_dt, r4, T, args = (sigma4, beta4, ro4))[:,2]
```

*Grafico*

```python
fig, ax = plt.subplots(2, 2, figsize = (10,8), subplot_kw = {"projection" : "3d"})

ax[0,0].plot(X1,Y1,Z1, 'b')
ax[0,0].set_title( "PVI #1" )
ax[0,0].set_xlabel( "eje x" )
ax[0,0].set_ylabel( "eje y" )
ax[0,0].set_zlabel( "eje z" )

ax[0,1].plot(X2,Y2,Z2, 'r')
ax[0,1].set_title( "PVI #2" )
ax[0,1].set_xlabel( "eje x" )
ax[0,1].set_ylabel( "eje y" )
ax[0,1].set_zlabel( "eje z" )

ax[1,0].plot(X3,Y3,Z3, 'g')
ax[1,0].set_title( "PVI #3" )
ax[1,0].set_xlabel( "eje x" )
ax[1,0].set_ylabel( "eje y" )
ax[1,0].set_zlabel( "eje z" )

ax[1,1].plot(X4,Y4,Z4, 'purple')
ax[1,1].set_title( "PVI #4" )
ax[1,1].set_xlabel( "eje x" )
ax[1,1].set_ylabel( "eje y" )
ax[1,1].set_zlabel( "eje z" )

fig.suptitle("Atractor de Lorentz")
plt.draw()
```
