```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
```

```python
f = lambda x : (x**2) - (3*x) + 1
df = lambda x : 2*x - 3
```

```python
def newton(x0, f, fp, epsilon):

    raiz = x0
    iteraciones = 0

    while abs(f(raiz)) >= epsilon:

        if iteraciones >= 100000:
            return "no converge"
        
        else:
            a = f(raiz)
            b = fp(raiz)
            raiz = (raiz) - (a/b)
            iteraciones += 1

    return raiz
```

```python
newton(0,f,df,1e-5)
```

```python
x = np.linspace(-10,10,1000)

plt.plot(newton(0,f,df,1e-5), f(newton(0,f,df,1e-5)), "o", label = "raíz")
plt.plot(x,f(x), label = "f(x)")
plt.plot(x, df(x), label = "f'(x)")
plt.title("f(x) = $x^2 - 3x$")
plt.ylabel("f(x)")
plt.xlabel("x")
plt.grid()
plt.legend()
plt.show()
```

```python
def secante(x0, x1 , f, epsilon):
    
    iteraciones = 0

    fp = lambda x0, x1:  x1 - f(x1) * (x1 - x0) / (f(x1) - f(x0))

    while abs(f(x1)) >= epsilon:

        if iteraciones >= 100000:
            return "no converge"
        
        else:
            x0, x1 = x1, fp( x0, x1 )
            iteraciones += 1

    return x1, (f(x1) - f(x0)) / (x1 - x0)
```

```python
secante(0, 1, f, 1e-5 )
```

```python
x = np.linspace(-10,10,1000)
y, mm = secante(0,1,f,1e-5)

m = lambda x: mm*x 

plt.plot(y, f(y), "o", label = "raíz")
plt.plot(x,f(x), label = "f(x)")
plt.plot(x, m(x), label = "f'(x)")
plt.title("f(x) = $x^2 - 3x$")
plt.ylabel("f(x)")
plt.xlabel("x")
plt.grid()
plt.legend()
plt.show()
```
