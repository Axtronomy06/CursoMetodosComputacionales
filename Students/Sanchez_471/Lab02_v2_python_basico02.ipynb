````python
def Lista_alicuota(n):

  if n <= 0:
    print("Insertar numero entero positivo")

  elif n == 1:
    return 1

  elif n > 1:

    Contador = 2
    Sumador = 1 + n
    Listador = [1]

    while Contador < n:

      k = n // Contador

      if k * Contador == n:

          Sumador = Sumador + k
          Contador = Contador + 1
          Listador.append(k)

      elif k * Contador != n:
        Contador = Contador + 1

    return Listador

def Suma_alicuota(n):

  Listador = Lista_alicuota(n)
  Sumador = n

  for i in Listador:
    Sumador += i
  return Sumador
````

````python
%run -i {pathL}/test02.py

def numeros_defectivos(n):

  if n * 2 > Suma_alicuota(n):
    return True

  elif n * 2 <= Suma_alicuota(n):
    return False
````

````python
def quicksort(Lista):

  if len(Lista) <= 1:
    return Lista

  pivote = Lista[len(Lista)//2 ]
  Lista_izquierda = []
  Lista_medio = []
  Lista_derecha = []

  for n in Lista:
    if n < pivote:

      Lista_izquierda.append(n)
    if n > pivote:

      Lista_derecha.append(n)

    if n==pivote:
      Lista_medio.append(n)

  return quicksort(Lista_izquierda) + Lista_medio + quicksort(Lista_derecha)

````
