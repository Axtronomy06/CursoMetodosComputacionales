```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def modulo(Vector):

  return np.sqrt((Vector[0]**2)+(Vector[1]**2))

def normalizador(Vector):

  if modulo(Vector) != 0:
    return [Vector[0]/modulo(Vector),Vector[1]/modulo(Vector)]
  else:
    return [0,0]

def resta_vectores(V1,V2):

  return [V1[0]-V2[0],V1[1]-V2[1]]

def suma_vectores(V1,V2):

  return [V1[0]+V2[0],V1[1]+V2[1]]
```

```python
def campo_dipolo(R1,R2,r,Q1,Q2):

  k = 9E+9

  r1 = resta_vectores(r,R1)
  r2 = resta_vectores(r,R2)

  u1 = normalizador(r1)
  u2 = normalizador(r2)

  if modulo(r1) != 0:
    e1 = (k*Q1)/(modulo(r1)**2)

  else:
    e1 = 0

  if modulo(r2) != 0:
    e2 = (k*Q2)/(modulo(r2)**2)

  else:
    e2 = 0

  E1 = [u1[0]*e1,u1[1]*e1]
  E2 = [u2[0]*e2,u2[1]*e2]

  E = suma_vectores(E1,E2)

  return E
```
