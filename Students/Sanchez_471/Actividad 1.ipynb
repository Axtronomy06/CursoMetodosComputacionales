```python
def factorial(n):

  if type(n) != int:
    print("Insertar numero entero no negativo")

  elif n < 0:
    print("Insertar numero entero no negativo")

  elif n > 0:

    Contador = n
    Multiplicador = n

    while Contador > 1:
      Contador = Contador - 1
      Multiplicador = Multiplicador * Contador
    return Multiplicador

  elif n == 0:
    return 1
```

```python
factorial(8)
```

```python
def doble_factorial(n):

  if type(n) != int:
    print("Insertar numero entero no negativo")

  elif n < 0:
    print("Insertar numero entero no negativo")

  elif n > 0:

    Contador = n
    Multiplicador = n

    while Contador > 1 and Contador != 2:
      Contador = Contador - 2
      Multiplicador = Multiplicador * Contador
    return Multiplicador

  elif n == 0:
    return 1
```

```python
doble_factorial(8)
```

```python
def lucas(n):

  if type(n) != int:
    print("Insertar numero entero positivo")

  elif n <= 0:
    print("Insertar numero entero positivo")

  else:

    Contador = 1
    Listador = [2,1]

    while Contador < (n - 1):

      Listador.append(Listador[Contador - 1] + Listador[Contador])
      Contador = Contador + 1

    return Listador[Contador]

```

```python
lucas(78)
```

```python
def binomial(n,k):

    Coeficiente = factorial(n) / (factorial(k)*factorial(n-k))
    return Coeficiente
```

```python
binomial(45,30)
```
