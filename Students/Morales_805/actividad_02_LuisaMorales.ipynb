1. Escribir un programa en python en el que ingresada la profundidad máxima, cálcule punto a punto la variación de presión con la profundidad. Para ello el usuario deberá definir un valor máximo y en metros y el programa deberá retornar la grafica de Presión como función de y



```python
import numpy as np #Libreria numerica
import matplotlib.pylab as plt #Libreria de graficacion
import json
```

```python
#Aumento de la presión
rho = 997 #kg/m³
g = 9.8 #m/s²
Pa = 101300 #Pa

y = np.linspace(1, 30, 5)


def presion(Pa, rho, g, y):
  return Pa+rho*g*y

F = presion(Pa, rho, g, y)

hecto = (max(F))/100

plt.figure(figsize=(6,4))
plt.title("Aumento de la presion con la profundidad")
plt.plot(F,y,"-", color='mediumpurple', label=f"Profundidad máxima: {max(y)} m\nPresión máxima: {hecto} hPa")
plt.xlabel("Presion (Pa)", fontsize=10)
plt.ylabel("Metros de profundidad (m)", fontsize=10)
plt.legend(shadow=True)
plt.grid(True)
plt.show()
plt.savefig('actividad_02_graf.png')
```

2. Un jugador de baloncesto decide lanzar su balón a una distancia d del aro, con una rapidez v y ángulo theta respecto a la horizontal. Si la cesta tiene una altura de 3m, y el movimiento que realiza la pelota es un movimiento parabólico, construya un programa que determine si el jugador de baloncesto encesto o no encesto.

```python
#Tiro parabólico
theta = 40 #45 grados
v0 = 9.2 #m/s
d = 7.0 #m
y0 = 2
g = 9.8

def mov_parabolico(theta, v0,d):
  t = (d)/(v0*np.cos(np.deg2rad(theta)))
  yf = round(y0 + v0*(np.sin(np.deg2rad(theta)))*t - (1/2)*(9.8)*(t)**2, 1)

  if yf == 3.0:
   print('ENCESTA')
  else:
   print('NO ENCESTA')

  return yf

mov_parabolico(theta, v0,d)
```

```python
t1 = (d)/(v0*np.cos(np.deg2rad(theta)))
t = np.linspace(0,t1,100)
x1 = v0*np.cos(np.deg2rad(theta))*t
y1 = y0 + (v0*np.sin(np.deg2rad(theta))*t)-(1/2)*g*(t**2)

plt.figure()
plt.plot(x1,y1,"o", color='mediumpurple', label="Trayectoria del balón")
plt.xlabel("Distancia al aro (m)", fontsize=10)
plt.ylabel("Altura del aro desde la altura del jugador (m)", fontsize=10)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.legend()
plt.show()
```
