```python
import numpy as np
import sympy as sp
import scipy as sy
```

```python
A = np.matrix([[2,1, -1, 5],
        [4,1, 1, 1],
        [6, 2, 1,4]])

A
```

```python
# funcion que intercambia filas
def intercambio(A,i,j):
  B = A.copy()
  B[i] = A[j]
  B[j] = A[i]
  return B
```

```python
# Funcion de la multiplicacion por esclar
def mul_escalar(A,i,x):
  B = A.copy()
  B[i] = x*A[i]
  return B
```

```python
# Funcion de suma y resta entre filas
def suma_resta(A,i,j,x):
  B = A.copy()
  B[j] = x*A[i] + A[j]
  return B
```

```python
C = suma_resta(A,0,1,-2)
C
```

```python
# i es la fila que le voy a sumar o restar a j
# j es la fila que voy a modificar
# x escalar que me defina si se suma o se resta
C = suma_resta(C,0,2,-3)
C
```
