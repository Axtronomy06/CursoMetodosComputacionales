# MÃ©todo de la Newton

$x_{n+1} = x_n - ({{f(x_n)}/{f'(x_n)}})$

```python
import numpy as np
import scipy as sp
from scipy import optimize
import matplotlib.pyplot as plt
```

```python
f = lambda x:((x**2) - 1) - 10 # F(x)
g = lambda x: 2*x  # Derivada de F(x)
x = np.linspace(-10,20,100)

plt.figure()
plt.plot(x,f(x))
plt.ylim(-12,20)
plt.grid()
plt.show()
```

```python
def newton_raphson(xo , e , f , g , Nmax):
  n = 0

  while (abs(f(xo))>=e) &  (n <= Nmax):
    xo = xo - (f(xo)/g(xo))
    n += 1
  return xo , n

xo = newton_raphson(1 , 1e-12 , f , g , 1000)
print(xo)

#  COMPROBACION DEL CERO CON LA LIBRERIA OPTIMIZE

x = optimize.newton(f , 1 , g)
x
```
