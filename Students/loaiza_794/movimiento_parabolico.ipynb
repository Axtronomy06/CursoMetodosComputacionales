# Movimiento Parabolico
ecuaciones para el movimiento parabolico de una particula:

* *Movimiento en el eje X*



\begin{equation}
x = x_0*v_0*cos( θ )*t
\end{equation}

* *Movimiento en el eje Y*
\begin{equation}
y = y_0*v_0*sin( θ )*t + \frac{1}{2} * g*t^2
\end{equation}

* *Tiempo de vuelo*
\begin{equation}
Tv =  \frac{2 v_o sin(θ)}{g}
\end{equation}




```python
import numpy as np
import matplotlib.pylab as plt
import json
```

```python
# Ec de movimiento
global g
g = 9.8

#se usa open para abrir el archivo modo escritura
def read(namefile):
  with open(namefile , "r") as archivo:
    #se guarda el diccionario
    d = json.load(archivo)
    return d


#posicion en x
def posx(t,  xo, vo, theta):
  return xo+(vo*np.cos(theta))*t

#posicion  en y
def posy(t,  vo, theta):
  return (vo*np.sin(theta))*t-0.5*g*t**2

#Tiempo de vuelo
def tv(vo, theta):
  tvuelo = 2*vo*np.sin(theta)/g
  return tvuelo

#funcion que grafica
def grafica(x,y):
  fig, ax  = plt.subplots()
  ax.plot(x, y)
  ax.set_xlabel("x")
  ax.set_ylabel("y")
```

```python
d = read("condicionesiniciale.json")
vo=d["theta"]
theta = d["theta"]

theta = np.deg2rad(theta)
tvuelo = tv(vo, theta)

t=np.linspace(0, tvuelo, 100)
x = posx(t, 0, vo, theta)
y = posy(t,  vo, theta)

grafica = grafica(x,y)
```
