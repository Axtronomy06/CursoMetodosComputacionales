```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
```

```python
f = lambda x: x*(np.exp(x))-1
x = np.linspace(0,1,100)

plt.figure(figsize=(12,5))
plt.title("Grafíca de la Función")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x,f(x))
plt.show()
```

```python
a = 0.5
b = 0.6

def bisection(a, b, f, e):
  c =a
  iter = 0
  Nmaxiter = 1000

  while ((abs(f(c))>=e) & ((f(a) * f(b))<0) & (iter < Nmaxiter)):
    c = (a+b)*0.5

    if (f(c)) > 0:
        b = c

    else:
      a = c
    iter += 1

    if (iter == Nmaxiter):
      print(f"No alcnaza la precision deseada, Numero maximo de iteraciones {Nmaxiter}")

    return c

c = bisection(a, b, f, 1E-5)

c , f(c)




```

# Metodo por la libreria optimize

```python
a = 0.5
b = 0.6
raiz = optimize.bisect(f, a, b, xtol=1e-5)
raiz
```
