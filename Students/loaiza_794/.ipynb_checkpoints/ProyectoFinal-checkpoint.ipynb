```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

**El analema del sol**

se puede explicar considerando la órbita de la Tierra alrededor del sol y su rotación sobre su propio eje.

La Tierra orbita alrededor del sol en una trayectoria elíptica y al mismo tiempo gira sobre su eje inclinado en relación con su plano orbital. Esta combinación de movimientos crea la apariencia del sol en diferentes posiciones en el cielo a lo largo del año.

La inclinación axial de la Tierra (aproximadamente 23.5 grados) es crucial. Esta inclinación hace que los rayos del sol incidan de manera diferente en la superficie terrestre en diferentes épocas del año.

Cuando se observa la posición aparente del sol al mediodía todos los días durante un año desde un mismo lugar en la Tierra, se traza una figura en forma de ocho llamada analema. Esta figura resulta de la combinación de dos efectos:

**Variación de la altura del sol:**
Debido a la inclinación axial de la Tierra, la altura máxima del sol en el cielo (su cenit) varía a lo largo del año. En los solsticios de verano e invierno, esta altura es máxima o mínima respectivamente.

**Variación en la velocidad orbital de la Tierra:**
La órbita elíptica de la Tierra hace que su velocidad varíe a lo largo del año. Esto influye en la posición aparente del sol a la misma hora cada día.

El analema muestra estas variaciones en la posición del sol en el cielo a lo largo del año, representando los desplazamientos en la altura y la posición horizontal del sol respecto a un observador en la Tierra. Este fenómeno se puede modelar utilizando conceptos de mecánica celeste, considerando la geometría de la órbita terrestre y la inclinación axial.








Para obtener la Altitud y el Acimuth del Sol se udan las siguientes formulas:


*   $$\cos(\gamma_s) = \frac{\cos(\lambda) \sin(\delta) - \cos(\omega) \sin(\lambda) \cos(\delta)}{\cos(\alpha_s)}$$


   Si el ángulo horario $ω<0$ entonces $γs= γc$

   
   Si el ángulo horario $ω>0$ entonces $γs= 360-γc$






*   $$\sin(\alpha_s) = \cos(\lambda) \cdot \cos(\omega) \cdot \cos(\delta) + \sin(\lambda) \cdot \sin(\delta)$$







```python
lambda_ = float(input('Latitud: '))
lambda_ = lambda_ * np.pi / 180
N = int(input('Número de día: '))
# se toma h = 12 , porque es al a hora donde el sol esta en el cenit, que es cuando alcanza la altura maxima en el dia
h = 12

# Declinación
x = 2 * np.pi * (N - 1 - (h - 12) / 24) / 365

def declinacion(N):
    return -23.45 * np.cos((360 / 365) * (N + 10) * np.pi / 180)

Delta = declinacion(N)
delta = np.deg2rad(Delta)


# Ángulo en radianes del orto u ocaso
hr = np.arccos(-np.tan(lambda_) * np.tan(delta))

# Altitud
def alfa_s(w):
    return np.arcsin(np.cos(lambda_) * np.cos(w) * np.cos(delta) + np.sin(lambda_) * np.sin(delta))

# Acimut


def gamma_c(w):
    temp_value = np.cos(lambda_) * np.sin(delta) - np.cos(w) * np.sin(lambda_) * np.cos(delta)
    temp_value /= np.cos(alfa_s(w))

    # Verifica si algún valor absoluto está fuera del rango [-1, 1]
    invalid_values = np.abs(temp_value) > 1

    if np.any(invalid_values):
        # Maneja el error o imprime los valores para depurar
        print("Valores inválidos para el arcoseno/arccoseno:", temp_value[invalid_values])

    # Utiliza np.arccos en el array completo
    return np.arccos(np.clip(temp_value, -1, 1))



plt.figure(figsize=(10,5))


# Desde el orto al mediodía
h = np.arange(-np.floor(hr * 12 / np.pi), 1)
w = h * np.pi / 12
a_s = alfa_s(w) * 180 / np.pi
g_c = []

for wi in w:
    if wi < 0:
        g_s = gamma_c(wi)
    else:
        g_s = 2 * np.pi - gamma_c(wi)
    g_c.append(180 - g_s * 180 / np.pi)

plt.plot(g_c, a_s, '-ro', markersize=4, markerfacecolor='y')

# Desde el mediodía al ocaso
h = np.arange(0, np.floor(hr * 12 / np.pi) + 1)
w = h * np.pi / 12
a_s = alfa_s(w) * 180 / np.pi
g_c = []

for i in w:
    if wi < 0:
        g_s = gamma_c(i)
    else:
        g_s = 2 * np.pi - gamma_c(i)
    g_c.append(180 - g_s * 180 / np.pi)


plt.plot(g_c, a_s, '-ro', markersize=4, markerfacecolor='y')

plt.axis([-150, 150, 0, 90])

duracion_horas = 2 * hr * 12 / np.pi  # Calculando la duración del día en horas
horas = int(duracion_horas)
minutos = int((duracion_horas - horas) * 60)
segundos = int(((duracion_horas - horas) * 60 - minutos) * 60)

# Muestra datos con cuadro
bbox_props = dict(boxstyle="round,pad=0.3", fc="white", ec="b", lw=2)
plt.text(-150, 87, f'duración día: {horas}h {minutos}m {segundos}s', bbox=bbox_props)
plt.text(-150, 80, f'altitud máxima: {90 - (lambda_ - delta) * 180 / np.pi:.3f}°', bbox=bbox_props)
plt.text(-150, 70, f'declinación: {delta * 180 / np.pi:.3f}°', bbox=bbox_props)
plt.xlabel("Azimut")
plt.ylabel("Altitud")
plt.title("posicion del sol a lo largo del dia")
plt.grid()
plt.show()
```

```python
fig = plt.figure()  # Aquí creamos la figura
ax = fig.add_subplot(111, projection='3d')

# Desde el orto al mediodía
h = np.arange(-np.floor(hr * 12 / np.pi), 1)
w = h * np.pi / 12
a_s = alfa_s(w)
g_c = gamma_c(w)
x = np.cos(a_s) * np.sin(g_c)
y = np.cos(a_s) * np.cos(g_c)
z = np.sin(a_s)
ax.plot3D(x, y, z, '-ro', markersize=6, markerfacecolor='y')

# Desde el mediodía al ocaso
h = np.arange(0, np.floor(hr * 12 / np.pi) + 1)
w = h * np.pi / 12
a_s = alfa_s(w)
g_c = 2 * np.pi - gamma_c(w)
x = np.cos(a_s) * np.sin(g_c)
y = np.cos(a_s) * np.cos(g_c)
z = np.sin(a_s)
ax.plot3D(x, y, z, '-ro', markersize=6, markerfacecolor='y')

ax.plot3D([0, 0], [-0.4, 0.6], [0, 0], 'k')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Posición del Sol')
ax.view_init(32, -126)

plt.show()
```

```python
# Ecuación del tiempo en minutos
eq_time = lambda x: 229.18 * (0.000075 + 0.001868 * np.cos(x) - 0.032077 * np.sin(x) - 0.014615 * np.cos(2 * x) - 0.040849 * np.sin(2 * x))

# Convierte minutos a grados
N = np.arange(1, 366)
h = 12
delta = lambda x: 0.006918 - 0.399912 * np.cos(x) + 0.070257 * np.sin(x) - 0.006758 * np.cos(2 * x) + 0.000907 * np.sin(2 * x) - 0.002697 * np.cos(3 * x) + 0.001480 * np.sin(3 * x)
x = 2 * np.pi * (N - 1 - (h - 12) / 24) / 365
alfa = 90 - (lambda_ - delta(x)) * 180 / np.pi
gamma = eq_time(x) * 15 / 60  # 60 minutos, 1 hora son 15 grados

plt.figure(figsize=(10,5))
plt.plot( alfa,gamma, '-ro', markersize=4, markerfacecolor='y')
plt.ylim([0, 90])
plt.xlabel('ángulo')
plt.ylabel('Altitud')
plt.title('Analema')
plt.ylim(-30,30)
plt.grid()
plt.show()
```
