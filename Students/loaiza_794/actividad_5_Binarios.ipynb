### Activity 2.0:

With the binary representation please try to implement the formula to recover the number.

Hint: Use as input the binary representation as a string and invert its order

```
number32(BIN):

```

Where BIN is a string with the binary number

Example:


Test your solution for the next number:

number32("00111110001000000000001000010000")

````python
def number32(BIN):

    decimal = 0
    power = len(BIN) - 1
    i = 0
    while i < len(BIN):
        if BIN[i] == '1':
            decimal += 2 ** power
        power -= 1
        i += 1

    # Convierte el número entero en un número de doble precisión
    sign = -1 if BIN[0] == '1' else 1
    exponent = (decimal >> 52) - 1023
    fraction = (decimal & ((1 << 52) - 1)) | (1 << 52)

    double = sign * fraction * (2 ** exponent)

    return double

# Ejemplo de uso
binario = "0100000000001001001000011111101101010100010001000010110100011"
numero_doble = number32(binario)
print("Número de doble precisión:", numero_doble)
````

### Activity 3.0:
1. Write a python script that calculates the double precision number represented by a 64-bits binary.

the name of function should be:
```
number64(BIN):

```

2. What is the number represented by:

BIN="0 10000000011 1011100100001111111111111111111111111111111111111111"

````python
def number64(binario):
    # Verificar si la cadena binaria tiene la longitud correcta
    if len(binario) != 64:
        print("el numero debe ser de 64 bits")
    else:
        exponente = 0  # Inicializar la variable exponente

        # Obtener el signo (el primer bit)
        for bit in binario[:1]:
            if bit == '1':
                signo = -1
            break

        # Obtener el exponente
        i = 11
        while i >= 1:
            bit = binario[i]
            if bit == '1':
                exponente += 2**(11 - i)
            i -= 1

        fraccion = binario[12:]

        # Calcular el valor del número en formato de doble precisión
        valor_fraccion = 0

        for i, bit in enumerate(fraccion):
            if bit == '1':
                valor_fraccion += 2**(-i)

        valor = signo * (1 + valor_fraccion) * 2**(exponente - 1023)

    return valor


binario = "1100000000111011100100001111111111111111111111111111111111111111"
n = number64(binario)
print("Número de doble precisión:", n)
````
