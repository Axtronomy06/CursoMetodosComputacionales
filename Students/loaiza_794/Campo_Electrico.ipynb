# Campo electrico generado por un dipolo

\begin{equation}
E = \frac{1}{4\piΕ_0} * \frac{1}{|Δ r|^2} \hat{u_r}  = E = K_e \frac{1}{|Δ r|^2} \hat{u_r}
\end{equation}

$Δ \vec{r} = (Δx , Δy)$

$|Δ \vec{r}| = \sqrt{(Δx^2 + Δy^2)}$

# Vcetores unitarios

$\hat{u_r}_x = \frac{Δx}{|Δ \vec{r}|}$

$\hat{u_r}_y = \frac{Δy}{|Δ \vec{r}|}$

#Componentes del campo

\begin{equation}
E_x = K_e \frac{Q}{Δ r^2} \hat{u_r}_x
\end{equation}

\begin{equation}
E_y = K_e \frac{Q}{Δ r^2} \hat{u_r}_y
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt

# Campo Electrico en un punto


def campo_electrico(xp, yp, x1, y1, q1):
  k = 1
  DeltaX = xp -x1
  DeltaY = yp -y1
  Deltar = np.sqrt(DeltaX**2+DeltaY**2)
  ux = DeltaX/Deltar
  uy = DeltaY/Deltar

  Ex = (k*q1/Deltar**2)*ux
  Ey = (k*q1/Deltar**2)*uy
  E=np.array([Ex, Ey])
  return E

```

```python
# Parámetros de los dipolos eléctricos
x1, y1 = -1, 0  # Coordenadas del dipolo 1
q1 = 0.5  # Carga del dipolo 1

x2, y2 = 1, 0 # Coordenadas del dipolo 2
q2 = -0.5  # Carga del dipolo 2

x = np.linspace(-1, 1, 10)
y = np.linspace(0.1, 2, 10)
plt.figure()

for xp1 in x:
  for yp1 in y:

    E1 = campo_electrico(xp1,yp1,x1,y1,q1)
    E2 = campo_electrico(xp1,yp1,x2,y2,q2)
    E = E1 + E2
    plt.plot(x1, y1, "ro", label="Dipolo 1")
    plt.plot(x2, y2, "bo", label="Dipolo 2")
    plt.quiver(xp1, yp1, E[0], E[1], color='green')
    plt.xlabel("x (m)")
    plt.ylabel("y (m)")




#plt.ylim(-0.1, 1.5)
plt.show
```
