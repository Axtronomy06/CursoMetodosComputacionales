```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
import pandas as pd
import scipy as sc
```

```python
x = np.array([0.9, 1.3, 1.9, 2.1, 2.6, 3.0, 3.9, 4.4, 4.7, 5.0, 6.0, 7.0, 8.0, 9.2, 10.5, 11.3, 11.6, 12.0, 12.6, 13.0, 13.3])
y = np.array([1.3, 1.5, 1.85, 2.1, 2.6, 2.7, 2.4, 2.15, 2.05, 2.1, 2.25, 2.3, 2.25, 1.95, 1.4, 0.9, 0.7, 0.6, 0.5, 0.4, 0.25])

datos = {"X":x,"Y":y}
datosdf = pd.DataFrame(datos)
datosdf
```

```python
x[-1]
```

```python
# Poliomio de Lagrange

inter_lag = sc.interpolate.lagrange(x,y)

intervalo = np.linspace(min(x),max(x))
plt.figure(figsize=(10,5))
plt.title("Polinomio de Lagrange")
plt.plot(intervalo,inter_lag(intervalo))
plt.grid()
plt.show()
```

```python
# Interpolación lagrange

inter_lag = sc.interpolate.lagrange(x,y)

plt.figure(figsize=(10,5))
plt.title("Pato segun Lagrange")
plt.scatter(datosdf.X,datosdf.Y,color="black")
plt.plot(x,inter_lag(x),color="orange")
plt.grid()
plt.show()
```

```python
# Poliniomo Cubico

inter_cubic = sc.interpolate.CubicSpline(x,y)

intervalo = np.linspace(min(x),max(x))
plt.figure(figsize=(10,5))
plt.title("Polinomio cubico")
plt.plot(intervalo,inter_cubic(intervalo))
plt.grid()
plt.show()
```

```python
# Interpolación Cubica

inter_cubic = sc.interpolate.CubicSpline(x,y)

plt.figure(figsize=(13,4))
plt.title("Pato Cubico")
plt.scatter(datosdf.X,datosdf.Y,color="black")
plt.plot(x,inter_cubic(x),color="purple")
plt.grid()
plt.show()
```

```python
# Comparación

plt.figure(figsize=(10,5))
plt.title("Comparación")
plt.scatter(datosdf.X,datosdf.Y,label="Puntos Originales",color="black")
plt.plot(x,inter_cubic(x), label="Pato Cubico",color="purple")
plt.plot(x,inter_lag(x), label="Pato de Lagrange",color="orange")
plt.legend()
plt.grid()
plt.show()
```

```python
# Metodo cubico con derivada

m = np.diff(x)/np.diff(y)

inter_cubic_her = sc.interpolate.CubicHermiteSpline(x[:-1], y[:-1], m)

plt.figure(figsize=(12,5))
plt.plot(x[:-1],inter_cubic_her(x[:-1]))
plt.scatter(x,y)
plt.grid()
plt.show()
```
