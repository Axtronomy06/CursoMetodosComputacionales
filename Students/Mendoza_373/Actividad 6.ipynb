```python
def aproximación_raices(funcion,a,b): # donde funcion(a) < 0 y funcion(b) > 0 (convensión) (Problemas con otra funcion distinta a x*e^(x))?)

  if funcion(a)*funcion(b) < 0:
    if (funcion(a)>0 and funcion(b)>0) or (funcion(a)<0 and funcion(b)<0):
      return "No hay raices en el intervalo (a,b)"
    else:

     i = 0
     while abs(funcion(a)-funcion(b)) > 1e-16 and i < 10000:
       c = (a+b)/2
       i += 1
       if (funcion(c)<0):
        a = c
       elif (funcion(c)>0):
        b = c
     return c

aproximación_raices(f,-40,40)
```
