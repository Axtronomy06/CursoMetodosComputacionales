```python

import numpy as np
import matplotlib.pyplot as plt
import scipy as sc
```

# Método Newton-Raphon

```python
def newton_raphson(funcion,derivada,x0):

 x1 = x0 - (funcion(x0)/derivada(x0))
 i = 0

 while (abs(funcion(x1)) > 1e-12) and i < 1000:

  x2 = x1 - (funcion(x1)/derivada(x1))

  x1 = x2
  i +=1

 return x2
```

```python
f = lambda x: x**3 - 1
df = lambda x: 3*x**2

raiz = newton_raphson(f,df,-2)

x = np.linspace(-2,5)
y1 = f(x)
y2 = df(x)

plt.plot(x,y1,label="Función")
plt.plot(x,y2,label="Derivada")
plt.scatter(raiz,0,color="red",label="Cero de la función")
plt.legend()
plt.grid()
plt.show()
```

```python
def metodo_secante(funcion,x0,x1):

  x2 = x1 - ((x1-x0)/(funcion(x1)-funcion(x0)))*funcion(x1)
  i = 0

  while (abs(funcion(x2)) > 1e-12) and i < 1000:

    x3 = x2 - ((x2-x1)/(funcion(x2)-funcion(x1)))*funcion(x2)

    x2 = x3
    x1 = x2
    i +=1

  return x3
```

```python
f = lambda x: x**3 - 1
#df = lambda x: 3*x**2

raiz = metodo_secante(f,-2,-2)

x = np.linspace(-2,5)
y1 = f(x)
y2 = df(x)

plt.plot(x,y1,label="Función")
plt.plot(x,y2,label="Derivada")
plt.scatter(raiz,0,color="red",label="Cero de la función")
plt.legend()
plt.grid()
plt.show()
```
