```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
def suma_filas(A,i,j,k):
   # por convención, la suma de filas remplazara la fila k

   A[k] = A[i] + A[j]

   return A
```

```python
def escalar_filas(A,i,alpha):
  # por convención, la multiplicación por escalar de filas remplazara la fila i

  A[i] = alpha*A[i]
  return A
```

```python
def reduccion_matriz(A):
  for j in range(np.shape(A)[1]-1):
   for i in range(1,np.shape(A)[0]):
     alpha = (A[i][j]/A[j][j])

     A = escalar_filas(A,j,-alpha)
     A = suma_filas(A,j,i,i)
     A = escalar_filas(A,j,-1/alpha)

  return A
```

```python
S = np.array([[2,1,1,5],
              [4,1,1,1],
              [6,2,1,4]])


S_c = S.copy()
```
