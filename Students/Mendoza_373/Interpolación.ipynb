```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sc
```

```python
def recta_2puntos(x0,y0,x1,y1):

  m = (y1-y0)/(x1-x0)

  x = np.linspace(x0,x1)

  y = y0 + m*(x-x0)

  recta = np.array([x,y])

  return recta
```

```python
x0 = 1
y0 = 2

x1 = 3
y1 = 5

recta_inter = recta_2puntos(x0,y0,x1,y1)

plt.scatter(x0,y0,color="black")
plt.scatter(x1,y1,color="black")
plt.plot(recta_inter[0],recta_inter[1],linestyle="--")
plt.grid()
plt.show()
```

```python
def recta_2puntos(x0,y0,x1,y1):

  m = (y1-y0)/(x1-x0)

  x = np.linspace(x0,x1)

  y = y0 + m*(x-x0)

  recta = np.array((x,y))

  return recta
```

```python
def interpolacion_lineal(X,Y):

  c_r = np.zeros((len(X)))

  for i in range(0,len(X)-1):

   r_i = recta_2puntos(X[i],Y[i],X[i+1],Y[i+1])

   plt.scatter(X[i],Y[i],color="black")
   plt.plot(r_i[0],r_i[1],color="royalblue",linestyle="--")

  plt.scatter(X[len(X)-1],Y[len(Y)-1],color="black")
  plt.xlabel("X")
  plt.ylabel("Y")
  plt.grid()
  plt.show()
```

```python
X = np.arange(0,50,1)
Y = np.random.rand(50,1)

interpolacion_lineal(X,Y)

plineal = sc.interpolate.interp1d(X,X)
plt.plot(X,plineal(Y),color="royalblue",linestyle="--")
plt.scatter(X,Y,color="black")
plt.grid()
plt.show()
```
