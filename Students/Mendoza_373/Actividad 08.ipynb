```python
# Método punto fijo

def punto_fijo(x0,funcion,derivada):

  if derivada(x0) < -1 or derivada(x0) > 1:
      return "El x inicial no permite que el método converga"
  else:

   x1 = x0 - funcion(x0)
   i = 0

   while abs(x1-x0) > 1e-12 and i < 1000:

     x2 = x1 - funcion(x1)

     if derivada(x2) < -1 or derivada(x2) > 1:
      return "El x inicial no permite que el método converga"
     else:
      x0 = x1
      x1 = x2

  return x1
```
