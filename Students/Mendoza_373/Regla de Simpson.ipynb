```python
# MÃ©todo de Simpson 1/3 compuesto

def integracion_simpson(f,a,b,n):

  h = (b-a)/n

  x_i = np.linspace(a,b,n)
  y_i = f(x_i)

  par = y_i[0:n:2].sum()
  impar = y_i[1:n-1:2].sum()

  int_num = (h/3)*(y_i[0] + 2*par + 4*impar + y_i[-1])

  return int_num

f = lambda x: np.cos(x)

integracion_simpson(f,0,np.pi/2,50)
```
