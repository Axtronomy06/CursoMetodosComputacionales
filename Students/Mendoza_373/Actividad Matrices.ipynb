```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rotacion_vec(v,theta):

  r = np.array([[np.cos(theta),-np.sin(theta)],
               [np.sin(theta),np.cos(theta)]])
  u = np.matmul(v,r)
  return u
```

```python
v = np.array([1,1])
theta = np.deg2rad(-20)

u = rotacion_vec(v,theta)
u

plt.quiver(v[0],v[1],scale = 5, color = "red")
plt.quiver(u[0],u[1], scale = 5, color = "blue")
plt.grid()
plt.show()
```
