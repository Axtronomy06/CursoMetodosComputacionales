# **Graficación del campo electrico del dipolo**

Un dipolo electrico es un sistema compuesto por dos particulas con cargas $Q$ opuestas (una positiva y otra negativa), separadas un distancia $d$. A pesar de que su carga neta es 0, debido a la distancia entre las particulas el dipolo genera un campo electrico, para entender como se obtiene, recordemos como obtener el campo electrico de una particula individual de carga $Q$.

Por la ley de Coulomb, la magnitud del campo electrico en un punto $ P = (x,y)$ dado puede ser descrita como:

$ E = K_e\frac{Q}{r^2} $

Donde $K_e$ es la constante electrica, y r es la distancia entre la particula y el punto $P$. El vector del campo se obtiene haciendo el producto escalar entre la magnitud y el vector normal de la distancia, asi el campo electrico queda descrito de
forma vectorial como:

$ \vec{E} = K_e\frac{Q}{r^2} \hat{r} $

Dados los puntos donde se quiere evaluar el campo electrico, la coordenadas y la carga de la particula se puede determinar el campo electrico de esta. Para calcular el campo electrico del dipolo, aplicamos el principio de superposición, de modo que el campo electrico del dipolo queda descrito como:

$ \vec{E_d} = \vec{E_+} + \vec{E_-}$

Donde $ \vec{E_+}$ es el campo de la particula de carga $Q$ positiva y $\vec{E_-}$ la de carga $Q$ negativa.

```python
# Importación de librerias
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definicion de función para el calculo del campo electrico

def campo_electrico(x,y,x_f,y_f,q):
    u1 = ((x-x_f)**(2) + (y-y_f)**(2))**(1/2) # Distancia entre carga y punto
    E_mag = q / (u1)**2 # Magnitud del campo electrico

    E_x = E_mag*np.array((x-x_f)/u1) # Componente en X del campo electrico
    E_y = E_mag*np.array((y-y_f)/u1) # Componente en Y del campo electrico

    E = np.array([E_x, E_y]) # Campo electrico como vector

    return E
```

```python
# Definición de los parametros iniciales (Puntos del campo y valor de las cargas)

# Puntos del campo

x,y = np.meshgrid(np.linspace(-2,2,25),np.linspace(-2,2,25))

# Cargas del dipolo

q1 = 1
q2 = -q1

# Calculo de los campos del cada carga

E1 = campo_electrico(x,y,-1,0,q1) # Campo Q1
E2 = campo_electrico(x,y,1,0,q2) # Campo Q2
```

```python
# Graficación del campo

# Normalización de vectores

MAG = np.sqrt((E1[0] + E2[0])**2 +(E1[1] + E2[1])**2) # Valor de normalización

E1 = E1/MAG # Vectores del campo 1 normalizados
E2 = E2/MAG # Vectores del campo 2 normalizados

# Grafica del campo vectorial

plt.title("Campo Vectorial")
plt.quiver(x,y,E1[0] + E2[0],E1[1] + E2[1], scale=70)
plt.scatter(-1,0, color="red", label="Carga Q")
plt.scatter(1,0,color="blue", label="Carga -Q")
plt.legend(loc="upper right")
plt.show()

# Grafica de las lineas de campo

plt.title("Lineas de campo")
plt.streamplot(x,y,E1[0] + E2[0],E1[1] + E2[1], color="grey")
plt.scatter(-1,0, color="red", label="Carga Q")
plt.scatter(1,0,color="blue", label="Carga -Q")
plt.legend(loc="upper right")
plt.show()
```
