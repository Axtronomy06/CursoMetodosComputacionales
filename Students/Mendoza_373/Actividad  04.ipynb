```python
# Importaci√≥n de librerias

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#**1: Leer el dataset desde la url, con el comando pd.read_cvs().**

```python
# url es el link dado con los datos (formato .csv)

url = "https://raw.githubusercontent.com/hernansalinas/autogradesMetodosComputacionales/main/Dataset/Temperature_Bogota_Barcelona.csv"
temp_bog_bar = pd.read_csv(url) # Comando pandas para leerlo como DataFrame
temp_bog_bar
```

#**2: Adicionar una nueva columna AvgTempCelcius con la temperatura en grados celcius. $ C=(F-32)/1.8$.**

```python
temp_bog_bar["AvgTempCelcius"] = (temp_bog_bar.AvgTemperature - 32)/1.8 # Definimos una nueva columna en base a otra
temp_bog_bar
```

#**3: Realizar un gr√°fico de AvgTempCelcius como funci√≥n del a√±o para la ciudad de Bogot√°.**

```python
temp_bogota = temp_bog_bar[temp_bog_bar['City'] == "Bogota"].copy() # Creamos un nuevo DataFrame filtrando solo los datos de Bogot√°
temp_barcelona = temp_bog_bar[temp_bog_bar['City'] == "Barcelona"].copy() # Creamos un nuevo DataFrame filtrando solo los datos de Barcelona

# Grafico Temperaturo Promedio en Celcius para Bogot√°

x = temp_bogota.Year # A√±os
y = temp_bogota.AvgTempCelcius # Temperatura (Bogot√°)

plt.title("Temperatura Promedio en Celcius para Bogot√° en funci√≥n del a√±o")
plt.plot(x,y, color="blue")
plt.xlabel("A√±o")
plt.ylabel("Temperatura promedio en Celcius ($C¬∞$)")
plt.grid()
plt.show()
```

#**4: Realizar un gr√°fico de AvgTempCelcius como funci√≥n del a√±o para la ciudad de Barcelona.**

```python
# Grafico Temperaturo Promedio en Celcius para Barcelona

x = temp_barcelona.Year # A√±o
y = temp_barcelona.AvgTempCelcius # Temperatura (Barcelona)

plt.title("Temperatura Promedio en Celcius para Barcelona")
plt.plot(x,y, color="red")
plt.xlabel("A√±o")
plt.ylabel("Temeperatura promedio en Celcius ($C¬∞$)")
plt.grid()
plt.show()
```

#**5: Determine la mayor temperatura para Bogot√° y Barcelona**

```python
# Grafico Temperaturo Promedio en Celcius Bogot√° v.s Barcelona

x1 = temp_bogota.Year # A√±os
y1 = temp_bogota.AvgTempCelcius # Temperatura (Bogota)

x2 = temp_barcelona.Year # A√±os
y2 = temp_barcelona.AvgTempCelcius # Temperatura (Barcelona)

# Ahora hacemos un condiconal donde solo nos aparezcan los datos de la fila donde este la temperatura maxima de cada ciudad

year_barcelona = temp_barcelona.loc[temp_barcelona['AvgTempCelcius'] == temp_barcelona.AvgTempCelcius.max()] # Barcelona
year_bogota = temp_bogota.loc[temp_bogota['AvgTempCelcius'] == temp_bogota.AvgTempCelcius.max()] # Bogota

# Grafica

plt.title("Temperatura Promedio en Celcius Bogot√° v.s Barcelona")
plt.plot(x1,y1, label="Bogot√°", color="blue") # Temperatura / A√±o (Bogot√°)
plt.plot(x2,y2, label="Barcelona", color="red") # Temperatura / A√±o (Barcelona)
plt.scatter(year_barcelona.Year,temp_barcelona.AvgTempCelcius.max(),label="Temp. Maxima en Barcelona",color="red") # Temp.MAX Barcelona
plt.scatter(year_bogota.Year,temp_bogota.AvgTempCelcius.max(),label="Temp. Maxima en Bogot√°",color="blue") # Temp.MAX Bogot√°
plt.xlabel("A√±o")
plt.ylabel("Temperatura promedio en Celcius ($C¬∞$)")
plt.grid()
plt.legend()
plt.show()
```
