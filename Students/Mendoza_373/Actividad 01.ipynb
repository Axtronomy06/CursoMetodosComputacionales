```python
# 1): Función Doble factorial

# El algoritmo es calcado del factorial, con el cambio del tamaño de pasos y donde para el ciclo "WHILE" dependiedo de la paridad del número

def double_factorial(n):
  if n < 0 or type(n) != int:
    return "Número no valido"
  else:
    if (n%2 != 0): # Este condicional determina si el número es par o impar, usando "%" para calcular el residuo al ser divido por 2
     i = n + 2 # Al tener que cambiar el salto del factorial, el "i" inicial tambien cambia correspondientemente
     fact_num = 1
     while i != 1: # Para impares, la condicion de cierre es i = 1, aunque i = 3 tambien funciona.
       fact_num = fact_num*(i-2) # Para tomar solo los valores impares, el salto es de 2 en 2 en vez de 1 en 1
       i = i-2 # Para tomar solo los valores impares, el salto es de 2 en 2 en vez de 1 en 1
     return fact_num
    else:
     i = n + 2 # Al tener que cambiar el salto del factorial, el "i" inicial tambien cambia correspondientemente
     fact_num = 1
     while i != 2: # Para pares, la condición de cierre es i = 2, ya que es el ultimo termino para el doble factorial par siempre
       fact_num = fact_num*(i-2) # Para tomar solo los valores pares, el salto es de 2 en 2 en vez de 1 en 1
       i = i-2 # Para tomar solo los valores pares, el salto es de 2 en 2 en vez de 1 en 1
     return fact_num

double_factorial(-7)
```

```python
# 2): Función Sucesión de Lucas

# Como su recursividad es parecida a la del Fibonacci, usaremos metodos similares para calcular el enesimo termino de la sucesion

def lucas(n):
 if n <= 0:
  return "Número no valido"
 else:
 # Los primeros 2 condiconales devoleran 2 y 1 si n = 1 o n = 2 respectivamente
   if n == 1:
     return 2
   elif n == 2:
     return 1
   else: # Si n > 2, se ira al metodo recursivo

   # Definimos "i_1" y "i_2" como los respectivos primer y segundo terminos de la sucesión

     i_1 = 2
     i_2 = 1
     j = 2 # Nuestro contador empieza en 2 para ajustar con el termino a calcular (ya tenemos los 2 primeros terminos definidos)
     while j < n: # Cuando "j" se igual a n, ya tendremos el enesimo termino de la sucesión
       i_n = i_2 + i_1 # Definimos el termino "i_n" de forma recursiva

     # Redefinimos "i_1" como "i_2" y "i_2" como "i_n" en cada ciclo, esto para matener la recursividad

       i_1 = i_2
       i_2 = i_n
       j += 1
     return i_n

lucas(10)
```

```python
# 3): Función Binomial

# La formula del coeficiente binomial usa factoriales, reutilizaremos el codigo que ya tenemos del factorial

def factorial_numero(numero):
  if numero < 0:
    return "Número no valido"
  else:
   i = numero + 1 # i = numero + 1, ya que si numero = 0, el resultado es 1, lo que cumple 0! = 1, nos permite usar numero como contador sin alteralo ni omitirlo
   fact_num = 1 # fact_num = 1 ya que 0! = 1, nos sirve como base para empezar a generar los factoriales
   while i != 1: # A medida que el ciclo se ejecute, "i" convertira en 1, en ese momento, ya tendremos el factorial y el ciclo podra terminar
     fact_num = fact_num*(i-1) # En cada ciclo, "fact_num" se redefinira como la cuenta que llevemos, asi hasta finalizar el ciclo
     i = i-1 # En cada ciclo, redefinimos "i" como el número siguiente en nuestra cuenta regresiva, empezando con numero y terminando en 1
  return fact_num

# Ahora definiremos la función binomial usando la función "factorial_numero"

def binomial(n,k):

# Es importante que n >= k y ambos sean enteros mayores a 0, por lo que haremos un condicional para evitar errores de ejecución

 if n < k or type(n) != int or type(k) != int or n < 0 or k < 0:
  return "Números no validos"
 else:

  # Aplicamos la formula del coeficiente binomial (n!)/(k!)*(n-k)!

   binomial = (factorial_numero(n))/((factorial_numero(k))*(factorial_numero(n-k)))
   return binomial

binomial(5,3)
```
