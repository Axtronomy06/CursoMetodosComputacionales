````python
# Números Defectivos

def numeros_defectivos(n):

# Definimos los contadores, la lista y la suma de divisores

  i = 1
  list_div = [1] # Nuestro método para encontrar los divisores omite al 1, como es divisor de cualquier número esta de base en la lista
  coef = 0

# Este While obtendra todos lo divisores (a exepción del 1) del numero "n" y los agrega a la lista de divisores

  while i < n:
    if n%i != 0:
      i += 1
    else:
      list_div.append(n/i)
      i += 1

# Hallamos la suma de todos los divisores

  for i in list_div:
    coef += i

  if coef-n < n: # Si cumple la condición (ser menor que "n") retorna True, caso contrario False
    return True
  else:
    return False

numeros_defectivos(6)

# El resto de funciones usan el mismo metodo para encontrar divisores y sumarlos, por lo que se omitira el comentario sobre eso
````

````python
# Números Abundantes

def numeros_abundantes(n):

  i = 1
  list_div = [1]
  coef = 0 # Suma de los divisores de "n"

  while i < n:
    if n%i != 0:
      i += 1
    else:
      list_div.append(n/i)
      i += 1

  for i in list_div:
    coef += i

  if coef-n > n: # Si cumple la condición (ser mayor que "n") retorna True, caso contrario False
    return True
  else:
    return False

numeros_abundantes(28)
````

````python
# Números Perfectos

def numeros_perfectos(n):

  i = 1
  list_div = [1]
  coef = 0

  while i < n:
    if n%i != 0:
      i += 1
    else:
      list_div.append(n/i)
      i += 1

  for i in list_div:
    coef += i

  if coef-n == n: # Si cumple la condición (ser mayor que "n") retorna True, caso contrario False
    return True
  else:
    return False

numeros_perfectos(28)
````


7. Diseñar un programa en el que entrado dos números `a`  y `b` retorne una variable booleana **True** o **false**.

  [Números amigos](https://es.wikipedia.org/wiki/N%C3%BAmeros_amigos) `a` y `b` tales que a es la suma de los divisores propios de `b` y viceversa.
    La rutina se debera llamar numeros_amigos




P2. Retornar los 10 primeros pares de numeros perfectos, semiperfectos, amigos



````python
def numeros_amigos(n,k):

  # En este caso, aplicamos los mismo metodos para encotrar divisores y sumarlos, simplemente lo hacemos tanto para "n" como "k"

  i = 1
  list_div_n = [1]
  coef_n = 0

  while i < n:
    if n%i != 0:
      i += 1
    else:
      list_div_n.append(n/i)
      i += 1

  i = 1
  list_div_k = [1]
  coef_k = 0

  while i < k:
    if k%i != 0:
      i += 1
    else:
      list_div_k.append(k/i)
      i += 1

  for i in list_div_n:
    coef_n += i

  for j in list_div_k:
    coef_k += j

  if coef_n == coef_k: # Si cumple la condición (coef_n es igual a coef_k) retorna True, caso contrario False
    return True
  else:
    return False

numeros_amigos(17296,18416)
````

````python
# Funcion Quicksort

"""
Para este problema se usa listas comprensivas, las cuales nos permiten crear listas en base a otras listas y metodos recursivos.
La idea es la misma que la del algoritmo quicksort, elegiremos un pivote y compararemos los elementos de la lista con este, dependiento
si son menores, iguales o mayores iran a listas correspondientes, una vez termiando, usaremos la misma funcion quicksort para organizar las
listas que se generaron y asi hasta tener listas de un solo elemento, donde la función para y la lista ya esta organizada
"""

def quicksort(lista):

  if len(lista) <= 1: # Punto en el que la función para
    return lista

  pivote = lista[(len(lista)//2)] # Definición del pivote (elemento en la mitad de la lista)
  lista_menor = [i for i in lista if i < pivote] # Lista para números en la "lista" menores al pivote
  lista_igual = [i for i in lista if i == pivote] # Lista para números en la "lista" iguales al pivote
  lista_mayor = [i for i in lista if i > pivote] # Lista para números en la "lista" mayores al pivote

  # Ahora usamos la misma función quicksort en las listas menor y mayor para organizarlas.

  return quicksort(lista_menor) + lista_igual + quicksort(lista_mayor)

lista_prueba = [22, 32, 42, 12, 22, 31, 41, 11, 12, 232, 24, 12, 22]

quicksort(lista_prueba)
````
