## Ecuaciones de movimiento

### Ecuaciones de movimiento del centro de masa

De acuerdo con la segunda ley de Newton, la ecuación de movimiento a lo largo de la vertical ($y$) puede ser escrita como:

\begin{equation}
m\frac{d^2y}{dt^2} = N - mg
\end{equation}

donde $m$ es la masa de la lápiz, $N$ es la fuerza normal y $g$ es la aceleración de la gravedad. Escribiendo las ecuaciones en términos del ángulo $\theta$, tenemos que:

\begin{equation}
-m\frac{l}{2}\cos \theta \left(\frac{d\theta}{dt} \right)^2 - m\frac{l}{2}\sin \theta \frac{d^2\theta}{dt^2} = N - mg
\end{equation}

### Ecuaciones de rotación del centro de masa

La sumatoria de momentos de fuerza se puede escribir como sigue:

\begin{equation}
I_c\frac{d^2\theta}{dt^2} = N\frac{l}{2} \sin \theta
\end{equation}

donde $I_c = \frac{1}{12}ml^2$ es el momento de inercia de una lápiz de masa $m$ y longitud $l$ respecto a un eje perpendicular a la lápiz en su centro de masa.

Reemplazando el valor de la normal en la ecuación anterior, obtenemos que la ecuación de movimiento para la posición angular está dada por:

\begin{equation}
\frac{d^2 \theta}{dt^2} = \left( \frac{\frac{2g}{l} - \left(\frac{d\theta}{dt}\right)^2\cos \theta }{\frac{1}{3} + \sin^2 \theta} \right)\sin \theta
\end{equation}

Podemos resolver la ecuación anterior si conocemos las condiciones iniciales: $\theta = \theta_0$ en $t = 0$ y $\frac{d\theta}{dt} = \omega_0 = 0$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from scipy.interpolate import CubicSpline
from scipy.integrate import quad
from scipy.optimize import newton
```

## **1) Punto**

```python
L = 0.1
g = 9.8
theta0 = np.pi/12
omega0 = 0
dt = 0.0001
ta = 0
tb = 10
N = int((tb-ta)/dt)
t = np.linspace(ta,tb,N)
U0 = [theta0, omega0]
```

```python
def fun(U, t, g, L):
    """
    Calcula las derivadas del ángulo y la velocidad angular

    Args:
        theta (float): Ángulo de desplazamiento (en radianes).
        omega (float): Velocidad angular (en radianes/segundo).
        g (float): Aceleración debida a la gravedad (en m/s²).
        L (float): Longitud del hilo del péndulo (en metros).

    Returns:
        Array: Derivadas de theta y omega.
    """
    theta, omega = U
    dtheta_dt = omega
    domega_dt = (2*g/L - dtheta_dt**2*np.cos(theta))/((1/3) + np.sin(theta)**2) * np.sin(theta)
    return np.array([dtheta_dt, domega_dt])
```

```python
U = odeint(fun, U0, t, args=(g, L))
theta = U[:,0]
omega = U[:,1]
y_cm = L/2 * np.cos(theta) # Posición del centro de masa
```

## **2) Punto**

```python
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(t, theta, 'b-', label='Ángulo θ [rad]')
plt.xlabel('Tiempo [s]')
plt.ylabel('Ángulo [θ]')
plt.legend(loc='lower right')
plt.title('Ángulo [θ] vs Tiempos [s]')

plt.subplot(3, 1, 2)
plt.plot(t, y_cm, 'r-', label='Posición del centro de masa y [m]')
plt.xlabel('Tiempo [s]')
plt.ylabel('Posición del CM')
plt.legend(loc='lower right')
plt.title('Posición del centro de masa')

plt.subplot(3, 1, 3)
plt.plot(t, omega, 'g-', label='Rapidez angular ω [rad/s]')
plt.xlabel('Tiempo (s)')
plt.ylabel('Rapidez angular ω')
plt.legend(loc='lower right')
plt.title('Rapidez angular')


plt.tight_layout()
plt.show()
```

Cuando $\theta = \frac{\pi}{2}$ el lapiz tocara el suelo, por ende el tiempo que tarda en caer es el tiempo valido para este sistema fisico. Lo encontrare filtrando los datos que me retorna odeint de tal forma que una vez que $\theta = \frac{\pi}{2}$  se rompa el ciclo.

```python
U = odeint(fun, U0, t, args=(g, L))
theta_ = U[:,0]
omega_ = U[:,1]
y_cm = L/2 * np.cos(theta_) # Posición del centro de masa
```

```python
theta1 = np.array([])
for i in range(len(theta_)):
  if theta_[i] < np.pi/2:
    theta1 = np.append(theta1, theta_[i])
  else:
    theta1 = np.append(theta1, theta_[i]) #  theta_[i] = np.pi/2
    break
```

```python
omega1 = omega_[:len(theta1)]
y_cm1 = y_cm[:len(theta1)]
t = t[:len(theta1)]
```

```python
t_final = t[-1]
print(f"El tiempo que tarda en caer el lapiz es {t[-1]:.3f} s")
```

```python
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(t, theta1, 'b-', label='Ángulo θ [rad]')
plt.xlabel('Tiempo [s]')
plt.ylabel('Ángulo [θ]')
plt.legend(loc='lower right')
plt.title('Ángulo [θ] vs Tiempos [s]')

plt.subplot(3, 1, 2)
plt.plot(t, y_cm1, 'r-', label='Posición del centro de masa y [m]')
plt.xlabel('Tiempo [s]')
plt.ylabel('Posición del CM')
plt.legend(loc='lower right')
plt.title('Posición del centro de masa')

plt.subplot(3, 1, 3)
plt.plot(t, omega1, 'g-', label='Rapidez angular ω [rad/s]')
plt.xlabel('Tiempo (s)')
plt.ylabel('Rapidez angular ω')
plt.legend(loc='lower right')
plt.title('Rapidez angular')


plt.tight_layout()
plt.show()
```

## **5) Punto**

```python
L = 0.1
g = 9.8
theta0 = np.array([np.pi/12, np.pi/6, np.pi/4, np.pi/3])
omega0 = 0
dt = 0.0001
ta = 0
tb = t_final  # Calculado en el 3 punto
N = int((tb-ta)/dt)
t = np.linspace(ta,tb,N)
```

La raiz que queremos econtrar es el valor de t cuando $\theta = \frac{\pi}{2}$

```python
for i in range(len(theta0)):
    U0 = [theta0[i], omega0]
    U = odeint(fun, U0, t, args=(g, L))
    theta_spline = CubicSpline(t, U[:, 0])
    def f(t):
        return theta_spline(t) - (np.pi/2)
    def df(t):
        return theta_spline.derivative()(t)
    t_fall = newton(f, tb/2, fprime=df)
    print(f"Tiempo de caída para theta0 = {theta0[i]:.3f} rad: {t_fall:.3f} s")
    plt.plot(t,  U[:, 0], label=f"$\\theta_0 = {theta0[i]:.2f}$ rad")
plt.xlabel('Tiempo [s]')
plt.ylabel('Theta [rad]')
plt.axhline(np.pi/2, color='black', linestyle='--', label='$\\theta = \\pi/2$')
plt.title('Evolución del Ángulo Theta para Diferentes Ángulos Iniciales')
plt.legend()
plt.grid()
plt.show()
```

## **6) Punto**

```python
theta0 = np.array([np.pi/12, np.pi/6, np.pi/4, np.pi/3])
```

```python
def integral(theta, theta0):
    numerador = 1/3 + np.sin(theta)**2
    denominador = np.cos(theta0) - np.cos(theta)
    return np.sqrt(numerador / denominador)
```

```python
for i in range(len(theta0)):
    valor , _= quad(integral, theta0[i], np.pi/2,args=(theta0[i]))
    T = (1 / 2) * np.sqrt(L / g) * valor
    print(f"El tiempo que tarda en caer el lapiz para un ángulo inicial de {theta0[i]:.2f} es {T:.3f} s")
```

## **7) Punto**

```python
theta = np.linspace(0, np.pi/2, 100)
T_ = np.zeros(len(theta))
for i in  range(len(theta)):
  valor , _= quad(integral, theta[i], np.pi/2, args=(theta[i]))
  T = (1 / 2) * np.sqrt(L / g) * valor
  T_[i] = T
```

```python
plt.plot(theta, T_, 'b-', label='Ángulo θ [rad]')
plt.xlabel('Ángulo inicial θ [rad]')
plt.ylabel('Tiempo [s]')
plt.axvline(x=np.pi/12, color='r', linestyle='--', label='$\\theta_0 = \\pi/12$')
plt.legend(loc='lower right')
plt.title('Tiempo de caída en función del ángulo inicial')
plt.grid()
plt.show()
```

Se deduce que efectivamente entre más cerca el valor de $\theta$ a $\pi/2$ caera en un tiempo más corto y entre más cerca este a cero más tiempo le tomara en caer aunque cuando tiende a cero el tiempo crece de manera cada vez más rapido
