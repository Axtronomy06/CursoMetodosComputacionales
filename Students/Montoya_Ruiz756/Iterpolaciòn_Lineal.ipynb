```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pendiente(x,y):
  m = np.empty(0)
  for i in range(len(x)-1):
    m1 = (y[i+1]-y[i])/(x[i+1]-x[i]) #Ecuacion de la pendiente
    m = np.append(m,m1)
  return m
```

```python
def iterp(x,y):
  m = pendiente(x,y)
  iterx = np.empty(0)
  itery = np.empty(0)
  for i in range(len(x)-1):
    x_ = np.linspace(x[i],x[i+1])
    y_ = (m[i] * (x_- x[i])) + y[i] # Ecuacion punto pendiente
    iterx = np.append(iterx,x_)
    itery = np.append(itery,y_)
  iterx = np.append(iterx,x[-1])
  itery = np.append(itery,y[-1])
  return iterx,itery
```

```python
x = np.linspace(0,2*np.pi,60)
y = np.sin(x)
iterx, itery = iterp(x,y)
plt.plot(x,y, "g*", label = "Puntos")
plt.plot(iterx,itery, label = "Iterpolacion Lineal")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
```

```python
# Dados estos puntos de prueba
x_ = np.linspace(0, 2*np.pi, 100)
y_ = np.sin(x_)


# Partiendo de la eucuacion de una recta entre dos puntos
def recta(x0, y0, x1, y1):
  num = y1-y0
  den = x1-x0
  pendiente = num/den
  x = np.linspace(x0, x1, 10)
  y = pendiente*(x-x0)+y0
  return x, y

# Se realiza la interpolacion
plt.figure()
for i in range(0, len(x_)-1):
  _x_, _y_ = recta(x_[i], y_[i], x_[i+1],y_[i+1])
  plt.plot(_x_, _y_)
```
