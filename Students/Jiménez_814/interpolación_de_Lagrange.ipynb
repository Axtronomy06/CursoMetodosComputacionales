```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
puntos=np.array([[2,4],[5,1],[6,2]])
plt.figure()
#plt.plot(puntos[:,0],puntos[:,1],"o")
plt.plot(puntos.T[0],puntos.T[1],"o", c="pink")
plt.show()
```

```python
def L(puntos,j):
  n=len(puntos)
  def evaluar(x):
    resultado=1
    for i in range(n):
      if i != j:
        resultado*=(x-puntos[i][0])/(puntos[j][0]-puntos[i][0])
    return resultado

  return evaluar
```

```python
L0=L(puntos,0)
L1=L(puntos,1)
L2=L(puntos,2)
```

```python
Lsuprema=np.array([L0,L1,L2])
X=np.linspace(0,6)
plt.figure()
plt.plot(puntos.T[0],puntos.T[1],"o", c="pink")
for p in Lsuprema:
  Y=p(X)
  plt.plot(X,Y)
plt.show()
```

```python
def P(puntos,Lsuprema):
  n=len(puntos)
  def evaluar(x):
    salida=0
    for k in range(n):
      salida+=puntos[k][1]*Lsuprema[k](x)
    return salida
  return evaluar
```

```python
plt.figure()
plt.plot(puntos.T[0],puntos.T[1],"o", c="pink")
Y=Pol(X)
plt.plot(X,Y,c="magenta",alpha=0.2)
plt.grid(linestyle="dotted")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpolación de Lagrange")
plt.show()
```
