```python
import numpy as np
import matplotlib.pyplot as plt
import random
from scipy import interpolate
```

```python
def pendiente(p0,p1):
  x0=p0[0]
  y0=p0[1]
  x1=p1[0]
  y1=p1[1]
  return (y1-y0)/(x1-x0)
```

```python
#necesito valores aleatorios, que por el momento los daré
puntos=np.array([[3,5],[0,5],[4,3]])
#ordenarlos según la coordenada x
puntos=np.sort(puntos,axis=0)
#puntos[:,i] así separo la columna, i la columna que quiero
m=np.array([pendiente(puntos[0],puntos[1]),pendiente(puntos[1],puntos[2])])
#plt.figure()
#plt.plot(puntos)
print(m)
```

```python
def pendientegrupo(longitud,puntos):
  m=np.zeros(longitud-1)
  for n in range(longitud-1):
    p0=puntos[n]
    p1=puntos[n+1]
    m[n]=pendiente(p0,p1)
  return m
def grafica(longitud):
  x=np.random.rand(longitud)
  y=np.random.rand(longitud)
  index=np.argsort(x)
  x=x[index]
  puntos=np.array([x,y])
  puntos=puntos.T
  #ahora necesito generalizar la pendiente
  m=pendientegrupo(longitud,puntos)
  plt.subplot()
  #graficando puntos
  for n in range(len(puntos)):
    plt.plot(puntos[n][0],puntos[n][1],"o")
  #graficar lineas
  for n in range(len(m)):
    x=np.linspace(puntos[n][0],puntos[n+1][0])
    y=m[n]*(x-puntos[n][0])+puntos[n][1]
    plt.plot(x,y)
  plt.grid()
  #plt.title("Conectando puntos (mi función)")
  plt.show()
```
