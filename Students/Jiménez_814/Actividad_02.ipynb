```python
#valores iniciales
#carga positiva
x_1=-1
y_1=0
r1=np.array([x_1,y_1])
#carga negativa
x_2=1
y_2=0
r2=np.array([x_2,y_2])


q=1
#constante
k=1

x=10
y=10
r3=np.array([x,y])
```

```python
def modulo(r):
  return np.sqrt(r[0]**2+r[1]**2)

def unitario(r):
  return r*(1/modulo(r))

def distancia(r1,r2):
  return np.sqrt((r1[0]-r2[0])**2+(r1[1]-r2[1])**2)

def campo(r,r1,q):
  return (k*q)/(distancia(r,r1)**2)*unitario(r-r1)

def superposicion(r,r1,r2,q):
  E1=campo(r,r1,q)
  E2=campo(r,r2,-q)
  return E1+E2

```

```python
X=np.linspace(-5,5,11)
Y=np.linspace(-5,5,11)
R=[]
for x in X:
    for y in Y:
      if y!=0 or x!=0:
        R.append(np.array([x,y]))
```

```python
for r in R:
  E=superposicion(r,r1,r2,q)
  plt.quiver(r[0],r[1],E[0],E[1],color='pink')
  #iuda

plt.plot(x_1,y_1,"ro")
plt.plot(x_2,y_2,"ro")
```
