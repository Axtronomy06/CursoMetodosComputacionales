Hacer un programa que ingresando un angulo tetha y una valociada inicial (V0), me devuelva la trayectoria de la particula desde un punto fijo (0,0)


Posición horizontal:  $$X = V_0*cos(\theta)*t $$

Posición vertical con $$Y_o = 0$$ y $$g = 9.8$$:  $$Y = y_o + V_0*sin(\theta)*t - \frac12gt² $$

Velocidad en X  $$ V_x = V*cos(\theta) $$

Velocidad en Y  $$ V_y = V*cos(\theta) $$

Además V está definido así:  $$V = \sqrt((V_x)^2 + (V_y)^2)$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Posición punto a punto
global g
g = 9.8

def tiempo_vuelo(V_0,theta):
  t = 2*V_0*np.sin(np.deg2rad(theta))/g
  return t

def posición_X(V_0,theta, t):
  theta = np.deg2rad(theta)
  X = V_0*np.cos(theta)*t
  return X

def posicion_Y(V_0,theta, t):
  theta = np.deg2rad(theta)
  Y = (V_0*np.sin(theta))*t-0.5*g*t**2
  return Y
```

```python
vo=2 #m/s
theta = 30 #deg

theta = np.deg2rad(theta)
tvuelo = tiempo_vuelo(2, 30)

t=np.linspace(0, tvuelo, 100)
x = posición_X(2, 30, t)
y = posicion_Y(2, 30, t)
#print(tvuelo,x,y)
```

```python
fig, ax  = plt.subplots()
ax.plot(x, y)
ax.set_xlabel("x")
ax.set_ylabel("y")
```

```python
def read_file(namefile):
  with open(namefile, 'r') as archivo:
    d = json.load(archivo)
  return d
```

```python
def time(d):
  t = 2*d['Vo']*np.sin(np.deg2rad(d['Theta']))/g
  return t
def mov_p(d,t):
  X = d['Vo']*np.cos(np.deg2rad(d['Theta']))*t
  Y = (d['Vo']*np.sin(np.deg2rad(d['Theta'])))*t-0.5*g*t**2
  return X,Y
```

```python
d = read_file('graficar_p.json')
print(d)
t1 = time(d)
t=np.linspace(0, t1, 100)
x,y = mov_p(d,t)

fig, ax  = plt.subplots()
ax.plot(x, y)
ax.set_xlabel("x")
ax.set_ylabel("y")
```
