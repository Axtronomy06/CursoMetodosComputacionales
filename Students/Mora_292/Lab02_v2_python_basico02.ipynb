````python
def numeros_defectivos(n):
  sum = 0
  for i in range(1,n):
    if n%i == 0:
      sum += i
  if sum < n:
    return True
  else:
    return False
````

  [Números amigos](https://es.wikipedia.org/wiki/N%C3%BAmeros_amigos) `a` y `b` tales que a es la suma de los divisores propios de `b` y viceversa.
    La rutina se debera llamar numeros_amigos




P2. Retornar los 10 primeros pares de numeros perfectos, semiperfectos, amigos



````python
#Suma de divisores propios
def sum_div_prop(n):
  sum = 0
  for i in range(1,n):
    if n%i == 0:
      sum += i
  return sum
````

````python
def numeros_amigos(a,b):
  an = sum_div_prop(a)
  print(an)
  bn = sum_div_prop(b)
  print(bn)
  if an == b and bn == a:
    return True
  else:
    return False
````

````python
numeros_amigos(63020, 76084)
````
