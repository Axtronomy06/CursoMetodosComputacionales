```python
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
  return (np.sin(x) + np.tan(x))**2
x_ = np.linspace(0,1)
```

```python
d = sp.misc.derivative(f, x_, 1e-6)
d
```

```python
x = np.linspace(0,1)
plt.plot(x, f(x))
plt.plot(x_, d)
plt.legend(['x','dx'])
```
