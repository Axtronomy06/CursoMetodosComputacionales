```python
def bisection(a, b, f, e):
  c = (a+b)/2
  while abs(f(c)) >= e and f(a)*f(b) < 0:
    c = (a+b)/2
    y = f(c)
    if y > 0:
      b = c
    else:
      a = c
  return c, f(c)
```

```python
def New_rap(f, x0, df, e):
  xn = x0 - f(x0)/df(x0)
  while abs(f(xn)) >= e:
    xn1 = xn - f(xn)/df(xn)
    xn = xn1
  return xn
```
