```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/hernansalinas/autogradesMetodosComputacionales/main/Dataset/Temperature_Bogota_Barcelona.csv'
df = pd.read_csv(url)
df
```

```python
df['AvgTempCelcius']=(df.AvgTemperature-32)/1.8
df
```

```python
dfBog = df.copy()
Bog = (dfBog.City=='Bogota')
dfBog = dfBog[Bog]
dfBog
```

```python
fig, ax = plt.subplots()
ax.plot(dfBog['Year'], dfBog['AvgTempCelcius'])
plt.title('Temperatura (C) vs año en Bogotá')
plt.xlabel('Año')
plt.ylabel('Temperatura (C)')
```

```python
dfBar = df.copy()
Bar = (dfBar.City=='Barcelona')
dfBar = dfBar[Bar]
dfBar
```

```python
fig, ax = plt.subplots()
ax.plot(dfBar['Year'], dfBar['AvgTempCelcius'])
plt.title('Temperatura (C) vs año en Barcelona')
plt.xlabel('Año')
plt.ylabel('Temperatura (C)')
```

```python
Bogmax = dfBog['AvgTempCelcius'].max()
print(f'La tempertura en celcius máxima en la ciudad de Bogotá es: {Bogmax}')
```

```python
Barmax = dfBar['AvgTempCelcius'].max()
print(f'La tempertura en celcius máxima en la ciudad de Barcelona es: {Barmax}')
```
