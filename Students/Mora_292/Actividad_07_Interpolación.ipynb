```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
#puntos
x = np.array([2,4,6])
y = np.array([1,3,7])

#pendiente
m1 = (y[1]-y[0])/(x[1]-x[0])

#Primera recta
x_ = np.linspace(x[0], x[1], 100)
y_ = y[0] + m1 * (x_ - x[0])

#pendiente recta 2
m2 = (y[2]-y[1])/(x[2]-x[1])

#recta 2
x2_ = np.linspace(x[1], x[2], 100)
y2_ = y[1] + m2 * (x2_ - x[1])
```

```python
plt.plot(x[0],y[0], marker = 'o',  color='Orange')
plt.plot(x[1],y[1],  marker = 'o', color='Pink')
plt.plot(x[2],y[2],  marker = 'o')
plt.plot(x_, y_)
plt.plot(x2_, y2_)
```

```python
#generalización
def interpolación(x, y):
  puntos = len(x)
  for i in range(0, puntos-1):
    m1 = (y[i+1]-y[i])/(x[i+1]-x[i])
    x_ = np.linspace(x[i], x[i+1], 100)
    y_ = y[i] + m1 * (x_ - x[i])

    plt.plot(x[i],y[i], marker = 'o',  color='Orange')

    plt.plot(x_, y_)
  plt.xlabel('X')
  plt.ylabel('Y')
  plt.title('Interpolación mediante rectas')
  plt.grid()
```

```python

x = np.array([0,1,4,10,15])
y = np.array([0,2,3.5, 7, 11.5])

interpolación(x, y)
```

```python
x = np.array([0,1,4,10,15])
y = np.array([0,2,3.5, 7, 11.5])

interpolación(x, y)
```
