```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy as sc
import sympy as sy
import urllib.request
```

```python
#datos
x=np.array([0.9,1.3,1.9,2.1,2.6,3.0,3.9,4.4,4.7,5.0,6.0,7.0,8.0,9.2,10.5,11.3,11.6,12.0,12.6,13.0,13.3])
y=np.array([1.3,1.5,1.85,2.1,2.6,2.7,2.4,2.15,2.05,2.1,2.25,2.3,2.25,1.95,1.4,0.9,0.7,0.6,0.5,0.4,0.25])
```

```python
#Pato
pato='https://3.bp.blogspot.com/-8_LTrCIxVSQ/VxU-5Gq2ffI/AAAAAAAAIto/LhCDEwHKTiAbLr7IhizQwGV13YTwgpbDwCLcB/s400/pato1.png' #URL
urllib.request.urlretrieve(pato, 'pato.jpg')#Decargar la imagen
imagen_pato=plt.imread('pato.jpg') #Variable con la imagen del pato


plt.figure()
plt.title('Representación del pato')
plt.imshow(imagen_pato,extent=(0.4, 14, -6.8, 3.45))
plt.plot(x,y,'.b')
plt.xlim(0, 14)
plt.ylim(-7,4)

plt.grid()
```

```python
#lagrange
pl = sc.interpolate.lagrange(x,y)
```

```python
x_int = np.linspace(min(x),max(x),100)
plt.figure(figsize=(9,3))
plt.plot(x_int,pl(x_int),label='Lagrange')
plt.title('Polinomio de Lagrange')
plt.ylabel('p(x)')
plt.xlabel('x')
plt.legend()
plt.grid()
```

```python
#cubicsplin
pc = sc.interpolate.CubicSpline(x,y)
```

```python
plt.figure(figsize=(9,3))
plt.plot(x_int,pc(x_int),label='Cubicsplin')
plt.title('Polinomio de Cubicsplin')
plt.ylabel('p(x)')
plt.xlabel('x')
plt.legend()
plt.grid()
```

```python
plt.figure(figsize=(9,3))
plt.plot(x_int,pl(x_int),label='Lagrange')
plt.plot(x_int,pc(x_int),label='Cubicsplin')
plt.plot(x,y,'.k',label='Puntos')
```
