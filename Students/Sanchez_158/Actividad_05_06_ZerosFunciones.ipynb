  $b\cos(e^x-2) - a e^{x} + 2 = 0 $


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

a = float(input("Ingresa el valor de a: "))
b = float(input("Ingresa el valor de b: "))    #Estos valores los definí al azar

f = lambda x: b*np.cos(np.exp(x)-2)-a*np.exp(x)+2
```

```python
x = np.linspace(-1,1, 50)
plt.ylabel("F(x)")
plt.xlabel("(x)")

plt.title(r"$Grafica\ de\ la\ Función \ :\  b\cos(e^x-2) - a e^{x} + 2 = 0$")
plt.plot( x, f(x), "r-", label= "Función")
plt.grid()
plt.show()
```

```python
#Metodo de la Bisección
#a = 4
#b = 3

f = lambda x: b*np.cos(np.exp(x)-2)-a*np.exp(x)+2

xa = -1  #Defino el intervalo
xb = 1

Tol = 1E-9
Nmax = 100

def biseccion(f, xa, xb, Tol, Nmax):

  if (f(xa)*f(xb)>0):
    print("La función no cambia de signo")
  else:
    i = 0
    while ((f(xa)*f(xb)<0) and i < Nmax):
      i=i+1
      xc = (xa + xb)/2
      if (f(xc) < 0):
        xb = xc
      elif (f(xc) > 0):
        xa = xc
    return xc

xr=biseccion(f,xa,xb,Tol,Nmax)

print(f"La raiz de la función f(x) por el Metodo de la Bisección es x = {xr}")
```

```python
x = np.linspace(-2,2, 50)

plt.plot( x, f(x), "r-", label= "Función")
plt.plot(xr,f(xr),"bo" ,label= "F(xr)")
plt.legend()
plt.ylabel("F(x)")
plt.xlabel("(x)")

plt.title(r"$Grafica\ de\ la\ Función \ :\  b\cos(e^x-2) - a e^{x} + 2 = 0$")
plt.grid()
plt.show()
```

```python
#Metodo Newton-Rhapson

#a= 4
#b= 3
f= lambda x: b*np.cos(np.exp(x)-2)-a*np.exp(x)+2
df= lambda x: -b*np.exp(x)*np.sin(np.exp(x)-2)-a*np.exp(x)


def newton_rapson(f, df, x0, epsilon= 1E-9,Nmax=100):
  iter=0
  while(abs(f(x0))>epsilon  and iter<=Nmax ):
    x1= x0 - f(x0)/df(x0)
    x0=x1
    iter+=1
  return x0

x0=newton_rapson(f,df,1)
print(f"La raiz de la función f(x) por el Metodo de Newton-Rhapson es x = {x0}")
```
