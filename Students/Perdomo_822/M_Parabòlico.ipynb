```python
#Librerias:
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Condiciones iniciales:
#v_0 el cual ondica la velocidad inicial del problema.
# a  valor que equivale al ángulo

#Se define una función tal que de acuerdo a las ecuaciones cinemáticas nos de un tiempo de vuelo, posición en x & y del plano cartesiano.

def mov(v_0,a):
  tv= (2*v_0*np.sin(np.deg2rad(a)))/9.8

# Se define la linealidad del tiempo, y las variables x & "y" temporalmente dependientes.
  t = np.linspace(0, tv, 100)
  x =v_0* np.cos(np.deg2rad(a))* t
  y = (v_0 * np.sin(np.deg2rad(a))*t) - (9.8*t**2/2)

  return x , y

```

```python
x,y = mov(10,45)

plt.plot(x,y)
plt.xlabel("Desplazamiento en X")
plt.ylabel("Posición en Y ")
plt.title("Movimiento Parabólico con V")
plt.legend(['10 m/s & 45°'])

plt.show()
```
