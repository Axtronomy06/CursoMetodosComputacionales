```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def funcion(r,t,sigma,beta,rho):
  x,y,z = r

  dx = sigma*(y-x)

  dy = x * (rho-z) - y

  dz = (x * y) - (beta*z)


  return dx,dy,dz
```

```python
#Condiciones iniciales
sigma = 10.0
beta= 8/3
rho = 28

#Puntos iniciales:
r = [1.5, 1.5, 1.5]

#Tiemppo de integraci√≥n:
t=np.linspace(0,20,10000)

#Resolviendo el sistema a travez de odeint:
argums=(sigma,beta,rho)
sol = odeint(funcion,r,t,argums)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection="3d")
ax.plot(sol[:,0],sol[:,1],sol[:,2])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.draw()
plt.title('Atractor de Lorents')
plt.show()
```

```python
#Para los dem√°s puntos:
#Condiciones iniciales:
sigma = 10.0
beta= 8/3
rho = 25
argums=(sigma,beta,rho)

#Definici√≥n de R
r=[[1.5,1.5,1.5],[1,1,1],[1.001,1.001,1.001]]

#soluci√≥n para los 3 puntos:
sol1 = odeint(funcion,r[0],t,argums)
sol2 = odeint(funcion,r[1],t,argums)
sol3 = odeint(funcion,r[2],t,argums)

#Grafica:
fig = plt.figure()
ax = fig.add_subplot(projection="3d")
ax.plot(sol1[:,0],sol1[:,1],sol1[:,2], label='R1')
ax.plot(sol2[:,0],sol2[:,1],sol2[:,2], label='R2')
ax.plot(sol3[:,0],sol3[:,1],sol3[:,2], label='R3')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.draw()
plt.legend()
plt.title('Atractor de Lorents')
plt.show()
```
