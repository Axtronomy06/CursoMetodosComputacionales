```python
# Importar Librer√≠as:

import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
```

```python
#Defino la funci√≥n a encontrar raices:
f = lambda x: (x*np.exp(x))-1
x= np.linspace(-1.2,1.2, 100)

#Gr√°fica de la funci√≥n dada:

plt.plot(x,f(x))
plt.xlabel("X")
plt.ylabel('f(x)')
plt.title('$xe^x - 1$')
plt.grid()
plt.show()
```

```python
from matplotlib import legend
#Algoritmo para  bisecci√≥n:

#Puntos iniciales:
a=0.5
b=0.6
c=(a+b)/4   #Condici√≥n de punto medio optimizado al doble.

#Definici√≥n de epsilon:
x=10e-8

#Contador de iteraciones para valor menor a epsilon.
i=0

#Ciclo para hallar la raiz por bisecci√≥n:
while (abs(f(c))>= x and (f(a)*f(b))<0):
  c = (a+b)*0.5
  if f(c)>0:    #Condiciones de parada al ciclo:
    b = c
  else:
    a = c
  i += 1

#Imprime el numero de iteraciones del ciclo y el punnto d√≥nde se encuentra la raiz:
print("Numero de iteraciones: ",i)
print('El punto encontrado es:', (c , (f(c))))

#Gr√°fica:
x= np.linspace(0.2,0.8, 100)
plt.plot(x,f(x), label='$xe^x - 1$')
plt.plot(c,f(c), '*', label='{ $c,f (c)$ }')
plt.title("Resultado de Bisecci√≥n en $xe^x - 1$")
plt.xlabel('X')
plt.ylabel('$f(x)$')
plt.grid()
plt.legend()
plt.show()
```

```python
#Generalizaci√≥n de bisecci√≥n a funci√≥n:

def bisection(f,a,b,tolerancia, Nmax):
  a=a
  b=b
  c=(a+b)/4   #Condici√≥n de punto medio optimizado al doble.

  #Definici√≥n de epsilon:
  x= tolerancia

  #Contador de iteraciones para valor menor a epsilon.
  i=0

  #Ciclo para hallar la raiz por bisecci√≥n:
  while (abs(f(c))>= x and (f(a)*f(b))<0 and i<Nmax):
    c = (a+b)*0.5
    if f(c)>0:    #Condiciones de parada al ciclo:
      b = c
    else:
      a = c
    i += 1

  return c
```

```python
#Prueba de la funci√≥n:

bisection(f, -1,1,10e-7,100)
```

```python
#Comparaci√≥n con scipy:

print('Por libreria Scipy nos dio: ', optimize.bisect(f,-1,1))
print('Nuestra funci√≥n nos dio: ', bisection(f, -1,1,10e-7,100))
print('La diferencia es de: ',(optimize.bisect(f,-1,1)-bisection(f, -1,1,10e-7,100)))
```
