1. . Un objeto que cae verticalmente en el aire sujeto a la resistencia del aire y también a la fuerza de gravedad.

````python
#Importando librerías:
import numpy as np
import matplotlib.pyplot as plt
import scipy as sc
import pandas as pd
import sympy as sp
````

````python
# Crear función caida:

def caida(t):
 #Definicòn de constantes:
  m=0.25    #kg
  g=9.8     #Gravedad
  k=0.148
  s0 = 900

  s = s0 -(m*g/k)*t + (((m**2) * g/(k**2))*(1-(np.exp(-k*t/m))))

  return s

#Hallamos tv:
s = lambda t: s0 -(m*g/k)*t + (((m**2) * g/(k**2))*(1-(np.exp(-k*t/m))))
tv= sp.optimize.newton(s,60)
print(tv)

t = np.linspace(0,tv,1000)

#Gráfico:
plt.plot(t,caida(t))
plt.title('Caidalibre con fricción')
plt.xlabel('t')
plt.ylabel('S(t)')
plt.grid()
plt.show()

````

b.Tiempo de vuelo:

````python
#Constantes:
m=0.25    #kg
g=9.8     #Gravedad
k=0.148
s0 = 900

#Definiciòn de funciòn:
s = lambda t: s0 -((m*g/k)*t) + (((m**2) * g/(k**2))*(1-(np.exp(-k*t/m))))

#Volvemos las variables simbòlicas para la derivada:

ds = lambda t:  (((m) * g/(k)) * (np.exp((-k*t)/m))) - ((m*g)/k)

#Secante:
tv= sc.optimize.newton(s,60)
print('El tiempo por el metodo de la secante nos da un valor de: ', round(tv,5))

#Newton-Raphson:
tv1= sc.optimize.newton(s,60, ds)
print('El tiempo de vuelo por el metodo de Newton- Raphson es: ', round(tv1,5))

# método punto fijo:
#defino la funciòn a introdicir:
#h = lambda t: t - s
tv2 = sc.optimize.fixed_point(s,56)

print('El tiempo de vuelo por el metodo Punto fijo es: ', tv2)
````



c (30 Pts). Determine la el tiempo de caida para 10 valores de la masa entre 0.25Kg y 1Kg.

````python
#Constantes:
m= np.linspace(0.25, 1,10)   #kg
g=9.8     #Gravedad
k=0.148
s0 = 900
s=[]

for i in range(len(m)):
  s[i] = lambda t: s0 -((m[i]*g/k)*t) + (((m[i]**2) * g/(k**2))*(1-(np.exp(-k*t/m[i]))))

#Volvemos las variables simbòlicas para la derivada:

ds = lambda t:  (((m) * g/(k)) * (np.exp((-k*t)/m))) - ((m*g)/k)

#Secante:
tv= sc.optimize.newton(s,60)
print('El tiempo por el metodo de la secante nos da un valor de: ', round(tv,5))


````
