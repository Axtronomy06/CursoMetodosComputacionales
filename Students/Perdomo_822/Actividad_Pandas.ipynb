```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url="https://raw.githubusercontent.com/hernansalinas/autogradesMetodosComputacionales/main/Dataset/Temperature_Bogota_Barcelona.csv"

df = pd.read_csv(url)
df
```

```python
df["AvgTempCelcius"] = (df.AvgTemperature - 32)/1.8
```

```python
df
```

```python
tf = df.City == "Bogota"

df1=df[tf].copy()
df1
```

```python

plt.plot(df1["Year"],df1['AvgTempCelcius'] )
plt.title("Temperatura en funciòn del Año Bogota")
plt.xlabel("Años")
plt.ylabel("Temp. ºC")
```

```python
tf1 = df.City == "Barcelona"
df2=df[tf1].copy()
df2
```

```python
plt.plot(df2["Year"],df2['AvgTempCelcius'] )
plt.title("Temperatura en funciòn del Año Barcelona")
plt.xlabel("Años")
plt.ylabel("Temp. ºC")
```

```python
Maxbog = df1['AvgTempCelcius'].max()
Maxbar = df2['AvgTempCelcius'].max()

print("La temperatura max en Bogota es:", Maxbog)
print("La temperatura max en Barcelona es:", Maxbar)
```
