## Ecuaciones de movimiento

### Ecuaciones de movimiento del centro de masa

De acuerdo con la segunda ley de Newton, la ecuación de movimiento a lo largo de la vertical ($y$) puede ser escrita como:

\begin{equation}
m\frac{d^2y}{dt^2} = N - mg
\end{equation}

donde $m$ es la masa de la lápiz, $N$ es la fuerza normal y $g$ es la aceleración de la gravedad. Escribiendo las ecuaciones en términos del ángulo $\theta$, tenemos que:

\begin{equation}
-m\frac{l}{2}\cos \theta \left(\frac{d\theta}{dt} \right)^2 - m\frac{l}{2}\sin \theta \frac{d^2\theta}{dt^2} = N - mg
\end{equation}

### Ecuaciones de rotación del centro de masa

La sumatoria de momentos de fuerza se puede escribir como sigue:

\begin{equation}
I_c\frac{d^2\theta}{dt^2} = N\frac{l}{2} \sin \theta
\end{equation}

donde $I_c = \frac{1}{12}ml^2$ es el momento de inercia de una lápiz de masa $m$ y longitud $l$ respecto a un eje perpendicular a la lápiz en su centro de masa.

Reemplazando el valor de la normal en la ecuación anterior, obtenemos que la ecuación de movimiento para la posición angular está dada por:

\begin{equation}
\frac{d^2 \theta}{dt^2} = \left( \frac{\frac{2g}{l} - \left(\frac{d\theta}{dt}\right)^2\cos \theta }{\frac{1}{3} + \sin^2 \theta} \right)\sin \theta
\end{equation}

Podemos resolver la ecuación anterior si conocemos las condiciones iniciales: $\theta = \theta_0$ en $t = 0$ y $\frac{d\theta}{dt} = \omega_0 = 0$.

```python
# Librerias
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
#Condiciones iniciales
l = 0.1
g = 9.8
theta0 = np.pi/12
omega0 = 0
#thetaf= np.pi/2
ti =0
t_max = 10
delta_t = 0.01
n = int((t_max-ti)/delta_t)
t = np.linspace(ti,t_max,n)
y= l/2*np.cos(theta0)
```

\begin{equation}
\frac{d^2 \theta}{dt^2} = \left( \frac{\frac{2g}{l} - \left(\frac{d\theta}{dt}\right)^2\cos \theta }{\frac{1}{3} + \sin^2 \theta} \right)\sin \theta
\end{equation}

\begin{equation}
\frac{d\theta}{dt} = \omega
\end{equation}

\begin{equation}
\frac{d\omega}{dt} = \left( \frac{\frac{2g}{l} - \omega^2\cos \theta }{\frac{1}{3} + \sin^2 \theta} \right)\sin \theta
\end{equation}

```python
def lapiz(sistema, t, l, g):
  theta, omega= sistema
  dtheta_dt = omega
  domega_dt=(((2*g/l)-(omega**2)*np.cos(theta))/((1/3)+np.sin(theta)**2))*np.sin(theta)
  return dtheta_dt, domega_dt
```

```python
parametros=(theta0, omega0)
solucion = odeint(lapiz, parametros, t, args=(l, g))
print(solucion)
```

```python
#thetaf= np.pi/2
def theta_final(solucion,t):
  theta= solucion[:,0]
  indice=np.argmax(theta>=np.pi/2)
  if theta[indice]>=np.pi/2:
    return solucion[:indice+1], t[:indice+1]
  else:
    return solucion, t
```

2. Empleando `matplotlib`, construya tres gráficos con la información del punto anterior, asegurándose de que tengan las unidades respectivas.



```python
# Nueva solucion con thetaf= np.pi/2
solucion_f, t_f = theta_final(solucion,t)

theta = solucion_f[:, 0]
omega = solucion_f[:, 1]
y= (l/2)*np.cos(theta)

# Graficas
plt.figure(figsize=(8, 8))

#Posición angular
plt.subplot(3, 1, 1)
plt.plot(t_f, theta, label='Posición angular θ(t)',color='blue')
plt.axhline(y=np.pi/2, color='r', linestyle='--', label='θ = π/2')
plt.xlabel('Tiempo [s]')
plt.ylabel('θ(t) [rad]')
plt.legend()
plt.grid()

#Rapidez angular
plt.subplot(3, 1, 2)
plt.plot(t_f, omega, label='Rapidez angular ω(t)', color='magenta')
plt.xlabel('Tiempo [s]')
plt.ylabel('ω(t) [rad/s]')
plt.legend()
plt.grid()

#Posición del centro de masa
plt.subplot(3, 1, 3)
plt.plot(t_f, y, label='Posición del centro de masa y(t)', color='orange')
plt.xlabel('Tiempo [s]')
plt.ylabel('y(t) [m]')
plt.legend()
plt.grid()


plt.tight_layout()
plt.show()
```

6. El Tiempo que tarda el lápiz en caer $T$, puede ser deducido de la ecuación de conservación de la energía, el cual viene dado por la siguiente integral:  


\begin{equation}
T = \frac{1}{2}\sqrt{\frac{l}{g}} \int_{\theta_0}^{\pi/2} \sqrt{\frac{1/3 + \sin^2 \theta}{\cos \theta_0 - \cos \theta}} \, \mathrm{d}\theta
\end{equation}

Valide sus soluciones del tiempo de caída utilizando la siguiente ecuación.

```python
import numpy as np
from scipy.integrate import quad, trapezoid, simpson, dblquad
thetas=[np.pi/12,np.pi/6,np.pi/4,np.pi/3]
t=lambda theta, theta0: np.sqrt(((1/3)+np.sin(theta)**2)/(np.cos(theta0)-np.cos(theta)))
T=[]
for theta0 in thetas:
    c = (1/2) * np.sqrt(l/g)
    it, error = quad(t, theta0, np.pi/2, args=(theta0,))
    T.append(c * it)

print("Tiempos de caída:", T)
```

7. Finalmente, realice un gráfico de $T$ como función de $\theta_0$. ¿Qué puede concluir?

```python
plt.figure(figsize=(6,4))
plt.plot(thetas, T, 'o-', color='orange', label=r'$T(\theta_0)$')
plt.xlabel(r'Ángulo inicial [rad]')
plt.ylabel(r'Tiempo de caída [s]')
plt.title(r'$T$ en función de $\theta_0$')
plt.grid(True)
plt.legend()
plt.show()
```
