```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#1er punto leer el archivo csv
url ="https://raw.githubusercontent.com/hernansalinas/autogradesMetodosComputacionales/main/Dataset/Temperature_Bogota_Barcelona.csv"
df=pd.read_csv(url)
df
```

```python
#segundo punto agregar una columna
df["AvgTempCelcius"]=((df.AvgTemperature)-32)/1.8
df
```

```python
# 3er Realizar un gráfico de AvgTempCelcius como función del año para la ciudad de Bogotá.
# se realiza el filtro para la ciudad de bogota

tf= df.City =="Bogota"
df1= df[tf].copy()
df1
```

```python
# se construye el grafico de temperatura en funcion del tiempo

plt.figure()
plt.plot(df1.Year,df1.AvgTempCelcius,"Blue")
plt.xlabel("Año")
plt.ylabel("Temperatura en °C")
plt.title("Temperatura de Bogota")
plt.ylim(8, 15)
plt.xlim(1994, 2021)

plt.show()


```

```python
# 4to Realizar un gráfico de AvgTempCelcius como función del año para la ciudad de Barcelona
# se realiza primero el filtro para la ciudad de barcelona
tf= df.City =="Barcelona"
df2= df[tf].copy()
df2
```

```python
# se construye el grafico de temperatura en funcion del tiempo
plt.figure()
plt.plot(df2.Year,df2.AvgTempCelcius,"Red")
plt.xlabel("Año")
plt.ylabel("Temperatura en °C")
plt.title("Temperatura de Barcelona")
plt.ylim(12, 18)
plt.xlim(1994, 2021)

plt.show()
```

```python
# 5to Determine la mayor temperatura para Bogotá y Barcelona.
m_bo= df1["AvgTempCelcius"].max()
m_ba= df2["AvgTempCelcius"].max()

mi_bo= df1["AvgTempCelcius"].min()
mi_ba= df2["AvgTempCelcius"].min()

print(f"La máxima temperatura registrada en Bogotá fue de {m_bo:.2f}°C")
print(f"La máxima temperatura registrada en Barcelona fue de {m_ba:.2f}°C")
print(f"La minima temperatura registrada en Bogotá fue de {mi_bo:.2f}°C")
print(f"La minima temperatura registrada en Barcelona fue de {mi_ba:.2f}°C")
```
