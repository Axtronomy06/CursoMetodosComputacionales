# 3. Movimiento de proyectiles en presencia de viento.

Supongamos que un proyectil es lanzado desde el origen con una rapidez $v=vo( \cos \theta , \sin \theta $. Supongamos además que el efecto del viento sobre el proyectil es el de una fuerza proporcional a la velocidad relativa entre el viento y el proyectil. Teniendo presente que si u es la velocidad del viento y $\alpha$ el ángulo que hace dirección de esta velocidad con el eje x, se puede entrontrar las siguientes ecuaciones diferenciales, asociada al estado de movimiento del [proyectil](https://iopscience.iop.org/article/10.1088/0143-0807/36/2/025016/meta):


\begin{equation}
m\frac{dv_x}{dt} = -mb(v_x-u\cos \alpha )
\end{equation}

\begin{equation}
m\frac{dv_y}{dt} = -mg-mb(v_x-u\sin \alpha ),
\end{equation}

donde b es el párametro de rozamiento. La soluciones a las ecuaciones anteriores vienen dadas por:

Posición en x
\begin{equation}
x=(u\cos \alpha) t + \frac{1}{b} (v_0\cos \theta_0 -u \cos\alpha)(1-\mathrm{exp}(-bt))
\end{equation}

Posicion en y
\begin{equation}
y= \frac{1}{b}\left(\frac{g}{b}+v_0 \sin \theta_0 -u\sin\alpha \right)(1-\mathrm{exp}(-bt))-\left( \frac{g}{b}-u\sin\alpha\right)t
\end{equation}

# 4. Fixed point methods.

When a new planet is discovered, there are different methods to estimate its physical properties. Many times is only possible to estimate either the planet mass or the planet radius and the other property has to be predicted through computer modelling.

If one has the planet mass, a very rough way to estimate its radius is to assume certain composition (mean density) and a homogeneous distribution (a very bad assumption!). For example, for the planet [Gliese 832c](http://es.wikipedia.org/wiki/Gliese_832_c) with a mass $M= 5.40 M_{\oplus}$, if we assume an earth-like composition, i.e. $\bar \rho_{\oplus} = 5520\ kg/m^3$, we obtain:

$$R_{g832c} = \left( \frac{3 M_{g832c}}{ 4 \pi \bar\rho_{\oplus} } \right)^{1/3} \approx 1.75 R_{\oplus}$$

That would be the planet radius if the composition where exactly equal to earth's.

A more realistic approach is assuming an internal one-layer density profile like:

$$\rho(r) = \rho_0 \exp\left( -\frac{r}{L} \right)$$

where $\rho_0$ is the density at planet centre and $L$ is a characteristic lenght depending on the composition. From numerical models of planet interiors, the estimated parameters for a planet of are $M= 5.40 M_{\oplus}$ are approximately $\rho_0 = 18000\ kg/m^3$ and $L = 6500\ km$.

Integrating over the planet volume, we obtain the total mass as

$$M = 4\pi \int_0^R \rho(r)r^2dr$$

This is a function of the mass in terms of the planet radius.

Solving the equation $M(R) = M_{g832c}$ it would be possible to find a more realistic planet radius. However when using numerical models, it is not possible to approach the solution from the left side as a negative mass makes no sense.

# Metodo de la Biseccion:
1. Inicializar a y b con los límites iniciales del intervalo [a, b].
2. Mientras (b - a) sea mayor que la tolerancia deseada:
     * Calcular el punto medio c = (a + b) / 2.
     * Evaluar la función f en los puntos a, b y c.
     * Si f(a) y f(c) tienen diferente signo, asignar b = c.
     * Si f(b) y f(c) tienen diferente signo, asignar a = c.
3. El resultado aproximado es (a + b) / 2, que es la raíz buscada.

La principal diferencia entre el Método de Newton-Raphson y el Método de la Secante radica en la forma en que se calcula la derivada de la función:

- Método de Newton-Raphson:
  - Requiere conocer la derivada de la función.

- Método de la Secante:
  - No requiere conocer la derivada de la función.


# Metodo de Punto Fijo

1. Inicializar x con una suposición inicial.
2. Mientras no se alcance una aproximación suficiente:

     - Calcular el siguiente valor $x_{n}$ utilizando la función $g(x_n) = x_n+1$.
     - Evaluar la diferencia absoluta entre x_n y x.
     - Si la diferencia es menor que la tolerancia deseada, detenerse.
     - Asignar x = x_n y continuar desde el paso 2.
3. El resultado aproximado es x, que es la raíz buscada.

````python
# a) Pseudocódigo para el método de la bisección:
Función bisección(f, a, b, tolerancia):
    Mientras (b - a) / 2 > tolerancia:
        c = (a + b) / 2
        si f(c) es cercano a 0:
            devolver c como aproximación de la raíz
        si no:
            si f(a) * f(c) < 0:
                b = c
            sino:
                a = c
    devolver c como aproximación de la raíz

#Este pseudocódigo representa una función bisección que toma una función f, un intervalo [a, b], y una tolerancia como entrada, y devuelve una aproximación de la raíz de la función f dentro del intervalo [a, b].


# d) Pseudocódigo para el método del Punto Fijo:
Función puntoFijo(g, x0, tolerancia, maxIteraciones):
    x = x0
    iteraciones = 0
    Mientras iteraciones < maxIteraciones:
        x_nuevo = g(x)
        si abs(x_nuevo - x) < tolerancia:
            devolver x_nuevo como aproximación de la raíz
        x = x_nuevo
        iteraciones = iteraciones + 1
    devolver "No se alcanzó convergencia después de maxIteraciones iteraciones"

````

````python
#punto 2
import numpy as np

# Definición de la función f(x)
def f(x):
    return np.exp(x) - 2 - np.cos(np.exp(x) - 2)

# Derivada de f(x)
def f_prime(x):
    return np.exp(x) + np.sin(np.exp(x) - 2) * np.exp(x)

# Método de la Bisección
def bisection(f, a, b, tolerance, max_iterations):
    iterations = 0
    while iterations < max_iterations:
        c = (a + b) / 2
        if abs(f(c)) < tolerance:
            return c
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
        iterations += 1
    return "No se alcanzó convergencia después de max_iterations iteraciones"

# Método de Newton-Raphson
def newton_raphson(f, f_prime, x0, tolerance, max_iterations):
    x = x0
    iterations = 0
    while iterations < max_iterations:
        x_new = x - f(x) / f_prime(x)
        if abs(x_new - x) < tolerance:
            return x_new
        x = x_new
        iterations += 1
    return "No se alcanzó convergencia después de max_iterations iteraciones"

# Método del Punto Fijo
def punto_fijo(g, x0, tolerance, max_iterations):
    x = x0
    iterations = 0
    while iterations < max_iterations:
        x_new = g(x)
        if abs(x_new - x) < tolerance:
            return x_new
        x = x_new
        iterations += 1
    return "No se alcanzó convergencia después de max_iterations iteraciones"

# Definir intervalo para el método de la bisección
a = 0
b = 2

# Valor inicial para los métodos de Newton-Raphson y Punto Fijo
x0 = 0.5

# Tolerancia y máximo de iteraciones
tolerance = 1e-6
max_iterations = 1000

# Aplicar los métodos
resultado_biseccion = bisection(f, a, b, tolerance, max_iterations)
resultado_newton = newton_raphson(f, f_prime, x0, tolerance, max_iterations)
resultado_punto_fijo = punto_fijo(lambda x: 2 + np.cos(np.exp(x) - 2), x0, tolerance, max_iterations)

print("Resultado de la Bisección:", resultado_biseccion)
print("Resultado de Newton-Raphson:", resultado_newton)
print("Resultado del Punto Fijo:", resultado_punto_fijo)
````

````python
import numpy as np
from scipy.optimize import bisect, newton, fixed_point

# Definición de la función f(x)
def f(x):
    return np.exp(x) - 2 - np.cos(np.exp(x) - 2)

# Intervalo para el método de la bisección
a = 0
b = 2

# Valor inicial para los métodos de Newton y Punto Fijo
x0 = 0.5

# Tolerancia
tolerance = 1e-6

# Encontrar la raíz utilizando los métodos de SciPy

# Método de la Bisección
resultado_biseccion = bisect(f, a, b, xtol=tolerance)
print("Resultado de la Bisección:", resultado_biseccion)

# Método de Newton-Raphson
resultado_newton = newton(f, x0, tol=tolerance)
print("Resultado de Newton-Raphson:", resultado_newton)

# Método del Punto Fijo
resultado_punto_fijo = fixed_point(lambda x: 2 + np.cos(np.exp(x) - 2), x0, xtol=tolerance)
print("Resultado del Punto Fijo:", resultado_punto_fijo)
````

````python
# punto 3 movimiento de proyectiles

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy as sc

def pos_x(t,tetha ,alpha, v0 , b , u , g):

  T1 = u*(np.cos(alpha))*t
  T2 = (1/b)*(v0*np.cos(tetha))- u*np.cos(alpha)
  T3 = (1-np.exp(-b*t))
  x = T1 + T2 +T3

  return x

def  pos_y(t,tetha, alpha, v0 , b , u ,g):

  T1 = (1/b)*((g/b) + v0*np.sin(tetha) - u*(np.sin(alpha)))
  T2 = (1-np.exp(-b*t))
  T3 = (g/b - u*np.sin(alpha))*t
  y = T1*T2 - T3

  return y

def trayectoria(pos_y , x0 , params):

  tv = sc.optimize.newton(pos_y, 12, args=params)
  tiempo = np.linspace(0,tv,100)
  x = pos_x(tiempo, *params)
  y = pos_y(tiempo, *params)

  dic = {"x":x , "y":y , "tiempo":tiempo}
  df = pd.DataFrame(dic)


  return df
````
