# Diagonalizaciòn

Una matriz cuadrada **A** es diagonalizable si y solo si cada autovalor $\lambda$ de multiplicidad **m** , produce exactamente **m** utovectores básicos; es decir, si y solo si la solución general del sistema $(A-I\lambda)\vec{x}=0$  tiene exactamente **m** parámetros.

Para diagonalizar una matriz **A**, se necesita encontrar una matriz invertible **P** tal que $PAP^{-1}$
 sea una matriz diagonal **D**. Esto significa que las columnas de **P**
 son los autovectores de **A**  y los elementos de la diagonal de **D**
 son los autovalores correspondientes de **A**

Considere la ecuación cuadrática $x^2-4xy+3y^2=6$, esta ecuación cuadrática se corresponde a una hiperbola, la cual puede ser escrita en forma matricial como sigue:


$A\vec{v} \cdot \vec{v}=6$, donde $\vec{v}=(x, y)$ y A es una matrix simetrica,
\begin{equation}
A=\begin{bmatrix}
1 & -2 \\
-2 & 3 \\
\end{bmatrix}
\end{equation}

1. Realice un grafico que represente la hyperbola
* Diagonalice A, para ello construya la matrix Q de autovectores ortoganales
tal que $D=Q^{T}AQ$
, en este caso se dice que A es diagonalizable ortogonalmente, donde $A=QDQ^T$
.

* Aplique el resultado de la siguiente demostración para encontrar una nueva hyperbola en una nueva base
\begin{equation}
QDQ^T \vec{v} \cdot \vec{v} = d
\end{equation}
\begin{equation}
Q(DQ^T\vec{v}) \cdot \vec{v} = (DQ^T\vec{v}) \cdot (Q^T\vec{v})=d
\end{equation}
\begin{equation}
(DQ^T\vec{v}) \cdot (Q^T\vec{v})=d
\end{equation}

Sea $u=Q^T\vec{v}$

Luego:

$\begin{equation}
(D\vec{u}) \cdot \vec{u}=d
\end{equation}$


Donde $\vec{u}=(x', y')$
, tal que $D\vec{u} \cdot \vec{u}=d$
 es una nueva hyperbola

```python
#diagonalizacion de la matriz 1



# Definir la matriz
A = np.array([[1, -1, 4],
              [3, 2, -1],
              [2, 1, -1]])

# Calcular los valores y vectores propios
eigenvalues, eigenvectors = np.linalg.eig(A)

# Crear la matriz de vectores propios inversa
P = eigenvectors

# Crear la matriz diagonal con los valores propios en la diagonal
D = np.diag(eigenvalues)

# Calcular la matriz inversa de la matriz de vectores propios
P_inv = np.linalg.inv(P)

# Verificar que P * D * P_inv sea igual a la matriz original A (debería serlo)
reconstructed_A = np.dot(np.dot(P, D), P_inv)

print("Matriz original A:")
print(A)

print("\nMatriz de vectores propios P:")
print(P)

print("\nMatriz diagonal D:")
print(D)

print("\nMatriz inversa de P (P_inv):")
print(P_inv)

print("\nMatriz reconstruida P * D * P_inv:")
print(reconstructed_A)
```

```python
# Grafico de la Hiperbola
x = np.linspace(-10,10,400)
y = np.linspace(-10,10,400)

f = lambda x,y: x**2 - 4*x*y + 3*y**2 - 6

X, Y = np.meshgrid(x, y)

Z = f(X, Y)

plt.figure(figsize = (7,7))
plt.contour(X, Y, Z,levels=[0] ,colors='r')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title('Gráfica de la Hipérbola: $x^2 - 4xy + 3y^2 = 6$')
plt.show()
```
