```python
A= np.matrix([[2,1,-1,5],[4,1,1,1],[6,2,1,4]])
A
```

```python
B=A[0]*3
B
```

```python
def intercambio(M,i,j):
  B=M.copy()
  B[i]=M[j]
  B[j]=M[i]
  return B
```

```python
def mul_escalar(M,i,x):
  B=M.copy()
  B[i]=x*M[i]
  return B
```

```python
def suma(M,i,j,x):
  B=M.copy()
  B[j]= x*M[i]+M[j]
  return B
```
