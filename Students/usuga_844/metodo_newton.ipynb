```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy import optimize
```

```python

f= lambda x: ((x**2) -1)-10
d= lambda x: 2*x
x=np.linspace(-10,20,100)
```

```python

x0= 1
e=1E-5
N=0
while (abs(f(x0) >= e)) or N<=1000:
  x0= x0- (f(x0)/d(x0))
  N +=1

print(x0)


plt.figure()
plt.plot(x, f(x))
plt.xlabel('x')
plt.ylabel('f(x)')
plt.ylim(-20,15)
plt.grid()

plt.show()

```

```python
def newton_raphson(x0 , e , f , d , Nmax):
    n = 0

    while (abs(f(x0))>=e) &  (n <= Nmax):
       x0= x0- (f(x0)/d(x0))
       n +=1

    return x0 , n
x0 = newton_raphson(1 , e , f , d , 1000)
print(x0)
```
