dada una funcion calcular su derivada a lo largo de todo un dominio
utilizando el comando misc.derivative de scipy

```python
import numpy as np
import scipy as sp
import sympy as sy
import matplotlib.pyplot as plt
from scipy import misc
```

```python
epsilon=0.2
f= lambda x,a: a*np.sin(x)+np.tan(x)**2
x= np.linspace(0,1,100)

plt.figure()
plt.grid()
plt.plot(x,f(x,1))
plt.show()
```

```python
a=1
g= misc.derivative(f, x ,dx=1E-5, args=(a, ))
g
```

```python
plt.figure()
plt.plot(x,g)
plt.grid()
plt.show()
```
