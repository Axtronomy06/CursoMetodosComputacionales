```python

import numpy as np
import matplotlib.pylab as plt
import sympy as sy
```

```python

x_ = np.array([2,5,6,8])
y_ = np.array([4,1,2,5])
x = sy.symbols("x")
n = len(x_)
```

```python
n= len(x_)
l0=1
j=0
for i in range (0,n):
  if(i!=j):
    l0*= (x-x_[i])/(x_[j]-x_[i])
l0
```

```python
def Lagrange(j,n):
  l0 = 1
  for i in range(0,n):
    if (i != j):
      l0 *= (x-x_[i])/(x_[j]-x_[i])
  return l0
```

```python
def p(x,x_,y_):
  poly=0
  for i in range (0,len(y_)):
    poly += y_[i]* Lagrange(i,n)

  return poly
```

```python
sy.simplify(p(x,x_,y_))
```

Para determinar la interpolación lineal que pasa por los puntos (2,4) y (5,1), podemos utilizar las fórmulas de la pendiente **m* y la ordenada al origen *b** de una línea recta:

$m = \frac{1 - 4}{5 - 2} = -1$

$b = 4 - (-1) \cdot 2 = 6$

Por lo tanto, la ecuación de interpolación lineal es:

$y = -x + 6$

Esta es la ecuación de la línea que pasa por los puntos (2,4) y (5,1) utilizando la interpolación lineal.
