```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x: np.exp(-x) - x
g = lambda x: np.exp(-x)
x = np.linspace(0,2,10)

plt.plot(x,g(x))
plt.grid()
plt.show
```

```python
def encontrar_punto_fijo(f, g, po, e):

    n = 0

    while n <= 1000:
        p1 = g(po)
        if abs(p1 - po) < e:
            return p1
        po = p1
        n += 1

    raise Exception("El método no convergió después de 1000 iteraciones.")


punto_fijo = encontrar_punto_fijo(f,g, 1.0, 1e-6)
print("Punto fijo encontrado:", punto_fijo)
```
