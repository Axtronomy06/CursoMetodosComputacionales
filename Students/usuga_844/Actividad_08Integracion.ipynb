```python
import numpy as np
import scipy as sp
import sympy as sy
import matplotlib.pyplot as plt
from scipy import misc
```

```python
def regla_del_trapecio(f, a, b, n):

    h = (b - a) / n


    suma = f(a) + f(b)


    for i in range(1, n):
        x_i = a + i * h
        suma += 2 * f(x_i)


    A = (h / 2) * suma

    return A


g= lambda x:x**2


a = 0
b = 2


n = 1000


integral = regla_del_trapecio(g, a, b, n)

print(f"Resultado de la integral: {integral}")
```

```python
def integral_trapz(f,a,b,N):
 x = np.linspace(a,b,int(N))
 h = (b-a)/N
 A = (f(x[0]) + f(x[-1]) + 2*f(x[1:-1]).sum())*h*0.5
 return A
```

```python
f = lambda x : x**2
N = np.linspace(10,1000,100)
areas = np.zeros(len(N))

for i,n in enumerate(N):
 areas[i] = integral_trapz(f,0,1,n)
plt.plot(N,areas)
```

**regla de simpson simple**

\begin{equation}
{\displaystyle \int _{a}^{b}f(x)\,dx\approx {\frac {b-a}{6}}\left[f(a)+4f\left({\frac {a+b}{2}}\right)+f(b)\right]}.
\end{equation}

**regla de simpson compuesta**
\begin{equation}
{\displaystyle \int _{a}^{b}f(x)\,dx\approx {\frac {h}{3}}{\bigg [}f(a)+2\sum _{j=1}^{{\frac {n}{2}}-1}f(x_{2j})+4\sum _{j=1}^{\frac {n}{2}}f(x_{2j-1})+f(b){\bigg ]-{\displaystyle (b-a)\,{\frac {h^{4}}{180}}\,\left(f^{(4)}(\mu )\right).}}}
\end{equation}
