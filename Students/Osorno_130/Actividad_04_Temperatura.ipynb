Para los siguientes datos de temperatura media como función del tiempo de los últimos años de Bogotá y Barcelona (Temperature_Bogota_Barcelona) realice lo siguiente:

1. Leer el dataset desde la url, con el comando pd.read_csv.
2. Adicionar una nueva columna AvgTempCelcius con la temperatura en grados celcius. $ C=(F-32)/1.8$
3. Realizar un gráfico de AvgTempCelcius como función del año para la ciudad de Bogotá.
4. Realizar un gráfico de AvgTempCelcius como función del año para la ciudad de Barcelona.
5. Determine la mayor temperatura para Bogotá y Barcelona.

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/hernansalinas/autogradesMetodosComputacionales/main/Dataset/Temperature_Bogota_Barcelona.csv"
df=pd.read_csv(url)
df
```

```python
df["AvgTempCelcius"] = (df.AvgTemperature - 32)/1.8
df
```

```python
tf = df.City=="Bogota"
bogota = df[tf].copy()
tf = df.City=="Bogota"
barcelona = df[tf==False].copy()
```

```python
maxBog = bogota[bogota.AvgTempCelcius == bogota.AvgTempCelcius.max()]
maxBar = barcelona[barcelona.AvgTempCelcius == barcelona.AvgTempCelcius.max()]
print("Máxima temperatura de Bogotá", maxBog.AvgTempCelcius.values[0])
plt.figure()
plt.plot(bogota.Year, bogota.AvgTempCelcius)
plt.plot(maxBog.Year, maxBog.AvgTempCelcius, "ro")
plt.title("Temperatura Ciudad Bogotá")
plt.xlabel("Año")
plt.ylabel("Temperatura [C]")
plt.show()
```

```python
maxBar = barcelona[barcelona.AvgTempCelcius == barcelona.AvgTempCelcius.max()]
print("Máxima temperatura de Barcelona", maxBar.AvgTempCelcius.values[0])
plt.figure()
plt.plot(barcelona.Year, barcelona.AvgTempCelcius)
plt.plot(maxBar.Year, maxBar.AvgTempCelcius, "ro")
plt.title("Temperatura Ciudad Barcelona")
plt.xlabel("Año")
plt.ylabel("Temperatura [C]")
plt.show()
```
