#Integraci贸n regla de simpson

Para la integraci贸n de simpson, se considera la aproximaci贸n de la funci贸n $f(x)$ como un polinomio de orden 2, en un rango $[a,b]$ y el valor medio entre estos $m=\frac{a+b}{2}$ y usando el polinomio de lagrange para 3 puntos

$f(x) \approx P(x) = f(a)\frac{(x-m)}{(a-m)} \frac{(x-b)}{(a-b)} +  f(m)\frac{(x-a)}{(m-a)} \frac{(x-b)}{(m-b)} + f(b)\frac{(x-a)}{(b-a)} \frac{(x-m)}{(b-m)}$

$P(x) = f(a)\frac{x^2 -x(b+m) + bm}{(a-m)(a-b)}  +  f(m)\frac{x^2 -x(a+b) + ab}{(m-a)(m-b)} + f(b)\frac{x^2 -x(a+m) + am}{(b-a)(b-m)}$


$P(x) = f(a)\frac{x^2 -x(b+\frac{a+b}{2}) + b\frac{a+b}{2}}{(a-\frac{a+b}{2})(a-b)}  +  f(\frac{a+b}{2})\frac{x^2 -x(a+b) + ab}{(\frac{a+b}{2}-a)(\frac{a+b}{2}-b)} + f(b)\frac{x^2 -x(a+\frac{a+b}{2}) + a\frac{a+b}{2}}{(b-a)(b-\frac{a+b}{2})}$


$P(x) = f(a)\frac{x^2 -x(\frac{a+3b}{2}) + b\frac{a+b}{2}}{(\frac{a-b}{2})(a-b)}  +  f(\frac{a+b}{2})\frac{x^2 -x(a+b) + ab}{(\frac{b-a}{2})(\frac{a-b}{2})} + f(b)\frac{x^2 -x(\frac{3a+b}{2}) + a\frac{a+b}{2}}{(b-a)(\frac{b-a}{2})}$

$P(x) = f(a)\frac{2x^2 -x(a+3b) + b(a+b)}{(a-b)(a-b)}  +  4f(\frac{a+b}{2})\frac{x^2 -x(a+b) + ab}{(b-a)(a-b)} + f(b)\frac{2x^2 -x(3a+b) + a(a+b)}{(b-a)(b-a)}$


$P(x) = \frac{f(a)}{(a-b)(a-b)}(2x^2 -x(a+3b) + b(a+b))  +  4\frac{f(\frac{a+b}{2})}{(b-a)(a-b)}(x^2 -x(a+b) + ab) + \frac{f(b)}{(b-a)(b-a)}(2x^2 -x(3a+b) + a(a+b))$


\begin{equation}
\int_{a}^{b}f(x)dx \approx \int_{a}^{b}P(x)dx = A
\end{equation}



$A = \int_{a}^{b}\left(\frac{f(a)}{(a-b)(a-b)}(2x^2 -x(a+3b) + b(a+b))  +  4\frac{f(\frac{a+b}{2})}{(b-a)(a-b)}(x^2 -x(a+b) + ab) + \frac{f(b)}{(b-a)(b-a)}(2x^2 -x(3a+b) + a(a+b))\right) dx$


$A=\frac{f(a)}{(a-b)(a-b)}(2\frac{x^3}{3} -\frac{x^2}{2}(a+3b) + xb(a+b))\Big|_a^b  +  4\frac{f(\frac{a+b}{2})}{(b-a)(a-b)}(\frac{x^3}{3} -\frac{x^2}{2}(a+b) + xab)\Big|_a^b + \frac{f(b)}{(b-a)(b-a)}(2\frac{x^3}{3} -\frac{x^2}{2}(3a+b) + xa(a+b))\Big|_a^b$


$A = \frac{f(a)}{(a-b)(a-b)}(2\frac{(b^3 - a^3)}{3} -\frac{(b^2 - a^2)}{2}(a+3b) + (b-a)b(a+b))  +  4\frac{f(\frac{a+b}{2})}{(b-a)(a-b)}(\frac{(b^3 - a^3)}{3} -\frac{(b^2 - a^2)}{2}(a+b) + (b-a)ab) + \frac{f(b)}{(b-a)(b-a)}(2\frac{(b^3 - a^3)}{3} -\frac{(b^2 - a^2)}{2}(3a+b) + (b-a)a(a+b))$

$A = \frac{f(a)}{(a-b)(a-b)}(2\frac{(b - a)(a^2 + ab + b^2)}{3} -\frac{(b - a)(b + a)}{2}(a+3b) + b(b-a)(a+b))  +  4\frac{f(\frac{a+b}{2})}{(b-a)(a-b)}(\frac{(b - a)(a^2 + ab + b^2)}{3} -\frac{(b - a)(b + a)}{2}(a+b) + (b-a)ab) + \frac{f(b)}{(b-a)(b-a)}(2\frac{(b - a)(a^2 + ab + b^2)}{3} -\frac{(b - a)(b + a)}{2}(3a+b) + a(b-a)(a+b))$



$A =(b-a)\left(\frac{f(a)}{(a-b)^2}(2\frac{(a^2 + ab + b^2)}{3} -\frac{(b + a)}{2}(a+3b) + b(a+b))  +  4\frac{f(\frac{a+b}{2})}{-(a-b)^2}(\frac{(a^2 + ab + b^2)}{3} -\frac{(b + a)}{2}(a+b) + ab) + \frac{f(b)}{(b-a)^2}(2\frac{(a^2 + ab + b^2)}{3} -\frac{(b + a)}{2}(3a+b) + a (a+b))\right)$


$A = (b-a)\left(\frac{f(a)}{(a-b)^2}(\frac{(a-b)^2}{6})  +  4\frac{f(\frac{a+b}{2})}{-(a-b)^2}(\frac{-(a-b)^2}{6}) + \frac{f(b)}{(a-b)^2}(\frac{(a-b)^2}{6})\right)$

$A = (b-a)\left(\frac{f(a)}{6} +  4\frac{f(\frac{a+b}{2})}{6} + \frac{f(b)}{6}\right)$

$A = \frac{(b-a)}{6}\left(f(a) +  4f(\frac{a+b}{2}) + f(b)\right)$


Por lo que la formula final de la regla de Simpson:

$$\int_{a}^{b}f(x)dx \approx \int_{a}^{b}P(x)dx = \frac{(b-a)}{6}\left(f(a) +  4f(\frac{a+b}{2}) + f(b)\right)$$
