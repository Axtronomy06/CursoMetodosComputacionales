1. Disenar un programa que permita sumar dos matrices.


```python
def sumaMatrices(A,B):
  dimA = np.shape(A)
  dimB = np.shape(B)
  if((len(dimA) <= 1) | (len(dimB) <= 1)):
    print("Los parámetro A y B deben ser bidimensionales, ejemplo: 4x3, 4x1, 1x4, 3x3")
    return
  if((dimA[0] != dimB[0]) | (dimA[1] != dimB[1])):
    print("El tamaño de A y B deben ser iguales")
    return
  S = np.zeros(dimA)
  for i in range(0,dimA[0]):
    for j in range(0,dimA[1]):
      S[i,j] =  A[i,j] +  B[i,j]
  return S

s = sumaMatrices(M,N)
print(s)

```

3. Determinar la transpuesta de una matrix

```python
def transpuesta(A):
  dimA = np.shape(A)
  if(len(dimA) <= 1):
    print("La matriz A debe ser bidimensional, ejemplo: 4x3, 4x1, 1x4, 3x3")
    return
  t = np.zeros((dimA[1],dimA[0]))
  for i in range(0,dimA[0]):
    for j in range(0,dimA[1]):
      t[j,i] = A[i,j]
  return t

Mt = transpuesta(M)
print(M)
print(Mt)
```

4.La matriz de rotación en 2D. Es una matriz que se usa para rotar puntos en el plano xy en sentido antihorario alrededor del origen. La matriz tiene la siguiente forma:

  \begin{equation}
    R(\theta) = \begin{bmatrix}
    \cos \theta & -\sin \theta \\
    \sin \theta & \cos \theta \\
    \end{bmatrix}
  \end{equation}


Construir un programa en python que permita realizar la rotación de un vector un ángulo theta, Mostrar la representación vectorial en el plano indicando la rotacion a realizar.


Pruebe su solución para el vector $\\vec{v}=(1,1)$ rotado 20 grados en sentido antihorario. Muestre el vector antes de la rotación y después de la rotación

```python
import matplotlib.pyplot as plt
```

```python
v = np.array([[1],[1]])
angulo = 20

#theta en grados
def rotar(theta, v):
  t = np.deg2rad(theta)
  v = np.asmatrix(v)
  R = np.asmatrix(np.array([[np.cos(t), -np.sin(t)],[np.sin(t), np.cos(t)]]))
  rot = R@v
  return rot

vrot = rotar(angulo,v)
vrot
```

```python
plt.figure(figsize=(10,10))
plt.gca().set_aspect('equal')
plt.grid(linestyle="--")
plt.quiver(*v,angles="xy", scale_units="xy", scale=1, color="r", label="vector")
plt.quiver(*vrot,angles="xy", scale_units="xy", scale=1,label="vector rotado")
plt.legend()
plt.title("Gráfica del vector")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-2, 2)
plt.ylim(-2, 2)
```
