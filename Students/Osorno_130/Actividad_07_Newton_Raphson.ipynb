#Método de la Newton

1. Diseñar un programa que permita calcular los ceros de una función empleando el metodo Newton-Raphson. La función deberá recibir como parámetros la funcion f, la derivada, la tolerancia y el punto de evaluación

```python
import matplotlib.pyplot as plt
from scipy import optimize as op
import numpy as np

f = lambda x: ((x-1)**2)- 5
df = lambda x: 2*(x-1)
x = np.linspace(-3,5,100)
y = f(x)
plt.figure
plt.title("Función $(x-1)^2-5$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x,y)
plt.grid()

```

```python
def newton(x0,f,df,epsilon=1E-12, nMax=1000):
  x = x0
  n = 0
  while (abs(f(x)) > epsilon) & (n <= nMax):
    try:
      x = x -(f(x)/df(x))
    except ZeroDivisionError as e:
      if x <= x0:
        x = x - 1E-8
      if x > x0:
        x = x + 1E-8
    n += 1
  return x


x0 = 2
epsilon = 1E-12
nMax = 100
x = newton(x0,f,df)
fx = f(x)
print("x = ", x , ", f(x) = ", fx)
```

2. Comparar su resultado empleando la libreria optmize de scipy


```python
opx = op.newton(f,x0, fprime=df, tol = epsilon)
print(opx)

print("Diferencia entre el valor de x obtenido por el método propio y el método de optimaze",abs(x-opx))
```
