```python
import numpy as np
import matplotlib.pyplot as plt

f =  lambda x: (x-1)**2
epsilon = 1E-10
def puntoFijo (f, p0, epsilon=1E-7, nMax=1000):
  g =  lambda x: x - f(x)
  x = np.linspace(0.5,2,100)
  yr = x
  yf = f(x)
  yg = g(x)
  plt.figure
  plt.plot(x,yr,label="recta y=x")
  plt.plot(x,yf,label="f(x)")
  plt.plot(x,yg,label="g(x)")
  plt.legend(bbox_to_anchor=(1.35, 1))
  plt.grid()
  plt.gca().set_aspect('equal')
  plt.show()
  pi = p0
  print("P0 = ", p0, ", g(P0) = ",g(p0))
  pi_1 = g(pi)
  iter = 1
  while(abs(pi_1 - pi)>epsilon)& (abs(pi_1 - pi)<=10) & (iter<nMax):
    #print(f"P{iter} = ", pi_1, f", g(P{iter}) = ",g(pi_1))
    pi = pi_1
    pi_1 = g(pi)
    iter +=1
    print(iter)
  print("P"+str(iter)+" = ", pi_1, ", g(P"+str(iter)+") = ",g(pi_1))
puntoFijo(f,3, epsilon=epsilon)
```

```python
from scipy import optimize as op
g =  lambda x: x - f(x)
xFixed = op.fixed_point(g,3,xtol=epsilon)
print("x = ",xFixed,", f(x) = ", f(xFixed))
```
