```python
def bisection(a,b,f, epsilon):
  c = (a+b)/2
  fc = f(c)
  iter = 0
  maxIter = 1000
  while (abs(fc) > epsilon) & (f(a)*f(b) < 0) & (iter < maxIter):
    if(fc < 0 ):
      a = c
    elif(fc > 0):
      b = c
    c = (a+b)/2
    fc = f(c)
    iter = iter + 1
    if(iter == maxIter):
      print(f"No se alcanza la presición deseada, se supero el número máximo de iteraciones ({maxIter})")
  return c

a = 0.5
b = 0.6
epsilon = 0.0000000000000001
x = bisection(a, b,f, epsilon)
y = f(x)

print("Valor encontrado con en metodo creado x = ",x)
```
