
| Comando | Descripción |
| --- | --- |
| `import pandas as pd` | Importa la librería pandas con el alias pd |
| `pd.Series(data, index, dtype)` | Crea una serie de una dimensión con los datos, índices y tipo de dato especificados |
| `pd.DataFrame(data, index, columns)` | Crea un DataFrame de dos dimensiones con los datos, índices y columnas especificados |
| `pd.read_csv(file)` | Lee un archivo CSV y devuelve un DataFrame |
| `pd.to_csv(file)` | Escribe un DataFrame a un archivo CSV |
| `df.head(n)` | Devuelve las primeras n filas del DataFrame df |
| `df.tail(n)` | Devuelve las últimas n filas del DataFrame df |
| `df.loc[row, col]` | Accede al elemento del DataFrame df en la fila y columna especificadas por etiqueta |
| `df.iloc[row, col]` | Accede al elemento del DataFrame df en la fila y columna especificadas por posición |
| `df.drop(labels, axis)` | Elimina las filas o columnas del DataFrame df según las etiquetas y el eje especificados |
| `df.assign(col=value)` | Crea una nueva columna en el DataFrame df con el nombre y valor especificados |
| `df.rename(columns=dict)` | Renombra las columnas del DataFrame df según el diccionario especificado |
| `df.sort_index(axis, ascending)` | Ordena el DataFrame df por índices a lo largo del eje especificado en orden ascendente o descendente |
| `df.sort_values(by, ascending)` | Ordena el DataFrame df por valores según la columna o lista de columnas especificada en orden ascendente o descendente |
| `df.rank(axis)` | Asigna rangos a los elementos del DataFrame df a lo largo del eje especificado |
| `df.describe()` | Devuelve un resumen estadístico de las columnas numéricas del DataFrame df |
| `df.info()` | Devuelve información sobre el DataFrame df, como el número de filas, columnas, tipos de datos y valores no nulos |
| `df.sum(axis)` | Devuelve la suma de los valores del DataFrame df a lo largo del eje especificado |
| `df.apply(func, axis)` | Aplica una función a lo largo de un eje del DataFrame df |
| `pd.merge(df1, df2, on, how)` | Combina dos DataFrames df1 y df2 según la columna o lista de columnas especificada en on y el tipo de combinación especificado en how |


```python
import pandas as pd

url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vT9mPBmgWIg1LFVKFQsrhrEsjCHHG7D4FCItqYeb6TEWUPnkPV8Tj6XzRT33DUl72zI7tI_iJbJFEXH/pub?output=csv"
df = pd.read_csv(url)
df.info()
```

```python
df=df.rename(columns={"Año":"Ano"}) #cambio de nombre a un campo se busca el nombre actual y el nombre nuevo
```

```python
#a las columnas generalmente se les llaman campos
# a las filas se les llama registros
tf = df.Ano>2010 #realiza un filtro en año donde
df[tf]
```

```python
df1=df[tf].copy()
df1.Ano.unique()
```

```python
tf = ( (df.Ano>2010) & (df.Ano<2015))
df1=df[tf].copy()
df1 = df1.reset_index()
```
