```python
import numpy as np
import matplotlib.pylab as plt
import json

global g
g = 9.8

def readFile(nameFile):
  with open(nameFile, "r") as archivo:
    #Lee el arrchivo que est√° en formato json
    d = json.load(archivo)
  return d

def writeFile(nameFile, dicc):
  #usar with open para abrir el archivo en modo escritura
  with open(nameFile, "w") as archivo:
    #Guarda el diccionario en el archivo como json
    json.dump(dicc, archivo, indent=2)

def posx (x0, v0x, t):
  x = x0 + (v0x*t)
  return x

def posy (y0, v0y, t):
  y = y0 + (v0y*t)-(0.5*g*(t**2))
  return y

def graficarMovimientoParabolico(v0,theta,x0=0,y0=0):
  tv = (2*v0*np.sin(np.deg2rad(theta)))/g
  tv = round(tv,4)
  print("Tiempo de vuelo = ", tv)
  v0x = round(v0*np.cos(np.deg2rad(theta)),5)
  v0y = round(v0*np.sin(np.deg2rad(theta)),5)
  print("V0x = ", v0x)
  print("V0y = ", v0y)
  v0xu = v0x/v0
  v0yu = v0y/v0
  t = 0
  t = np.linspace(0,tv, 1000)
  x = posx(x0,v0x,t)
  y = posy(y0,v0y,t)
  fig, ax  = plt.subplots()
  ax.set_title("Trayectoria movimiento parab√≥lico")
  ax.plot(x,y)
  ax.set_xlabel("x")
  ax.set_ylabel("y")
  ax.quiver(x0,y0, v0x,v0y,angles='xy', scale_units='xy', color=['r'], width=0.005)
  ax.annotate("v0", (v0xu,v0yu+2), fontsize=10)
  plt.gca().set_aspect('equal')
  X = np.array([x,y])
  np.savetxt("movimientoParabolico.txt",X.T)

try:
 data = readFile("diccionario.json")
except FileNotFoundError as e:
  data = { "velocidad":26, "angulo":40}
  print("Archivo diccionario.json no encontrado!")
  print("Se crear√° el diccionario por defecto =  { 'velocidad':26, 'angulo':40}")
v0 = data["velocidad"]
theta = data["angulo"]

graficarMovimientoParabolico(v0, theta)
```

```python
r = np.loadtxt("movimientoParabolico.txt")
plt.plot(r[:,0], r[:,1])
plt.title("Trayectoria movimiento parab√≥lico")
plt.xlabel("x")
plt.ylabel("y")
plt.gca().set_aspect('equal')
plt.savefig("movimientoParabolico.png")
```
