```python
import scipy.misc as misc
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp


a = sp.Symbol("a")
b = sp.Symbol("b")
x = sp.Symbol("x")
epsilon = 0.1
a_=1
b_=1
x_min = 0
x_max = (np.pi/2)-epsilon
Np = 10000
x_ = np.linspace(x_min,x_max,Np)
f = lambda x, a, b: a*sp.sin(x) + b*(sp.tan(x)**2)
fun = sp.lambdify(x,f(x,a_,b_),'numpy')
f(x,a,b)
```

```python
def fprim(x,a,b):
  fp = sp.diff(f(x,a,b),x,1)
  return fp
fprima = fprim(x,a,b)
fprima
```

```python
funp = sp.lambdify(x,fprim(x,a_,b_),'numpy')
```

```python
y_ = fun(x_)
plt.plot(x_, y_)
plt.title("Gráfica f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(linestyle="--")
```

```python
der = misc.derivative(fun, x_, dx=1E-2)

fig, axs = plt.subplots(1,2, figsize=(12,6))
axs[0].plot(x_, der)
axs[0].set_title("Gráfica f'(x) por derivative ")
axs[0].set_xlabel("x")
axs[0].set_ylabel("f'(x)")
axs[0].grid(linestyle="--")

der2 = funp(x_)
axs[1].plot(x_, der2)
axs[1].set_title("Gráfica f'(x) por sp.diff")
axs[1].set_xlabel("x")
axs[1].set_ylabel("f'(x)")
axs[1].grid(linestyle="--")
```

```python
#Derivada analítica
fp_analitica = lambda x,a,b: a*sp.cos(x) + b*sp.tan(x)*(2*sp.tan(x)**2 + 2)
fp_analitica(x,a,b)
```

Calcular la derivada usando el método de los 3 puntos

```python
def derivada3Puntos(f_, x_, a, b):
  Np = len(x_)
  fun = sp.lambdify(x,f_(x,a,b),'numpy')
  h = (x_[Np-1]-x_[0])/Np
  der3 = np.zeros(Np-2)
  for i in np.arange(1,Np-1,1):
    der3[i-1] = (1/(2*h))*(fun(x_[i]+h)-fun(x_[i]-h))
  return der3


xp = x_[1:len(x_)-1]
der3 = derivada3Puntos(f, x_,a_,b_)
plt.plot(xp,der3)
plt.title("f'(x) método 3 puntos")
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.grid(linestyle="--")
```
