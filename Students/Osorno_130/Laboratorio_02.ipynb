````python
def divisores_propios (n):
  esInt = isinstance(n, int)
  clase = type(n).__name__
  esKindOfInt = "int" in clase
  if (esInt == False) | esKindOfInt == False:
    print ("La variable ingresada como parámetro debe ser un entero positivo")
    print ("tipo de variable = ", type(n))
    return 0
  if n < 0:
    print ("Solo se pueden ingresar enteros números positivos")
    return 0
  div_propios = []
  for x in range(2,n):
    if (n%x) == 0:
      div_propios.append(x)
  return div_propios
````

````python
def numeros_defectivos(n):
  div_propios = divisores_propios(n)
  if(div_propios == 0):
    return False
  if n == 1:
    return True
  suma = 1
  for x in div_propios:
    suma = suma + x
  return (suma < n)
````

````python
import numpy as np
def numeros_semiperfectos(n):
  div_propios = divisores_propios(n)
  if div_propios == 0:
    return False
  suma = 0
  i = len(div_propios)-1
  while i >= 0 :
    suma = suma + div_propios[i]
    if (suma == n):
      return True
    elif suma > n:
      suma = suma - div_propios[i]
    i = i - 1
  suma = suma + 1
  if (suma == n):
    return True
  return False

print(numeros_semiperfectos(72))
````


7. Diseñar un programa en el que entrado dos números `a`  y `b` retorne una variable booleana **True** o **false**.

  [Números amigos](https://es.wikipedia.org/wiki/N%C3%BAmeros_amigos) `a` y `b` tales que a es la suma de los divisores propios de `b` y viceversa.
    La rutina se debera llamar numeros_amigos




P2. Retornar los 10 primeros pares de numeros perfectos, semiperfectos, amigos



````python
def numeros_amigos(a, b):
  div_a = divisores_propios(a);
  div_b = divisores_propios(b);
  if (div_a == 0) | (div_b == 0):
    return False
  suma_div_a = 1
  suma_div_b = 1
  for x in div_a:
    suma_div_a = suma_div_a + x
  for x in div_b:
    suma_div_b = suma_div_b + x
  if (suma_div_a == b) & (suma_div_b == a):
    return True
  return False
````
