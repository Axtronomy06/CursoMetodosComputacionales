Para campos magneticos

```python
import numpy as np
import scipy as sc
import sympy as sp
```

```python
Bx = sp.Symbol("Bx")
By = sp.Symbol("By")
Bz = sp.Symbol("Bz")
```

```python
H = sp.Matrix([[Bz, Bx-1j*By],[Bx+1j*By, -Bz]])
H
```

```python
H.subs({Bx:1, By:1, Bz:1})
```

```python
E1, E2 = H.eigenvals()
```

```python
E1
```

```python
E2
```

```python
v1, v2 = H.eigenvects()
print(v1)
print(v2)
```
