```python
#Importar librerías
import numpy as np
import matplotlib.pylab as plt
import json
```

```python
#variables y formulas presentes en el movimiento parabólico
theta=np.pi/4 #np.pi es para definir a pi
x_o=0
v_o= 2
g=9.8
t_v=2*v_o*np.sin(theta)/g

t= np.linspace(0,t_v,100)
x=x_o+v_o*np.cos(theta)*t
y=(v_o*np.sin(theta)*t)-(1/2)*g*(t**2)

plt.figure()
plt.plot(x,y,"bo")
plt.title("movimiento parabólico")
plt.show()
```

```python
X = np.array([x,y])
np.savetxt("parabolico.txt",X.T)
```
