## Decaimiento Radioactivo

Primero, defina la función que representa la derivada de N con respecto al tiempo:

Parámetros:
- N: valor actual de N
- lamb: parámetro lambda en la ecuación diferencial

Retorna:
- La derivada de N con respecto al tiempo

Ahora, se implementa la función euler que realizará el método de Euler para resolver la ecuación diferencial:

Parámetros:
- funcion:Funcuón que devuelve la derivada de N
- N0: Condición inicial para N en el tiempo inicial
- lamb: Parametro lambda en la ecuación diferencial
- Inicio: Tiempo inicial
- Fin: Tiempo final
- Paso: Tamño del paso en el método de Euler

La función euler realiza

1. Crea un array para los valores de tiempo llamado v
2. crea un array para almacenar los valores de N llamado y
3. Establecemos la condición incial N0 en el pimer elemento de y
4. Realice un bucle for para iterar sobre los vaores de tiempo en v
5. Calculamos la variación en el tiempo "$\Delta v$" en cada paso
6. Aplicamos el método de Euler para actualizar el valor de N en cada paso de tiempo, en cada iteración, el nuevo valor de N se calcula utilizando la derivada de N con respecto al tiempo $(\frac{dN}{dt})$ evaluada en el valor actual de N multiplicada por la variación en el tiempo

Retorna:
- t= Array con los valores de tiempo
- N: Array con los valores de N en cada punto de tiempo
Note que est retorna los dos arrays que contienen los valores de tiempo y los valores correspondientes de N en cada punto de tiempo

Graficación

Resuelta la ecuación diferencial utilizando el método de Euler, se grafica la solución utilizando matplotlib para graficar la solución númerica obtenida.

El gráfico resultante muestra como la cantidad N evoluciona a lo largo del tiempo de acuerdo con la solución númerica de la ecuación diferencial

````python
#Parametros iniciales
N0=1
lamb=0.1
inicio=0
fin=10
paso=0.1
````

````python
def funcion(N,lamb):
    return -lamb*N
````

````python
def euler(funcion,N0,lamb,inicio,fin,paso):
    v=np.arange(inicio,fin+paso,paso)
    y=np.zeros(len(v))
    y[0]=N0

    for i in range(1,len(v)):
        delta_v=v[i]-v[i-1]
        y[i]=y[i-1]+delta_v*funcion(y[i-1],lamb)
    return y,v
````

````python
y,v=euler(funcion,N0,lamb,inicio,fin,paso)
print(f"valor tiempo:{v}")
print(f"decaimiento:{y}")
````

````python
#Gráfico
plt.figure(figsize=(6,6))
plt.plot(v,y,label="solución númerica")
plt.xlabel("Tiempo")
plt.ylabel("N")
plt.legend()
plt.grid()
plt.show()
````

## Oscilador armónico

````python
gamma=0.0
omega=100 #rad/s
params=(gamma,omega)
def fun( t, y,params):
    gamma=params[0]
    omega=params[1]

    dydt = y[1]
    dvdt= -gamma*y[1]-omega*y[0]
    return np.array([dydt,dvdt])
````

````python
#ciclo
y0 = 10
v0=1E-12
y_ci=np.array([y0,v0])
t= np.linspace(0,5,10000)
h=t[1]-t[0]
N=len(t)
y=np.zeros(N)
v=np.zeros(N)

y[0]=y_ci[0]
v[0]=y_ci[1]
for i in range(0,N-1):
    y[i+1]=y[i]+h*fun(t,y_ci,params)[0]
    v[i+1]=v[i]+h*fun(t,y_ci,params)[1]
    y_ci=np.array([y[i+1],v[i+1]])
````

````python
plt.figure(figsize=(8,6))
plt.plot(t,y,"k", label="posición y(t)")
plt.plot(t,v,"orange", label="velocidad v(t)")
plt.xlabel("Tiempo")
plt.ylabel("Posición y velocidad")
plt.title("Soución del Oscilador armónico, ecuación diferencial de segundo orden")
plt.legend()
plt.grid(True)
plt.show
````

## **Oscilador** **armónico**

````python
def oscilador(t,y,v,w,b):
    dydt=v
    dvdt=-w**2*y-b*v
    return np.array([dydt,dvdt])
````

````python
def rk4_armonico(oscilador,t,y0,v0,w,b):
    h=t[1]-t[0]
    y=np.zeros(len(t))
    v=np.zeros(len(t))
    y[0]=y0

    for i in range(0,len(t)-1):
        k1=h*oscilador(t[i],y[i],v[i],w,b)
        k2=h*oscilador(t[i]+h/2,y[i]+k1[0]/2,v[i]+k1[1]/2,w,b)
        k3=h*oscilador(t[i]+h/2,y[i]+k2[0]/2,v[i]+k2[1]/2,w,b)
        k4=h*oscilador(t[i]+h,y[i]+k3[0],v[i]+k3[1],w,b)
        y[i+1]=y[i]+1/6*(k1[0]+2*k2[0]+2*k3[0]+k4[0])
        v[i+1]=v[i]+1/6*(k1[1]+2*k2[1]+2*k3[1]+k4[1])
    return y,v
````

````python
#Parámetros
w=1.0
b=0.1
y0=1.0
v0=0.0
t1=np.arange(0,10,0.01)
y1,v1=rk4_armonico(oscilador,t1,y0,v0,w,b)
````

````python
fig,ax=plt.subplots(nrows=1,ncols=1)
ax.plot(t1,y1,"b-",label="posición")
ax.plot(t1,v1,"r.",ms=2,label="velocidad")
ax.set_title("Método RK4 en el oscilador")
ax.legend()
ax.set_xlabel("t")
ax.set_ylabel("f(t)")
plt.grid()
plt.show()
````
