2. Operaciones matricial con regresión lineal.

Cuando se desea hacer una regresión lineal multivariada, existen diferentes formas de realizar la regresión, una de ellas es encontrar el error cuadrático medio y miniminzar el conjunto de parámetros de la regresión emplando algún metodo de minimización, otra forma es encontrar la expresión analítica en forma matricial, para determinar el conjunto de parámetros de la regresión. En este último caso, la regresión lineal viene dada por la siguiente expresion generalizada:


- $\Theta = (X^T X)^{-1} X^T y$

donte $\Theta$ es la matriz de parametros asociados a la regresión lineal, X la matriz de características y $y$ los valos esperados, así, si tenemos una conjunto de datos en N dimensiones; la recta, plano o hyperplano puede ser calculado como:

$Y=\Theta^T X$


En el caso de dos dimensiones, supongamos que X, y Y presentan un comportamiento lineal, y vienen dados por ( En este caso el super indice es :


- Sea $\vec{X}=x^{(1)}, x^{(2)}, x^{(3)}, x^{(4)}, ..., x^{(m)}$
- Sea $\vec{Y}=y^{(1)}, y^{(2)}, y^{(3)}, y^{(4)}, ..., y^{(m)}$

Para determinar la ecuación de la recta, hay que conocer los párametros $\theta_0 $, $\theta_1$, que cumplen que :   $y^{(i)} = \theta_1 x^{(i)} + \theta_0$ la anterior expresión puede ser escrita como:

\begin{equation}
\begin{bmatrix}
y^{(1)}\\
y^{(2)}\\
\vdots\\
y^{(m)}\\
\end{bmatrix}=
\begin{bmatrix}
\theta_0 & \theta_1 \\
\end{bmatrix}
\begin{bmatrix}
1 & 1 & \cdots & 1 \\
x_1^{(1)} & x_1^{(2)} & \cdots & x_1^{(m)} \\
\end{bmatrix}
\end{equation}


Para determinar la matriz de theta entonces se puede aplicar la ecuación:

- $\Theta = (X^T X)^{-1} X^T y$

donde:
\begin{equation}
X=
\begin{bmatrix}
1 & x_1^{(1)} \\
1 & x_1^{(2)} \\
\vdots & \vdots \\
1 & x_1^{(m)}
\end{bmatrix}
\end{equation}

y


\begin{equation}
y=
\begin{bmatrix}
y^{(1)} \\
y^{(2)} \\
\vdots\\
y^{(m)}
\end{bmatrix}
\end{equation}

````python
#Nociones preliminares
N = 10
x1 = np.linspace(0, 10, N)
noise=2*np.random.random(N)-1
y =0.9*x1 + noise
````

````python
x0 = np.ones((N,1))
X = np.zeros((N,2))
X
````

````python
X[:,0]=1
X[:,1]=x1
X
````

````python
#Desenglosando la formula brindada
#Transpuesta
X_t=np.transpose(X)
X_t
````

````python
#Multiplicación de matrices
M2=np.dot(X_t,X)
M2
````

````python
#Inversa de la matriz
inver= np.linalg.inv(M2)
inver
````

````python
#Multiplicación de matrices entre la inversa y la traspuesta
Mtzi_X=np.dot(inver,X_t)
Mtzi_X
````

````python
theta=np.dot(Mtzi_X,y)
theta
````

````python
#Recta
theta_trp=np.transpose(theta)
Y=theta_trp*X
Y[:,1]
````

````python
#Gráfico
plt.figure(figsize=(5,5))
plt.plot(x1,y,"o",label="Datos originales")
plt.plot(x1,Y[:,1],color="blue", label="Regresión lineal")
plt.xlabel("X1")
plt.ylabel("y")
plt.title("Regresión lineal unidimensional")
plt.legend()
plt.grid()
plt.show()
````

````python
#Nociones preliminares
N = 100
x1 = np.linspace(-1, 1, N)
x2 = np.linspace(-2, 1, N)
noise1=2*np.random.random(N)-1
noise2=2*np.random.random(N)-1
y = 2*(x1+noise1) - 3*(x2+noise2) + 0.0


N = 100
X1, X2 = np.meshgrid(x1,x2)
Y = 2*X1 - 3*X2 + 0.0
#fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
#surf = ax.plot_surface(X1, X2, Y)
````

````python
x0b = np.ones((N,1))
Xb = np.zeros((N,3))
Xb
````

````python
Xb[:,0]=1
Xb[:,1]=x1
Xb[:,2]=x2
Xb
````

````python
Xb_t=np.transpose(Xb)
Xb_t
````

````python
#Multiplicación de matrices
Mb=np.dot(Xb_t,Xb)
Mb
````

````python
thetab=np.dot(Mtzi_Xb,y)
thetab=np.matrix(thetab)
````

````python
#Plano
Yb=thetab[0,1]*X1 + thetab[0,2]*X2+ thetab[0,0]
Yb
````

Generalización

````python
def regresion_bidimensional(x1,x2,y):
    if len(x1) != len(x2) and len(x2) != len(y):
        print("Las lista x1 y x2 deben tener la misma dimensión")

    X1,X2=np.meshgrid(x1,x2)

    x0b = np.ones((N,1))
    Xb = np.zeros((N,3))
    Xb[:,0]=1
    Xb[:,1]=x1
    Xb[:,2]=x2
    Xb

    #Coeficientes de la regresión
    Xb_t=Xb.T
    Mb=np.dot(Xb_t,Xb)
    inv_Mb=np.linalg.inv(Mb)
    Mtzi_Xb=np.dot(inv_Mb,Xb_t)
    thetab=np.dot(Mtzi_Xb,y)
    thetab=np.matrix(thetab)

    #Calcular el plano de regresión
    Yb=thetab[0,1]*X1 + thetab[0,2]*X2+ thetab[0,0]

    #Gráfico
    X1, X2 = np.meshgrid(x1,x2)
    fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
    surf = ax.plot_surface(-X1, X2, Y)
    ax.scatter(x1,x2,y,"o",color="red",label="Datos originales")
    ax.set_xlabel("x1")
    ax.set_ylabel("x2")
    ax.set_zlabel("Y")
    ax.legend()
    ax.grid()
    plt.show()
    return surf
````

````python
#Ejemplo de uso de la generalización
N = 100
x1 = np.linspace(-1, 1, N)
x2 = np.linspace(-2, 1, N)
noise1=2*np.random.random(N)-1
noise2=2*np.random.random(N)-1
y = 2*(x1+noise1) - 3*(x2+noise2) + 0.0

regresion_bidimensional(x1,x2,y)
````
