$\textbf{Introducci$\acute{o}$n:}$

Para conocer la posición de un planeta en movimiento, es común utilizar métodos numéricos para resolver las ecuaciones del movimiento de dicho planeta bajo la influencia gravitatoria de un centro de fuerzas fijo, como una estrella, en nuestro caso, el Sol.

- $\textbf{Las leyes del Movimiento Planetario: Las leyes de Kepler}$

La leyes de Kepler describen el movimiento de los planetas alrededor del sol de la siguiente manera:

    1. Los planetas describen órbitas elípticas con el sol en uno de los focos.
    2. El vector posición de cualquier planeta con respecto al sol, barre áreas iguales de la elipse en tiempos iguales.
    3. El cuadrado del periodo orbital de un planeta es proporcional al cubo de su semieje mayor

Además, las leyes de Newton amplian la comprensión del movimiento celestial al predecir no solo órbitas elípticas, sino también parabólicas e hiperbólicas. En términos generales, un cuerpo bajo la influencia de la atracción gravitoria describe una trayectoria plana que es una cónica.

Finalmente, para modelar estas trayectorias se utilizan ecuaciones de movimiento. En este enfoque, en lugar de expresar estas ecuaciones en coordenadas polares, se calculan las componentes de la aceleración a lo largo de los ejes X e Y. Esto resulta en un sistema de dos ecuaciones diferenciales de segundo orden y que se resuelven mediante métodos numéricos. Este enfoque permite predecir y simular las trayectorias de los planetas con precisión.

- $\textbf{Soluci$\acute{o}$n n$\acute{u}$merica de las ecuaciones del movimiento}$

Imaginemos una partícula con masa $(m)$, que representa un plaenta, está sometida a la atracción gravitatoria de un cuerpo mucho más Masivo $(M)$, que en nuestro caso sería el Sol. En este escenario, consideramos que la influencia de la partícula(planeta) en el cuerpo masivo(el sol) es despreciable, y por lo tanto, el sol permanece en reposo en el origen de nuestro sistema de coordenadas.

La partícula esta sometida a una fuerza atractiva cuya dirección es radial y apuntando hacia el centro del sol. El módulo de la fuerza viene dado por la ley de Gravitación Universal

\begin{equation}
F=G\frac{Mm}{r^2}
\end{equation}
 Donde r es la distancia entre la partícual y el centro de fuerza, X e Y representan su posición respecto del sistema de referencia cuyo origen está situado en el sol.

\begin{equation}
r=\sqrt{x^2+y^2}
\end{equation}

Las componentes de fuerza son:

\begin{equation}
F_{x}=-F\cos(\theta)=-F \frac{x}{r}\\
F_{y}=-F\sin(\theta)=-F \frac{y}{r}
\end{equation}

Aplicando la segunda ley de Newton, la cual relaciona la fuerza con la masa y la aceleración, obtenemos un sistema de dos ecuaciones diferenciales de segundo orden al expresar la aceleración como la segunda derivada de la posición. Esto es fundamental para describir el movimiento de la partícula(el planeta) bajo la influencia gravitatoria del cuerpo masivo(el Sol)


\begin{equation}
m\frac{d^2x}{dt^2}=-G\frac{mM}{r^2}\frac{x}{r}\\
m\frac{d^2y}{dt^2}=-G\frac{mM}{r^2}\frac{y}{r}
\end{equation}

Dadas las condiciones iniciales(posición y velocidad) el sistema de dos ecuaciones diferenciales se puede resolver aplicando un procedimiento numérico.

La ecuación de la dinámica del movimiento circular uniforme:

\begin{equation}
G\frac{Mm}{r_0^2}=m\frac{v_0^2}{r_0} ⟹ v_0=\sqrt\frac{GM}{r_0}
\end{equation}

En sistema de ecuaciones diferenciales se transforma en:

\begin{equation}
\frac{d^2x}{dt^2}=-G\frac{M}{r^2}\frac{x}{r}\\
\frac{d^2X}{d\tau^2}\frac{r_0}{P^2}=-GM\frac{X}{(X^2+Y^2)^{3/2}}\frac{1}{r_0^2}\\

Del mismo modo para la ecuación diferencial en y

\end{equation}

Volviendo a la notación x e y para la posición y t para el tiempo en el nuevo sistema de unidades. El sistema de ecuaciones se escribe

\begin{equation}

Con estos datos, podemos calcular la velocidad:

$v_y=\sqrt\frac{2GMr_2}{r_1(r_1+r_2)}$

# **Método de Rugen-Kutta**

La elección del método de Runge-Kutta para resolver el problema de las órbitas planetarias se justifica por su eficacia y precisión en la aproximación de soluciones a ecuaciones diferenciales ordinarias (EDO). En el contexto de mecánica celeste, donde las órbitas de los planetas están modeladas mediante EDO, es crucial seleccionar un método numérico confiable que garantice resultados precisos

El método de Rugen-Kutta es particularmente adecuado para este propósito debido a su versatilidad y capacidad de manejar sistemas de EDO no lineales, pues este proporciona una mayor exactitud al dividir el paso de tiempo en varios subintervalos y calcular promedios ponderados de pendientes en cada subintervalo

Desglosando el método de Runge-Kutta de cuarto orden, el proceso implica calcular $k_1,k_2,k_3,k_4$ que son incrementos de la función en distintos puntos del intervalo de tiempo. Luego, se utiliza una combinación ponderada de estos incrementos para obtener la aproximación final del siguiente paso. Este enfoque permite minimizar errores y proporciona una solución más precisa en comparación con otros métodos de menor orden

En el contexto específico de la simulación de órbitas planetarias la precisión del método de Runge-Kutta hace que sea una elección sólida permitiendo obtener resultados numéricos confiables y visualmente precisos de las trayectorias planetarias en función del tiempo

Definiendo un problema de valor inicial como:
$y'=f(x,y)$, $y(x_0)=y_0$

Entonces el método RK4 para este problema está dado por la siguiente ecuación:

$y_{i+1}=y_i+\frac{h}{6}*(k_1+2k_2+2k_3+k4)$

Donde:

$k_1=f(x_i,y_i)$

$k_2=f(x_i+\frac{h}{2}, y_i+\frac{1}{2}k_1h)$

$k_3=f(x_i+\frac{h}{2},y_i+\frac{1}{2}k_2h)$

$k_4=f(x_i+h,y_i+k_3h)$


Así el siguiente valor (y_{n+1}) es determinado por el presente valor (y_n) más el producto del tamaño del intervalo (h) por una pendiente estimada

$pendiente= \frac{k_1+2k_2+2k_3+k4}{6}$

```python
def orbita(condI,t):
    x, vx, y, vy = condI
    dxdt = vx
    dydt = vy
    dvxdt=-4 * np.pi**2 * x/((x**2+y**2)**(3/2))
    dvydt=-4 * np.pi**2 * y/((x**2+y**2)**(3/2))

    return [dxdt,dvxdt,dydt,dvydt]
```

```python
def RK4(condI,t):
    h=t[1]-t[0]
    resultado=np.zeros((len(t),4))
    resultado[0,:]=condI
    for i in range(1,len(t)):
        k1= h*np.array(orbita(resultado[i-1,:],t[i-1]))
        k2= h*np.array(orbita(resultado[i-1,:]+k1/2,t[i-1]+h/2))
        k3= h*np.array(orbita(resultado[i-1,:]+k2/2,t[i-1]+h/2))
        k4= h*np.array(orbita(resultado[i-1,:]+k3,t[i-1]+h))
        resultado[i,:] = resultado[i-1,:] + (k1+2*k2+2*k3+k4)/6
    return resultado
```

```python
t2=np.linspace(0,1000,1000000)
```

```python
fig,axs=plt.subplots(4,2,figsize=(12,16))
#fig.suptitle("órbita Planetaria para Varios Planetas",y=0.94)

for i,ax in enumerate(axs.flatten()):
    planeta=df.iloc[i]
    condi_inicial=[planeta['x0'],planeta['vx0'],planeta['y0'],planeta['vy0']]
    solucion_RK4=RK4(condi_inicial,t2)

    ax.plot(solucion_RK4[:,0],solucion_RK4[:,2])
    ax.set_title(planeta["Planeta"])
    ax.set_xlabel("x")
    ax.set_ylabel("y")

plt.tight_layout()
plt.show()
```

```python
fig,axs=plt.subplots(4,2,figsize=(12,16))
#fig.suptitle("órbita Planetaria para Varios Planetas",y=0.94)

for i,ax in enumerate(axs.flatten()):
    planeta=df.iloc[i]
    condi_inicial=[planeta['x0'],planeta['vx0'],planeta['y0'],planeta['vy0']]
    solucion_RK4=RK4(condi_inicial,t2)

    ax.plot(t2,solucion_RK4[:,0],"greenyellow", label="X0")
    ax.plot(t2,solucion_RK4[:,2],"darkseagreen",label="Y0")
    ax.set_title(planeta["Planeta"])
    ax.set_xlabel("x")
    ax.set_ylabel("y")

plt.tight_layout()
plt.show()
```
