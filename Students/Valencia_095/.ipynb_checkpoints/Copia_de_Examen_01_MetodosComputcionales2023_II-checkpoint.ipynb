````python
#Importación de librerías
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
import sympy as sp
from scipy import optimize
````

````python
#Parámetros iniciales
m=0.25    #Masa
s0=900   #Altura inicial
g=9.8     #Gravedad
k=0.148   #coeficiente de resistencia con el aire
````

a. Realizar un gráfico de $s$ como función del tiempo, para ello cree una función para  $s(t)$  llamada caida:

````python
#Divida la formula dada para evitar errores
def caida(t):
    T1 = s0
    T2 = (m * g / k) * t
    T3= ((m**2)*g)/k**2
    T4=(1-np.exp(-k*t/m))
    return T1-T2+(T3*T4)
#caida(t)
````

````python
#Gráfico
t=np.linspace(0,10,1000)
plt.figure(figsize=(10,6))
plt.plot(t,caida(t))
plt.title("S en función del tiempo",fontsize=15)
plt.xlabel("t", fontsize=15)
plt.ylabel("s(t)",fontsize=15)
plt.grid(True)
plt.show()
````
