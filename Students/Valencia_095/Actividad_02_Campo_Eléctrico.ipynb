Elaborar un programa para calcular el campo eléctrico producido por un dipolo eléctrico, realizar una representación del campo para todo el dominio elegido para calcular el campo. Para encontrar el campo, se sugiere hacer lo siguiente:

**a.** Encontrar el campo en un punto producido por dos cargas q1 y q2

**b.** Emplee el comando plt.quiver(x, y, Ex, Ey) para mostrar el campo producido por q1 y q2. Puedes intentarlo con el comando, plt.stremplot.

**c.** Generalice el problema para varios puntos en el espacio.
Se recomienda emplear numpy array.

```python
import numpy as np
import matplotlib.pylab as plt
```

```python
#Caso particular
#CONDICIONES INICIALES
xp=0
yp=1
k=1
#posición de carga y magnitud

#Carga 1
q1=2 #microcoulomb
x1=-1
y1=0

#Carga 2
q2=-2
x2=1
y2=0
```

```python
#PERMITE DETERMINAR EL CAMPO ELECTRICO
def campo_electrico(xp,yp,x1,y1,q1):
    Deltax=xp-x1
    Deltay=yp-y1
    Deltar=np.sqrt(Deltax**2+Deltay**2)


    ux=Deltax/Deltar
    uy=Deltay/Deltar


    Ex=(k*q1/Deltar**2)*ux
    Ey=(k*q1/Deltar**2)*uy
    E=np.array([Ex,Ey])
    return E
```

```python
#PUNTO DE EVOLUCIÓN
E1=campo_electrico(xp,yp,x1,y1,q1)
E2=campo_electrico(xp,yp,x2,y2,q2)
E=E1+E2
print(E1)
print(E2)
```

```python
#GRÁFICO
plt.figure()
plt.plot(x1,y1,"ro")
plt.plot(x2,y2,"ro")
plt.plot(xp,yp,"bo")
plt.quiver(xp,yp,E[0],E[1])
plt.ylim(-0.1,1.5)
plt.title("Campo eléctrico inducido por un dipolo")
plt.show()
```

```python
#Manera más generalizada del código

#CONDICIONES INICIALES
xp=0
yp=1
k=1
#variables que ingresa el usuario
q1=int(input("ingrese el valor de la carga 1: "))
q2=int(input("ingrese el valor de la carga 2: "))
x1=int(input("ingrese el valor de la coordenada x de q1: "))
y1=int(input("ingrese el valor de la coordenada y de q1: "))
x2=int(input("ingrese el valor de la coordenada x de q2: "))
y2=int(input("ingrese el valor de la coordenada y de q2: "))

def campo_electrico(xp,yp,x1,y1,q1):

    Deltax=xp-x1
    Deltay=yp-y1
    Deltar=np.sqrt(Deltax**2+Deltay**2)


    ux=Deltax/Deltar
    uy=Deltay/Deltar


    Ex=(k*q1/Deltar**2)*ux
    Ey=(k*q1/Deltar**2)*uy
    E=np.array([Ex,Ey])
    return E

x=np.arange(-5,5,1) #genera un conjunto de números entre los valores inicial y final que va en incremento de 1 en 1
y=np.arange(-5,5,1)

plt.figure(figsize=(10,8))
plt.plot(x1,y1,"ro")
plt.plot(x2,y2,"ro")

for xi in x:
    for yj in y:
        E1=campo_electrico(xi,yj,x1,y1,q1)
        E2=campo_electrico(xi,yj,x2,y2,q2)
        E=E1+E2
        print(E1)
        print(E2)
        plt.quiver(xi,yj,E[0],E[1])

plt.ylim(-5,5)
plt.xlim(-5,5)
plt.title("Campo eléctrico inducido por un dipolo")
plt.show()
```
