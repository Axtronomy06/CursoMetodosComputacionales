# b) Deducción del método de Newton-Raphson
La expansión en serie de Taylor de primer orden de $f(x)$ alrededor de $x_0$ es:


> $f(x)$ $\approx$ $f(x_0)+f'(x_0)(x-x_0)$

considere la función $f(x)$ donde $(\alpha)$ es una raiz. Tomando un valor $x_n$ muy cercano a $(\alpha)$. Se puede expresar la serie de taylor de $f(x)$ así:

\begin{equation}
f(x)=f({x}_n)+f'({x_n})(x-{x_n})+\frac{f''({\xi})}{2} (x-{x_n}^{2})
\end{equation}
con $\xi \in [x,{x}_n]$
Iguale a cero al lado izquierdo de la ecuación, así:

\begin{equation}
0=f({x}_n)+f'({x_n})(x-{x_n})+\frac{f''({\xi})}{2} (x-{x_n}^{2})
\end{equation}

Despejando $\alpha$ del segundo termino de la serie de taylor obtenemos:
\begin{equation}
-f'({x}_n)(\alpha-{x}_n)=f({x}_n)+ \frac{f''({\xi})}{2} (\alpha-{x}_n)^{2}
\end{equation}

\begin{equation}
f'({x}_n)(\alpha-{x}_n)=-f({x}_n)-\frac{f''({\xi})}{2} (\alpha-{x}_n)^{2}
\end{equation}

\begin{equation}
(\alpha-{x}_n)=-\frac{(f({x}_n)}{f'({x}_n)}-\frac{f''({\xi})}{2} \frac{1}{f'({x}_n)}(\alpha-{x}_n)^2
\end{equation}

\begin{equation}
(\alpha)= {x}_n-\frac{(f({x}_n)}{f'({x}_n)}-\frac{f''({\xi})}{2} \frac{1}{f'({x}_n)}(\alpha-{x}_n)^2
\end{equation}

Tenga en cuenta que si $\alpha \approx{x}_0$ entonces los términos $(\alpha-x_0)$ y $(\alpha-x_1)$ serán muy cercanos a cero. Por tanto se deduce que
\begin{equation}
\frac{f''({\xi})}{2} \frac{1}{f'({x}_n)}(\alpha-{x}_n)^2
\end{equation}
Es despreciable
De esta manera,
\begin{equation}
(\alpha) \approx x_1= {x}_0-\frac{(f({x}_0)}{f'({x}_0)}
\end{equation}

se repite el proceso con $x_2$
\begin{equation}
(\alpha) \approx x_2 =x_1- \frac{f(x_1)}{f'(x_1)}
\end{equation}

Esta es la fórmula del método de Newton_Raphson. En cada iteración comenzamos con una estimación de $x_0$, y luego actualizamos nuestra estimación de la raíz utilizando la formula $x =x_0- \frac{f(x_0)}{f'(x_0)}$. Repetimos el proceso hasta que alcancemos la presición deseada o el número máximo de iteraciones.


# c) ¿Cuál es la diferencia del metodo de Newton-Rapson con respecto al de la secante?
La principal diferencia radica en la forma en que se calculan las aproximaciones sucesivas de la raíz. El método de Newton-Raphson utiliza la derivada de la función para calcular la tangente en cada iteración, mientras el método de la secante utiliza una linea recta que pasa por dos puntos cercanos en la curva


Posición en x
\begin{equation}
x=(u\cos \alpha) t + \frac{1}{b} (v_0\cos \theta_0 -u \cos\alpha)(1-\mathrm{exp}(-bt))
\end{equation}

Posicion en y
\begin{equation}
y= \frac{1}{b}\left(\frac{g}{b}+v_0 \sin \theta_0 -u\sin\alpha \right)(1-\mathrm{exp}(-bt))-\left( \frac{g}{b}-u\sin\alpha\right)t
\end{equation}

````python
# Posicion en x
def pos_x(t, theta, alpha, v0, b, u,g):
    h= u*np.cos(alpha)*t
    i= 1/b*(v0*np.cos(theta)-u*np.cos(alpha))
    j= 1-np.exp(-b*t)
    x = h+i*j
    return x

# Posicion en y
def pos_y(t, theta, alpha, v0, b, u,g):
    k=1/b*(g/b+v0*np.sin(theta)-u*np.sin(alpha))
    l=1-np.exp(-b*t)
    m=(g/b-u*np.sin(alpha) )*t
    y = k*l-m
    return y
````

````python
#Cálculo del tiempo de vuelo
tv=sc.optimize.newton(pos_y,12,args=Params)
tv
````

````python
def trayectoria(pos_y,Params):
    tv=sc.optimize.newton(pos_y,12,args=Params)
    tiempo=np.linspace(0,tv,1000)
    x=pos_x(tiempo,*Params)
    y=pos_y(tiempo,*Params)

    datos={"x":x ,"y":y , "tiempo":tiempo}
    df=pd.DataFrame.from_dict(datos)
    return df
df= trayectoria(pos_y,Params)
````

# 4. Fixed point methods.

When a new planet is discovered, there are different methods to estimate its physical properties. Many times is only possible to estimate either the planet mass or the planet radius and the other property has to be predicted through computer modelling.

If one has the planet mass, a very rough way to estimate its radius is to assume certain composition (mean density) and a homogeneous distribution (a very bad assumption!). For example, for the planet [Gliese 832c](http://es.wikipedia.org/wiki/Gliese_832_c) with a mass $M= 5.40 M_{\oplus}$, if we assume an earth-like composition, i.e. $\bar \rho_{\oplus} = 5520\ kg/m^3$, we obtain:

$$R_{g832c} = \left( \frac{3 M_{g832c}}{ 4 \pi \bar\rho_{\oplus} } \right)^{1/3} \approx 1.75 R_{\oplus}$$

That would be the planet radius if the composition where exactly equal to earth's.

A more realistic approach is assuming an internal one-layer density profile like:

$$\rho(r) = \rho_0 \exp\left( -\frac{r}{L} \right)$$

where $\rho_0$ is the density at planet centre and $L$ is a characteristic lenght depending on the composition. From numerical models of planet interiors, the estimated parameters for a planet of are $M= 5.40 M_{\oplus}$ are approximately $\rho_0 = 18000\ kg/m^3$ and $L = 6500\ km$.

Integrating over the planet volume, we obtain the total mass as

$$M = 4\pi \int_0^R \rho(r)r^2dr$$

This is a function of the mass in terms of the planet radius.

Solving the equation $M(R) = M_{g832c}$ it would be possible to find a more realistic planet radius. However when using numerical models, it is not possible to approach the solution from the left side as a negative mass makes no sense.

````python
#Datos planeta
M_g832=5.40 #Masa en masas de la Tierra
rho0=18000  #densidad en el centro del planeta en kg/m^3
L=6500000   #Longitud característica en metros

R=sp.symbols("R")#R se utiliza para representar el radio del planeta en las ecuaciones

#Densidad calcula la densidad en función de la distancia r desde el centro del planeta
def densidad(r):
    return rho0*sp.exp(-r/L)

#Calcula la masa total del planeta(integrando la densidad a lo largo del radio r desde cero hasta R)
M_total=sp.integrate(4*sp.pi*densidad(r)*r**2, (r,0,R))

#sustituir los valores conocidos en la ecuación (relaciona la masa total con la masa conocida)
ecuacion=M_total-M_g832

#convierte la ecuación simbolica en una funcion numerica,"numpy" quiere decir que debe ser compatible con Numpy
funcion_numerica=sp.lambdify(R,ecuacion,"numpy")

#Resuelve la ecuación de forma númerica y obtiene un valor numérico para el radio
solucion=fsolve(funcion_numerica,1.0)

#Imprimir la solución numérica encontrada
print(f"radio estimado del planeta(usando sympy):{solucion[0]:.2f} metros")
````

````python
import numpy as np
import matplotlib.pylab as plt
from PIL import Image

Npx_X = 480 #Numero de pixeles en x
Npx_Y = 480 #Numero de pixeles en y

#Creación de una imagen en blanco con dimensiones Npx_X x Npx_Y
image = Image.new("RGB", (Npx_X, Npx_Y)) #Creacion de la imagen inicial

npx = np.arange(0, Npx_X)
npy = np.arange(0, Npx_Y)

xmin = -1
xmax = 1
ymin = -1
ymax = 1
# Transformación al plano complejo en un rectangulo (xmin, xmax), (ymin, ymax)
zx = npx * (xmax - xmin) / (Npx_X - 1)   + xmin
zy = npy * (ymax-ymin) / (Npx_Y - 1) + ymin

# Definicion  la funcion y las derivada
def f(z):
  return z**3 - 1

def fp(z):
  return 3*z**2

h = 1E-4
def fpn(z, h):
  return (f(z+h)-f(z))/h


a = 0.2
Niteraciones = 120
eps = 1E-8
#iterar a través de los pixeles de la imagen
x = - 1
for zx_ in zx:
  x = x + 1
  y = -1
  for zy_ in zy:
      y = y + 1
      Z_0 = complex(zx_, zy_)
#Aplicar el método de Newton para encontrar las raíces de f(z)
      for i in range(0, Niteraciones):
        Z = Z_0 - a * f(Z_0)/fp(Z_0)
        if( abs(Z-Z_0)<= eps):
          break
        Z_0 = Z
      #print(i)
      #print(i)
      #Asignar colores basados en el número de iteraciones
      r, g, b = i % 4 *64 , i % 8 *16, i% 16* 16
      image.putpixel((x,y), (r,g,b) )
#mostrar la imagen generada
plt.imshow(image)
# plt.xlim(400, 600)
# plt.ylim(400, 523)
````
