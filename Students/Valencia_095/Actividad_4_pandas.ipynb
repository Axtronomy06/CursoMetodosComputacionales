```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url="https://raw.githubusercontent.com/hernansalinas/autogradesMetodosComputacionales/main/Dataset/Temperature_Bogota_Barcelona.csv"
```

Leer el dataset desde la url, con el comando pd.read_csv.

```python
df=pd.read_csv(url)
```

```python
df
```

Adicionar una nueva columna AvgTempCelcius con la temperatura en grados celcius. $ C=(F-32)/1.8$

```python
df["Celcius"]=(df.AvgTemperature-32/1.8)
```

```python
df
```

Realizar un gráfico de AvgTempCelcius como función del año para la ciudad de Bogotá.

```python
tf=df.City =="Bogota"
dfBogota = df[tf].copy()
```

```python
#GRAFICO
x=dfBogota['Year']
y=dfBogota['Celcius']
plt.title("Temperatura de Bogotá a través de los años", fontsize=20)
plt.xlabel("Years/Años", fontsize=10)
plt.ylabel("T° Celcius",fontsize=10)
plt.bar(x,y)
```

Realizar un gráfico de AvgTempCelcius como función del año para la ciudad de Barcelona.

```python
tf1=df.City =="Barcelona"
dfBarcelona=df[tf1].copy()
```

```python
#GRAFICO
x=dfBarcelona['Year']
y=dfBarcelona['Celcius']
plt.title("Temperatura de Barcelona a través de los años", fontsize=20)
plt.xlabel("Years/Años", fontsize=10)
plt.ylabel("T° Celcius",fontsize=10)
plt.bar(x,y)
```
