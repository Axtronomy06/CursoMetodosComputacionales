````python
import numpy as np

# Definición de constantes
h = 6.62607015e-34  # Constante de Planck en J·s
c = 2.99792458e8    # Velocidad de la luz en m/s
k_B = 1.380649e-23  # Constante de Boltzmann en J/K
T = 3000            # Temperatura en Kelvin (puedes ajustar este valor)

# Constante reescalada
hc_kB = h * c / k_B

# Definir la función usando logaritmos
def calcular_exponencial(lambda_metros):
    exponent = hc_kB / (lambda_metros * T)
    return exponent

# Generar valores de lambda desde 1 nanómetro (1e-9 metros) hasta 8 micrómetros (8e-6 metros)
lambdas = np.linspace(1e-9, 8e-6, 100)

# Calcular la función exponencial reescalada para cada valor de lambda
resultados_log = [calcular_exponencial(l) for l in lambdas]

# Convertir de nuevo a la escala exponencial solo para mostrar resultados manejables
# Establecemos un límite para el exponente, más allá del cual retornaremos 'inf'
limite_exponente = 700
resultados = [np.exp(res) if res < limite_exponente else float('inf') for res in resultados_log]

# Imprimir resultados
for l, res, res_log in zip(lambdas, resultados, resultados_log):
    print(f"lambda = {l:.2e} m, exp(log_result) = {res:.2e}, log_result = {res_log:.2e}")

````

### I. Radiación de Cuerpo Negro:

 La radiación de cuerpo negro, se refiere a la radiación electromagnética emitida por un objeto que absorbe toda la radiación incidente sin reflejar nada. La cual  puede ser modelada según la ley de Planck que relaciona la intensidad de la radiación de un cuerpo negro con la temperatura del objeto. A medida que la temperatura aumenta, la longitud de onda máxima de la radiación se desplaza hacia longitudes de onda más cortas. Para generar una intuición,  piensa en  trozo  de carbón, el cual representa un ejemplo cercano a lo que se conoce como un cuerpo negro en física. Este elemento tiene la particularidad de reflejar mínimamente la luz, capturando prácticamente toda la radiación que recibe. Al alcanzar temperaturas elevadas, similar a las de las brasas ardientes, el carbón comienza a emitir radiación tanto en el rango visible como en el infrarrojo del espectro.

  La radiación de cuerpo negro puede ser modelada empleando la ley de Planck para describir la densidad  espectral  de energía de la radiación emitida por un cuerpo negro a una temperatura (T). La ley de Planck establece que la densidad espectral de energía $u(\lambda, T)$  para un cuerpo negro está dada por:

\begin{equation}
u(\lambda, T) = \frac{{8\pi hc}}{{\lambda^5}} \cdot \frac{1}{{e^{\frac{{hc}}{{\lambda k_B T}}} - 1}}
\end{equation}

 donde:

$h = 6.62607015  \times 10^{-34} \mathrm{J} s$ : es la constante de Planck.

$c = 299792458 \mathrm{m/s}$ : es la velocidad de la luz.


$k_B = 1.380649 \times 10^{-23}\mathrm{J/K}$ : es la constante de Boltzmann.

$ \lambda$ es la longitud de onda.


Podemos integrar la densidad espectral de energía sobre todas las longitudes de onda para obtener la potencia total emitida por unidad de área de un cuerpo negro a una temperatura (T) (ley  de Stefan-Boltzmann):


\begin{equation}
P_{\text{total}} = \sigma T^4
\end{equation}

donde:

$\sigma$ es la constante de Stefan-Boltzmann.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fmin_powell

# Constantes físicas
h = 6.62607015e-34  # Constante de Planck [J s]
c = 299792458       # Velocidad de la luz [m/s]
K_B = 1.380649e-23  # Constante de Boltzmann [J/K]

# Función de densidad espectral de energía para una longitud de onda dada y temperatura dada
def u_lambda_T(lambda_, T):
    a1 = (8 * np.pi * h * c) / (lambda_ ** 5)
    exponent = (h * c) / (lambda_ * K_B * T)
    e_ = np.where(exponent > 700, np.inf, np.exp(exponent))
    a2 = 1 / (e_ - 1)
    return -a1 * a2  # Negativo porque queremos minimizar

# Transformación para mantener lambda en el rango [1 nm, 8 µm]
def transform(x):
    lambda_min, lambda_max = 1e-9, 8e-6
    return lambda_min + (lambda_max - lambda_min) / (1 + np.exp(-x))

# Función objetivo transformada
def objective(x, T):
    lambda_ = transform(x)
    return u_lambda_T(lambda_, T)

# Valores de temperatura
T_values = np.array([1000, 2000, 4000, 5000])  # [K]

# Graficar densidad espectral de energía vs. longitud de onda
lambda_values = np.linspace(1e-9, 8e-6, 1000)  # [m]
plt.figure(figsize=(10, 6))
for T in T_values:
    # Encontrar el valor óptimo de lambda usando fmin_powell
    x0 = 0  # Valor inicial
    result = fmin_powell(objective, x0, args=(T,), disp=False)
    lambda_opt = transform(result)
    u_values = u_lambda_T(lambda_values, T)

    # Graficar
    plt.plot(lambda_values * 1e9, -u_values, label=f'T = {T} K')  # Nota: revertir el signo para graficar la función original
    plt.scatter(lambda_opt * 1e9, -u_lambda_T(lambda_opt, T), color='red')

    # Imprimir valores máximos y verificar la ley de desplazamiento de Wien
    lambda_opt_scalar = lambda_opt.item()
    print(f"Para T = {T} K, el valor de lambda que minimiza la función es {lambda_opt_scalar * 1e9:.2e} nm")
    print(f"Verificación de la ley de Wien: lambda_max * T = {lambda_opt_scalar * T * 1e6:.2f} µm K")
    print()

plt.xlabel('Longitud de Onda (nm)')
plt.ylabel('Densidad Espectral de Energía (W/m^3)')
plt.title('Densidad Espectral de Energía vs. Longitud de Onda')
plt.legend()
plt.grid(True)
plt.show()
````

````python
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt

# Constantes físicas
h = 6.62607015e-34  # Constante de Planck [J s]
c = 299792458       # Velocidad de la luz [m/s]
K_B = 1.380649e-23  # Constante de Boltzmann [J/K]
sigma = 5.67e-8     # Constante de Stefan-Boltzmann [W m^-2 K^-4]

# Función integrando después de sustitución
def integrand(x):
    return (x**3) / (np.exp(x) - 1)

# Calcular la integral de 0 a ∞
integral_value, _ = quad(integrand, 0, np.inf)

# Verificar que el valor de la integral es aproximadamente π^4 / 15
expected_value = (np.pi**4) / 15
print(f"Valor calculado de la integral: {integral_value}")
print(f"Valor esperado de la integral: {expected_value}")

# Calcular el término a y verificar que es igual a sigma
a = (2 * np.pi * (K_B**4)) / ((h**3) * (c**2)) * integral_value
print(f"\nEl término calculado es: {a:.5e} W m^-2 K^-4")
print(f"Constante de Stefan-Boltzmann (sigma): {sigma} W m^-2 K^-4")
print(f"Diferencia relativa: {abs(a - sigma) / sigma}")

# Mostrar que la potencia total es proporcional a T^4
T_values = np.linspace(100, 5000, 100)  # Rango de temperaturas [K]
P_total_values = sigma * T_values**4  # Potencia total por unidad de área [W/m^2]

# Graficar P_total vs. T^4
plt.figure(figsize=(10, 6))
plt.plot(T_values**4, P_total_values, label='$P_{total} = \sigma T^4$', color='blue')
plt.xlabel('T^4 (K^4)')
plt.ylabel('Potencia Total Emitida por Unidad de Área (W/m^2)')
plt.title('Potencia Total Emitida por Unidad de Área vs. Temperatura')
plt.legend()
plt.grid(True)
plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt

def procesar_datos():
    """
    Carga y preprocesamiento de los datos de anomalías de temperatura terrestre.

    Devuelve:
        DataFrame con los datos procesados.
    """
    url = "https://data.giss.nasa.gov/gistemp/graphs/graph_data/Global_Mean_Estimates_based_on_Land_and_Ocean_Data/graph.txt"
    datos = pd.read_csv(url, sep='\s+', header=None, skiprows=5)
    datos.columns = ["Año", "NoSuavizado", "Promedio_5"]
    return datos

def graficar_anomalias(datos):
    """
    Crea un gráfico combinado de anomalías de temperatura terrestre.

    Parámetro:
        datos: DataFrame con los datos procesados.
    """
    plt.figure(figsize=(12, 6))

    plt.plot(datos["Año"], datos["Promedio_5"], label="Promedio Móvil 5 Años")
    #plt.scatter(datos["Año"], datos["NoSuavizado"], alpha=0.7, label="No Suavizado")
    plt.plot(datos["Año"], datos["NoSuavizado"], alpha=0.7, label="No Suavizado")


    plt.xlabel("Año")
    plt.ylabel("Anomalía de Temperatura (°C)")
    plt.title("Anomalías de Temperatura Terrestre (1880 - 2021)")
    plt.grid(True)
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Carga y preprocesamiento de datos
datos = procesar_datos()

# Gráfico de anomalías de temperatura
graficar_anomalias(datos)
````

````python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Selección de datos del período 2000-2020
datos_2000_2020 = datos[(datos["Año"] >= 2000) & (datos["Año"] <= 2020)]

# Extracción de datos sin suavizar y años
anio_2000_2020 = datos_2000_2020["Año"].to_numpy()
temperatura_sin_suavizar_2000_2020 = datos_2000_2020["NoSuavizado"].to_numpy()

def linear_func(x, m, b):
  return m * x + b

# Ajuste de modelo lineal
params, _ = curve_fit(linear_func, anio_2000_2020, temperatura_sin_suavizar_2000_2020)

# Parámetros del ajuste
m_lineal, b_lineal = params

# Generación de datos de x para la línea de ajuste
anio_fit = np.linspace(min(anio_2000_2020), max(anio_2000_2020), 100)
temperatura_fit_lineal = linear_func(anio_fit, m_lineal, b_lineal)

# Gráfico de datos y ajuste lineal
plt.figure(figsize=(10, 6))
plt.scatter(anio_2000_2020, temperatura_sin_suavizar_2000_2020, label='Datos sin Suavizar (2000-2020)')
plt.plot(anio_fit, temperatura_fit_lineal, label='Ajuste Lineal', color='red')
plt.xlabel('Año')
plt.ylabel('Anomalía de Temperatura (°C)')
plt.title('Ajuste Lineal de Anomalías de Temperatura (2000-2020)')
plt.grid(True)
plt.legend()
plt.show()
````

````python
from scipy.interpolate import interp1d

# Interpolación de datos
interpolador = interp1d(anio_2000_2020, temperatura_sin_suavizar_2000_2020)

# Generación de datos de año para la interpolación
anio_interpolacion = np.linspace(min(anio_2000_2020), max(anio_2000_2020), 1000)

# Cálculo de temperatura interpolada
temperatura_interpolada = interpolador(anio_interpolacion)

# Gráfico de datos y curva interpolada
plt.figure(figsize=(12, 6))
plt.scatter(anio_2000_2020, temperatura_sin_suavizar_2000_2020, label='Datos sin Suavizar (2000-2020)')
plt.plot(anio_interpolacion, temperatura_interpolada, label='Interpolación', color='blue')
plt.plot(anio_fit, temperatura_fit_lineal, label='Ajuste Lineal', color='red')
plt.xlabel('Año')
plt.ylabel('Anomalía de Temperatura (°C)')
plt.title('Comparación: Interpolación vs Ajuste Lineal (2000-2020)')
plt.grid(True)
plt.legend()
plt.show()
````
