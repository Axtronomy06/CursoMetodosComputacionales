Disenar un programa, en el que entrada la función y su derivada, definidas como funciones lambda de python, se determine el valor mínimo de esa función. Prueba tu algoritmo para la función f(x)=x2

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# POR FAVOR ESCRIBA SU FUNCION DESCENDENTE Y SU CORRESPONDIENTE DERIVADA, Y LUEGO HAGA CLICK EN EJECUTAR TODO
funcion = lambda x: x**2

# Derivada de la función
derivada = lambda x:2*x

# Punto inicial
x0 = 2

# Tasa de aprendizaje
alfa = 0.1

# Número de iteraciones
Np = 100

# Arrays para almacenar los resultados
xf = np.zeros(Np)
valores = np.zeros(Np)
num_iteraciones = np.zeros(Np)

# Bucle para iterar el método del gradiente descendiente
for i in range(0, Np):
    # Actualización del valor de x
    xf[i] = x0 - alfa * derivada(x0)

    # Evaluación de la función en el nuevo valor de x
    valores[i] = funcion(x0)

    # Almacenamiento del número de iteración
    num_iteraciones[i] = i

    # Actualización del valor de x para la siguiente iteración
    x0 = xf[i]


```

```python
# Impresión del mínimo de la función
print("El mínimo de la función descendente es ", round(np.min(valores), 2))

```

Realiza el grafico de la funcion y la derivada, especifica la leyenda de cada curva empleando el metodo legend() de matplotlib

```python
# Gráfica de la función y su derivada
x2 = np.arange(-4, 4, 0.01)
eje_y_f = funcion(x2)
eje_y_d = derivada(x2)


```

```python
plt.plot(x2, eje_y_f, label="Funcion")
plt.plot(x2, eje_y_d, label="Derivada")
plt.title("Función y su Derivada")
plt.xlabel("Dominio")
plt.ylabel("Rango")
plt.grid()
plt.legend()
plt.show()
```

Analiza como se minimiza la función, gráficando de la funcion en cada punto a iterar, como funcion del numero de iteraciones. Para dos valores diferentes de la constante alpha (prueba con  α=0.1 ,  α=0.01 )

```python
# Valores para alfa2 = 0.01

alfa2 = 0.01

# Valor inicial igual al anterior de x02=2, Np=1000 igual anterior

x02 = 2

Np = 100
xf2 = np.zeros(Np)

v_funcion_2 = np.zeros(Np)

for i in range(0,Np):
  xf2[i] = x02 - alfa2 * (derivada(x02))
  v_funcion_2[i] = funcion (x02)
  x02 = xf2[i]

```

```python
# NUMERO DE ITERACCIONES

num_valores = np.zeros(len(valores))
num_valores2 = np.zeros(len(v_funcion_2))

for i in range(len(valores)):
  num_valores[i] = i

for i in range(len(v_funcion_2)):
  num_valores2[i] = i

```

```python
plt.plot(num_valores,valores, label = "alfa = 0.1" )
plt.plot(num_valores2,v_funcion_2, label = f"alfa = {alfa2}" )
plt.title(" F(xi) vs Numero de iteraciones")
plt.ylabel("F(xi)")
plt.xlabel("Numero de iteraciones")
plt.grid()
plt.legend()
```

Analisis: Para  diferentes valores de alfa (taza de aprendizaje) la función se minimiza en  un mismo  numero. En el caso de X^2 las dos curvas se minimizan en cero,  pero lo hacen con diferentes inclinaciones(Ver grafica anterior).

Por ejemplo, en el caso particular de X^2,  para un alfa=0.1, se alcanza el valor minimo con un menor numero de iteraciones(un numero menor de 20 ) mientras que para un alfa = 0.01, se el alcanza despues de  80 iteraciones, en consecuencia,  la curva tienen un descenso más suave que la de alfa=0.1.

a. Programar la ecuación para el mapa logístico, y realizar una grafica del valor de xn como funcion de n.

```python
def mapa_logistico (r,x0,n):
  x = np.zeros(n)
  for i in range(0,n):
    xn = r * x0 * (1-x0)
    x[i] =x0
    x0 = xn
  return x

mapa = mapa_logistico(1,0.5,100)
n=np.arange(0,100,1)
```

```python
plt.plot(n,mapa , "ro")
plt.xlabel=("Iteraciones")
plt.ylabel = (" F")
plt.show()
```
