```python
import numpy as np
import matplotlib.pyplot as plt

def interpolacion(x, y):
    # Comprobamos que la longitud de x e y sea la misma y mayor que 1
    if len(x) != len(y) or len(x) < 2:
        raise ValueError("Las listas x e y deben tener la misma longitud y al menos 2 puntos.")

    plt.figure()
    for i in range(len(x) - 1):
        # Cálculo de la pendiente (m) entre los puntos (x[i], y[i]) y (x[i+1], y[i+1])
        m = (y[i+1] - y[i]) / (x[i+1] - x[i])

        # Generación de puntos intermedios en el intervalo [x[i], x[i+1]]
        x_interpolado = np.linspace(x[i], x[i+1], 100)

        # Cálculo de los valores interpolados utilizando la ecuación de la recta
        recta_interpolada = y[i] + m * (x_interpolado - x[i])

        # Dibujar los puntos de datos originales y la interpolación
        plt.plot(x, y, "ro")
        plt.plot(x_interpolado, recta_interpolada)

    plt.legend()
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Interpolación Lineal")
    plt.grid(True)
    plt.show()

# Datos de ejemplo
x = np.linspace(1, 10, 10)
y = np.array([1, 3, 2, 4, 5, 7, 6, 8, 9, 10])

interpolacion(x, y)
```
