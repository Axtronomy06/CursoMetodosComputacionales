
CAMPO ELECTRICO
---



```python
import numpy as np
import matplotlib.pyplot as plt

def campo_electrico(xp, yp, x1, y1, q):
    # Resta
    delta_x = xp - x1
    delta_y = yp - y1
    # Magnitud
    magnitud = np.sqrt(delta_x**2 + delta_y**2)
    # Vector unitario
    unitario_x = delta_x / magnitud
    unitario_y = delta_y / magnitud
    k = 1
    Ex = (k * q * unitario_x) / (magnitud)**2
    Ey = (k * q * unitario_y) / (magnitud)**2
    return Ex, Ey

# cuadrícula de puntos
x_vals, y_vals = np.meshgrid(np.linspace(-2, 2, 25), np.linspace(-2, 2, 25))

# Cargas del dipolo
carga_positiva = 1
carga_negativa = -carga_positiva

# campos eléctricos de cada carga
campo_1 = campo_electrico(x_vals, y_vals, -1, 0, carga_positiva)
campo_2 = campo_electrico(x_vals, y_vals, 1, 0, carga_negativa)

# Normalizar vectores
magnitud = np.sqrt((campo_1[0] + campo_2[0])**2 + (campo_1[1] + campo_2[1])**2)
campo_1 = campo_1 / magnitud
campo_2 = campo_2 / magnitud

# Graficar campo vectorial
plt.figure()
plt.title("Campo Vectorial")
plt.quiver(x_vals, y_vals, campo_1[0] + campo_2[0], campo_1[1] + campo_2[1], scale=70)
plt.scatter([-1, 1], [0, 0], color=["red", "blue"], label=["Carga Q", "Carga -Q"])
plt.legend(loc="upper right")
plt.show()

# Graficar líneas de campo
plt.figure()
plt.title("Líneas de Campo")
plt.streamplot(x_vals, y_vals, campo_1[0] + campo_2[0], campo_1[1] + campo_2[1], color="grey")
plt.scatter([-1, 1], [0, 0], color=["red", "blue"], label=["Carga Q", "Carga -Q"])
plt.legend(loc="upper right")
plt.show()
```
