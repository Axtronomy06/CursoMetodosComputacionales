```python
from scipy.misc import derivative
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-1,1,2000000)

f=lambda x,a,b: a*np.sin(x)+ (b*np.tan(x))**2
a =1
b = 2
de= derivative(f,x,dx=1e-7,args=(a,b))
de
```

```python
y = f(x,a,b)
plt.grid()
plt.ylim(-1,1)
plt.plot(x,y)
plt.plot(x,de)
```

```python
@np.vectorize
def fpri(f,x):
  delta =0.00001
  der = 1/(2+delta)*(f(x+delta)-f(x-delta))
  return der
```

```python
g = lambda x:x**3
x = np.linspace(-2,2,100)
plt.plot(x,g(x))
plt.plot(x,fpri(g,x))
```
