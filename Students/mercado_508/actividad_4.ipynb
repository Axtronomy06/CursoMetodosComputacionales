```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("https://raw.githubusercontent.com/hernansalinas/autogradesMetodosComputacionales/main/Dataset/Temperature_Bogota_Barcelona.csv")
df.head(5)
```

```python
df['AvgTempCelcius'] = (df['AvgTemperature'] - 32) / 1.8
```

```python
# Filtrar datos para Bogotá
bogota_data = df[df['City'] == 'Bogota']
plt.figure(figsize=(10, 5))
plt.plot(bogota_data['Year'], bogota_data['AvgTempCelcius'], label='Bogota', color='red')
plt.xlabel('Año')
plt.ylabel('Temperatura Promedio (°C)')
plt.title('Temperatura Promedio en Bogotá por Año')
plt.legend()
plt.grid()
print(bogota_data.head())
# Filtrar datos para Barcelona
barcelona_data = df[df['City'] == 'Barcelona']
plt.figure(figsize=(10, 5))
plt.plot(barcelona_data['Year'], barcelona_data['AvgTempCelcius'], label='Barcelona', color='orange')
plt.xlabel('Año')
plt.ylabel('Temperatura Promedio (°C)')
plt.title('Temperatura Promedio en Barcelona por Año')
plt.legend()
plt.grid()
plt.show()
```
