1.multiplicar dos matrices.

```python
l = int(input("ingrese el numero de filas de la matriz1: "))
f = int(input("ingrese el numero de columnas de la matriz2: "))
o = int(input("ingrese el numero de columnas de la matriz1/numero de filas matriz2: "))
print("ingrese los valores de la  matriz1: ")
def multiplicacion_matrices(p,q,n,matrix1,matrix2):
  resul = [[0 for i in range(q)] for j in range(p)]
  for i in range(p):
    for j in range(q):
      for k in range(n):
        resul[i][j] = resul[i][j] + matrix1[i][k] * matrix2[k][j]
  print("resultado es: ")
  for i in range(p):
    print("\n")
    for j in range(q):
      print(format(resul[i][j],"<10"),end="")

  return
matriz1 = [[int(input()) for i in range(o) ] for j in range(l)]
print("")
print("matriz1:")
for i in range(l):
 print("\n")
 for j in range(o):
    print(format(matriz1[i][j],"<10"),end="")
print("")
print("ingrese los valores de la matriz2: ")
matriz2 = [[int(input()) for i in range(f)] for j in range(o)]
print("matriz2: ")
for i in range(o):
  print("\n")
  for j in range(f):
    print(format(matriz2[i][j],"<10"),end="")
print("")
multiplicacion_de_matrices = multiplicacion_matrices(l,f,o,matriz1,matriz2)
print(multiplicacion_de_matrices)
```

2.sumar dos matrices.

```python
def suma_matrices(a,b,matrix3,matrix4):
  result = [[0 for i in range(b)] for j in range(a)]
  for i in range(a):
    for j in range(b):
      result[i][j] = matrix3[i][j] + matrix4[i][j]
  print("resultado")
  for i in range(a):
    print("\n")
    for j in range(b):
      print(format(result[i][j],"<3"),end="")
  print("")
  return
fila = int(input("ingrese numero de filas: "))
columna = int(input("ingrese numero de columnas: "))

print(" ingrese elementos a la matriz1: ")
matriz3 = [[int(input()) for i in range(columna)] for j in range(fila)]
print("matriz1: ")
for i in range(fila):
  print("\n")
  for j in range(columna):
    print(format(matriz3[i][j],"<3"),end="")
print()
print("ingrese  elementos a la matriz2: ")
matriz4 = [[int(input()) for i in range(columna)] for j in range(fila)]
print("matriz2: ")
for i in range(fila):
  print("\n")
  for j in range(columna):
    print(format(matriz4[i][j],"<3"),end="")
print()
suma_de_matrices = suma_matrices(fila,columna,matriz3,matriz4)
print(suma_de_matrices)
```

```python
#matriz transpuesta
def matriz_transpuesta(a,b,matrix7):
  result2 = [[0 for i in range(a)] for j in range(b)]
  for i in range(b):
    for j in range(a):
      result2[i][j] = matrix7[j][i]
  print("resultado")
  for i in range(b):
    print("\n")
    for j in range(a):
      print(format(result2[i][j],"<4"),end="")
  print()
  return
P = int(input("ingrese numeros de filas:"))
Q = int(input("ingrese numeros de columnas:"))
print("ingrese valores a la matriz1")
matriz7 = [[int(input()) for  i in range(Q)] for j in range(P)]
print("matriz1")
for i in range(P):
  print("\n")
  for j in range(Q):
    print(format(matriz7[i][j],"<3"),end=" ")
print()
transpuesta_de_matriz = matriz_transpuesta(P,Q,matriz7)
print(transpuesta_de_matriz)
```

4. La matriz de rotación en 2D. Es una matriz que se usa para rotar puntos en el plano xy en sentido antihorario alrededor del origen. La matriz tiene la siguiente forma:
   \begin{equation}
 \begin{bmatrix}
     \cos \theta & -\sin \theta \\
     \sin \theta & \cos \theta
 \end{bmatrix}
 \end{equation}


 Construir un programa en python que permita realizar la rotación de un vector un ángulo theta, Mostrar la representación vectorial en el plano indicando la rotacion a realizar.

Pruebe su solución para el vector $\\vec{v}=(1,1)$
 rotado 20grados en sentido antihorario. Muestre el vector antes de la rotación y después de la rotación


    

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el vector original (coordenadas x e y)
vector = np.array([1, 0])

# Ángulo de rotación en grados
theta_degrees = 20

# Convertir el ángulo a radianes
theta_radians = np.deg2rad(theta_degrees)

# Definir la matriz de rotación
rotation_matrix = np.array([[np.cos(theta_radians), -np.sin(theta_radians)],
                            [np.sin(theta_radians), np.cos(theta_radians)]])

# Realizar la rotación del vector
rotated_vector = np.dot(rotation_matrix, vector)

# Mostrar el vector original y el vector rotado
plt.figure()
plt.quiver(0, 0, vector[0], vector[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector Original')
plt.quiver(0, 0, rotated_vector[0], rotated_vector[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector Rotado')
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.title(f'Rotación de {theta_degrees} grados en sentido antihorario')
plt.grid(True)
plt.show()
```
