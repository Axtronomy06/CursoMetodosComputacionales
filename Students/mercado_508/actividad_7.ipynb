```python
import matplotlib.pyplot as plt
import sympy as sym
import numpy as np

# Datos de entrada
xi = np.array([2, 5, 6])
fi = np.array([4, 1,2])
x = sym.Symbol('x')

# Inicialización del polinomio
polinomio = 0
n = len(xi)

# Calcular los productos de diferencias divididas
for i in range(n):
    numerador = 1
    denominador = 1
    for j in range(n):
        if j != i:
            numerador *= (x - xi[j])
            denominador *= (xi[i] - xi[j])
           #  print(denominador)
    L = numerador / denominador
    polinomio += L * fi[i]

# Imprimir el polinomio resultante
print("Polinomio resultante:", polinomio)
```

```python
polisimp = polinomio.expand()
print(polisimp)
```

```python
sym.simplify(polinomio)
```

```python
#grafica de polinomio
pol = sym.lambdify(x,polisimp)

a = min(xi)
b = max(xi)

pxi = np.linspace(a,b,100)
pyi = pol(pxi)
plt.figure()
plt.scatter(xi,fi)
plt.plot(pxi,pyi,color='red')
plt.grid()
plt.show()
```
