Dada el siguiente sistema de ecuaciones, asociadas a un modelo simplificado de la atomosfera, encontrar los valores de x, y, z.

\begin{equation}
\frac{\mathrm{d}x}{\mathrm{d}t} = \sigma (y-x)
\end{equation}


\begin{equation}
\frac{\mathrm{d}y}{\mathrm{d}t} = x(\rho-z)-y
\end{equation}

\begin{equation}
\frac{\mathrm{d}z}{\mathrm{d}t} = xy-\beta z
\end{equation}

El modelo anteror permite describir de un a manera simplificada el movimiento atmosférico de masas de aire, los valores x, y, z se corresponden a posiciones de un volumen que contiene  una masa de aire, y los párametros $\sigma$, $\beta$ y  $\rho$ hacen referencia a, las siguientes cantidades fisicas:

\begin{equation}
\sigma = \frac{\mathrm{viscosidad}}{\mathrm{conductividad\ térmica }},
\end{equation}

$\rho$ hace refencia al término de convección entre el suelo y el techo del volumen discretizado.

\begin{equation}
\rho =\frac{g\beta}{v \alpha} (T_s-T_{\infty})L^3
\end{equation}

y $\beta=\frac{\mathrm{Ancho\ del\ cubo}}{\mathrm{Altura\ de\ la\ caja}}$ es un termino geométrico.

````python
import numpy as np
from scipy.integrate import odeint
import sympy as sy
import matplotlib.pyplot as plt
````

````python
#Defienido la ecuación
def edo(pvi,t,sigma,ro,beta):
  dxdt=sigma*(pvi[1]-pvi[0])
  dydt=pvi[0]*(ro-pvi[2])-pvi[1]
  dzdt=pvi[0]*pvi[1]-beta*pvi[2]

  u=np.array([dxdt,dydt,dzdt])

  return u
````

- $\sigma = 10.0$, $\beta = 8./3$ ,$\rho = 28$  con condiciones iniciales,  $x_0,y_0,z_0 = [1.5,1.5,1.5]$

````python
#Defiendo los valores inciales
pvi0=np.array([1.5,1.5,1.5])

#Constantes
sigma0=10.0
beta0=8/3
ro0=28

t=np.linspace(-30,30,10000)
````

````python
sol0=odeint(edo,pvi0,t,args=(sigma0,ro0,beta0))
````

- $\sigma = 10.0$, $\beta = 8./3$ ,$\rho = 25$  con condiciones iniciales,  $x_0,y_0,z_0 = [1.5,1,5,1.5]$

````python
#Defiendo los valores inciales
pvi1=np.array([1.5,1.5,1.5])

#Constantes
sigma1=10.0
beta1=8/3
ro1=25
````

````python
sol1=odeint(edo,pvi1,t,args=(sigma1,ro1,beta1))
````

- $\sigma = 10.0$, $\beta = 8./3$ ,$\rho = 25$  con condiciones iniciales,  $x_0,y_0,z_0 = [1,1,1]$

````python
#Definiendo los valores inicales
pvi2=np.array([1,1,1])

#Constantes
sigma2=10.0
beta2=8/3
ro2=25
````

````python
sol2=odeint(edo,pvi2,t,args=(sigma2,ro2,beta2))
np.shape(sol2)
````

- $\sigma = 10.0$, $\beta = 8./3$ ,$\rho = 25$  con condiciones iniciales,  $x_0,y_0,z_0 = [1001,1.001,1.001]$

````python
#Definiendo los valores inicales
pvi3=np.array([1.001,1.001,1.001])

#Constantes
sigma3=10.0
beta3=8/3
ro3=25
````

````python
sol3=odeint(edo,pvi3,t,args=(sigma3,ro3,beta3))
````

##**Gráfico del atractor de Lorentz**

````python
fig,ax=plt.subplots(2, 2, figsize=(10,8),subplot_kw={'projection': '3d'})
ax[0,0].plot(sol0[:,0],sol0[:,1],sol0[:,2], color='green')
ax[0,0].set_title('Solución PVI 0')
ax[0,1].plot(sol1[:,0],sol1[:,1],sol1[:,2], color='blue')
ax[0,1].set_title('Solcuión PVI 1')
ax[1,0].plot(sol2[:,0],sol2[:,1],sol2[:,2], color='purple')
ax[1,0].set_title('Solución PVI 2')
ax[1,1].plot(sol3[:,0],sol3[:,1],sol3[:,2], color='red')
ax[1,1].set_title('Solcuión PVI 3')
fig.suptitle('Atractor de Lorentz')
plt.draw()
plt.show()
````

5. ¿Cómo puede estar seguro de que la solución es la esperada?

Para las diferentes condiciones inciales, los puntos resultantes no describen el mismo comportamiento, efectivamente es caótico. Pero como lo propone el atractor, pronto forma la figura de un mariposa con líneas que no se tocan.

Cada condición, hace que el atractor desarolle el caos de formas diferentes, pero siempre tiende a formar la figura de la mariposa.

Por tanto, esto es una solcuión.
