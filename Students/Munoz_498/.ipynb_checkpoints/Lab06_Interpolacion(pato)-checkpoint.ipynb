```python
#Librerías
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy as sc
import sympy as sy
import urllib.request
```

```python
#Listas de valores para el gráfico
x=np.array([0.9,1.3,1.9,2.1,2.6,3.0,3.9,4.4,4.7,5.0,6.0,7.0,8.0,9.2,10.5,11.3,11.6,12.0,12.6,13.0,13.3])
y=np.array([1.3,1.5,1.85,2.1,2.6,2.7,2.4,2.15,2.05,2.1,2.25,2.3,2.25,1.95,1.4,0.9,0.7,0.6,0.5,0.4,0.25])
print(min(y))
```

```python
#Creando un dataframe con los datos
data={'x':x,'y':y}
df=pd.DataFrame(data)
df
```

###**Representación gráfica del pato**

```python
#Representación gráfica del Pato
pato='https://3.bp.blogspot.com/-8_LTrCIxVSQ/VxU-5Gq2ffI/AAAAAAAAIto/LhCDEwHKTiAbLr7IhizQwGV13YTwgpbDwCLcB/s400/pato1.png' #URL donde está el pato
urllib.request.urlretrieve(pato, 'pato.jpg')#Decargar la imagen donde está el pato
imagen_pato=plt.imread('pato.jpg') #Variable con la imagen del pato

#Finalmente para mostrar el pato
plt.figure()
plt.title('Representación del pato')
plt.imshow(imagen_pato,extent=(0.4, 14, -6.8, 3.45)) #Para crear un similaridad con la imagen adjuntada en esta actividad
plt.plot(df['x'],df['y'],'.b')
plt.xlim(0, 14)
plt.ylim(-7,4)

plt.grid()
```

```python
plt.figure(figsize=(10,3))
plt.title('Paloma con los puntos de la evaluación')
plt.plot(df['x'],df['y'],'.m',markersize=15)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

###**Polinomio de Lagrange**

```python
#Utilzando la función definida por scipy
p=sc.interpolate.lagrange(x,y)
type(p) #Para observar el tipo de variable que es p
```

```python
#Gráfico, polinomio de Lagrange
x_intervalo=np.linspace(min(x),max(x),100)
plt.figure(figsize=(10,3))
plt.plot(x_intervalo,p(x_intervalo),'r',label='p(x) con Lagrange')
plt.title('Paloma con polinomio de Lagrange')
plt.ylabel('p(x)')
plt.xlabel('x')
plt.legend()
plt.grid()
```

###**Polinomio de CubicSpline**

```python
#Utilizando la función de scipyCubicSpline
p_c=sc.interpolate.CubicSpline(x,y)
type(p_c)
```

```python
#Grafico, polinomio de CubicSpline
plt.figure(figsize=(10,3))
plt.plot(x_intervalo,p_c(x_intervalo),'g',label='p(x) de CubicSpline')
plt.title('Paloma con polinomio de CubicSpline')
plt.ylabel('p(x)')
plt.xlabel('x')
plt.legend()
plt.grid()
```

###**Polinomio de CubicHermiteSpline**

```python
#Obteniendo una derivada númerica, a través de una aproximación con la pendiente
#Pencientes usando comandos de numpy
m=np.diff(y)/np.diff(x)


#Para comprobar que el resultado de pendientes 'm' anterior es correcto.
pendientes=np.zeros(len(x)-1)
for i in range(0,len(x)-1):
  mi=(y[i+1]-y[i])/(x[i+1]-x[i]) #Para obtener los valores de las pendientes
  pendientes[i]=mi

print(m-pendientes)
```

```python
#Para tener la misma cantidad de puntos que tiene el vector de pendientes
x1=np.delete(x,20)
y1=np.delete(y,20)
#Interpolación con CubicHermiteSpline
p_cd=sc.interpolate.CubicHermiteSpline(x1,y1,m)
```

```python
#Gráfico con CubicHermiteSpline
#Intervalo con el nuevo valor de x
x1_intervalo=np.linspace(min(x1),max(x1),100)
plt.figure(figsize=(10,3))
plt.plot(x1_intervalo,p_cd(x1_intervalo),'m',label='p(x) de CubicHermiteSpline ')
plt.title('Paloma con polinomio de CubicHermiteSpline')
plt.ylabel('p(x)')
plt.xlabel('x')
plt.legend()
plt.grid()
```

###**Comparación de los resultados obtenidos con los distintos métodos de interpolación**

```python
#Gráfico con superposicón de los resultados
plt.figure(figsize=(10,3))
plt.plot(df['x'],p(df['x']),'r',label='p(x) de Lagrange')
plt.plot(df['x'],p_c(df['x']),'g',label='p(x) de CubicSpline')
plt.plot(df['x'],df['y'],'.k',label='Puntos originales')
plt.plot(df['x'],p_cd(df['x']),'m',label='p(x) de CubicHermiteSpline ')
plt.title('Paloma con comparación de métodos de interpolación')
plt.ylabel('p(x)')
plt.xlabel('x')
plt.legend()
plt.grid()
```
