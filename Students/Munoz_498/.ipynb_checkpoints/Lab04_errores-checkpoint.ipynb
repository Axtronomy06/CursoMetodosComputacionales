#**exp(-x)**

Definiendo exp(-x) a partir de la serie de Taylor, para luego realizar una comparación con el resultado obtenido por np.exp(-x). Y analizando los términos que $x\approx 10$

````python
#Serie para exponencial de x

def exponencial_menosx(x,N):

  exponencial_suma= 0 #Para iniciar la sumatoria

  for i in range(0,N):
    exponencial_suma= exponencial_suma + (((-1)**i)*((x**i)/math.factorial(i)))

  return exponencial_suma

z=exponencial_menosx(10,500)
z
````

````python
#Comparación, función definida y función construida.

def valor_n(x,N):
  exponencial_sum=0

  cota=np.absolute((((-1)*x)**(N+1))/(math.factorial(N+1)))

  for i in range(0,N):
    exponencial_sum= exponencial_sum + (((-1)**i)*((x**i)/math.factorial(i)))

  abs_exponencial_sum=(10**(-7))*np.absolute(exponencial_sum)

  if cota<=abs_exponencial_sum:
    return N, 'es un buen valor para n'
  else:
    return 'No es un valor acorde para N'
````

#**${e^{-x}=\frac{1} {e^x} }$**
Se calculará la exp(x), con la serie de Taylor:

\begin{equation}
e^{x}=\sum_{n=0}^{N}  \frac{x^{n}}{n!}
\end{equation}

Y 1 será dividido por este valor, para luego comparar con np.exp(-x) y exponencial_menox

````python
#Se intenta usanto el 1/exp(x), para reducir el error que pueda presentar la fución exp(-x) con el término alternane

def exponencial_menosxcambio(x,N):

  exponencial_suma= 0 #Para iniciar la sumatoria

  for i in range(0,N):
    exponencial_suma= exponencial_suma + ((x**i)/math.factorial(i))

  return 1/exponencial_suma #exp(-x)

z=exponencial_menosxcambio(0.5,15)
print(z/np.exp(-0.5))
````

```

  r =z/np.exp(-0.5)
```

Mientras que para, la función exponecial_menosx, se tiene que:
```
  a=exponencial_menosx(0.5,15)
```

# **Cancelación sustractiva**



**serieSn**. Un ejemplo de la cancelacion substractiva es en la suma de series alternantes. Considere la serie $\mathrm{S}_N$,  escrita  matematicamente de tres formas diferentes:

-
\begin{equation}
\mathrm{S}_N^{(1)} = \sum_{n = 1}^{2N} (-1)^n \frac{n}{n+1}
\end{equation}

- Sumando números pares e impares.

\begin{equation}
\mathrm{S}_N^{(2)} = -\sum_{n = 1}^{N}\frac{2n-1}{2n} + \sum_{n=1}^{N} \frac{2n}{2n+1}
\end{equation}

- Combinando las series de una forma analítica

\begin{equation}
\mathrm{S}_N^{(3)} = \sum_{n = 1}^{N}\frac{1}{2n(2n+1)}
\end{equation}



Las tres sumas $\mathrm{S}_N^{(1)}$, $\mathrm{S}_N^{(2)}$ y $\mathrm{S}_N^{(3)}$ son matemáticamente iguales, pero los resultados  pueden tener resultados numéricos diferentes.


- Escribir un programa que realice el cálculo de $\mathrm{S}_N^{(1)}$, $\mathrm{S}_N^{(2)}$ y $\mathrm{S}_N^{(3)}$
- Estudiar la convergencia de la serie para los tres casos.
- Considere  que $S_N^{(3)}$ es la solución exacta, realizar dos graficos log-log, del error relativo en función
del número de términos ($\mathrm{Log_{10}} E$ VS $\mathrm{Log_{10}} N$), para $S_N^{(1)}$ y $S_N^{(2)}$. Donde,  $E = |(S_N^{(i)} - S_N^{(3)})/S_N^{(3)}|$, i = 1, 2. Considere valores de $N = 1\times 10^{6}$.
- Analizar y discutir los resultados. Tenga en cuenta la discusión realizada sobre teoría de errores en la página
39 del libro del libro [Computational Physics. ](https://www.dropbox.com/s/n06ul2r4l65khl6/Computational%20Physics%20-%20Problem%20Solving%20with%20Computers%2C%203527406263.pdf?dl=0)

````python
#Cálculo de series
#Se utilizan los for para iterar sobre los indices y obtener las sucesiones de sumas
def S_1N(N):
  suma_1N=0
  for n in range(1,(2*N)+1):
    suma_1N= suma_1N + (((-1)**n)*(n/(n+1))) #Aplica formula

  return suma_1N

def S_2N(N):
  pares=0
  for m in range(1,N+1):
    pares= pares + (((2*m)-1)/(2*m))

  impares=0
  for n in range(1,N+1):
    impares=impares + ((2*n)/((2*n)+1)) #Aplica formula

  return impares-pares

def S_3N(N):
  suma_3N=0
  for n in range(1,N+1):
    suma_3N= suma_3N+((1)/((2*n)*((2*n)+1))) #Aplica formula

  return suma_3N


print(S_1N(2))
print(S_2N(2))
print(S_3N(2))
````

Para la serie
  \begin{equation}
\mathrm{S}_N^{(1)} = \sum_{n = 1}^{2N} (-1)^n \frac{n}{n+1}
\end{equation}

Usando el criterio de divergencia, tomando el ${a_n}={(-1)^n \frac{n}{n+1}}$. Se tiene que:

$\lim_{n \to \infty} {(-1)^n \frac{n}{n+1}}$ este límite no existe, por tan la **serie diverge**.

````python
serie_1N=np.array([S_1N(n) for n in N]) #El vector presenta las componentes para cada iteración

plt.figure()
plt.plot(N, serie_1N, '.g')
plt.title('$\mathrm{S}_N^{(1)}$')
plt.xlabel('''N
Figura 5. Muestra la evaluación de la serie $\mathrm{S}_N^{(1)}$ para ${1 \leq N \leq 200}$.''')
plt.grid()
````

Para la serie
\begin{equation}
\mathrm{S}_N^{(2)} = -\sum_{n = 1}^{N}\frac{2n-1}{2n} + \sum_{n=1}^{N} \frac{2n}{2n+1}
\end{equation}

Por separado, la serie para pares diverge, al igual que la serie para impares. Entonces reecribiendo tenemos que:
\begin{equation}
\mathrm{S}_N^{(2)} = \sum_{n = 1}^{N}\frac{1} {4n^{2}+{2n}}
\end{equation}

Por criterio de compración de límite, comparando con la p-serie convergente $\sum_{n = 1}^{N}\frac{1} {n^{2}}$. Se obtiene el limite L:

$L = \lim_{n \to \infty} {\frac{n^{2}}{4n^{2}+{2n}}}= \frac{1}{4}$, como ${0 < L < \infty}$ **la serie converge**.

````python
serie_2N=np.array([S_2N(n) for n in N]) #El vector presenta las componentes para cada iteración

plt.figure()
plt.plot(N, serie_2N, '.m')
plt.title(' $\mathrm{S}_N^{(2)}$')
plt.xlabel('''N
Figura 6. Muestra la evaluación de la serie $\mathrm{S}_N^{(2)}$ para ${1 \leq N \leq 200}$.''')
plt.grid()
````

Utilizando el mismo desarrollo que se usa para la serie $\mathrm{S}_N^{(2)}$, se puede ver por criterio de comparación de limite se puede probar que $\mathrm{S}_N^{(3)} = \sum_{n = 1}^{N}\frac{1}{2n(2n+1)}$ **converge**.

````python
serie_3N=np.array([S_3N(n) for n in N]) #El vector presenta las componentes para cada iteración

plt.figure()
plt.plot(N, serie_3N, '.b')
plt.title(' $\mathrm{S}_N^{(3)}$')
plt.xlabel('''N
Figura 6. Muestra la evaluación de la serie $\mathrm{S}_N^{(3)}$ para ${1 \leq N \leq 200}$.''')
plt.grid()
````

````python
#Asumiendo que la solución exacta se da con la tercera serie

fig, axs = plt.subplots(2, 1, figsize=(7,7))

N_slog=np.arange(1,800)
N_log=[]

for n in N_slog:
  N_log.append(np.log10(n)) #Obtener el logaritmo de las iteraciones


Log_E1=[]
Log_E2=[]

for n in N_slog: #Para obetener los valores de los logaritmos de cada valor de error relativo en función de las iteraciones N
  log_en1=np.log10(abs((S_1N(n)-S_3N(n))/S_3N(n)))
  log_en2=np.log10(abs((S_2N(n)-S_3N(n))/S_3N(n)))

  Log_E1.append(log_en1)
  Log_E2.append(log_en2)

axs[0].plot(N_log,Log_E1,'.m-',label='Error Relativo $S_N^{1}$')
axs[0].set_title('Error Relativo en Función de N para $S_N^{1}$')
axs[0].set_xlabel('$\mathrm{Log}_{10} N$')
axs[0].set_ylabel('$\mathrm{Log}_{10} E$')
axs[0].legend()
axs[0].grid()

axs[1].plot(N_log,Log_E2,'.g-',label='Error Relativo $S_N^{2}$')
axs[1].set_title('Error Relativo en Función de N para $S_N^{2}$')
axs[1].set_xlabel('$\mathrm{Log}_{10} N$')
axs[1].set_ylabel('$\mathrm{Log}_{10} E$')
axs[1].legend()
axs[1].grid()

fig.text(0.5, 0.00, 'Figura 7. Grafico Log-Log para el error relativo y el valor de iteraciones N', ha='center', fontsize=10)

fig.tight_layout()
````

A medida que N, aumenta el error tambipen lo hace.
