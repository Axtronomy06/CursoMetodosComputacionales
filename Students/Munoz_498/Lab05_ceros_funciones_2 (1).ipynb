###**Pseudocódigo método de la bisección**

1. Se define la función f y los puntos a y b, que representan el intervalo $[a,b]$
2. Definir la tolerancia.
3. Repetir hasta que $|f(c) \leq tolerancia|$ y siempre que $f(a) f(b)<0$:

*  Calcular c como la bisección del intervalo, es decir, el punto medio de $[a,b]$: $c=\frac{a+b}{2}$
* Calcular f(c)
* Si $f(c) ≈ 0$, retornar c como el valor de la raiz
* Si $f(c)<0$:
  $c=a$
*Si $f(x)>0$:
  $c=b$

4. Retornar el valor de c como la raíz de la función

###**Deducción del metodo de Newton-Rapshon**

Considere la función $f(x)$, donde $f(α)=0$, entonces $\alpha$ es una raíz.

Tomando un valor ${x_{n}}$, muy cercano $\alpha$. Se puede tomar la serie de Taylor de f(x) entorno a  ${x_{n}}$:

\begin{equation}
f(x)= f({x}_n) + f'({x_n})(x-{x_n}) + \frac{f''({\xi})}{2} (x-{x_n})^{2}
\end{equation}

Con $\xi \in [x, {x}_n] $

Luego si se hace $x=\alpha$,obteniendo un cero al lado izquierdo de la ecuación. Se tiene que:
\begin{equation}
0=f(\alpha)= f({x}_n) + f'({x}_n)(\alpha - {x}_n) + \frac{f''({\xi})}{2} (\alpha - {x}_n)^{2}
\end{equation}

Entonces despejando $\alpha$ del segundo término de la serie de Taylor para f(x):
\begin{equation}
-f'({x}_n)(\alpha - {x}_n)=f({x}_n) + \frac{f''({\xi})}{2} (\alpha - {x}_n)^{2}
\end{equation}

\begin{equation}
f'({x}_n)(\alpha - {x}_n)=-f({x}_n) - \frac{f''({\xi})}{2} (\alpha - {x}_n)^{2}
\end{equation}

\begin{equation}
(\alpha - {x}_n)=-\frac{f({x}_n)}{f'({x}_n)} - \frac{f''({\xi})}{2} \frac{1}{f'({x}_n)} (\alpha - {x}_n)^{2}
\end{equation}

\begin{equation}
\alpha= {x}_n-\frac{f({x}_n)}{f'({x}_n)} - \frac{f''({\xi})}{2} \frac{1}{f'({x}_n)} (\alpha - {x}_n)^{2}
\end{equation}

Ahora si $\alpha \approx {x}_0$ entonces los términos $(\alpha - {x}_0)$ y $(\alpha - {x}_1)$, son muy pequeños, es decir, que tienen valores muy cercanos a cero. Por tanto el término $ \frac{f''({\xi})}{2} \frac{1}{f'({x}_n)} (\alpha - {x}_n)^{2}$ es depreciable.

Así,

\begin{equation}
\alpha \approx {x}_1= {x}_0-\frac{f({x}_0)}{f'({x}_0)}
\end{equation}

Y repitiendo este proceso, se obtiene una aproximación mejorada:
\begin{equation}
\alpha \approx {x}_2= {x}_1-\frac{f({x}_1)}{f'({x}_1)}
\end{equation}

Así, obteniendo una sucesión que converge a la raíz de f(x), $\alpha$.

###**¿Cuál es la diferencia del metodo de Newton-Rapson con respecto al de la secante?**

El método de Newton- Rapson necesita conocer el valor de la derivada de la función, mientras que el método de la secante no. Pues tomando dos puntos cercanos a la raíz de la función, se puede aproximar la función por una recta y utilizar una proximación a la definición de derivada.

###**Pseudocódigo método Punto Fijo**.

1. Definir la función f(x)
2. Definir g(x), que es restarle a la función identidad, la función que se definió en el punto anterior. Es decir, g(x)= x- f(x).
3. Se define el punto en el que se desea comenzar a evaluar la g(x), p0. Luego p1,será g(p0).
4. Se define la tolerancia.
5. Repetir hasta que $|g(p0)-p0|\leq toleranica $:
*   Calcular p1=g(p0)
*   Actualizar p0=p1
6. Retornar p1 como la aproximación de la raíz.








# 2. Ceros de funciones
*Encontrar* los ceros de $e^{x}-2= \cos(e^x-2)$ por diferentes metodos, bisection, newton-rapshon, punto fijo.

````python
import matplotlib.pyplot as plt
import numpy as np
from scipy import optimize
````

````python
#Igulando la función a cero para conocer el valor de x en el cual se presentá la raíz, se definen la funció y su derivada como:
f=lambda x: ((np.exp(x)-2)-np.cos(np.exp(x)-2))
df= lambda x: (np.exp(x)+ np.sin(np.exp(x)-2)*np.exp(x))
````

````python
#Analizando dónde se podrían encontrar las raíces gráficamente.
intervalo=np.linspace(-4,4,200) #Los valores entre los que se encuentra x

#Grafico
plt.figure()
plt.title('Prueba para los valores de la función')
plt.xlabel('''x
Figura 1.Gráfico de la función, para observar donde pueden
encontrarse sus raices. ''')
plt.ylabel('f(x)')
plt.plot(intervalo,f(intervalo),'.m-')
plt.grid()
````

````python
#En base a lo visto en la figura 1, se hace una evaluación de f en números cercanos a donde se espera este la raíz
x=np.arange(0.5,1.4,0.01) #Para mirar como se comportala función evaluada en esos datos
f(x)
````

###**Método de Bisección**.

````python
#Puntos cercanos a la raíz
a=0.5
b=1.4
#Uso de scipy para calcular la raíz
raiz_bisection=optimize.bisect(f,a,b)
#Resultado
print(f'La raíz es {raiz_bisection} y la función evaluada en ese punto es {f(raiz_bisection)}')
````

###**Método Newton-Rapson**.

````python
#A partir de la derivada, la función, un punto cercano a la raíz definido antes y usando una función de scipy, se cálcula la raíz
raiz_newton=optimize.newton(f,b,df)
#Resultado
print(f'La raíz es {raiz_newton} y la función evaluada en ese punto es {f(raiz_newton)}')
````

###**Método punto fijo**.

````python
#Se define g(x)

g=lambda x: 1-f(x)

#Usando scipy con un valor cercano a la raíz y la función g(x)
raiz_punto_fijo=optimize.fixed_point(g,b)

print(f'La raíz es {raiz_punto_fijo} y la función evaluada en ese punto es {f(raiz_punto_fijo)}')
````

\begin{equation}
m\frac{dv_x}{dt} = -mb(v_x-u\cos \alpha )
\end{equation}

\begin{equation}
m\frac{dv_y}{dt} = -mg-mb(v_x-u\sin \alpha ),
\end{equation}

Posición en x
\begin{equation}

\begin{equation}
y= \frac{1}{b}\left(\frac{g}{b}+v_0 \sin \theta_0 -u\sin\alpha \right)(1-\mathrm{exp}(-bt))-\left( \frac{g}{b}-u\sin\alpha\right)t

##**Alcance máximo R**

````python
#Pasando los valores de theta a radianes para los intervalos pedidos
thetadeg=np.arange(10,80,5)
theta_radianes=np.deg2rad(thetadeg)
````

````python
#Para graficar las trayectorias

def trayectorias(a, alpha, v0, b, u, g):
    # Para calcular las raíces de esta función
    y_valor=lambda t: (1/b) * ((g/b) + v0 * np.sin(a) - u * np.sin(alpha)) * (1 - np.exp(-b * t)) - ((g/b) - u * np.sin(alpha)) * t

    # La raíz es el tiempo de vuelo
    tfi=optimize.newton(y_valor, 10)

    # Intervalo de tiempo
    tin=np.linspace(0, tfi, 1000)

    # Evaluación en x y y
    xe=pos_x(tin, a, alpha, v0, b, u, g)
    ye=pos_y(tin, a, alpha, v0, b, u, g)

    # Diccionario con los datos
    dicc={'x': xe, 'y': ye, 'tiempo': tin}

    df=pd.DataFrame.from_dict(dicc)
    return df
````

````python
#Para graficar las trayectorias, se va autilizar una iteración
plt.figure(figsize=(20,20))

for i in theta_radianes:
    df1 = trayectorias(i, alpha, v0, b, u, g)
    plt.plot(df1['x'],df1['y'],label=f'Ángulo {i} radianes')

plt.title('Trayectorias del proyectil para distintos valores de theta: y en función x')
plt.ylabel('y')
plt.xlabel('''x
Figura 5. Valores de de la posición en
y en función de la posición x''')
plt.legend()
plt.grid()
````

````python
#Ahora construyendo una gráfica de alcance máximo en función del ángulo
def alcance_maximoR(a, alpha, v0, b, u, g):
    # Para calcular las raíces de esta función
    y_valor=lambda t: (1/b) * ((g/b) + v0 * np.sin(a) - u * np.sin(alpha)) * (1 - np.exp(-b * t)) - ((g/b) - u * np.sin(alpha)) * t

    # La raíz es el tiempo de vuelo
    tfi=optimize.newton(y_valor, 10)

    # Evaluación de x en el punto tfi
    xe=pos_x(tfi, a, alpha, v0, b, u, g)

    return xe
````

````python
#Iteraciones para lograr obtner los múltiples valores
plt.figure(figsize=(12,8))

for i in theta_radianes:
    x=alcance_maximoR(i, alpha, v0, b, u, g)
    plt.plot(i,x,'.', label=f'Ángulo {i} radianes',markersize=17)

plt.title('Alcence máximo del proyectil para distintos valore de theta')
plt.ylabel('y')
plt.xlabel('''x
Figura 6. Muestra el alcance máximo del proyectil,
teniedo en cuenta el tiempo final para $0\leq \theta \leq 100$''')
plt.legend()
plt.grid()
````

# 4. Fixed point methods.

When a new planet is discovered, there are different methods to estimate its physical properties. Many times is only possible to estimate either the planet mass or the planet radius and the other property has to be predicted through computer modelling.

If one has the planet mass, a very rough way to estimate its radius is to assume certain composition (mean density) and a homogeneous distribution (a very bad assumption!). For example, for the planet [Gliese 832c](http://es.wikipedia.org/wiki/Gliese_832_c) with a mass $M= 5.40 M_{\oplus}$, if we assume an earth-like composition, i.e. $\bar \rho_{\oplus} = 5520\ kg/m^3$, we obtain:

$$R_{g832c} = \left( \frac{3 M_{g832c}}{ 4 \pi \bar\rho_{\oplus} } \right)^{1/3} \approx 1.75 R_{\oplus}$$

That would be the planet radius if the composition where exactly equal to earth's.

A more realistic approach is assuming an internal one-layer density profile like:

$$\rho(r) = \rho_0 \exp\left( -\frac{r}{L} \right)$$

where $\rho_0$ is the density at planet centre and $L$ is a characteristic lenght depending on the composition. From numerical models of planet interiors, the estimated parameters for a planet of are $M= 5.40 M_{\oplus}$ are approximately $\rho_0 = 18000\ kg/m^3$ and $L = 6500\ km$.

Integrating over the planet volume, we obtain the total mass as

$$M = 4\pi \int_0^R \rho(r)r^2dr$$

This is a function of the mass in terms of the planet radius.

Solving the equation $M(R) = M_{g832c}$ it would be possible to find a more realistic planet radius. However when using numerical models, it is not possible to approach the solution from the left side as a negative mass makes no sense.

Reemplazando $$\rho(r) = \rho_0 \exp\left( -\frac{r}{L} \right)$$ en $$M = 4\pi \int_0^R \rho(r)r^2dr$$

Se tiene que:
$$4\pi \int_0^R \rho_0 \exp\left( -\frac{r}{L} \right)r^2dr$$

Entonces $$M=4\pi \rho_0 \int_0^R \exp\left( -\frac{r}{L} \right)r^2dr$$

Tomando la integral $$S=\int \exp\left( -\frac{r}{L} \right)r^2dr$$. E integrando por partes, se tiene que:

$$S=-L \exp\left( -\frac{r}{L} \right)r^2 +2L \int \exp\left( -\frac{r}{L} \right)rdr$$

Ahora $$S1= \int \exp\left( -\frac{r}{L} \right)rdr$$ Nuevamete integrando por partes:
$$S1=-Lr\exp\left( -\frac{r}{L} \right) +L \int \exp\left( -\frac{r}{L} \right)dr$$
$$S1=-Lr\exp\left( -\frac{r}{L} \right) -L^{2}\exp\left( -\frac{r}{L} \right)$$

Asi, $$S=-L \exp\left( -\frac{r}{L} \right)r^2 + 2L (-Lr\exp\left( -\frac{r}{L} \right) -L^{2}\exp\left( -\frac{r}{L} \right))$$

$$S=-L \exp\left( -\frac{r}{L} \right)r^2 -2L^{2}r\exp\left( -\frac{r}{L} \right)-2L^3\exp\left( -\frac{r}{L} \right)$$


Entonces, utilizando el teorema fundamental del cálculo y demás se tiene que:
$$M(R)=-4\pi \rho_0 L \exp\left( -\frac{R}{L} \right)(R^{2}+{2}LR+2L^2-2L^2\exp\left( \frac{R}{L} \right))$$






````python
import sympy as sp

# Definir las variables y parámetros
rho_0= sp.symbols('rho_0')
R=sp.symbols('R')
L=sp.symbols('L')
r=sp.symbols('r')
M=4*sp.pi*rho_0*(sp.exp(-r / L))*(r**2)

resultado=sp.simplify(sp.integrate(M,(r,0,R)))
print(resultado)
````

Se acaba de probar con sympy que el resultado de la Integral es $$4\pi L \rho_0(2L^2exp(R/L) - 2L^2 - 2LR - R^2)exp(-R/L)$$

````python
#Normalización de la datos
rho_0_normalizado=(rho_0/5520) #Normalizando con la densidad de la Tierra
L_normalizada=(L/6.31e6) #Normalizada con el radio de la Tierra
M_g832c_normalizada=5.4 #Normalizada con la masa de la Tierra
````

````python
#Función para obtener la raíz, que este caso representará el radio
def radio(R,rho_0,L,M_g832c):
  #Sacando de Fator común un menos
  M_R= 4*np.pi*L*rho_0*(2*L**2+ 2*L*R +R**2 -(2*L**2)*np.exp(R/L))*np.exp(-R/L) + M_g832c
  return M_R

radio(1.84,rho_0_normalizado,L_normalizada,M_g832c_normalizada)
````

````python
#Función para analizar con el fixed point

def g(R,rho_0,L,M_g832c):
  g=R-radio(R,rho_0,L,M_g832c)
  return g
````

````python
import numpy as np
import matplotlib.pylab as plt
from PIL import Image
````

````python
Npx_X = 80 #Numero de pixeles en x
Npx_Y = 80 #Numero de pixeles en y

image = Image.new("RGB", (Npx_X, Npx_Y)) #Creacion de la imagen inicial

npx = np.arange(0, Npx_X)    #Imagen vacía con el tamaño especificado
npy = np.arange(0, Npx_Y)

xmin = -1
xmax = 1
ymin = -1
ymax = 1
# Transformación (función) al plano complejo en un rectangulo (xmin, xmax), (ymin, ymax)
zx = npx * (xmax - xmin) / (Npx_X - 1)   + xmin
zy = npy * (ymax-ymin) / (Npx_Y - 1) + ymin

#plt.plot(zx,zy)
````

````python
# Definicion  la funcion y las derivada
def f(z):
  return z**3 - 1

def fp(z):
  return 3*z**2

h = 1E-4
def fpn(z, h):
  return (f(z+h)-f(z))/h
````

````python
a = 0.2
Niteraciones = 120 #Número máximo de iteraciones para cada cuadricula
eps = 1E-8 #Toleranica

x = - 1
for zx_ in zx:
  x = x + 1
  y = -1
  for zy_ in zy:
      y = y + 1
      Z_0 = complex(zx_, zy_)

      for i in range(0, Niteraciones):
        Z = Z_0 - a * f(Z_0)/fp(Z_0) #Método de Newton-Rhapson
        if( abs(Z-Z_0)<= eps):
          break
        Z_0 = Z
      #print(i)
      #print(i)
      r, g, b = i % 4 *64 , i % 8 *16, i% 16* 16
      image.putpixel((x,y), (r,g,b) )


plt.imshow(image)
#plt.xlim(400, 600)
#plt.ylim(400, 523)
````
