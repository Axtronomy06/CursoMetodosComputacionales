```python
#Librerias
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy as sc
import sympy as sy
```

###**Derivadas con scipy y sympy**

```python
#Función de prueba 1

f=lambda x,m: x**m
#Derivada análitica
derivada_= lambda x,m: m*x**(m-1)
#Dominio
x_intervalo=np.linspace(0,2,10)
```

```python
#Derivada con sympy
x__=sy.Symbol('x__')
m_=sy.Symbol('m_')
f__= x__**m_

derivada_f_=sy.diff(f__,x__)

derivada_f_
```

```python
#Derivada con scipy
m=2
derivada=sc.misc.derivative(f,x_intervalo,dx=1E-5, args=(m,))
derivada
```

```python
#Figura para anailizar la relación de las derivadas, análitica y con scipy
plt.figure()
plt.title('Función y derivada')
plt.plot(x_intervalo,f(x_intervalo,2),'.g-', label='$x^2$')
plt.plot(x_intervalo,derivada_(x_intervalo,2),'.r-', label='$2x$,analítica')
plt.plot(x_intervalo,derivada,'.m-', label='$2x$,scipy')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.grid()
plt.legend()
```

```python
#La derivada de la función de prueba 2, con Sympy
x=sy.Symbol('x')
a=sy.Symbol('a')
b=sy.Symbol('b')

funcion1= a*sy.cos(x)+ b*sy.tan(x)**2

derivada=sy.diff(funcion1,x)
derivada
```

##**Derivada para tres puntos**

```python
f1= lambda x: x**2
derivadaf1= lambda x: 2*x

x_intervalo=np.linspace(0,2,100)
derivada1=np.zeros(len(x_intervalo))

h=(2-0)/3

for i in range(0, len(x_intervalo)):
  derivada=(1/(2*h))*(f1(x_intervalo[i]+h)- f1(x_intervalo[i]-h))
  derivada1[i]=derivada



derivada1

```

```python
#Figura para comparar los resultados de la derivadas
plt.figure()
plt.title('Función y derivada')
plt.plot(x_intervalo,derivadaf1(x_intervalo), '.g-', label='$2x$, analítica')
plt.plot(x_intervalo,f1(x_intervalo), '.r-', label='$x^2$')
plt.plot(x_intervalo,derivada1, '.-',label='$2x$, con método de derivada con tres puntos' )
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend()
plt.grid()
```

```python
#Vectorizado, para cualquier función

def derivada_1(funcion,xmin,xmax,salto):

  x_intervalo=np.linspace(xmin,xmax,salto)
  h=(xmax-xmin)/salto
  derivada1=np.zeros(len(x_intervalo))

  for i in range(0, len(x_intervalo)):
    derivada=(1/(2*h))*(funcion(x_intervalo[i]+h)- funcion(x_intervalo[i]-h))
    derivada1[i]=derivada

  return derivada1
```
