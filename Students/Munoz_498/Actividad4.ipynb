```python
#librerías
import pandas as pd
import matplotlib.pyplot as plt
```

###Punto 1
Lectura de archivo a partir de url

```python
#a. Lectura del archivo
url='https://raw.githubusercontent.com/hernansalinas/autogradesMetodosComputacionales/main/Dataset/Temperature_Bogota_Barcelona.csv'
df=pd.read_csv(url)
df
```

###Punto 2
Nueva Columna para la temperatura

```python
#b. Nueva columna
df['Celsius']=(df.AvgTemperature-32/1.8)
df
```

###Punto 3
Filtros para separar los datos de Bogotá y graficarlos

```python
#Filtro para organizar un data frame con los datos de Bogota
tf=df['City']=='Bogota' #Para Bogotá
df1=df[tf].copy()
df1
```

```python
#Filtros
grados=df1.AvgTemperature.unique() #Nuevamente un filtro que retorna como array los valores de una columna, en este caso de AvgTemperature
years=df1.Year.unique() #Un filtro para retornar los de AvgTemperature como un array

#Grafico AvgTempCelcius como función del año para la ciudad de Bogotá
plt.figure()
plt.plot(years,grados,'.m-')
plt.title('Temperatura Bogotá',fontsize=30)
plt.xlabel('Años',fontsize=15)
plt.ylabel('Celsius',fontsize=15)
plt.grid()
```

###Punto 4
Filtro para organizar los datos de temperatura por año para Barcelona y graficarlos

```python
#Filtro para definir un data frame para Barcelona
tf1=(df.City=='Barcelona') #Para Barcelona
df2=df[tf1].copy()
df2
```

```python
#Filtros
grados1=df2.AvgTemperature.unique() #Para definir los valores de la columna como array
years1=df2.Year.unique() #Para definir los valores de la columna como array

#Grafico
plt.figure()
plt.plot(years1,grados1,'.r-')
plt.title('Temperatura Barcelona',fontsize=30)
plt.xlabel('Años',fontsize=15)
plt.ylabel('Celsius',fontsize=15)
plt.grid()
```

```python
#Ambas Gráficas con el uso de subplots
fig, axs = plt.subplots(2,1, figsize=(5,5))

axs[0].plot(years1,grados1,'.b-')
axs[0].set_title('Temperatura Barcelona')
axs[0].set_xlabel('Años',fontsize=15)
axs[0].set_ylabel('Temperatura [C]')
axs[0].grid()

axs[1].plot(years,grados,'.m-')
axs[1].set_title('Temperatura Bogota')
axs[1].set_xlabel('Años')
axs[1].set_ylabel('Temperatura [C]')
axs[1].grid()

fig.tight_layout()
```
