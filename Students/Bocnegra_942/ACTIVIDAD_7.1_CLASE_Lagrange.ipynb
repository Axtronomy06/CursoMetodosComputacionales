```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from IPython.display import display, Math


x =sp.Symbol('x')
puntos = np.array([[2,4],[5,1],[6,1],[10,1],[7,3],[0,1],[1,2],[8,5],[9,2]])


def l(j):

    f =1
    for i in range(0,len(puntos)):
        if i != j:
            f = f*(x-puntos[i][0])/(puntos[i][0]-puntos[j][0])

    return f

print(puntos)

pl =0
for k in range(0,len(puntos)):

    print(l(k))
    pl = pl + l(k)*puntos[k][1]





# Convertir a formato LaTeX
pl_latex = sp.latex(pl)

# Mostrar el resultado en formato LaTeX
display(Math('Polinomio de lagrange: {}'.format(pl_latex)))
#print(pl)

pl_numerica = sp.lambdify(x, pl, 'numpy')

valores_x = np.linspace(0,10,1000)

#Grafica

plt.plot(valores_x,pl_numerica(valores_x))
for i in range(0,len(puntos)):
    plt.scatter(puntos[i][0],puntos[i][1])
```
