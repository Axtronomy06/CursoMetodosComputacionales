#SOLUCIÓN DEL POZO CUADRADO INFINITO

 La solución del pozo cuadrado infinito en mecánica cuántica se aborda utilizando el concepto de valores y vectores propios (autovalores y autovectores) porque la ecuación de Schrödinger es un problema de autovalores.

La ecuación de Schrödinger describe el comportamiento de una partícula cuántica y toma la forma general:

$$H \psi = E \psi$$

Donde:

- $H$ es el operador hamiltoniano, que representa la energía total de la partícula.
- $\psi$ es la función de onda, que describe el estado cuántico de la partícula.
- $E$ es la energía total asociada con el estado cuántico descrito por $\psi$.

Para el caso del pozo cuadrado infinito, $ H $ toma una forma específica que depende del potencial en la región del pozo y fuera del pozo.

Se supone  un potencial $V(x)$ que es cero dentro del pozo y infinito fuera del pozo. En este caso, la ecuación de Schrödinger se ve así dentro del pozo:

$$\frac{-\hbar^2}{2m} \frac{d^2 \psi}{dx^2} = E \psi $$

Esta es una ecuación diferencial ordinaria de segundo orden. Ahora, expresamos la función de onda $ \psi $ como una combinación lineal de funciones de base, y asumimos que $ E $ y $ \psi $ pueden escribirse en términos de ciertos parámetros:

$$ \psi(x) = A \sin(kx) + B \cos(kx) $$

Donde $$ k = \frac{\sqrt{2mE}}{\hbar} $$

Al sustituir esta función de onda en la ecuación de Schrödinger, obtenemos una ecuación que puede ser resuelta para los valores permitidos de $ E $. Este proceso implica la aplicación de condiciones de contorno (por ejemplo, que $\psi $ debe ser continua en los límites del pozo) y la obtención de los valores y funciones propias adecuadas.

Cuando resolvemos la ecuación, obtenemos varios valores posibles de energía $ E_n $ y las funciones de onda correspondientes $\psi_n(x)$, que son los autovectores y autovalores respectivamente.

Para poder obtener los valores de la energía podemos usar las condiciones de frontera .
$\psi(0)=0$


$\psi(L)=0$

Donde cero es el inicio del pozo y $L$ es el final del pozo

notemos que al aplicar la condición

$\psi(0)=0$


$$ \psi(0) = A \sin(k0) + B \cos(k0) = 0 $$


Se cumple  si $B =0$

Lo cual nos reduce la ecuación a:

$$ \psi(x) = A \sin(kx)  $$


Ahora si aplicamos

$\psi(L)=0$

$$ \psi(L) = A \sin(kL) =0 $$

Esto se cumpliría cuando $$kL=n\pi$$

esto es $$k=\frac{n\pi}{L}$$


entoces:

$$ k = \frac{\sqrt{2mE}}{\hbar} = \frac{n\pi}{L} $$

$$E = \frac{(n\hbar\pi)^2}{2mL^2}$$



donde $n=1,2,3,.....,n$

De esta manera por cada valor de $E$ (auto valor) se obtiene una Ed  lo que conlleva una solucón $\psi$ (auto vector)


Luego la solución etáría dada por

$$ \psi(x) = A \sin( \frac{\sqrt{2m \frac{(n\hbar\pi)^2}{2mL^2}}}{\hbar}x)  $$



En este punto solo  falta hallar la constante $A$ Y esto se puede hacer por normalización

$$A\int_{0}^{L}[\sin( \frac{\sqrt{2m \frac{(n\hbar\pi)^2}{2mL^2}}}{\hbar}x)]^2 =1$$

Al hacer esta integral llegamos a que
$A= \sqrt{2/L}$


Por lo cual la solución última estaría dada por

$$ \psi(x) = \sqrt{\frac{2}{L}} * \sin( \frac{\sqrt{2m \frac{(n\hbar\pi)^2}{2mL^2}}}{\hbar}x) $$





**A continuación se hallan algunos valores de la Energía o auto-valores usando la formula teórica**

```python
import numpy as np

hbar = 1.054571817e-34
m = 0.067 * 9.109e-31
c_e =-1.6022e-19
L = 10e-9

def Energia(n):
    return(n*hbar*np.pi)**2/(2*m*L**2)

for i in range(1,7):
    print('La energía del estado ',i,'es : ', Energia(i)/c_e)
```

**Ahora se realiza una gráfica usando la formula teórica para varios valores de n**

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
L = 10e-9  # Ancho del pozo
hbar = 1.054571817e-34
m = 0.067 * 9.109e-31
# Valores de x
x = np.linspace(0, L, 1000)

# Valores de n
n_values = np.arange(1, 8)

# Función de onda
def wave_function(x, n):
    return np.sqrt(2 / L) * np.sin(np.sqrt(2 * m * (n * hbar * np.pi) ** 2 / (2 * m * L ** 2)) / hbar * x)

# Graficar para diferentes valores de n
for n in n_values:
    psi = wave_function(x, n)
    plt.plot(x, psi, label=f'n = {n}')

# Configuración de la gráfica
plt.xlabel('x')
plt.ylabel('$\psi(x)$')
plt.title('Funciones de onda para diferentes valores de n')
plt.legend()
plt.grid(True)
plt.show()
```

##**verificación numérica de lo anterior**

La ecuación de Schrödinger para un sistema unidimensional en un potencial $ V(x) $ puede aproximarse numéricamente mediante la discretización del espacio.Se puede suponer que se está trabajando con un sistema en una caja unidimensional (pozo cuadrado infinito).

La discretización implica dividir el espacio $ x $ en una serie de puntos $x_i$ y aproximar las derivadas espaciales como diferencias finitas. La ecuación de Schrödinger discretizada toma la forma:

$$\frac{\hbar^2}{2m}\frac{\psi_{i+1} - 2\psi_i + \psi_{i-1}}{\Delta x^2} + V_i \psi_i = E \psi_i $$

Donde $\psi_i$ es la función de onda en el punto $ x_i $,mientras que $ V_i$ es el potencial en $ x_i$, $ E $ es la energía, y $ \Delta x $ es la distancia entre puntos vecinos.

Esto se puede reorganizar en la forma matricial $$ H \psi = E \psi $$, donde $ H $ es una matriz tridiagonal hermítica que surge de la discretización. Resolver este problema implica encontrar los autovalores (energías permitidas) y autovectores (funciones de onda asociadas) de la matriz $ H $.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


#***************************************************
N = 200

hbar = 1.054571817e-34
m = 0.067 * 9.109e-31
c_e = 1.6022e-19
V = 10                    # profundidad del pozo

L = 10e-9

dx = L/(N)

#***************************************************

H = np.zeros((N,N))

x = np.arange( 0, L+dx , dx )


#******************************************
# Llenado de la matriz Hamiltoniano

for i in range(0,len(x)-1):

    #LLenando la primera fila
    if i ==0:
        H[i,i] = -hbar**2/(m*dx**2)
        H[i,i+1] = hbar**2/(2*m*dx**2)

    #Llenando la ultima fila
    if i == N-1:
        H[i,i] = -hbar**2/(m*dx**2)
        H[i,i-1] = hbar**2/(2*m*dx**2)

    #Llenando las demás filas de la matriz
    if i > 0 and i < N-1:

        H[i,i] = -hbar**2/(m*dx**2)
        H[i,i+1] = hbar**2/(2*m*dx**2)
        H[i,i-1] = hbar**2/(2*m*dx**2)

#************************************************



autovalores, autovectores = np.linalg.eig(H)


energias = np.sort(autovalores/c_e)[::-1]

print('Las primeras energias en eV son : ',energias[:6])




for i in range(0,3):
    plt.plot( x[:len(autovectores)] , autovectores[:,i] , 'k,' )
```

**Notemos que las energías encontradas son iguales a las encontradas teóricamente**

```python
                    #*****************************************************
                    #*****************************************************

                            # Solución de cada una de las ED asociadas
                            # a cada auto-valor (Energía)

                    #*****************************************************
                    #*****************************************************

k =energias[:6]


N=2000
L = 10e-9 #(Ancho del pozo)
delta = L/N



x = np.arange(0,L,delta)
psi_sig = np.zeros(len(x))
u_sig = np.zeros(len(x))



def euler( x  , psi_sig , u_sig , k ,delta ):



    #***************************************************
                #Definición de parámetros
    #***************************************************

    hbar = 1.054571817e-34
    m0 = 0.067 * 9.109e-31
    c_e =-1.6022e-19
    b = 0 #(parámetor que en  el caso del pozo es cero)
    m = (hbar**2)/(2*m0*c_e)
    #******************************************************


    data = np.zeros((len(x), len(k)))

    for j in range(0,len(k)):


        #Condiciones iniciales
        psi_ant = 0  #(valor de la función de onda en el extremodel pozo)
        u_ant = 0.001  #(valor de la derivada en el extremo del pozo)


        for i in range(0,len(x)):

            psi_sig[i] = psi_ant + delta * u_ant
            u_sig[i] = u_ant + delta * (1/m )*( -b * u_ant - k[j] * psi_ant )

            u_ant = u_sig[i]
            psi_ant = psi_sig[i]

            data[i, j] = psi_sig[i]


    return data


data  = euler( x ,  psi_sig , u_sig , k ,delta)

# Crear el DataFrame
df = pd.DataFrame(data, columns=[f'Energía={i} eV' for i in k])


plt.plot(x,df)

plt.legend(df.columns)

plt.xlabel('x')
plt.ylabel('$\psi$')

plt.grid()
plt.show()
```
