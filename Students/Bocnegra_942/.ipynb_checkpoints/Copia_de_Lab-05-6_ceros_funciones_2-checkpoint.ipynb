# 3. Movimiento de proyectiles en presencia de viento.

Supongamos que un proyectil es lanzado desde el origen con una rapidez $v=vo( \cos \theta , \sin \theta $. Supongamos además que el efecto del viento sobre el proyectil es el de una fuerza proporcional a la velocidad relativa entre el viento y el proyectil. Teniendo presente que si u es la velocidad del viento y $\alpha$ el ángulo que hace dirección de esta velocidad con el eje x, se puede entrontrar las siguientes ecuaciones diferenciales, asociada al estado de movimiento del [proyectil](https://iopscience.iop.org/article/10.1088/0143-0807/36/2/025016/meta):


\begin{equation}
m\frac{dv_x}{dt} = -mb(v_x-u\cos \alpha )
\end{equation}

\begin{equation}
m\frac{dv_y}{dt} = -mg-mb(v_x-u\sin \alpha ),
\end{equation}

donde b es el párametro de rozamiento. La soluciones a las ecuaciones anteriores vienen dadas por:

Posición en x
\begin{equation}
x=(u\cos \alpha) t + \frac{1}{b} (v_0\cos \theta_0 -u \cos\alpha)(1-\mathrm{exp}(-bt))
\end{equation}

Posicion en y
\begin{equation}
y= \frac{1}{b}\left(\frac{g}{b}+v_0 \sin \theta_0 -u\sin\alpha \right)(1-\mathrm{exp}(-bt))-\left( \frac{g}{b}-u\sin\alpha\right)t
\end{equation}

**Método de la bisección**
Es un método para deterinar los ceros de ecuaciones de una sola variable
que se apoya en en el hecho que toda función continua  en un intervalo cerrado [a,b] toma todos los valores que se encuentran en f(a) y f(b) Además se usa el hecho que si f(a) y f(b) tienen signos contrarios entonces fue que la función cruzo el eje x por lo tanto en el intervalo [a,b] se encuentra un cero de la función. Seguida mente se procede a nombrar un punto &ensp;$c=\frac{a+b} {2}$ &ensp;y &ensp;nuevamente se evalua si $f(c)*f(a) < 0 $ &ensp; o &ensp; $f(c)*f(b)< 0 $ &ensp; si el primer caso es cierto entonces se hace &ensp;$b=c$&ensp;y si el segundo es cierto se hace&ensp; $a=c$ este proceco se repite hasta que $f(c)< toleracia$ donde el valor de la toleracia la elegimos como un valor muy pequeño que indicara la precisión con la cual queremos hallar la raiz.   
**Metodo de newton**
Es te métod se puede deducir a partir de la expansión en series de Talilor de &ensp;$f(x)$ pero solo hasta la primera derivada , es decir;&ensp;$f(x)=f(x_0)+f'(x_0)(x-x_0)$&ensp; pero lo que nos interesa es que &ensp;$f(x)=0$ por lo tanto &ensp;
<br>
<br>
$$0=f(x_0)+f'(x_0)(x-x_0)$$
$$0=f(x_0)+f'(x_0)(x)-f'(x_0)(x_0)$$
$$0=\frac{f(x_0)}{f'(x_0)}+x-(x_0)$$
$$x=(x_0)-\frac{f(x_0)}{f'(x_0)}$$
La ultima sera nuestra formula de iteración.


**Metodo de la secante** Este método se basa en hallar la recta que pasa por dos puntos , a dicha recta se le llama secante . Para obtener la formula de iteracion del método de la secante solo es necesario calcular la pendiente de dicha recta y remplazar $f'(x)$ en al formula de iteración el método de newton, es decir;
&ensp;$f'(x)=\frac{f(x_1)-f(x_0)}{x_1-x_0}$


**Diferencia entre le metodo  de la secante y mewton**
Una diferencia es que para iterar el método de newton solo se necesita un punto  mientras que para la secante se hace necesario dos puntos

````python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

def g(x):
    return np.log(np.cos( np.exp(x) - 2 ) + 2 )

def f(x):
    return np.cos( np.exp(x) - 2 ) - np.exp(x) + 2

#*************************************************************
#Método de bisección

a =-2
b = 2
c = a
tol= 1e-4
max_inter = 5000
contador = 0

while (abs(f(c)) > tol) and (f(a)*f(b) < 0) and  (contador < max_inter) :

    c = (a+b)/2

    if f(c)*f(b) < 0 :
        a = c
    else:
        b = c


    contador += 1


print('Raiz por método de la bisección : ',c)
#Fin del método de la bisección
#*************************************************************************

#************************************************************************
#metodod de newton

t = sp.Symbol('t')

# Definir la función


fs = sp.Function('fs')(t)
fs = sp.cos( sp.exp(t) - 2 ) - sp.exp(t) + 2
derivada = sp.Function('derivada')(t)
derivada = sp.diff(fs, t)
#print(sp.diff(fs, t ))


x0=0.5

tol = 1e-3
contador=0
max_inter=50

#print(fs.subs(t, x0).evalf())


while (abs(fs.subs(t, x0)).evalf(10) > tol)  and  (contador < max_inter) :

    xsig = x0 - fs.subs(t, x0).evalf(10)/derivada.subs(t, x0).evalf(10)
    x0 = xsig
    contador +=1
    #print('contador = ',contador)
print('Raiz por el método de newton : ',x0)

#Fin del método de newton
#*********************************************************************

#*********************************************************************
#Metodo del punto fijo

x_ini = 0.5
tol =1e-4
max_iteraciones = 2000
cont = 0

while abs(f(x_ini)) > tol  and cont< max_iteraciones:
    x_sig = g(x_ini)
    x_ini = x_sig
    cont += 1

print('Raiz por el metodo del punto fijo : ',x_ini)


#fin del metodo del punto fijo
#************************************************************************

x = np.linspace(0,2,1000)

plt.plot(x,g(x),label='g(x)')
plt.plot(x,x,label='y=x')
plt.plot(x,f(x),label='f(x)')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.legend()
````

````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import optimize

#condiciones iniciales
#*********************
v0 = 60
alfa = np.pi
teta = np.deg2rad(45)
u = v0/2
b = 0.05
g = 9.8
#*********************

#funcion para hallar el tiempo   que usaremos par ael alcnce maximo
#**************************************************************************************************************************
def f(t):
    return (1/b) * ( g/b + v0 * np.sin(teta) - u * np.sin(alfa) ) * ( 1 - np.exp( -b * t) ) - ( g/b - u * np.sin(alfa) ) * t

#***************************************************************************************************************************

#funcion para haller el alcance maximo evaluando el timepo  que hace cero la función anterior
#**************************************************************************************************************
def cordenada_x(t):
    return ( u * np.cos(alfa) ) * t + (1/b) * (v0 * np.cos(teta) - u * np.cos(alfa)) * ( 1 - np.exp(-b * t))
#**************************************************************************************************************

#Funcion para determinar las coordenas a lo largo del movimiento
#************************************************************************************************************************
def coordenadas(*args):

    v0 , alfa , teta , u , b , g , t  = args

    x = ( u * np.cos(alfa) ) * t + (1/b) * (v0 * np.cos(teta) - u * np.cos(alfa)) * ( 1 - np.exp(-b * t))

    y = (1/b) * ( g/b + v0 * np.sin(teta) - u * np.sin(alfa) ) * ( 1 - np.exp( -b * t) ) - ( g/b - u * np.sin(alfa) ) * t

    return x,y,t
#*************************************************************************************************************************


t = np.arange(0,10,1)

#tupla con CI para pasar a las funciones
ci = ( v0 , alfa , teta , u , b , g , t )

#x , y = (coordenadas(*ci))

diccionario = {'X':[],'Y':[],'Tiempo':[]}

df = pd.DataFrame(diccionario)



#llenando el dataframe con las coordenadas
df.X , df.Y , df.Tiempo = coordenadas(*ci)

raiz = optimize.bisect(f,7,10)

alcance_maximo = cordenada_x(raiz)

print('El tiempo para el alcance máximo es : ',raiz,' segundos')
print('El alcance máximo es : ' ,alcance_maximo,' metros')

#graficos
fig , graficos = plt.subplots(3,figsize =(5,5))

graficos[0].plot( df.X , df.Y,'k-' )
graficos[0].set_xlabel('X')
graficos[0].set_ylabel('Y')

graficos[1].plot(df.Tiempo , df.X )
graficos[1].set_xlabel('Tiempo')
graficos[1].set_ylabel('X')

graficos[2].plot(df.Tiempo , df.Y)
graficos[2].set_xlabel('Tiempo')
graficos[2].set_ylabel('Y')
graficos[2].set_xlim(0,15)
graficos[2].grid()


plt.tight_layout()
plt.show()

df
````

# 4. Fixed point methods.

When a new planet is discovered, there are different methods to estimate its physical properties. Many times is only possible to estimate either the planet mass or the planet radius and the other property has to be predicted through computer modelling.

If one has the planet mass, a very rough way to estimate its radius is to assume certain composition (mean density) and a homogeneous distribution (a very bad assumption!). For example, for the planet [Gliese 832c](http://es.wikipedia.org/wiki/Gliese_832_c) with a mass $M= 5.40 M_{\oplus}$, if we assume an earth-like composition, i.e. $\bar \rho_{\oplus} = 5520\ kg/m^3$, we obtain:

$$R_{g832c} = \left( \frac{3 M_{g832c}}{ 4 \pi \bar\rho_{\oplus} } \right)^{1/3} \approx 1.75 R_{\oplus}$$

That would be the planet radius if the composition where exactly equal to earth's.

A more realistic approach is assuming an internal one-layer density profile like:

$$\rho(r) = \rho_0 \exp\left( -\frac{r}{L} \right)$$

where $\rho_0$ is the density at planet centre and $L$ is a characteristic lenght depending on the composition. From numerical models of planet interiors, the estimated parameters for a planet of are $M= 5.40 M_{\oplus}$ are approximately $\rho_0 = 18000\ kg/m^3$ and $L = 6500\ km$.

Integrating over the planet volume, we obtain the total mass as

$$M = 4\pi \int_0^R \rho(r)r^2dr$$

This is a function of the mass in terms of the planet radius.

Solving the equation $M(R) = M_{g832c}$ it would be possible to find a more realistic planet radius. However when using numerical models, it is not possible to approach the solution from the left side as a negative mass makes no sense.

````python
import matplotlib.pyplot as plt

import sympy as sp
import numpy as np
from scipy import optimize
from scipy.optimize import root_scalar



# Definir las variables simbólicas
R = sp.symbols('R')
M = sp.symbols('M')
rho0 = 18000  # Densidad en kg/m^3
L = 6500000  # Longitud característica en metros

# Definir la expresión de la densidad
rho = rho0 * sp.exp(-R / L)

# Definir la expresión para la masa
mass_expr = 4 * sp.pi * sp.integrate(rho * R**2, (R, 0, R))

# Definir la masa conocida de Gliese 832c
M_g832c = 5.40 * 5.972e24  # Masa en kg

# Definir la ecuación M(R) - M_g832c = 0
equation = mass_expr - M_g832c

# Convertir la ecuación en una función numerica de Python
equation_func = sp.lambdify(R, equation, 'numpy')


print(f"El radio estimado de Gliese 832c es aproximadamente {optimize.bisect(equation_func,1e7,1.2e7)} metros")
print(f'El radio de  Gliese 832c es aproximadamente {round(optimize.bisect(equation_func,1e7,1.2e7)/6371000,1)} veces el radio de la tierra')

x_valores = np.linspace(5e-6,15e6,1000000)


plt.plot(x_valores,equation_func(x_valores))
plt.xlabel('RADIO')
plt.ylabel('$f(r)$')
plt.grid()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

Npx_X = 80  # Numero de pixeles en x
Npx_Y = 80  # Numero de pixeles en y

image = Image.new("RGB", (Npx_X, Npx_Y))  # Creación de la imagen inicial

npx = np.arange(0, Npx_X)
npy = np.arange(0, Npx_Y)

xmin = -1
xmax = 1
ymin = -1
ymax = 1

# Transformación al plano complejo en un rectángulo (xmin, xmax), (ymin, ymax)
zx = npx * (xmax - xmin) / (Npx_X - 1) + xmin
zy = npy * (ymax - ymin) / (Npx_Y - 1) + ymin

# Definición de la función y las derivadas
def f(z):
    return z**3 - 1

def fp(z):
    return 3 * z**2

h = 1E-4

def fpn(z, h):
    return (f(z + h) - f(z)) / h

a = 0.2
Niteraciones = 120
eps = 1E-8

x = -1
for zx_ in zx:
    x = x + 1
    y = -1
    for zy_ in zy:
        y = y + 1
        Z_0 = complex(zx_, zy_)

        for i in range(0, Niteraciones):
            Z = Z_0 - a * f(Z_0) / fp(Z_0)
            if abs(Z - Z_0) <= eps:
                break
            Z_0 = Z

        r, g, b = i % 4 * 64, i % 8 * 16, i % 16 * 16
        image.putpixel((x, y), (r, g, b))

plt.imshow(image)
plt.show()
````
