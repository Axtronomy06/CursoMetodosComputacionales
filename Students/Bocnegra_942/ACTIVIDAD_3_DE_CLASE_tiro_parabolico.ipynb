**SEGUNDA SULUCIÓN USANDO ARCHIVOS**

```python
#Importando librerías

import numpy as np
import matplotlib.pyplot as plt
import json

#*****************************************************************
#función para definir el tiempo en el cual se da el alcance maximo

def alcance_maximo(teta,v0,y0):

    #coeficientes de la cuadratica
    a = 0.5*9.8
    b = -v0 * np.sin(np.deg2rad(teta))
    c = y0
    #dos raices de la cuadratica usando la formula del estudiante
    t1= (-b + np.sqrt(b**2-4*a*c))/(2*a)
    t2= (-b - np.sqrt(b**2-4*a*c))/(2*a)

    print(t1,t2)

    #roots = np.roots([a, b, c])

    #print("Raíces:", roots)

    #condiciones para elegir cual tiempo es el correcto
    if t1 > 0 and t1 > t2:
        return t1
    if t2 > 0 and t2 > t1:
        return t2
#FIN DE LA FUCION  alcance_maximo
#**********************************************************************

# Condiciones iniciales del problema
x0 = 0
y0 = 0
v0 = 10
teta = 10

tmax = alcance_maximo(teta, v0, y0) # calculando el tiempo de alcance maximo

#arreglos de numpy con los calculos
t = np.linspace(0.0, tmax, num=100)
x_values = x0 + v0 * np.cos(np.deg2rad(teta)) * t
y_values = y0 + v0 * np.sin(np.deg2rad(teta)) * t - 0.5 * 9.8 * t ** 2

# Guardar los datos en un archivo JSON
data = {
    "x_values": x_values.tolist(),
    "y_values": y_values.tolist()
}

with open("datos.json", "w") as json_file:
    json.dump(data, json_file)

# Cargar los datos desde el archivo JSON
with open("datos.json", "r") as json_file:
    loaded_data = json.load(json_file)

x_loaded = loaded_data["x_values"]
y_loaded = loaded_data["y_values"]

# Graficar los datos cargados
plt.plot(x_loaded, y_loaded, ".r", label="parabolico")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Tiro parabólico")
plt.legend()
plt.grid()
plt.show()
```
