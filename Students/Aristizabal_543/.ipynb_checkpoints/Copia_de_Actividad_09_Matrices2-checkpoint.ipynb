Operaciones matricial con regresión lineal.

Cuando se desea hacer una regresion lineal multivariada, existen diferentes formas de realizar la regresión, una de ellas es encontrar el error cuadrático medio y miniminzar el conjunto de parámetros de la regresión emplando algún metodo de minimización, otra forma es encontrar la expresión analítica en forma matricial para enoncontrar el conjunto de parametros de la regresión. En este último caso, la regresion lineal viene dada por la siguiente expresion generalizada:


- $\Theta = (X^T X)^{-1} X^T y$

donte $\Theta$ es la matriz de parametros asociados a la regresión lineal, X la matriz de características y y los valos esperados, así, si tenemos una conjunto de datos en N dimensiones la recta, plano o hyperplano puede ser calculado como:

$Y=\Theta^T X$


En el caso de dos dimensiones, tenemos lo siguiente, Supongamos que X, y Y presentan un comportamiento lineal, y vienen dados por:


- Sea $\vec{X}=x^1, x^2, x^3, x^4, ..., x^m$
- Sea $\vec{Y}=y^1, y^2, y^3, y^4, ..., y^m$

Para determinar la ecuacion de la recta, hay que conocer los párametros $\theta_0 $, $\theta_1$, que cumplen que :   $y^{i} = \theta_1 x^{i} + \theta_0$ la anterior expresion puede ser escrita como:

\begin{equation}
\begin{bmatrix}
y^{1}\\
y^{2}\\
\vdots\\
y^{m}\\
\end{bmatrix}=
\begin{bmatrix}
\theta_0 & \theta_1 \\
\end{bmatrix}
\begin{bmatrix}
1 & 1 & \cdots & 1 \\
x_1^{1} & x_1^{2} & \cdots & x_1^{m} \\
\end{bmatrix}
\end{equation}


Para determinar la matriz de theta entonces se puede aplicar la ecuación:

- $\Theta = (X^T X)^{-1} X^T y$

donde:
\begin{equation}
X=
\begin{bmatrix}
1 & x_1^{1} \\
1 & x_1^{2} \\
\vdots & \vdots \\
1 & x_1^{m}
\end{bmatrix}
\end{equation}

y


\begin{equation}
y=
\begin{bmatrix}
y^1 \\
y^2 \\
\vdots\\
y^m
\end{bmatrix}
\end{equation}

4. Construir un programa que permita realizar eliminacion de Gauss- Jordan de una matriz



````python
#Punto 1
#Recuérdese que el producto está definido siempre que el número de filas de la matriz1 sea igual al número de columnas de la matriz2

def multiplicacionM(matriz1, matriz2):
    filas_matriz1 = len(matriz1)
    columnas_matriz1 = len(matriz1[0])
    filas_matriz2 = len(matriz2)
    columnas_matriz2 = len(matriz2[0])

    if columnas_matriz1 != filas_matriz2:
        print("No se pueden multiplicar las matrices.")
        return

    resultado = [[0 for _ in range(columnas_matriz2)] for _ in range(filas_matriz1)]

    for i in range(filas_matriz1):
        for j in range(columnas_matriz2):
            for k in range(columnas_matriz1):
                resultado[i][j] += matriz1[i][k] * matriz2[k][j]

    return resultado



matriz1 = [[1, 2, 3], [4, 5, 6]]
matriz2 = [[3, 2], [1, 0], [5, 7]]

resultado = multiplicacionM(matriz1, matriz2)

for fila in resultado:
    print(fila)
````

````python
#Punto 2

N = 10
x1 = np.linspace(0, 10, N)
noise = 2 * np.random.random(N) - 1
y = 0.9 * x1 + noise

X = np.vstack((np.ones(N), x1)).T
theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

y_reg = X.dot(theta)

print("Parámetros theta de la regresión lineal:")
print(theta)
print("\nRecta de regresión (La recta que pasa por los puntos):")
for i in range(N):
    print(f"x={x1[i]:.2f}, y_regresión={y_reg[i]:.2f}")
````

````python
#Punto 3
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Datos de entrada
x = np.array([1, 2, 3, 4, 5])
y = np.array([6, 7, 8, 9, 10])

A = np.vstack([x, np.ones(len(x))]).T
m, c = np.linalg.lstsq(A, y, rcond=None)[0]


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x_surf = np.linspace(0, 6, 100)
y_surf = np.linspace(0, 7, 100)
x_surf, y_surf = np.meshgrid(x_surf, y_surf)
z_surf = m*x_surf + c

ax.plot_surface(x_surf, y_surf, z_surf, alpha=0.5)


ax.scatter(x, y, np.dot(np.array([x, np.ones(len(x))]).T, [m, c]))

plt.show()
````

````python
#Punto que está antes del 4

N = 10
x1 = np.linspace(-1, 1, N)
x2 = np.linspace(-2, 2, N)
noise1 = 2*np.random.random(N) - 1
noise2 = 2*np.random.random(N) - 1
y = 2*(x1 + noise1) - 3*(x2 + noise2) + 0.0


A = np.vstack([x1, x2, np.ones(N)]).T

coefficients = np.linalg.lstsq(A, y, rcond=None)[0]

X1, X2 = np.meshgrid(x1, x2)
Y = coefficients[0]*X1 - coefficients[1]*X2 + coefficients[2]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X1, X2, Y, cmap='viridis')
plt.scatter(x1, x2, y, c='r', marker='o')
plt.show()
````

````python
#Generalización a k dimensiones

k = 3
N = 10
x = [np.linspace(-k, k, N) for _ in range(k)]
noises = [2*np.random.random(N) - 1 for _ in range(k)]
y = sum([i*(x[j] + noises[j]) for j, i in enumerate(range(k, 0, -1))]) + 0.0

A = np.vstack([x[j] for j in range(k)] + [np.ones(N)]).T

coefficients = np.linalg.lstsq(A, y, rcond=None)[0]

meshgrid = np.meshgrid(*x)
Y = sum([coefficients[j]*meshgrid[j] for j in range(k)]) + coefficients[k]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(meshgrid[0], meshgrid[1], Y, cmap='viridis')
plt.scatter(*x, y, c='r', marker='o')
plt.show()

````

````python
#Punto 4

def row_lamb(A,i, l):
    B=np.copy(A)
    B=B.astype('float64')

    B[i,:] = B[i,:]*l
    return B

def row_comb(A, i, j, l):
    B=np.copy(A)
    B=B.astype('float64')

    B[i,:] += row_lamb(A,j,l)[j,:]
    return B

def row_swap(A, i, j):
    B=np.copy(A)
    B=B.astype('float64')

    B[[i,j]]=B[[j,i]]
    return B

A=[[1, 2, 1],
   [2, 4, 3],
   [3, 3, 1]]

def GaussJordan(A):
    B=np.copy(A)

    m=len(B[:,0]) #Número filas
    n=len(B[0,:]) #Número columnas

    k=min([m,n]) #Para matrices que no son cuadradas



    for j in range(k): #Corremos sobre las columnas

        for i in range(j,m): #Corremos sobre las filas
            if B[i,j]!=0: #Encontramos el pivote
                break

        B=row_swap(B,i,j)
        #print(i,j,'Intercambio \n',B)

        if B[j,j]==0: #Si no hay pivote pasar a la siguiente columna
            continue

        B=row_lamb(B,j,1/B[j,j])

        for i in range(m): #Corremos sobre TODAS las filas:
            if i==j:
                continue

            B=row_comb(B,i,j,-B[i,j]/B[j,j])

    return B

print('Original: \n',A)
print('Gauss Jordan: \n',GaussJordan(A))
````
