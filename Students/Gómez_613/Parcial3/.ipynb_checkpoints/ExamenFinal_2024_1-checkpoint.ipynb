## Ecuaciones de movimiento

### Ecuaciones de movimiento del centro de masa

De acuerdo con la segunda ley de Newton, la ecuación de movimiento a lo largo de la vertical ($y$) puede ser escrita como:

\begin{equation}
m\frac{d^2y}{dt^2} = N - mg
\end{equation}

donde $m$ es la masa de la lápiz, $N$ es la fuerza normal y $g$ es la aceleración de la gravedad. Escribiendo las ecuaciones en términos del ángulo $\theta$, tenemos que:

\begin{equation}
-m\frac{l}{2}\cos \theta \left(\frac{d\theta}{dt} \right)^2 - m\frac{l}{2}\sin \theta \frac{d^2\theta}{dt^2} = N - mg
\end{equation}

### Ecuaciones de rotación del centro de masa

La sumatoria de momentos de fuerza se puede escribir como sigue:

\begin{equation}
I_c\frac{d^2\theta}{dt^2} = N\frac{l}{2} \sin \theta
\end{equation}

donde $I_c = \frac{1}{12}ml^2$ es el momento de inercia de una lápiz de masa $m$ y longitud $l$ respecto a un eje perpendicular a la lápiz en su centro de masa.

Reemplazando el valor de la normal en la ecuación anterior, obtenemos que la ecuación de movimiento para la posición angular está dada por:

\begin{equation}
\frac{d^2 \theta}{dt^2} = \left( \frac{\frac{2g}{l} - \left(\frac{d\theta}{dt}\right)^2\cos \theta }{\frac{1}{3} + \sin^2 \theta} \right)\sin \theta
\end{equation}

Podemos resolver la ecuación anterior si conocemos las condiciones iniciales: $\theta = \theta_0$ en $t = 0$ y $\frac{d\theta}{dt} = \omega_0 = 0$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint,quad
from sympy import pi
```

```python
def fun(u,t,params):
    g=params[0]
    l=params[1]

    θ=u[0]
    w0=u[1]
    dθdt=w0
    dw0dt=((2*g/l-w0**2*np.cos(θ))*np.sin(θ))/((1/3)+(np.sin(θ))**2)

    dudt=np.array([dθdt,dw0dt])

    return dudt
```

```python
l=0.1
g=9.81

θ0=np.pi/12
w0=0

params=(g,l)

ta=0
tb=0.2
dt=0.001

n=int((tb-ta)/dt)
t=np.linspace(ta,tb,n)
```

## 1, 2 y 3

```python
u0=np.array([θ0,w0])

sol=odeint(fun,u0,t,args=(params,))

θ=sol[:,0]
w=sol[:,1]

y=l/2*np.cos(θ)
```

```python
_=len(θ[θ<np.pi/2])

fig,axes=plt.subplots(1,3,figsize=(10,4))

axes[0].plot(t[0:_],θ[θ<np.pi/2],label='θ(t)')      #gráfica con sentido físico
#axes[0].plot(t,θ,label='θ(t)')                     gráfica completa
axes[0].set_title('Posición angular')
axes[0].set_xlabel('t[s]')
axes[0].set_ylabel('θ[rad]')
axes[0].grid()
axes[0].legend()

axes[1].plot(t[0:_],y[0:_],label='y(t)')            #gráfica con sentido físico
#axes[1].plot(t,y,label='y(t)')                     gráfica completa
axes[1].set_title('Posición centro de masa')
axes[1].set_xlabel('t[s]')
axes[1].set_ylabel('y[m]')
axes[1].grid()
axes[1].legend()

axes[2].plot(t[0:_],w[0:_],label='$\omega(t)$')     #gráfica con sentido físico
#axes[2].plot(t,w,label='$\omega(t)$')              gráfica completa
axes[2].set_title('Rapidez angular')
axes[2].set_xlabel('t[s]')
axes[2].set_ylabel('$\omega$[rad/s]')
axes[2].grid()
axes[2].legend()

plt.tight_layout()
```

Para tiempos posteriores a 0.15s aproximadamente ya la solución deja de tener sentido porque ya pasó el suelo. Es decir, que el intervalo que tiene sentido es de 0 a 0.15s.

```python
θs=np.array([np.pi/12,np.pi/6,np.pi/4,np.pi/3])
__=np.array([pi/12,pi/6,pi/4,pi/3])
for i in θs:
    u0=np.array([i,w0])

    sol=odeint(fun,u0,t,args=(params,))
    θ=sol[:,0]
    w=sol[:,1]

    _=len(θ[θ<np.pi/2])

    ind=np.where(θs==i)[0][0]

    plt.plot(t[0:_],θ[θ<np.pi/2],label=f'$θ_0={__[ind]}$')
    plt.title('Posición angular para varios $θ_0$')
    plt.xlabel('t[s]')
    plt.ylabel('θ(t)[rad]')
    plt.legend()
plt.grid()
```

## 6 y 7

```python
θ0s=np.arange(np.pi/12,np.pi/2,dt)
T=np.zeros(len(θ0s))

def f(θ,θ0):
    num=(1/3)+np.sin(θ)**2
    den=np.cos(θ0)-np.cos(θ)

    return np.sqrt(num/den)

for i,θ in enumerate(θ0s):
    K=(1/2)*np.sqrt(l/g)

    T[i]=K*quad(f,θ,np.pi/2,args=(θ0,))[0]

plt.plot(θ0s,T,label='T($θ_0$)')
plt.title('Tiempo de caída en función del ángulo inicial')
plt.xlabel('θ[rad]')
plt.ylabel('T[s]')
plt.grid()
plt.legend()
```

Según la gráfica se podría concluir que a menor ángulo inicial con respecto a la vertical mayor es el tiempo de caída, y a mayor ángulo, menor tiempo.
