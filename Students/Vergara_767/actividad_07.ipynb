```python
import numpy as np
import matplotlib.pyplot as plt
import math as mt
import sympy as sp
```

```python
# Definición de la función Newton-Raphson para encontrar raíces de ecuaciones
def newton_raphson(f, df, x_0, eps, max_iter):
    # Crear un rango de valores para x y xx (para gráficos)
    x = np.arange(-10, 10, 0.1)
    xx = np.arange(-10, 10, 0.1)

    # Inicializar x_n con el valor inicial proporcionado
    x_n = x_0

    # Dibujar la función f(x) en el rango [-10, 10]
    plt.plot(x, f(x))

    # Inicializar el contador de iteraciones
    iterations = 0

    # Iniciar el bucle de Newton-Raphson hasta que se alcance la precisión deseada o el máximo de iteraciones
    while abs(f(x_n)) >= eps and iterations < max_iter:
        # Definir una función lineal y que representa la tangente en x_n
        y = lambda x: df(x_n)*(x - x_n) + f(x_n)

        # Dibujar la línea tangente como una línea punteada
        plt.plot(xx, y(xx), "--")

        # Calcular el siguiente punto x_m usando el método de Newton-Raphson
        x_m = x_n - f(x_n) / df(x_n)

        # Actualizar x_n para la siguiente iteración
        x_n = x_m

        # Incrementar el contador de iteraciones
        iterations += 1

    # Devolver la raíz encontrada
    return x_m

# Definir la función y su derivada
f = lambda x: np.sin(x)
df = lambda x: np.cos(x)

# Llamar a la función newton_raphson para encontrar la raíz de la función
root = newton_raphson(f, df, 1.0, 10e-7, 100)

# Imprimir la raíz encontrada
print("La raíz encontrada es:", round(root, 6))

# Generar gráfico de la función y las iteraciones
x = np.linspace(-2*np.pi, 2*np.pi, 400)
plt.plot(x, f(x), label="f(x) = sin(x)")
plt.plot(root, f(root), ".r", label="Raíz")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)

# Configuración de límites y leyenda
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.legend()
plt.title("Método de Newton-Raphson")
plt.show()
```
