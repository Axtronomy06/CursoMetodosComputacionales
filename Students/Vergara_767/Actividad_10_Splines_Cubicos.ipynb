```python
import matplotlib.pyplot as plt
from matplotlib.image import imread
import pandas as pd
import numpy as np
from scipy.interpolate import lagrange
from scipy.interpolate import CubicSpline
from scipy.interpolate import CubicHermiteSpline
```

```python
paloma = imread("Fig.png")
```

```python
# Crear una figura y un eje de gráficos
fig, ax = plt.subplots(1, 1)

# Desactivar la cuadrícula en el gráfico
ax.grid(False)

# Mostrar la imagen "paloma"
ax.imshow(paloma)

# Mostrar el gráfico
plt.show()
```

```python
x = np.array([0.9, 1.3, 1.9, 2.1, 2.6, 3.0, 3.9, 4.4, 4.7, 5.0, 6.0, 7.0, 8.0, 9.2, 10.5, 11.3, 11.6, 12.0, 12.6, 13.0, 13.3])
y = np.array([1.3, 1.5, 1.85, 2.1, 2.6, 2.7, 2.4, 2.15, 2.05, 2.1, 2.25, 2.3, 2.25, 1.95, 1.4, 0.9, 0.7, 0.6, 0.5, 0.4, 0.25])
```

```python
dicc = {"x":x, "y":y}
```

```python
df = pd.DataFrame(dicc)
```

```python
plt.plot(x, y, ".")
plt.xlim(0,14)
plt.ylim(-6, 4)
plt.grid(True)
```

```python
p_x = lagrange(x, y)
```

```python
w = np.linspace(x[0], x[-1], 100)
plt.plot(w,p_x(w))
plt.xlim(0,14)
plt.ylim(-6, 4)
plt.grid(True)
```

```python
x_vals = np.linspace(x[0], x[-1], 100)

cs = CubicSpline(x, y)

y_vals = cs(x_vals)
```

```python
plt.plot(x_vals, y_vals, label='Cubic Spline')
plt.xlim(0,14)
plt.ylim(-6, 4)
plt.grid(True)
```

```python
dydx = np.diff(y)/np.diff(x)


hermite_interp = CubicHermiteSpline(x[:-1], y[:-1], dydx)

y_valss = hermite_interp(x_vals)

plt.plot(x_vals, y_valss)
plt.xlim(0,14)
plt.ylim(-6, 4)
```

```python
plt.plot(x_vals, y_valss, label='Cubic Hermite Spline')
plt.xlim(0,14)
plt.ylim(-6, 4)
plt.grid(True)
```

```python
plt.plot(x_vals, y_vals, label='Cubic Spline')
plt.plot(w,p_x(w), label = "Lagrange")
plt.plot(x_vals, y_valss, label = "Hermite")
plt.plot(x, y, "r.")
plt.xlim(0,14)
plt.ylim(-6, 4)
plt.grid(True)
plt.legend()
```
