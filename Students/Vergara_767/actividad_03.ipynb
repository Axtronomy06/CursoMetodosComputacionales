```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = lambda x, v_0, a: x*np.tan(a)-(1/2)*9.8*(x**2)*(1/(v_0*np.cos(a)))**2
v_0 = float(input("Ingrese rapidez inicial en m/s: "))
a = float(input("Ingrese angulo de tiro en radianes: "))
t_xmax = 2*v_0*np.sin(a)/9.8
x = np.arange(0,v_0*np.cos(a)*t_xmax,0.1)
plt.plot(y(x,v_0,a))
plt.title('Tiro parabolico')
plt.xlabel('Coordenada X (m)')
plt.ylabel('Coordenada Y (m)')
plt.grid(True)
plt.show()
```

```python
y = lambda x, v_0, a: x*np.tan(a)-(1/2)*9.8*(x**2)*(1/(v_0*np.cos(a)))**2
v_0 = float(input("Ingrese rapidez inicial en m/s: "))
a = [np.pi/12,np.pi/6,np.pi/4,np.pi/3,5*np.pi/12]
t_xmax = []
x = []
etiquetas = ["Angulo de 15 grados","Angulo de 30 grados","Angulo de 45 grados","Angulo de 60 grados","Angulo de 75 grados"]
for i in range(5):
  t_xmax.append(2*v_0*np.sin(a[i])/9.8)
  x.append(np.arange(0,v_0*np.cos(a[i])*t_xmax[i],0.1))
  plt.plot(y(x[i],v_0,a[i]),label = etiquetas[i])
plt.title('Tiro parabolico para distintos angulos')
plt.xlabel('Coordenada X (m)')
plt.ylabel('Coordenada Y (m)')
plt.grid(True)
plt.legend(loc='upper right')
plt.show()
```
