```python
import numpy as np
import matplotlib.pyplot as plt
```

# Euler

```python
# Parámetros
λ = 0.5
t0 = 0
tf = 10

# Número de puntos y paso de tiempo
n = 100
dt = (tf - t0) / n

# Condición inicial
N_0 = 0.5

# Crear arreglo de tiempo
t = np.linspace(t0, tf, n)

# Inicializar arreglo para la solución numérica
N = np.zeros(len(t))
N[0] = N_0

# Método de Euler para resolver la EDO
for i in range(n - 1):
    N[i + 1] = N[i] - dt * λ * N[i]

# Solución analítica para comparación
N_analytical = N_0 * np.exp(-λ * t)

# Gráfica de las soluciones
plt.plot(t, N, label='Solución Numérica')
plt.plot(t, N_analytical, label='Solución Analítica', linestyle='dashed')
plt.grid(True)
plt.xlabel("t")
plt.ylabel("N")
plt.title("Decamiento radioactivo")

plt.legend()
plt.show()

# Salchicha
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del oscilador armónico simple
W = 1.5
t0 = 0
tf = 20

# Número de puntos y paso de tiempo
n = 100
dt = (tf - t0) / n

# Condiciones iniciales
y_0 = 1
v_0 = 0

# Crear arreglos de tiempo, velocidad y posición
t = np.linspace(t0, tf, n)
v = np.zeros(len(t))
y = np.zeros(len(t))
v[0] = v_0
y[0] = y_0

# Método de Euler para resolver el sistema de EDOs
for i in range(n - 1):
    v[i + 1] = v[i] - dt * (W ** 2) * y[i]
    y[i + 1] = y[i] + dt * v[i + 1]

# Solución analítica para comparación
y_analytical = y_0 * np.cos(W * t)

# Gráfica de las soluciones
plt.plot(t, y, label='Solución Numérica')
plt.plot(t, y_analytical, label='Solución Analítica', linestyle='dashed')
plt.grid(True)
plt.xlabel("t")
plt.ylabel("y")
plt.legend()
plt.title("Solución del Oscilador Armónico Simple")
plt.show()

# Salchicha
```

```python
# Parámetros del oscilador armónico amortiguado
Γ = 0.5
W = 1.5
t0 = 0
tf = 20

# Número de puntos y paso de tiempo
n = 100
dt = (tf - t0) / n

# Condiciones iniciales
y_0 = 1
v_0 = 0

# Crear arreglos de tiempo, velocidad y posición
t = np.linspace(t0, tf, n)
v = np.zeros(len(t))
y = np.zeros(len(t))
v[0] = v_0
y[0] = y_0

# Método de Euler para resolver el sistema de EDOs del oscilador amortiguado
for i in range(n - 1):
    v[i + 1] = v[i] - dt * (Γ * v[i] + (W ** 2) * y[i])
    y[i + 1] = y[i] + dt * v[i + 1]

# Solución analítica para comparación
omega = np.sqrt(W ** 2 - Γ ** 2)
y_analytical = y_0 * np.exp(-Γ / 2 * t) * np.cos(omega * t)

# Gráfica de las soluciones
plt.plot(t, y, label='Solución Numérica')
plt.plot(t, y_analytical, label='Solución Analítica', linestyle='dashed')
plt.grid(True)
plt.xlabel("t")
plt.ylabel("y")
plt.legend()
plt.title("Solución del Oscilador Armónico Amortiguado")
plt.show()

# Salchicha
```

# Runge-Kutta

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la EDO
λ = 0.5
t0 = 0
tf = 10

# Número de puntos y tamaño del paso
n = 100
h = (tf - t0) / n

# Condición inicial
N_0 = 0.5

# Crear arreglo de tiempo
t = np.linspace(t0, tf, n)

# Inicializar arreglo para la solución numérica
N = np.zeros(len(t))
k = np.zeros(4)
N[0] = N_0

# Función que representa la EDO: dN/dt = -λ * N
f = lambda t, N: -λ * N

# Método de Runge-Kutta de cuarto orden para resolver la EDO
for i in range(n - 1):
    k[0] = f(t[i], N[i])
    k[1] = f(t[i] + h/2, N[i] + (h/2)*k[0])
    k[2] = f(t[i] + h/2, N[i] + (h/2)*k[1])
    k[3] = f(t[i] + h, N[i] + h*k[2])

    N[i + 1] = N[i] + (h / 6) * (k[0] + 2 * k[1] + 2 * k[2] + k[3])

# Solución analítica para comparación
N_analytical = N_0 * np.exp(-λ * t)

# Gráfica de las soluciones
plt.plot(t, N, label='Solución Numérica')
plt.plot(t, N_analytical, label='Solución Analítica', linestyle='dashed')
plt.grid(True)
plt.xlabel("t")
plt.ylabel("N")
plt.legend()
plt.title("Decaimiento Radioactivo")
plt.show()

# Salchicha
```

```python
# Parámetros del oscilador armónico amortiguado
Γ = 0.5
W = 1.5
t0 = 0
tf = 20

# Número de puntos y tamaño del paso
n = 1000
h = (tf - t0) / n

# Condiciones iniciales
y_0 = 1
v_0 = 0

# Crear arreglos de tiempo, velocidad y posición
t = np.linspace(t0, tf, n)
v = np.zeros(len(t))
y = np.zeros(len(t))
k = np.zeros(4)
v[0] = v_0
y[0] = y_0

# Funciones que representan las EDOs del oscilador amortiguado
f = lambda y, v: -(Γ * v + (W ** 2) * y)
ff = lambda v, y: v

# Método de Runge-Kutta de cuarto orden para resolver el sistema de EDOs
for i in range(n - 1):
    k[0] = f(y[i], v[i])
    k[1] = f(y[i] + h/2, v[i] + (h/2)*k[0])
    k[2] = f(y[i] + h/2, v[i] + (h/2)*k[1])
    k[3] = f(y[i] + h, v[i] + h*k[2])

    v[i + 1] = v[i] + (h / 6) * (k[0] + 2 * k[1] + 2 * k[2] + k[3])

    ##############################################
    k[0] = ff(v[i], y[i])
    k[1] = ff(v[i] + h/2, y[i] + (h/2)*k[0])
    k[2] = ff(v[i] + h/2, y[i] + (h/2)*k[1])
    k[3] = ff(v[i] + h, y[i] + h*k[2])

    y[i + 1] = y[i] + (h / 6) * (k[0] + 2 * k[1] + 2 * k[2] + k[3])

# Solución analítica para comparación
omega = np.sqrt(W ** 2 - Γ ** 2)
y_analytical = y_0 * np.exp(-Γ / 2 * t) * np.cos(omega * t)

# Gráfica de las soluciones
plt.plot(t, y, label='Solución Numérica')
plt.plot(t, y_analytical, label='Solución Analítica', linestyle='dashed')
plt.grid(True)
plt.xlabel("t")
plt.ylabel("y")
plt.legend()
plt.title("Solución del Oscilador Armónico Amortiguado")
plt.show()

# Salchicha
```
