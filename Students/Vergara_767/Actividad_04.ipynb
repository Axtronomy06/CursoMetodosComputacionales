```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/hernansalinas/autogradesMetodosComputacionales/main/Dataset/Temperature_Bogota_Barcelona.csv")
df
```

```python
df["AvgTempCelcius"] = (df['AvgTemperature']-32)/1.8
df
```

```python
dff = df.loc[df['City'] == "Barcelona"]
dff2 = df.loc[df['City'] == "Bogota"]
```

```python
plt.plot(dff['Year'],dff['AvgTempCelcius'])
plt.title("Temperatura vs Año de la ciudad de Barcelona")
plt.xlabel("Año")
plt.ylabel("Cº")
plt.show()
```

```python
plt.plot(dff2['Year'],dff2['AvgTempCelcius'])
plt.title("Temperatura vs Año de la ciudad de Bogota")
plt.xlabel("Año")
plt.ylabel("Cº")
plt.show()
```

```python
max_bog = dff2['AvgTempCelcius'].max()
max_bar = dff['AvgTempCelcius'].max()
print("La temperatura maxima de Bogota es: ", max_bog, "grados celcius, y la temperatura maxima de Barcelona es: ", max_bar,"grados celcius")
```
