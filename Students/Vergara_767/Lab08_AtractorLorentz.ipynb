```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
# Definir las ecuaciones diferenciales del atractor de Lorenz
def lorentz_system(i_c, t, sigma, rho, beta):
    x, y, z = i_c
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return [dxdt, dydt, dzdt]

# Parámetros del sistema
sigma = 10
rho = 28
beta = 8 / 3

# Condiciones iniciales
initial_conditions = [1.5, 1.5, 1.5]

# Tiempo de integración
t = np.linspace(0, 25, 10000)

# Resolver el sistema de ecuaciones diferenciales usando odeint
sol = odeint(lorentz_system, initial_conditions, t, args=(sigma, rho, beta))

# Gráfica de las soluciones en el espacio tridimensional
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(sol[:, 0], sol[:, 1], sol[:, 2], label='Atractor de Lorenz')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
plt.title('Atractor de Lorenz')
plt.show()
```
