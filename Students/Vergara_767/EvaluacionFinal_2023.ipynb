```python
import numpy as  np
import matplotlib.pyplot as plt
from scipy.optimize import newton
```

## Todo en uno

```python
# Parámetros
T_inf = 300
Ac = 5*10**(-5)
p = 0.102
k = 200
T_b = 313
h = 20
a = h*p/(k*Ac)

L = 3 #longitud de la aleta
x0 = 0
xf = L

# Número de puntos y paso de longitud
n = 10000
dx = (xf - x0) / n

# Condiciones iniciales
theta_0 = T_b - T_inf
dtheta_0 = -(T_b - T_inf) * a**(1/2) #derivando la expresión analitica e evaluando en cero se obtiene esta expresión

#Crear arreglos de temperatura, su derivada y posición
x = np.linspace(x0, xf, n)
theta = np.zeros(n)
dtheta = np.zeros(n)

theta[0] = theta_0
dtheta[0] = dtheta_0

# Método de Euler para resolver el sistema de EDOs
for i in range(n - 1):
    dtheta[i + 1] = dtheta[i] + dx * (a**2)*theta[i]
    theta[i + 1] = theta[i] + dx * dtheta[i + 1]

# Solución númerica en T
T_numerical = theta + T_inf
# Solución analítica para comparación
T_analytical = T_inf + (T_b - T_inf)*np.exp(-x*np.sqrt(a))

# Gráfica de las soluciones
plt.plot(x, T_numerical, label='Solución Numérica')
plt.plot(x, T_analytical, label='Solución Analítica', linestyle='dashed')
plt.grid(True)
plt.xlabel("x")
plt.ylabel("T")
plt.legend()
#plt.title("Solución del Oscilador Armónico Simple")
plt.show()
```
