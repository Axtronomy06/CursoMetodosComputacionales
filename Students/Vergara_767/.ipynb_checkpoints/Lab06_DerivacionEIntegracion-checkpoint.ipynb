```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline
import sympy as sp
```

## Parte 1

```python
# Creación de arrays i y t
i = np.array([3.1, 3.12, 3.14, 3.18, 3.24])
t = np.array([1, 1.01, 1.02, 1.03, 1.04])

# Creación de un diccionario con los arrays i y t
dicc = {"t": t, "i": i}

# Creación de un DataFrame a partir del diccionario
df = pd.DataFrame(dicc)

# Creación de un array didt de ceros
didt = np.zeros(len(i) - 1)

# Cálculo de la derivada didt
for j in range(len(i) - 1):
    didt[j] = (df.loc[j+1]['i'] - df.loc[j]['i']) / (df.loc[j+1]['t'] - df.loc[j]['t'])

# Definición de una función para calcular el promedio
def promedio(array):
    return (array[0:len(array) - 1] + array[1:len(array)]) / 2

# Cálculo de los promedios de i y t
i_p = promedio(i)
t_p = promedio(t)

# Definición de constantes L y R
L = 0.98
R = 0.142

# Cálculo de la variable epsilon (ε)
ε = L * didt + R * i_p
```

```python
# Generación de valores de tiempo t_vals usando la función linspace
t_vals = np.linspace(t_p[0], t_p[-1], 1000)

# Creación de una función de interpolación cúbica (CubicSpline) para ε en función de t
cs = CubicSpline(t_p, ε)

# Cálculo de los valores interpolados de ε en t_vals
ε_vals = cs(t_vals)

# Creación de un gráfico que muestra ε en función del tiempo
plt.plot(t_vals, ε_vals, label='ε(t)')  # Línea de la función interpolada
plt.plot(t_p, ε, ".", label="puntos")    # Puntos de datos originales
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Generación de valores de tiempo t_valss usando la función linspace
t_valss = np.linspace(t[0], t[-1], 1000)

# Creación de una función de interpolación cúbica (CubicSpline) para i en función de t
css = CubicSpline(t, i)

# Cálculo de los valores interpolados de i en t_valss
i_vals = css(t_valss)

# Creación de un gráfico que muestra i en función del tiempo
plt.plot(t_valss, i_vals, label='i(t)')  # Línea de la función interpolada
plt.plot(t, i, ".", label="puntos")    # Puntos de datos originales
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Cálculo de la derivada de i_vals con respecto a t_valss usando np.diff
didt_f = np.diff(i_vals) / np.diff(t_valss)

# Cálculo de los promedios de i_vals y t_valss
i_vals_p = promedio(i_vals)
t_valss_p = promedio(t_valss)

# Cálculo de la variable ε (ε_f) usando la fórmula dada
ε_f = L * didt_f + R * i_vals_p

# Creación de un gráfico que muestra ε en función del tiempo
plt.plot(t_valss_p, ε_f, label='ε(t)')  # Línea de la función calculada
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Creación de un gráfico que muestra ε_f en función del tiempo
plt.plot(t_valss_p, ε_f, label='ε_f(t)')  # Línea de la función ε_f
plt.plot(t_vals, ε_vals, label='ε_p(t)')  # Línea de la función interpolada ε_p
plt.plot(t_p, ε, ".", label="puntos")    # Puntos de datos originales
plt.grid(True)
plt.legend()
plt.show()
```

# Parte 2

1. **Regla del Trapecio:**
   La expresión para calcular una integral con la regla del trapecio es:

   $$ \int_{a}^{b} f(x) \,dx \approx \frac{h}{2} \left[ f(a) + 2f(x_1) + 2f(x_2) + \ldots + 2f(x_{n-1}) + f(b) \right] $$

   Donde $ h = \frac{b - a}{n} $ es el paso de integración y $ x_i = a + ih $ son los nodos de integración.

2. **Regla del Simpson:**
   La expresión para calcular una integral con la regla del Simpson es:

   $$ \int_{a}^{b} f(x) \,dx \approx \frac{h}{3} \left[ f(a) + 4f(x_1) + 2f(x_2) + 4f(x_3) + \ldots + 2f(x_{n-2}) + 4f(x_{n-1}) + f(b) \right] $$

   Donde $ h = \frac{b - a}{n} $ es el paso de integración y $x_i = a + ih $ son los nodos de integración.

3. **Método de Cuadratura:**
   La expresión general para calcular una integral con el método de cuadratura es:

   $$ \int_{a}^{b} f(x) \,dx \approx \sum_{i=1}^{n} w_i f(x_i) $$

   Donde  $x_i$ son los nodos de integración, $ w_i $ son los pesos asociados a cada nodo, y $n$ es el número de nodos. Los nodos y pesos dependen del método de cuadratura específico utilizado (por ejemplo, cuadratura de Gauss).

```python
# Define una función llamada 'trap' que calcula la aproximación numérica de una integral mediante el método del trapecio.
def trap(f, a, b, Np):
    # Calcula el ancho de cada subintervalo.
    dx = (b - a) / Np

    # Calcula la suma ponderada de los valores de la función 'f' en los puntos del intervalo.
    s = (dx/2) * (-f[0] - f[-1] + 2*np.sum(f))

    # Retorna la aproximación numérica de la integral.
    return s
```

```python
def simpson(a, b, n, f):
  # Calcula una aproximación de la integral definida utilizando la regla compuesta de Simpson.

  h = (b - a)/n  # Calcula el ancho de cada subintervalo.
  x = np.linspace(a, b, n)  # Genera los puntos equidistantes en el intervalo [a, b].

  s_1 = 0
  s_2 = 0
  for j in range(1, int((n/2))):
    s_1 += f(x[2*j])  # Suma las evaluaciones de la función en puntos pares.
    s_2 += f(x[2*j - 1])  # Suma las evaluaciones de la función en puntos impares.

  s_2 += f(x[n-1])  # Agrega la evaluación de la función en el punto final.

  s = (h/3)*(f(a) + 2 * s_1 + 4 * s_2 + f(b))  # Fórmula de la regla compuesta de Simpson.
  return s  # Devuelve la aproximación de la integral.
```

```python
from scipy.integrate import trapezoid, quad

valor_real = np.zeros(3)

# Primer integral
e = 1e-7
a = 0 + e
b = np.pi/2 - e
n = 10000
x = np.linspace(a, b, n)
f_1 = 1/np.sqrt(np.tan(x))
valor_real[0] = trapezoid(f_1, x)
print("El valor verdadero de la primera integral es:", valor_real[0])

# Segunda integral
inf = 10e10
a = -inf
b = inf
n = 10000
x = np.linspace(a, b, n)
f_2 = 1/(1 + x**2)
valor_real[1] = trapezoid(f_2, x)
print("El valor verdadero de la segunda integral es:", valor_real[1])

# Tercera integral
a = 0
b = np.pi/4
n = 1000
x = np.linspace(a, b, n)
S_x = np.zeros(n)

j = 0
for i in x:
    f_i = lambda y: 2*y*np.sin(i) + (np.cos(i))**2
    S_x[j] = quad(f_i, c(i), d(i))[0]
    j += 1

valor_real[2] = trapezoid(S_x, x)
print("El valor verdadero de la tercera integral es:", valor_real[2])
```

\begin{equation}
P = \frac{2P_0}{\pi} \int _0^{\pi/2} \frac{\mathrm{d\phi}}{\sqrt{1-k^2\sin^2\phi}}
\end{equation}

```python
# Función para determinar el periodo como función numérica de un péndulo:
def calcular_periodo(L, d, theta_i):
    theta_i_rad = np.deg2rad(theta_i)
    g = 9.8
    b = L  # Longitud del péndulo (m)
    param_geo = d  # Parámetro geométrico del sistema
    m = 1  # Masa del péndulo (kg)
    I0 = b**2 * m * param_geo  # Momento de inercia

    T = 2 * np.pi * (np.sqrt(b * param_geo / g))
    K = np.sin(theta_i_rad / 2)

    # Definimos los límites de la integral:
    a = 0
    b = np.pi / 2

    # Definimos phi como un array:
    phi = np.linspace(a, b, 1000)

    # Definimos la función numérica:
    p_num = 2 * T / (np.pi * np.sqrt(1 - (K**2 * np.sin(phi)**2)))

    return p_num

# Utilizamos la función:
L = 0.25  # Longitud de hilo (m)
d = 1  # Parámetro geométrico del sistema
theta_i_values = np.arange(0, 90, 1)
p_vals = np.zeros(len(theta_i_values))

# Se crea un ciclo para calcular el periodo en función del ángulo:
for i, theta_i_iter in enumerate(theta_i_values):
    p_vals[i] = trapezoid(calcular_periodo(L, d, theta_i_iter), np.linspace(0, np.pi / 2, 1000))

# Graficamos los resultados:
plt.plot(theta_i_values, p_vals, label="P(s)")
plt.title("Periodo vs Ángulo en un Péndulo Simple")
plt.xlabel("Ángulo (°)")
plt.ylabel("Periodo (s)")
plt.grid()
plt.show()
```
