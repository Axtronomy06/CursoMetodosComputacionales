```python
import math as mt
import numpy as np
import matplotlib.pyplot as plt
```

## Exponencial, serie alternante (serieEn)

```python
# Definir el valor de x
x = 0.5

# Inicializar la suma s y el contador n
s = 0
n = 0

# Imprimir encabezados para mostrar los resultados
print("Iteración  |  Aproximación  |  Error (%)")
print("----------------------------------------")

# Comenzar un bucle while que se ejecutará mientras el error calculado sea mayor que un valor umbral
while 10e-7 * abs(s) < abs((-x)**(n+1) / mt.factorial(n+1)):
    # Calcular el siguiente término de la serie y agregarlo a la suma s
    term = (-1)**n * x**n / mt.factorial(n)
    s += term

    # Calcular el error en la aproximación actual en comparación con exp(-x)
    error = abs(mt.exp(-x) - s) / mt.exp(-x) * 100

    # Imprimir el número de iteración, la aproximación actual y el error
    print(f"{n:9}  |  {s:.10f}  |  {error:.7f}%")

    # Incrementar el contador n para la siguiente iteración
    n += 1
```

```python
# Definir el valor de x
x = 10

# Inicializar la suma s y el contador n
s = 0
n = 0

# Iniciar un bucle while que se ejecutará mientras el error calculado sea mayor que un valor umbral
while 10e-7 * abs(s) < abs((-x)**(n+1) / mt.factorial(n+1)):
    # Verificar si n es igual a x - 1
    if n == x - 1:
        # Si n es igual a x - 1, imprimir el término de la iteración n
        print(f"Término de la iteración: {n}, es igual a: ", (-1)**n * 10**n / mt.factorial(n))
        # También imprimir el término de la iteración n+1
        print(f"Término de la iteración: {n+1}, es igual a: ", (-1)**(n+1) * 10**(n+1) / mt.factorial(n+1))

    # Calcular el siguiente término de la serie y agregarlo a la suma s
    term = (-1)**n * x**n / mt.factorial(n)
    s += term

    # Incrementar el contador n para la siguiente iteración
    n += 1
```

La cancelación sustractiva ocurre cuando los términos en la serie son muy grandes y se restan entre sí para dar como resultado un valor pequeño, lo que puede llevar a una pérdida de precisión en el cálculo. En este caso, cuando n es igual a x - 1, estamos aproximadamente en el punto donde la cancelación sustractiva es evidente. Esto se debe a que los términos x**(x-1) y x**x son significativos y de magnitudes similares, pero tienen signos opuestos en la serie.

Como resultado, cuando n es igual a x - 1, observamos en las impresiones que los términos (-1)**n * 10**n / mt.factorial(n) y (-1)**(n+1) * 10**(n+1) / mt.factorial(n+1) son casi iguales en magnitud pero con signos opuestos. Esto es un claro ejemplo de cancelación sustractiva, y la diferencia entre estos términos es pequeña en comparación con sus magnitudes individuales.

```python
# Función para calcular la aproximación de 1/exp(x)
def expp(x, N):
  if x == 0:
    return 1
  s = 0
  ss = 0
  for n in range(N):
    s += x**n / mt.factorial(n)
    ss = 1/s
  return ss

# Función para calcular la aproximación de exp(-x)
def nexp(x, N):
  if x == 0:
    return 1
  s = 0
  for n in range(N):
    s += (-1)**n * x**n / mt.factorial(n)
  return s

x = 10
N = 40

# Imprimir encabezados para mostrar los resultados
print("Iteraciones  |  Error de exp(-x)  |  Error de 1/exp(x)")
print("-----------------------------------------------------")

# Calcular y mostrar el error para cada número de iteraciones
for i in range(1, N + 1):
  error_nexp = abs(nexp(x, i) - mt.exp(-x)) / mt.exp(-x) * 100
  error_expp = abs(expp(x, i) - mt.exp(-x)) / mt.exp(-x) * 100
  print(f"{i:12}  |  {error_nexp:.6f}%         |  {error_expp:.6f}%")
```

Basándonos en el código anterior, podemos concluir que la función expp, que calcula la aproximación de 1/exp(x) utilizando una serie, converge más rápidamente al valor real en comparación con la función nexp, que calcula la aproximación de exp(-x) utilizando una serie. Esta conclusión se basa en el análisis de los errores relativos en las aproximaciones para diferentes números de iteraciones (N).

Al observar los resultados de la ejecución del código, notamos que el error relativo en las aproximaciones de 1/exp(x) utilizando expp disminuye más rápidamente a medida que aumenta el número de iteraciones (N). En otras palabras, se necesita menos cantidad de términos en la serie para obtener una aproximación precisa de 1/exp(x) en comparación con la aproximación de exp(-x) utilizando nexp.

```python
N = 40 # Número de términos para cada función

# Imprimir encabezados para mostrar los resultados
print("Valores de x  |  Error de exp(-x)  |  Error de 1/exp(x)")
print("-----------------------------------------------------")

# Iterar a través de valores de x del 1 al 10
for i in range(1, 10+1):
    # Calcular el error relativo en las aproximaciones de exp(-x) y 1/exp(x)
    error_nexp = abs(nexp(i, N) - mt.exp(-i)) / mt.exp(-i) * 100
    error_expp = abs(expp(i, N) - mt.exp(-i)) / mt.exp(-i) * 100

    # Imprimir los valores de x y los errores relativos
    print(f"{i:12}  |  {error_nexp:.10f}%         |  {error_expp:.10f}%")
```

A partir de x = 8, la función nexp empieza a ser imprecisa en la aproximación de exp(-x). Esto se debe a la presencia de cancelación sustractiva en la serie de Taylor utilizada para la aproximación, lo que resulta en errores significativos en los cálculos.

```python
N = 40 # Número de términos para cada función

# Imprimir encabezados para mostrar los resultados
print("Valores de x  |  Error de exp(-x)  |  Error de 1/exp(x)")
print("-----------------------------------------------------")

# Iterar a través de valores de x del 1 al 10
for i in range(10, 101, 10):
    # Calcular el error relativo en las aproximaciones de exp(-x) y 1/exp(x)
    error_nexp = abs(nexp(i, N) - mt.exp(-i)) / mt.exp(-i) * 100
    error_expp = abs(expp(i, N) - mt.exp(-i)) / mt.exp(-i) * 100

    # Imprimir los valores de x y los errores relativos
    print(f"{i:12}  |  {error_nexp:.10f}%         |  {error_expp:.10f}%")
```

En este análisis, hemos observado que la función expp comienza a volverse imprecisa después de la 20ava iteración, mientras que la función nexp simplemente no es adecuada para aproximar exp(-x) en ningún punto de las iteraciones, ya que su error es muy grande desde el principio.

```python
N = 100

# Número de iteraciones
num_iteraciones = range(0, 51, 5)

# Número de filas (una por cada iteración)
num_filas = len(num_iteraciones)

# Configurar subplots con dos columnas
fig, axs = plt.subplots(num_filas, 2, figsize=(10, 3*num_filas))

for i, iteracion in enumerate(num_iteraciones):
    error_nexp = np.zeros(N)
    error_expp = np.zeros(N)

    for j in range(1, N + 1):
        error_nexp[j-1] = abs(nexp(iteracion, j) - mt.exp(-iteracion)) / mt.exp(-iteracion)
        error_expp[j-1] = abs(expp(iteracion, j) - mt.exp(-iteracion)) / mt.exp(-iteracion)

    # Graficar el error de exp(-x) en la primera columna
    axs[i, 0].plot(range(1, N+1), error_nexp, label=f"Error de exp({-iteracion})")
    axs[i, 0].set_title("Error de exp(-x) vs Número de términos")
    axs[i, 0].legend()
    axs[i, 0].set_xlabel("N")  # Nombre del eje X
    axs[i, 0].set_ylabel("Error")  # Nombre del eje Y

    # Graficar el error de 1/exp(x) en la segunda columna
    axs[i, 1].plot(range(1, N+1), error_expp, label=f"Error de 1/exp({iteracion})", color = "r")
    axs[i, 1].set_title("Error de 1/exp(x) vs Número de términos")
    axs[i, 1].legend()
    axs[i, 1].set_xlabel("N")  # Nombre del eje X
    axs[i, 1].set_ylabel("Error")  # Nombre del eje Y

# Ajustar el espaciado entre subplots
plt.tight_layout()

plt.show()
```

## Cancelación sustractiva

```python
def S_1(N):
  s = 0
  for n in range(1, 2*N + 1):
    s += (-1)**n * n / (n+1)
  return s

def S_2(N):
  s = 0
  ss = 0
  for n in range(1, N + 1):
    s +=  (2*n - 1)/(2*n)
    ss += 2*n/(2*n + 1)
  return -s + ss

def S_3(N):
  s = 0
  for n in range(1, N + 1):
    s += 1/(2*n * (2*n + 1))
  return s
```

```python
N = 25

print("        Serie 1      |          Serie 2        |        Serie 3")
print("----------------------------------------------------------------------")
for i in range(1, N):
  print(f"{S_1(i):.17f}  |  {S_2(i):.17f}   |  {S_3(i):.17f}")
```

Las diferencias en la convergencia de las tres series se encuentran en los últimos tres decimales con precisión de 17 cifras decimales.

```python
# Definir el valor de N
N = 10000

# Calcular el valor de referencia utilizando la función S_3 con N
value = S_3(N)

# Inicializar dos arrays para almacenar los errores de S_1 y S_2
error_s1 = np.zeros(N)
error_s2 = np.zeros(N)

# Calcular los errores para diferentes valores de N
for i in range(1, N + 1):
    # Calcular el error relativo para S_1 y S_2 con respecto al valor de referencia
    error_s1[i - 1] = abs((S_1(i) - value) / value)
    error_s2[i - 1] = abs((S_2(i) - value) / value)

# Calcular el logaritmo en base 10 de N
logN = np.log10(np.arange(1, N + 1))

# Graficar log(error_s1) vs log(N)
plt.figure(figsize=(10, 6))
plt.plot(logN, np.log10(error_s1), label='Error S_1')
plt.xlabel('N (log scale)')
plt.ylabel('Error (log scale)')
plt.title('Error S_1 vs N')
plt.legend()
plt.grid(True)

# Graficar log(error_s2) vs log(N)
plt.figure(figsize=(10, 6))
plt.plot(logN, np.log10(error_s2), label='Error S_2')
plt.xlabel('N (log scale)')
plt.ylabel('Error (log scale)')
plt.title('Error S_2 vs N')
plt.legend()
plt.grid(True)

# Mostrar las gráficas
plt.show()
```

Ambas series convergen a S_3 prácticamente a la misma tasa, de manera que con pocos términos están relativamente alejadas del valor real, pero a medida que se aumentan los términos, convergen al valor real.
