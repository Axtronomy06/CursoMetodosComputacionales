```python
from scipy.misc import derivative
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definimos la función cuadrática f(x) = x^2
f = lambda x:x**2

# Generamos un arreglo de 100 puntos equidistantes en el intervalo [1, 10]
x = np.linspace(1, 10, 100)

# Calculamos la derivada de f(x) en cada punto del arreglo x
df = derivative(f, x)

# Trazamos f(x) y f'(x) respectivamente
plt.plot(x, f(x), label = "f(x)")
plt.plot(x, df, label = "f'(x)")

# Agregamos una cuadrícula al gráfico y una leyenda
plt.grid(True)
plt.legend()

# Mostramos el gráfico generado
plt.show()
```

```python
# Definimos los parámetros del intervalo y el número de puntos
x_min = 1
x_max = 10
N_p = 100

# Generamos un arreglo de 100 puntos equidistantes en el intervalo [x_min, x_max]
x = np.linspace(x_min, x_max, N_p)

# Calculamos la longitud de cada subintervalo
h = (x_max - x_min)/N_p

# Definimos la función cuadrática f(x) = x^2
f = lambda x: x**2

# Definimos tres funciones diferentes para calcular la aproximación de la derivada de f(x) en cada punto del arreglo x
def df_1(x):
  i = 0
  df = np.zeros(len(x))
  for x_0 in x:
    df[i] = 1/(2*h) * (f(x_0 + h) - f(x_0 - h))
    i += 1
  return df

def df_2(x):
  i = 0
  df = np.zeros(len(x))
  for x_0 in x:
    df[i] = 1/(h) * (f(x_0 + h) - f(x_0))
    i += 1
  return df

def df_3(x):
  i = 0
  df = np.zeros(len(x))
  for x_0 in x:
    df[i] = 1/(h) * (f(x_0) - f(x_0 - h))
    i += 1
  return df

# Calculamos la derivada de f(x) en cada punto del arreglo x utilizando cada uno de los tres métodos
df_1 = df_1(x)
df_2 = df_2(x)
df_3 = df_3(x)

# Calculamos el error relativo de cada método
err_1 = np.mean(abs(df - df_1(x))/abs(df))
err_2 = np.mean(abs(df - df_2(x))/abs(df))
err_3 = np.mean(abs(df - df_3(x))/abs(df))

# Convertimos los errores relativos a porcentajes
err = np.array([err_1, err_2, err_3]) * 100

# Imprimimos los errores relativos de cada método en la consola
print("El error del primer método es: ", err[0], "%", "El error del segundo método es: ", err[1], "%", "El error del tercer método es: ", err[2],"%")
```

```python
plt.plot(x, abs(df - df_1(x)) / abs(df))
plt.title('Plot 1')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()

plt.plot(x, abs(df - df_2(x)) / abs(df))
plt.title('Plot 2')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()

plt.plot(x, abs(df - df_3(x)) / abs(df))
plt.title('Plot 3')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```
