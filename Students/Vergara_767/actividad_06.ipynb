```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
def bisection(a, b, f, tolerancia, max_iter):
    x = np.linspace(-10, 10, 1000)  # Crear una serie de puntos en el rango [-10, 10] para graficar f(x)
    iterations = 0
    plt.plot(x, f(x), label=f"f(x) = {expr}")  # Graficar la función f(x)

    c = (a + b) / 2  # Inicializar el punto medio c
    cc = 0
    while abs(f(c)) > tolerancia and iterations <= max_iter:
        if f(c) < 0:
            a = c
        else:
            b = c
        plt.plot(c, f(c), ".g")  # Graficar el punto (c, f(c)) en verde (antes de la actualización)
        cc = c
        c = (a + b) / 2
        iterations += 1
    plt.plot(cc, f(cc), ".g", label = "Iteraciones")
    plt.plot(c, f(c), ".r", label="Raíz")  # Graficar el punto (c, f(c)) en rojo (después de encontrar la raíz)
    return c

x = sp.symbols('x')
f = lambda x: x**2 - 1
expr = sp.sympify(f(x))
a = 0
b = 20
eps = 10e-7

if f(a) < 0 and f(b) > 0:
    print("La raíz encontrada es: ", round(bisection(a, b, f, eps, 100), 6))

plt.legend()  # Agregar una leyenda al gráfico
plt.grid(True)  # Mostrar una cuadrícula en el gráfico
plt.show()  # Mostrar el gráfico
```

```python
from scipy.optimize import bisect

def mi_funcion(x):
    return x**2 - 1

a = 0
b = 20

raiz = bisect(mi_funcion, a, b)

print("La raíz encontrada es:", round(raiz, 6))
```
