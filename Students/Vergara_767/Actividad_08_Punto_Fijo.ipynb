```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x: (x-1)**2
g = lambda x: x - f(x)
x = np.linspace(-4, 4, 100)
plt.plot(x,f(x), label = "f(x)")
plt.plot(x, g(x), label = "g(x)")
plt.plot(x, x, label = "x")
plt.legend()
plt.grid(True)
```

```python
def punto_fijo(f, g, p_0, eps, N_max):
    # Inicializa un array para almacenar los valores de p_i
    p = np.zeros(N_max)
    p[0] = p_0  # Establece el valor inicial p_0
    i = 1

    # Realiza el método del punto fijo
    while i < N_max:
        p[i] = g(p[i - 1])  # Calcula el siguiente valor usando g(p_{i-1})
        if abs(f(p[i])) < eps:  # Verifica si se alcanza la convergencia
            return p[i]  # Retorna el valor si la convergencia es alcanzada
        i += 1

    return None  # Devuelve None si se alcanza el número máximo de iteraciones sin convergencia


# Define la función f(x)
f = lambda x: (x - 1) ** 2

# Define la función g(x) para el método del punto fijo
g = lambda x: x - f(x)

# Aplica el método del punto fijo para encontrar una raíz aproximada
resultado = punto_fijo(f, g, 1.2, 10e-7, 1000)

# Imprime el resultado
if resultado is not None:
    print("La raíz aproximada es:", resultado)
else:
    print("Se alcanzó el número máximo de iteraciones sin convergencia.")
```

```python
from scipy.optimize import fixed_point

# Aplica el método del punto fijo usando mi implementación
resultado_tu_metodo = punto_fijo(f, g, 1.2, 10e-7, 1000)

# Aplica el método del punto fijo usando la función fixed_point de SciPy
resultado_scipy = fixed_point(g, 1.2, xtol=10e-7, maxiter=1000)

# Imprime los resultados
print("Resultado con tu implementación:", resultado_tu_metodo)

# fixed_point devuelve un objeto que contiene información adicional, así que extraemos el resultado de la raíz
print("Resultado con SciPy:", resultado_scipy.item())
```
