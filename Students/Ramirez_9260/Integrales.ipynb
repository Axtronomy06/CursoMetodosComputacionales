```python
import scipy
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

a=0
b=10
N=10000000
X=np.linspace(a,b,N)
```

```python
def f(x):
  return 3*x**2
```

```python
dx=(b-a)/N
fin=dx*(f(X+1)+f(X).sum())
plt.plot(X,fin)
print("El area es: ",fin[0])
```

```python
print("El area con la formula es : ", fin[0])
inte=scipy.integrate.trapezoid(f(X),X)
print("El area con scipy es: ", inte)
e=np.absolute(inte-fin[0])
print("El el error ente ambos metodos es: ", e )
```
