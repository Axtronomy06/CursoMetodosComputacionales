```python
def fun(t,y,*params):
  lamda=params[0]
  f= -lamda*y
  return f

def rk4(fun,t,y,*params):
  for i in range(0,len(t-1)):
    k1=fun(t[i],y[i],*params)
    k2=fun(t[i]+h/2,y[i]+h/2*k1,*params)
    k3=fun(t[i]+h/2,y[i]+h/2*k2,*params)
    k4=fun(t[i]+h/2,y[i]+h/2*k3,*params)
    y[i+1]=y[i]+h/6*(k1+2*k2+2*k3+k4)
  return y


tini=0
h=0.01
n=100
lamda=0.5
params=(lamda,)
t0=0
tf=10

N=int((tf-t0)/h)

tmax=int(n*h)
t=np.linspace(tini,tmax,N)
y=np.zeros(len(t))

t[0]=t0
y[0]=100
```
