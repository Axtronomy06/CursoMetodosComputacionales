```python
import scipy
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

a=0
b=2
X=np.linspace(a,b,365)
```

```python
def f(x):
  return (((x-1)**3)*np.sin(x))/3+1
plt.grid()
plt.plot(X,f(X))
```

```python
#fp=sp.diff((((x-1)**3)*sp.sin(x))/3+1,x)

def fp(x):
  return (((x-1)**3)*np.cos(x)/3)+(x-1)**2*np.sin(x)
plt.grid()
plt.plot(X,fp(X))
```

```python
#fpp=sp.diff(fp,x)

def fpp(x):
  return (-((x-1)**3)*np.sin(x)/3)+2*((x-1)**2)*np.cos(x)+(2*x-2)*np.sin(x)
plt.grid()
plt.plot(X,fpp(X))
```

```python
n2=((2**3)/(12*(10**-5)))*fpp(0)
n=np.sqrt(n2)
n
```

```python
A=np.sin(0)
B=np.cos(np.pi/4)
```
