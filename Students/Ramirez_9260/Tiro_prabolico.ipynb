```python
import math
import matplotlib.pyplot as plt
import json

with open("CondIni.json", "r") as archivo:

  d = json.load(archivo)
d

def calcular_tiro_parabolico(angulo, velocidad_inicial, gravedad, intervalo_tiempo):
    # Convertir el ángulo a radianes
    angulo_radianes = math.radians(angulo)

    # Componentes de la velocidad inicial en x e y
    velocidad_inicial_x = velocidad_inicial * math.cos(angulo_radianes)
    velocidad_inicial_y = velocidad_inicial * math.sin(angulo_radianes)

    # Calcular el tiempo en que el proyectil alcanza el suelo (y = 0)
    tiempo_vuelo = (2 * velocidad_inicial_y) / gravedad

    # Calcular las posiciones en x e y en cada intervalo de tiempo
    posiciones_x = []
    posiciones_y = []

    for t in range(0, int(tiempo_vuelo*100)+1):
        t_actual = t * intervalo_tiempo
        posicion_x = velocidad_inicial_x * t_actual
        posicion_y = velocidad_inicial_y * t_actual - 0.5 * gravedad * t_actual ** 2
        posiciones_x.append(posicion_x)
        posiciones_y.append(posicion_y)

    return posiciones_x, posiciones_y, tiempo_vuelo

# Parámetros de entrada
angulo = d["angulo"]#float(input("Ingrese el ángulo en grados: "))
velocidad_inicial = d["velocidad"]#float(input("Ingrese la velocidad inicial: "))
gravedad = 9.81  # Aceleración debido a la gravedad en m/s^2
intervalo_tiempo = 0.01  # Intervalo de tiempo entre puntos

# Calcular la trayectoria del tiro parabólico y el tiempo de vuelo
posiciones_x, posiciones_y, tiempo_vuelo = calcular_tiro_parabolico(angulo, velocidad_inicial, gravedad, intervalo_tiempo)

# Mostrar el tiempo de vuelo calculado
print("Tiempo de vuelo:", tiempo_vuelo, "segundos")

# Graficar la trayectoria
plt.figure()
plt.plot(posiciones_x, posiciones_y)
plt.title('Tiro Parabólico')
plt.xlabel('Distancia (m)')
plt.ylabel('Altura (m)')
plt.grid(True)
plt.show()

```
