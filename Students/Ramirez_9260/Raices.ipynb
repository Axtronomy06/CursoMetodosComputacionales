```python
import numpy as np
import math as mt
from math import e
import matplotlib.pyplot as plt

f=lambda x: x*(np.exp(x))-1
x=np.linspace(-1,1,100)
```

```python
plt.figure()
plt.plot(x,f(x))
plt.grid()
```

```python
a = -0.1
b = 0.1
c=a
epsilon=1E-5
iter=0
NmaxIter=100
while( (abs(f(c))>=epsilon) & (f(a)*f(b)<0) & (iter<NmaxIter)):
  c=(a+b)/2

  if (f(c)>0):
    b=c

  else:
    a=c
  iter=iter+1
```
